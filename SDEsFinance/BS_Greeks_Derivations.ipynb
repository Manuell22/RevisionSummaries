# Black-Scholes Greeks — Full Derivations

All Greeks are derived by differentiating the BS call price directly.
The same identity underlies almost every derivation, so we prove it first.

**Setup.**  In Black-Scholes the time-0 call price is
$$C = S_0\,\Phi(d_1) - K e^{-rT}\Phi(d_2)$$
with
$$d_1 = \frac{\ln(S_0/K)+(r+\tfrac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}, \qquad
  d_2 = d_1 - \sigma\sqrt{T} = \frac{\ln(S_0/K)+(r-\tfrac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$$

and $\Phi,\varphi$ the standard normal CDF and PDF.

---

## 0. Key Identity: $S_0\,\varphi(d_1) = K e^{-rT}\varphi(d_2)$

This cancellation is used in **every** Greek derivation, so we prove it once.

$$\frac{S_0\,\varphi(d_1)}{K e^{-rT}\varphi(d_2)}
  = \frac{S_0 e^{rT}}{K}\cdot\frac{e^{-d_1^2/2}}{e^{-d_2^2/2}}
  = \frac{S_0 e^{rT}}{K}\cdot e^{(d_2^2-d_1^2)/2}$$

Compute $d_1^2 - d_2^2 = (d_1-d_2)(d_1+d_2)$.  We have $d_1-d_2 = \sigma\sqrt{T}$ and

$$d_1+d_2 = \frac{2\ln(S_0/K)+2rT}{\sigma\sqrt{T}}$$

so

$$d_1^2 - d_2^2 = \sigma\sqrt{T}\cdot\frac{2\ln(S_0/K)+2rT}{\sigma\sqrt{T}}
  = 2\ln\!\left(\frac{S_0 e^{rT}}{K}\right)$$

Therefore

$$\frac{S_0 e^{rT}}{K}\cdot e^{(d_2^2-d_1^2)/2}
  = \frac{S_0 e^{rT}}{K}\cdot e^{-\ln(S_0 e^{rT}/K)}
  = \frac{S_0 e^{rT}}{K}\cdot\frac{K}{S_0 e^{rT}} = 1 \qquad \checkmark$$

$$\boxed{S_0\,\varphi(d_1) = K e^{-rT}\varphi(d_2)}$$

---

## 1. Partial Derivatives of $d_1$ and $d_2$

These are needed as building blocks.  Write
$$d_1 = \frac{\ln(S_0/K)+rT}{\sigma\sqrt{T}}+\frac{\sigma\sqrt{T}}{2}, \qquad
  d_2 = \frac{\ln(S_0/K)+rT}{\sigma\sqrt{T}}-\frac{\sigma\sqrt{T}}{2}$$

| Derivative | $\partial d_1/\partial\,\cdot$ | $\partial d_2/\partial\,\cdot$ | Difference $\partial(d_1-d_2)/\partial\,\cdot$ |
|---|---|---|---|
| $\partial/\partial S_0$ | $\dfrac{1}{S_0\sigma\sqrt{T}}$ | $\dfrac{1}{S_0\sigma\sqrt{T}}$ | $0$ |
| $\partial/\partial\sigma$ | $-\dfrac{\ln(S_0/K)+rT}{\sigma^2\sqrt{T}}+\dfrac{\sqrt{T}}{2}$ | $-\dfrac{\ln(S_0/K)+rT}{\sigma^2\sqrt{T}}-\dfrac{\sqrt{T}}{2}$ | $\sqrt{T}$ |
| $\partial/\partial T$ | $-\dfrac{\ln(S_0/K)+rT}{2\sigma T^{3/2}}+\dfrac{\sigma}{4\sqrt{T}} + \dfrac{r}{2\sigma\sqrt{T}}$ | same minus sign on last | $\dfrac{\sigma}{2\sqrt{T}}$ |
| $\partial/\partial r$ | $\dfrac{\sqrt{T}}{\sigma}$ | $\dfrac{\sqrt{T}}{\sigma}$ | $0$ |

The **differences** $\partial d_1/\partial x - \partial d_2/\partial x$ are the only quantities that survive after applying the key identity.

---

## 2. Delta — $\Delta = \partial C/\partial S_0$

Differentiate $C = S_0\Phi(d_1)-Ke^{-rT}\Phi(d_2)$ with respect to $S_0$:

$$\Delta = \Phi(d_1)
  + S_0\,\varphi(d_1)\frac{\partial d_1}{\partial S_0}
  - Ke^{-rT}\varphi(d_2)\frac{\partial d_2}{\partial S_0}$$

Since $\partial d_1/\partial S_0 = \partial d_2/\partial S_0 = \tfrac{1}{S_0\sigma\sqrt{T}}$ (they differ by a constant), collect the $\varphi$ terms:

$$\Delta = \Phi(d_1)
  + \underbrace{\bigl[S_0\,\varphi(d_1) - Ke^{-rT}\varphi(d_2)\bigr]}_{=\;0\text{ by key identity}}
  \cdot\frac{1}{S_0\sigma\sqrt{T}}$$

$$\boxed{\Delta_{\text{call}} = \Phi(d_1) \in (0,1)}$$

For the **put** via parity $P = C - S_0 + Ke^{-rT}$:
$$\Delta_{\text{put}} = \Delta_{\text{call}} - 1 = \Phi(d_1)-1 = -\Phi(-d_1) \in (-1,0)$$

**Interpretation.** $\Delta$ is the number of units of stock to hold in the replicating portfolio, and equals the $Q$-probability of finishing in the money (approximately).

---

## 3. Gamma — $\Gamma = \partial^2 C/\partial S_0^2$

Differentiate Delta once more:

$$\Gamma = \frac{\partial}{\partial S_0}\Phi(d_1)
  = \varphi(d_1)\frac{\partial d_1}{\partial S_0}
  = \varphi(d_1)\cdot\frac{1}{S_0\sigma\sqrt{T}}$$

$$\boxed{\Gamma = \frac{\varphi(d_1)}{S_0\,\sigma\sqrt{T}} > 0}$$

Gamma is the same for a call and a put (parity adds a linear term whose second derivative is zero).

**Interpretation.**
- $\Gamma > 0$ always: delta-hedged portfolios have **convexity**, earning from large moves.
- Gamma is highest when the option is at-the-money and close to expiry (small $T$), where $\varphi(d_1)\approx\varphi(0)=1/\sqrt{2\pi}$ is maximal and $\sigma\sqrt{T}$ is small.

---

## 4. Vega — $\mathcal{V} = \partial C/\partial\sigma$

$$\mathcal{V} = S_0\,\varphi(d_1)\frac{\partial d_1}{\partial\sigma}
  - Ke^{-rT}\varphi(d_2)\frac{\partial d_2}{\partial\sigma}$$

Apply the key identity to factor:

$$\mathcal{V} = S_0\,\varphi(d_1)\left(\frac{\partial d_1}{\partial\sigma} - \frac{\partial d_2}{\partial\sigma}\right)$$

From the table in §1, $\partial d_1/\partial\sigma - \partial d_2/\partial\sigma = \sqrt{T}$, so:

$$\boxed{\mathcal{V} = S_0\,\varphi(d_1)\,\sqrt{T} > 0}$$

Vega is the same for a call and a put (parity removes the $\sigma$-dependence of the forward).

**Interpretation.**
- Vega $> 0$ always: any option (call or put) gains value as volatility rises.
- Maximum at ATM-forward: $d_1 = \sigma\sqrt{T}/2$, so $\varphi(d_1) \approx \varphi(0) = 1/\sqrt{2\pi}$ is near its maximum.
- Deep ITM/OTM options have near-zero vega (payoff essentially certain or zero regardless of $\sigma$).

---

## 5. Theta — $\Theta = \partial C/\partial t$

Theta is the sensitivity to calendar time $t$.  Let $\tau = T-t$ be time to maturity; then $\Theta = -\partial C/\partial\tau$.

Differentiate $C$ with respect to $\tau$ (treating $T \equiv \tau$ as time-to-maturity):

$$\frac{\partial C}{\partial\tau}
  = S_0\,\varphi(d_1)\frac{\partial d_1}{\partial\tau}
  + rKe^{-r\tau}\Phi(d_2)
  - Ke^{-r\tau}\varphi(d_2)\frac{\partial d_2}{\partial\tau}$$

Apply the key identity to collect the $\varphi$ terms:

$$\frac{\partial C}{\partial\tau}
  = S_0\,\varphi(d_1)\underbrace{\left(\frac{\partial d_1}{\partial\tau}-\frac{\partial d_2}{\partial\tau}\right)}_{=\;\sigma/(2\sqrt{\tau})}
  + rKe^{-r\tau}\Phi(d_2)$$

Therefore $\partial C/\partial\tau = \dfrac{S_0\,\varphi(d_1)\,\sigma}{2\sqrt{\tau}} + rKe^{-r\tau}\Phi(d_2)$, and:

$$\boxed{\Theta = -\frac{\partial C}{\partial\tau}
  = -\frac{S_0\,\varphi(d_1)\,\sigma}{2\sqrt{T}} - rKe^{-rT}\Phi(d_2)}$$

**Why is Theta negative?**

Both terms are negative (or zero), so $\Theta < 0$: a call loses value as time passes, all else equal.

- The first term is the **time-decay of the optionality**: less time means less chance of large moves.
- The second term is the **interest earned on the strike**: holding the call defers paying $K$, and the present value of that deferral shrinks as $T$ decreases.

For the **put**, via $P = C - S_0 + Ke^{-rT}$, differentiate:
$$\Theta_{\text{put}} = \Theta_{\text{call}} + rKe^{-rT} = -\frac{S_0\,\varphi(d_1)\,\sigma}{2\sqrt{T}} + rKe^{-rT}\Phi(-d_2)$$

The put theta can be positive (deep ITM put benefits from rate effects) but is usually negative.

---

## 6. Rho — $\rho = \partial C/\partial r$

$$\rho = S_0\,\varphi(d_1)\frac{\partial d_1}{\partial r}
  + KTe^{-rT}\Phi(d_2)
  - Ke^{-rT}\varphi(d_2)\frac{\partial d_2}{\partial r}$$

From the table, $\partial d_1/\partial r = \partial d_2/\partial r = \sqrt{T}/\sigma$, so the $\varphi$ terms cancel by the key identity:

$$\rho = \underbrace{\bigl[S_0\,\varphi(d_1)-Ke^{-rT}\varphi(d_2)\bigr]}_{=\;0}\frac{\sqrt{T}}{\sigma}
  + KTe^{-rT}\Phi(d_2)$$

$$\boxed{\rho_{\text{call}} = KTe^{-rT}\,\Phi(d_2) > 0}$$

For the **put** via parity $P = C - S_0 + Ke^{-rT}$:
$$\rho_{\text{put}} = \rho_{\text{call}} - KTe^{-rT} = -KTe^{-rT}\Phi(-d_2) < 0$$

**Interpretation.**
- Call rho $> 0$: higher rates make calls more valuable (carrying the stock costs more, so the call as a substitute gains).
- Put rho $< 0$: higher rates reduce put value (present value of the strike payout falls).
- Rho is relatively small for short-dated options (factor of $T$) and matters more for long-dated ones.

---

## 7. Summary Table

| Greek | Definition | Call | Put |
|---|---|---|---|
| **Delta** | $\partial C/\partial S_0$ | $\Phi(d_1)$ | $-\Phi(-d_1)$ |
| **Gamma** | $\partial^2 C/\partial S_0^2$ | $\dfrac{\varphi(d_1)}{S_0\sigma\sqrt{T}}$ | same as call |
| **Vega** | $\partial C/\partial\sigma$ | $S_0\,\varphi(d_1)\sqrt{T}$ | same as call |
| **Theta** | $\partial C/\partial t$ | $-\dfrac{S_0\varphi(d_1)\sigma}{2\sqrt{T}} - rKe^{-rT}\Phi(d_2)$ | $-\dfrac{S_0\varphi(d_1)\sigma}{2\sqrt{T}} + rKe^{-rT}\Phi(-d_2)$ |
| **Rho** | $\partial C/\partial r$ | $KTe^{-rT}\Phi(d_2)$ | $-KTe^{-rT}\Phi(-d_2)$ |

**Key steps that appear in every derivation:**
1. Product rule on $\Phi(d_i)$ gives a $\varphi(d_i)\,\partial d_i/\partial x$ term.
2. The key identity $S_0\varphi(d_1) = Ke^{-rT}\varphi(d_2)$ cancels those terms whenever $\partial d_1/\partial x = \partial d_2/\partial x$.
3. The difference $\partial d_1/\partial x - \partial d_2/\partial x$ is the only survivor when the partials differ (Vega: $\sqrt{T}$; Theta: $\sigma/(2\sqrt{T})$).
