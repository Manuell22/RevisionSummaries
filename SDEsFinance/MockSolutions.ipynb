## Mock Exam 1

### Problem 1 Solution â€” SDEs: $dX = m\,dt + \sigma X\,dW$

**a) Existence and uniqueness.**

The drift is $\mu(t,x) = m$ and the diffusion coefficient is $\sigma(t,x) = \sigma x$. Both are measurable. The initial condition $x_0$ is deterministic with $E[X_0^2] = x_0^2 < \infty$.

**Lipschitz condition:**
$$|\mu(t,x) - \mu(t,y)| + |\sigma(t,x) - \sigma(t,y)| = |m - m| + |\sigma x - \sigma y| = \sigma|x - y|$$
Satisfied with $K = \sigma$.

**Linear growth:**
$$|m| + \sigma|x| \leq \max(|m|, \sigma)(1 + |x|)$$
Satisfied with $K' = \max(|m|, \sigma)$.

Therefore the SDE admits a **unique global solution**.

**b) Mean and variance of $X_T$.**

Integrate the SDE: $X_T = x_0 + mT + \sigma \int_0^T X_t\,dW_t$.

Taking expectations and using $E[\text{Ito integral}] = 0$:
$$\boxed{E[X_T] = x_0 + mT}$$

For the variance, apply Ito's formula to $\varphi(x) = x^2$:
$$d(X_t^2) = (2mX_t + \sigma^2 X_t^2)\,dt + 2\sigma X_t^2\,dW_t$$

Let $m_2(T) = E[X_T^2]$. Taking expectations and differentiating:
$$\frac{dm_2}{dT} = 2m(x_0 + mT) + \sigma^2 m_2(T), \quad m_2(0) = x_0^2$$

This is a linear ODE. Solving:
$$m_2(T) = e^{\sigma^2 T}\left[x_0^2 + \int_0^T e^{-\sigma^2 u} 2m(x_0 + mu)\,du\right]$$

Then $\text{Var}(X_T) = m_2(T) - (x_0 + mT)^2$.

**c) SDE for $Y = X_t^3$.**

Apply Ito's formula with $\varphi(x) = x^3$: $\varphi'(x) = 3x^2$, $\varphi''(x) = 6x$.

$$dY_t = 3X_t^2\,dX_t + \tfrac{1}{2}\cdot 6X_t\,(dX_t)^2$$
$$= 3X_t^2(m\,dt + \sigma X_t\,dW_t) + 3X_t \cdot \sigma^2 X_t^2\,dt$$
$$\boxed{dY_t = 3X_t^2(m + \sigma^2 X_t)\,dt + 3\sigma X_t^3\,dW_t}$$

In terms of $Y$: drift $= 3Y^{2/3}(m + \sigma^2 Y^{1/3})$, diffusion $= 3\sigma Y$, $Y_0 = x_0^3$.

### Problem 2 Solution â€” Straddle Option Pricing (Black Scholes)

**a) Payoff plot and investor profile.**

The straddle payoff $Y = (S_T - K)^+ + (K - S_T)^+ = |S_T - K|$ is a V-shaped function with minimum 0 at $S_T = K$. It increases symmetrically as $S_T$ moves away from $K$ in either direction.

An investor buying a straddle expects **high volatility** â€” large stock movements in either direction. The investor profits from large moves and loses (the premium paid) if the stock stays near $K$. The investor **relies on volatility** to make money.

**b) Straddle price.**

The straddle is the sum of a call and a put with the same strike $K = S_0 e^{rT}$ (ATM-forward).

Call price (Black-Scholes formula):
$$V^{\text{Call}} = s_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)$$

Put price via put-call parity: $V^{\text{Put}} = V^{\text{Call}} - (s_0 - Ke^{-rT})$

With $K = s_0 e^{rT}$: $s_0 - Ke^{-rT} = 0$, so $V^{\text{Put}} = V^{\text{Call}}$.

Straddle price:
$$V^{\text{Str}} = 2V^{\text{Call}} = 2\left[s_0\Phi(d_1) - s_0\Phi(d_2)\right]$$

With ATM-forward strike: $d_1 = \sigma\sqrt{T}/2$, $d_2 = -\sigma\sqrt{T}/2$.

$$\boxed{V^{\text{Str}} = 2s_0\left[\Phi\left(\frac{\sigma\sqrt{T}}{2}\right) - \Phi\left(-\frac{\sigma\sqrt{T}}{2}\right)\right] = 2s_0\left[2\Phi\left(\frac{\sigma\sqrt{T}}{2}\right) - 1\right]}$$

**c) Vega of the straddle.**

$$\text{Vega} = \frac{\partial V^{\text{Str}}}{\partial \sigma} = 2s_0\sqrt{T}\,\varphi\left(\frac{\sigma\sqrt{T}}{2}\right) > 0$$

Since $\varphi > 0$ always, the Vega is **strictly positive**: the straddle price increases with volatility. This confirms the intuition that the straddle is a bet on volatility.

### Problem 3 Solution â€” Straddle Option Pricing (Displaced Diffusion)

**a) Straddle price in DD.**

In the DD model $dS_t = rS_t\,dt + \sigma(S_t - \alpha e^{rt})\,dW_t$, write $S_t = \alpha e^{rt} + X_t$ where
$$dX_t = rX_t\,dt + \sigma X_t\,dW_t, \quad X_0 = s_0 - \alpha$$

So $X_t$ follows GBM. The straddle on $S$ with strike $K = s_0 e^{rT}$ becomes a straddle on $X$ with strike $K' = K - \alpha e^{rT} = (s_0 - \alpha)e^{rT}$, which is ATM-forward for $X$.

Applying Black-Scholes to $X$:
$$\boxed{V^{\text{Str}}_{\text{DD}} = 2(s_0 - \alpha)\left[2\Phi\left(\frac{\sigma\sqrt{T}}{2}\right) - 1\right]}$$

**b) Limit $\alpha = 0$.**

When $\alpha = 0$: $V^{\text{Str}}_{\text{DD}} = 2s_0[2\Phi(\sigma\sqrt{T}/2) - 1] = V^{\text{Str}}_{\text{BS}}$. âœ“

**c) Impact of $\alpha$ on the straddle price.**

Write $A = 2\Phi(\sigma\sqrt{T}/2) - 1 > 0$. Then $V^{\text{Str}}_{\text{DD}} = 2(s_0 - \alpha)A$.

$$\frac{\partial V^{\text{Str}}_{\text{DD}}}{\partial \alpha} = -2A < 0$$

Increasing $\alpha$ **decreases** the straddle price. For $\alpha < 0$, the DD straddle is more expensive than BS; for $\alpha > 0$, it is cheaper. The shift $\alpha$ effectively changes the effective initial value from $s_0$ to $s_0 - \alpha$.

### Problem 4 Solution â€” Risk Measures: VaR for a stock position

**a) Definition of VaR.**

Value at Risk $\text{VaR}_{H,\alpha}$ for horizon $H$ and confidence level $\alpha$ is the loss level $q$ such that
$$P(L_H < q) = \alpha$$
where $L_H = V_0 - V_H$ is the loss over the horizon (positive = loss).

**b) VaR for $N$ units of equity in Black-Scholes.**

Under $P$: $S_H = S_0 \exp\left[(\mu - \sigma^2/2)H + \sigma\sqrt{H}\,Z\right]$ where $Z \sim N(0,1)$.

Loss: $L_H = N(S_0 - S_H) = NS_0\left[1 - \exp\left((\mu - \sigma^2/2)H + \sigma\sqrt{H}\,Z\right)\right]$

Setting $P(L_H < q) = \alpha$ and solving:

$$\boxed{\text{VaR}_{H,\alpha} = NS_0\left[1 - \exp\left(-\sigma\sqrt{H}\,\Phi^{-1}(\alpha) + (\mu - \sigma^2/2)H\right)\right]}$$

**c) Drawback of VaR.**

VaR is **not subadditive**: $\text{VaR}(A+B)$ can exceed $\text{VaR}(A) + \text{VaR}(B)$. This means VaR does not recognise diversification benefits. A portfolio split into two can appear less risky than the combined portfolio, which is economically unreasonable.

**d) Physical vs risk-neutral dynamics.**

No. VaR uses the **physical measure $P$** dynamics (drift $\mu$), while option pricing uses the **risk-neutral measure $Q$** dynamics (drift $r$). The two measures differ because VaR measures real-world risk, while option pricing removes the risk premium.

## Mock Exam 2

### Problem 1 Solution â€” SDEs: $dX = m(X-a)\,dt + \sigma(X-a)\,dW$

**a) Existence and uniqueness.**

Drift $\mu(t,x) = m(x-a)$ and diffusion $\sigma(t,x) = \sigma(x-a)$ are both linear in $x$, hence measurable. 

Lipschitz: $|m(x-a) - m(y-a)| + |\sigma(x-a) - \sigma(y-a)| = (|m| + \sigma)|x-y|$. âœ“

Linear growth: $|m(x-a)| + \sigma|x-a| \leq (|m|+\sigma)(|a| + |x|) \leq K'(1+|x|)$. âœ“

The SDE admits a **unique global solution**.

**b) Stratonovich form.**

The Ito-to-Stratonovich correction is $\frac{1}{2}\sigma(x)\frac{\partial \sigma}{\partial x} = \frac{1}{2}\sigma(x-a) \cdot \sigma = \frac{\sigma^2}{2}(x-a)$.

Stratonovich drift $= m(x-a) - \frac{\sigma^2}{2}(x-a) = (m - \sigma^2/2)(x-a)$.

$$\boxed{dX_t = \left(m - \frac{\sigma^2}{2}\right)(X_t - a)\,dt + \sigma(X_t - a) \circ dW_t}$$

**c) Finding a solution.**

**Method 1 (Substitution):** Let $Y_t = X_t - a$. Then $dY_t = mY_t\,dt + \sigma Y_t\,dW_t$ with $Y_0 = x_0 - a$. This is GBM:
$$Y_t = (x_0 - a)\exp\left[(m - \sigma^2/2)t + \sigma W_t\right]$$

$$\boxed{X_t = a + (x_0 - a)\exp\left[(m - \sigma^2/2)t + \sigma W_t\right]}$$

**Method 2 (Stratonovich separation):** From the Stratonovich form, $\frac{dX}{X-a} = (m-\sigma^2/2)dt + \sigma \circ dW_t$. Integrating gives $\ln(X_t - a) - \ln(x_0 - a) = (m-\sigma^2/2)t + \sigma W_t$, yielding the same solution.

**d) Probability $P\{X_t > x_0\}$.**

$X_t > x_0 \iff (x_0-a)e^{(m-\sigma^2/2)t + \sigma W_t} > x_0 - a \iff e^{(m-\sigma^2/2)t + \sigma W_t} > 1$ (since $x_0 > a$).

$(m-\sigma^2/2)t + \sigma W_t > 0 \iff W_t > -(m-\sigma^2/2)t/\sigma$.

Since $W_t \sim N(0,t)$:
$$\boxed{P\{X_t > x_0\} = \Phi\left(\frac{(m - \sigma^2/2)\sqrt{t}}{\sigma}\right)}$$

### Problem 2 Solution â€” Option pricing Black Scholes (2 stocks)

**a) Risk-neutral dynamics.**

Under $Q$, replace drift $\mu_i$ with $r$:
$$dS^{(1)}_t = rS^{(1)}_t\,dt + \sigma_1 S^{(1)}_t\,dW^{(1),Q}_t, \quad dS^{(2)}_t = rS^{(2)}_t\,dt + \sigma_2 S^{(2)}_t\,dW^{(2),Q}_t$$
with $dW^{(1),Q}dW^{(2),Q} = \rho\,dt$.

**b) Portfolio price.**

The payoff $Y = (S^{(1)}_T - K_1)^+ + (K_2 - S^{(2)}_T)^+$ decomposes into two independent option payoffs.

$$V_{\text{BS}} = V^{\text{Call}}_{\text{BS}}(s_0^{(1)}, K_1, T, \sigma_1, r) + V^{\text{Put}}_{\text{BS}}(s_0^{(2)}, K_2, T, \sigma_2, r)$$

Call: $V^{\text{Call}} = s_0^{(1)}\Phi(d_1^{(1)}) - K_1 e^{-rT}\Phi(d_2^{(1)})$

Put (via put-call parity $V^{\text{Put}} = V^{\text{Call}} - s_0^{(2)} + K_2 e^{-rT}$):
$$V^{\text{Put}} = K_2 e^{-rT}\Phi(-d_2^{(2)}) - s_0^{(2)}\Phi(-d_1^{(2)})$$

**c) Dependence on correlation $\rho$.**

The price **does not depend on $\rho$**. The payoff is additive in the two stocks, so the pricing expectation separates:
$$V = E^Q[e^{-rT}(S^{(1)}_T - K_1)^+] + E^Q[e^{-rT}(K_2 - S^{(2)}_T)^+]$$

Each term depends only on the marginal distribution of one stock (which is lognormal regardless of $\rho$).

$$\boxed{\frac{\partial V_{\text{BS}}}{\partial \rho} = 0}$$

**d) Sensitivity to $\sigma_2$.**

Only the put on $S^{(2)}$ depends on $\sigma_2$:
$$\frac{\partial V_{\text{BS}}}{\partial \sigma_2} = s_0^{(2)}\varphi(d_1^{(2)})\sqrt{T} > 0$$

At ATM-forward ($K_2 = s_0^{(2)}e^{rT}$): $d_1^{(2)} = \sigma_2\sqrt{T}/2$, so
$$\frac{\partial V_{\text{BS}}}{\partial \sigma_2}\bigg|_{\text{ATMF}} = s_0^{(2)}\sqrt{T}\,\varphi(\sigma_2\sqrt{T}/2) > 0$$

### Problem 3 Solution â€” Option pricing with smile (2 stocks)

**a) Model selection.**

- Stock $S^{(1)}$: constant (flat) smile â†’ **Black-Scholes** model: $dS^{(1)}_t = rS^{(1)}_t\,dt + \sigma_1 S^{(1)}_t\,dW^{(1),Q}_t$

- Stock $S^{(2)}$: decreasing smile â†’ **Displaced Diffusion** (DD) model: $dS^{(2)}_t = rS^{(2)}_t\,dt + \sigma_2(S^{(2)}_t - \alpha e^{rt})\,dW^{(2),Q}_t$ with $\alpha > 0$ for a decreasing smile (or **Bachelier** model, which also gives a purely decreasing smile).

**b) Portfolio price at time 0.**

The payoff is $Y = (S^{(1)}_T - K_1)^+ - (K_2 - S^{(2)}_T)^+$. At ATMF strikes $K_i = S_0^{(i)}e^{rT}$:

For $S^{(1)}$ (BS): $V^{\text{Call}} = s_0^{(1)}[2\Phi(\sigma_1\sqrt{T}/2) - 1]$

For $S^{(2)}$ (DD): Write $S^{(2)}_t = \alpha e^{rt} + X_t$ with $X_t$ GBM. The put on $S^{(2)}$ becomes a put on $X$ at ATMF:
$V^{\text{Put}}_{\text{DD}} = (s_0^{(2)} - \alpha)[2\Phi(\sigma_2\sqrt{T}/2) - 1]$

$$V_0 = s_0^{(1)}[2\Phi(\sigma_1\sqrt{T}/2) - 1] - (s_0^{(2)} - \alpha)[2\Phi(\sigma_2\sqrt{T}/2) - 1]$$

**c) V-shaped smile model.**

For a V-shaped smile with minimum at ATM-forward, use the **Mixture Dynamics** model:
$$dS^{(2)}_t = rS^{(2)}_t\,dt + \sigma_{(j)} S^{(2)}_t\,dW^{(2),Q}_t$$
where $\sigma_{(j)}$ takes value $\sigma_A$ with probability $\lambda$ and $\sigma_B$ with probability $1-\lambda$ (chosen once at $t=0$, fixed for all $t$). The mixture of two lognormals produces a V-shaped implied volatility smile.

### Problem 4 Solution â€” Risk measures (2 stocks & Bachelier model)

**a) VaR with $\rho = 0$.**

Under $P$ with Bachelier: $S^{(i)}_H = s_0^{(i)} + \mu_i H + \sigma_i W^{(i)}_H$.

Portfolio value at $H$: $V_H = N_1 S^{(1)}_H - N_2 S^{(2)}_H + N_B e^{-r(T-H)}$ (with $r=0$: $= N_B$).

Loss: $L_H = V_0 - V_H = N_1(s_0^{(1)} - S^{(1)}_H) - N_2(s_0^{(2)} - S^{(2)}_H)$
$= -N_1(\mu_1 H + \sigma_1 W^{(1)}_H) + N_2(\mu_2 H + \sigma_2 W^{(2)}_H)$

With $\rho = 0$, $-N_1\sigma_1 W^{(1)}_H + N_2\sigma_2 W^{(2)}_H \sim N(0, (N_1^2\sigma_1^2 + N_2^2\sigma_2^2)H)$.

So $L_H \sim N\big((-N_1\mu_1 + N_2\mu_2)H,\;(N_1^2\sigma_1^2 + N_2^2\sigma_2^2)H\big)$.

$$\boxed{\text{VaR}_{H,\alpha} = (-N_1\mu_1 + N_2\mu_2)H + \Phi^{-1}(\alpha)\sqrt{(N_1^2\sigma_1^2 + N_2^2\sigma_2^2)H}}$$

**b) Barings-style risk analysis.**

- **Large $N_1$, small $N_2, N_B$:** Dominated by long stock position. Risky if stock 1 drops significantly (loss $\approx N_1\sigma_1\sqrt{H}\Phi^{-1}(\alpha)$).

- **Large $N_2$, small $N_1, N_B$:** Dominated by short stock position. Dangerous if stock 2 **rises** (unlimited loss potential since short position). Similar to Barings â€” Nick Leeson's short straddle position on the Nikkei led to catastrophic losses when the index moved.

- **Large $N_B$, small $N_1, N_2$:** Bond position with $r=0$ has no risk. VaR $\approx 0$. Very safe.

### Problem 1 Solution â€” SDEs with cubic root

**a) Finding an explicit solution.**

Transform to Stratonovich. The correction is:
$$\frac{1}{2}\sigma(x)\frac{\partial \sigma}{\partial x} = \frac{1}{2}x^{2/3} \cdot \frac{2}{3}x^{-1/3} = \frac{1}{3}x^{1/3}$$

Since the Ito drift is also $\frac{1}{3}x^{1/3}$, the Stratonovich drift is $\frac{1}{3}x^{1/3} - \frac{1}{3}x^{1/3} = 0$.

Stratonovich form: $dX_t = X_t^{2/3} \circ dW_t$

Separation of variables: $\frac{dX}{X^{2/3}} = \circ\,dW_t$

$$\int_{x_0}^{X_t} X^{-2/3}\,dX = W_t \implies 3X_t^{1/3} - 3x_0^{1/3} = W_t$$

$$\boxed{X_t = \left(x_0^{1/3} + \frac{W_t}{3}\right)^3}$$

**b) A posteriori check.**

Let $Z_t = x_0^{1/3} + W_t/3$. Then $X_t = Z_t^3$.

By Ito: $dX_t = 3Z_t^2\,dZ_t + \frac{1}{2}\cdot 6Z_t\,(dZ_t)^2 = 3Z_t^2 \cdot \frac{1}{3}dW_t + 3Z_t \cdot \frac{1}{9}dt$

$= Z_t^2\,dW_t + \frac{1}{3}Z_t\,dt = X_t^{2/3}\,dW_t + \frac{1}{3}X_t^{1/3}\,dt$ âœ“

**c) Non-uniqueness when $X_0 = 0$.**

Two solutions exist:
1. $X_t = (W_t/3)^3$ (from the formula above)
2. $X_t \equiv 0$ (constant solution â€” check: $dX = 0 = \frac{1}{3}\cdot 0\,dt + 0\,dW_t$ âœ“)

**Why non-uniqueness is expected:** The coefficients $x^{1/3}$ and $x^{2/3}$ are **not Lipschitz** at $x = 0$ (their derivatives blow up). The Lipschitz condition fails, so uniqueness is not guaranteed by the existence-uniqueness theorem.

### Problem 2 Solution â€” Asset/Cash or Nothing options in Black Scholes

**a) Payoff descriptions.**

- ANO: $Y^{AN} = S_T\mathbf{1}_{\{S_T > K\}}$ â€” pays the full stock price if $S_T > K$, else nothing. Attractive for speculators very confident the stock will rise above $K$ (cheaper than owning the stock).

- CNO: $Y^{CN} = K\mathbf{1}_{\{S_T > K\}}$ â€” pays a fixed amount $K$ if $S_T > K$, else nothing. Attractive for investors wanting a binary bet: fixed payout if the stock finishes above a threshold.

**b) Portfolio equivalence.**

$Y^{AN} - Y^{CN} = S_T\mathbf{1}_{\{S_T > K\}} - K\mathbf{1}_{\{S_T > K\}} = (S_T - K)\mathbf{1}_{\{S_T > K\}} = (S_T - K)^+$

This is the **call option** payoff. So: ANO $-$ CNO $=$ Call.

**c) Price of the CNO.**

$V^{CN}_0 = e^{-rT}E^Q[K\mathbf{1}_{\{S_T > K\}}] = Ke^{-rT}Q(S_T > K) = Ke^{-rT}\Phi(d_2)$

where $d_2 = \frac{\ln(s_0/K) + (r - \sigma^2/2)T}{\sigma\sqrt{T}}$.

$$\boxed{V^{CN}_0 = Ke^{-rT}\Phi(d_2)}$$

**d) Price of the ANO.**

From b): ANO $=$ Call $+$ CNO.

$$V^{AN}_0 = V^{\text{Call}} + V^{CN}_0 - V^{CN}_0 + V^{CN}_0$$

Actually: Call = ANO - CNO, so ANO = Call + CNO:
$$V^{AN}_0 = [s_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)] + Ke^{-rT}\Phi(d_2) = s_0\Phi(d_1)$$

$$\boxed{V^{AN}_0 = s_0\Phi(d_1)}$$

### Problem 3 Solution â€” Asset/Cash or Nothing options with smile

**a) Model choice and CNO price with V-shaped smile.**

For a V-shaped smile with minimum at ATM-forward, use the **Mixture Dynamics** model with $N=2$ components:

The stock density is a mixture: $p_{S_T}(y) = \lambda\, p^{\text{LN}}_{\sigma_1}(y) + (1-\lambda)\,p^{\text{LN}}_{\sigma_2}(y)$

CNO price in mixture model:
$$V^{CN}_{\text{mix}} = \sum_i \lambda_i\, Ke^{-rT}\Phi(d_2(\sigma_i))$$

**b) Delta of the CNO.**

$$\Delta^{CN}_{\text{mix}} = \sum_i \lambda_i\, Ke^{-rT}\varphi(d_2(\sigma_i))\frac{1}{s_0 \sigma_i \sqrt{T}} > 0$$

**c) Sign of Delta.**

Delta is always **positive**: higher $S_0$ makes $S_T > K$ more likely, so the CNO is worth more. The CNO price is increasing in $S_0$.

**d) ATM-forward specialisation ($K = S_0 e^{rT}$).**

At ATMF: $d_2(\sigma_i) = -\sigma_i\sqrt{T}/2$

$$V^{CN}_{\text{mix}}\big|_{\text{ATMF}} = \sum_i \lambda_i\, s_0\Phi(-\sigma_i\sqrt{T}/2)$$

$$\Delta^{CN}_{\text{mix}}\big|_{\text{ATMF}} = \sum_i \lambda_i\frac{\varphi(-\sigma_i\sqrt{T}/2)}{\sigma_i\sqrt{T}}$$

This is independent of $S_0$ and always positive.

### Problem 4 Solution â€” Risk Measures: ES for a stock in Bachelier

**a) Risk-neutral dynamics.**

Under $Q$ with $r = 0$: $dS_t = \sigma\,dW_t^Q$ (drift becomes $r = 0$).

**b) Definition of Expected Shortfall.**

$$\text{ES}_{T,\alpha} = E[L_T \mid L_T > \text{VaR}_{T,\alpha}] = \frac{E[L_T \cdot \mathbf{1}_{\{L_T > \text{VaR}\}}]}{1 - \alpha}$$

ES is the average loss in the worst $(1-\alpha)$ fraction of scenarios.

**c) ES for $N$ units of equity.**

Under $P$: $S_T = S_0 + \mu T + \sigma\sqrt{T}\,Z$, $Z \sim N(0,1)$.

Loss: $L_T = N(S_0 - S_T) = N(-\mu T - \sigma\sqrt{T}\,Z)$

VaR: $\text{VaR}_{T,\alpha} = N(-\mu T + \sigma\sqrt{T}\,\Phi^{-1}(\alpha))$

Computing the conditional expectation:
$$\boxed{\text{ES}_{T,\alpha} = -N\mu T + \frac{N\sigma\sqrt{T}\,\varphi(\Phi^{-1}(\alpha))}{1-\alpha}}$$

**d) Drawback of ES.**

ES averages the tail but doesn't show the full tail structure. It is also **homogeneous** in portfolio size, neglecting **liquidity risk**: in practice, larger positions are more than proportionally risky because they are harder to unwind.

**e) Physical vs risk-neutral dynamics.**

No. ES uses the **physical measure $P$** (drift $\mu$), while option pricing uses $Q$ (drift $r = 0$). Different measures for different purposes: real-world risk vs arbitrage-free pricing.

## Mock Exam 4

### Problem 1 Solution â€” SDEs: $dX = v^2 X(1+X^2)\,dt + v(1+X^2)\,dW$

**a) Finding a formal solution.**

Ito-to-Stratonovich correction: $\frac{1}{2}v(1+x^2) \cdot 2vx = v^2 x(1+x^2)$

Stratonovich drift $= v^2x(1+x^2) - v^2x(1+x^2) = 0$.

Stratonovich form: $dX_t = v(1+X_t^2) \circ dW_t$

Separation: $\frac{dX}{1+X^2} = v\,\circ dW_t$

$\int_{x_0}^{X_t} \frac{dX}{1+X^2} = vW_t \implies \arctan(X_t) - \arctan(x_0) = vW_t$

$$\boxed{X_t = \tan(\arctan(x_0) + vW_t)}$$

**b) A posteriori check.**

Let $Z_t = \arctan(x_0) + vW_t$, so $X_t = \tan(Z_t)$.

$\frac{d}{dz}\tan(z) = 1 + \tan^2(z) = 1 + X^2$, $\frac{d^2}{dz^2}\tan(z) = 2\tan(z)(1+\tan^2(z)) = 2X(1+X^2)$.

By Ito: $dX = (1+X^2)v\,dW + \frac{1}{2}\cdot 2X(1+X^2)v^2\,dt = v(1+X^2)\,dW + v^2X(1+X^2)\,dt$ âœ“

**c) Problems with the solution.**

The tangent function has vertical asymptotes at $\arctan(x_0) + vW_s = \pm\pi/2 + n\pi$. Since Brownian motion is unbounded, there exists a **positive probability** that $vW_s$ reaches $\pi/2$ in finite time.

This means the solution **explodes in finite time** with positive probability. The drift and diffusion coefficients have **super-linear growth** (cubic drift, quadratic diffusion), which violates the linear growth condition. The theorem guaranteeing global existence does not apply.

### Problem 2 Solution â€” Option pricing: no arbitrage & put-call parity

**a) Arbitrage when $\text{Call}_0 = \text{Put}_0 + X$, $X > 0$.**

The call is overpriced relative to the put. Strategy at $t=0$:
- **Sell** (write) the call: receive $\text{Call}_0 = \text{Put}_0 + X$
- **Buy** the put: pay $\text{Put}_0$
- **Enter a long forward** at strike $K$ (costs 0 at inception)

Net cost at $t=0$: receive $\text{Call}_0 - \text{Put}_0 = X > 0$ (invest at rate $r$).

At maturity $T$, the combined payoff of short call + long put + long forward:
$$-(S_T - K)^+ + (K - S_T)^+ + (S_T - K) = 0$$

(This is zero regardless of $S_T$: if $S_T > K$: $-(S_T-K) + 0 + (S_T-K) = 0$; if $S_T \leq K$: $0 + (K-S_T) + (S_T-K) = 0$.)

Total profit = $Xe^{rT} > 0$ â€” riskless profit. **Arbitrage.** âœ“

**b) Model independence.**

This arbitrage argument uses **only** the payoff definitions and the existence of a riskless rate $r$. It does **not** require Black-Scholes or any specific model. It works for Bachelier, DD, CEV, mixture dynamics, or any other model. Put-call parity is **model-free**.

**c) Opposite violation: $\text{Call}_0 = \text{Put}_0 - X$.**

Now the put is overpriced. Reverse the strategy:
- **Buy** the call, **sell** the put, **enter a short forward**.
- Net receipt at $t=0$: $\text{Put}_0 - \text{Call}_0 = X > 0$.
- Payoff at $T$: $(S_T-K)^+ - (K-S_T)^+ - (S_T-K) = 0$.
- Total profit: $Xe^{rT} > 0$. **Arbitrage.** âœ“

### Problem 3 Solution â€” Smile modeling: butterfly spread in BS & Bachelier

**a) Butterfly payoff.**

$Y = (S_T - L)^+ - 2(S_T - M)^+ + (S_T - H)^+$ with $L = S_0 - X$, $M = S_0$, $H = S_0 + X$.

Piecewise:
- $S_T \leq L$: $Y = 0$
- $L < S_T \leq M$: $Y = S_T - L$
- $M < S_T \leq H$: $Y = H - S_T$
- $S_T > H$: $Y = 0$

This is a **triangular** payoff, non-zero only in $[L, H]$, with maximum value $X$ at $S_T = M = S_0$.

**b) Investor profile.**

An investor buying a butterfly expects **low volatility** â€” the stock stays near $S_0$. Maximum profit $X$ if $S_T = S_0$. Limited risk: the premium paid is the maximum loss.

**c) Black-Scholes price ($r = 0$).**

$$V^{\text{Butterfly}}_{\text{BS}} = C_{\text{BS}}(S_0, L) - 2C_{\text{BS}}(S_0, M) + C_{\text{BS}}(S_0, H)$$

where $C_{\text{BS}}(s_0, K) = s_0\Phi(d_1(K)) - K\Phi(d_2(K))$ with $r = 0$.

**d) Butterfly delta in BS.**

$$\Delta = \Phi(d_1(L)) - 2\Phi(d_1(M)) + \Phi(d_1(H))$$

**e) Bachelier price ($r = 0$).**

Using Bachelier call formula $V^{\text{Call}} = (s_0-K)\Phi\left(\frac{s_0-K}{\sigma\sqrt{T}}\right) + \sigma\sqrt{T}\,\varphi\left(\frac{s_0-K}{\sigma\sqrt{T}}\right)$:

$$V^{\text{Butterfly}}_{\text{BaM}} = X[2\Phi(X/(\sigma\sqrt{T})) - 1] - 2\sigma\sqrt{T}[\varphi(0) - \varphi(X/(\sigma\sqrt{T}))]$$

**f) Delta in Bachelier.**

$$\Delta_{\text{BaM}} = \Phi\left(\frac{X}{\sigma\sqrt{T}}\right) - 2\Phi(0) + \Phi\left(\frac{-X}{\sigma\sqrt{T}}\right) = 0$$

**Zero delta** in Bachelier â€” the butterfly price does not depend on $s_0$.

**g) Limit $X \downarrow 0$.**

As $X \to 0$, the triangular payoff shrinks to zero height at a single point. Price â†’ 0. Confirmed: $V_{\text{BaM}}(X=0) = 0 \cdot [2\Phi(0)-1] - 2\sigma\sqrt{T}[\varphi(0) - \varphi(0)] = 0$. âœ“

### Problem 4 Solution â€” Risk measures: long Bond & short forward in B&S

### Q3 Solution â€” Option pricing Bachelier (long Straddle)

**a) Straddle price.**

Straddle $= V^{\text{Call}} + V^{\text{Put}} = 2V^{\text{Call}} - (s_0 - K)$

With $K = s_0$:
$$V^{\text{Call}} = \sigma\sqrt{T}\,\varphi(0) = \frac{\sigma\sqrt{T}}{\sqrt{2\pi}}$$

$$V^{\text{Put}} = V^{\text{Call}} = \frac{\sigma\sqrt{T}}{\sqrt{2\pi}}$$

$$\boxed{V_0^{\text{Str}} = \frac{2\sigma\sqrt{T}}{\sqrt{2\pi}} = \sigma\sqrt{\frac{2T}{\pi}}}$$

**b) Delta.**

Starting from the general straddle formula:
$$\frac{\partial V^{\text{Str}}}{\partial s_0} = 2\Phi\left(\frac{s_0-K}{\sigma\sqrt{T}}\right) - 1$$

At $K = s_0$: $\Delta = 2\Phi(0) - 1 = 0$.

### Q4 Solution â€” ES of a stock position in Bachelier

**a) ES definition.**

$$\text{ES}_{T,\alpha} = E[L_T \mid L_T > \text{VaR}_{T,\alpha}] = \frac{E[L_T \cdot \mathbf{1}_{\{L_T > \text{VaR}\}}]}{1-\alpha}$$

ES is the **average loss** in the worst $(1-\alpha)$ fraction of scenarios. It is always $\geq$ VaR.

**b) ES computation.**

Under $P$: $S_T = S_0 + \mu T + \sigma\sqrt{T}Z$, $Z \sim N(0,1)$.

Loss: $L_T = N(S_0 - S_T) = N(-\mu T - \sigma\sqrt{T}Z)$

VaR: $\text{VaR}_{T,\alpha} = N(-\mu T + \sigma\sqrt{T}\Phi^{-1}(\alpha))$

Computing the conditional tail expectation:
$$\boxed{\text{ES}_{T,\alpha} = -N\mu T + \frac{N\sigma\sqrt{T}\,\varphi(\Phi^{-1}(\alpha))}{1-\alpha}}$$

**c) Drawback of ES.**

ES is **homogeneous** in portfolio size: doubling the portfolio exactly doubles ES. This ignores **liquidity risk** â€” in practice, larger positions are disproportionately harder (and costlier) to unwind, so the true risk grows faster than linearly.

**d) Physical vs risk-neutral.**

No. ES uses the **physical measure $P$** (drift $\mu$) to capture real-world loss probabilities. Option pricing uses the **risk-neutral measure $Q$** (drift $r = 0$ in Bachelier). Different purposes: risk measurement vs arbitrage-free pricing.

### Q5 Solution â€” Mastery: Black-Scholes PDE and risk neutral measure

**a) Black-Scholes model.**

The economy has two assets:
- **Bank account:** $dB_t = rB_t\,dt$, $B_0 = 1$ (deterministic growth at rate $r$)
- **Stock:** $dS_t = \mu S_t\,dt + \sigma S_t\,dW_t$, $S_0 = s_0$ (GBM under $P$)

Assumptions: constant $r, \mu, \sigma$; no transaction costs; continuous trading; no dividends; no default risk; short selling allowed; infinitely divisible assets.

**b) BS PDE derivation (sketch).**

Assume claim price $V(t,S_t)$ is $C^{1,2}$. By Ito:
$$dV = \left(\frac{\partial V}{\partial t} + \mu S\frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2 S^2\frac{\partial^2 V}{\partial S^2}\right)dt + \sigma S\frac{\partial V}{\partial S}dW$$

Construct self-financing portfolio: $\Pi = V - \frac{\partial V}{\partial S}S$ (hedge the Brownian motion).

$d\Pi = dV - \frac{\partial V}{\partial S}dS = \left(\frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2 S^2\frac{\partial^2 V}{\partial S^2}\right)dt$

Since $\Pi$ is riskless: $d\Pi = r\Pi\,dt$, giving:

$$\boxed{\frac{\partial V}{\partial t} + rS\frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2 S^2\frac{\partial^2 V}{\partial S^2} = rV}$$

with terminal condition $V(T,S) = (S-K)^+$.

**c) Feynman-Kac and expectations.**

The **Feynman-Kac theorem** connects the PDE to a probabilistic representation:
$$V(t,S) = e^{-r(T-t)}E^Q_{t,S}[(S_T - K)^+]$$

This gives the price as a discounted expectation of the payoff under a new probability measure $Q$ (the risk-neutral measure), where the stock drift is $r$ instead of $\mu$.

**d) Girsanov's theorem.**

**Girsanov's theorem** provides the change of measure from $P$ to $Q$:

$$\frac{dQ}{dP} = \exp\left(-\frac{1}{2}\lambda^2 T - \lambda W_T\right)$$

where $\lambda = \frac{\mu - r}{\sigma}$ is the **market price of risk** (Sharpe ratio).

Under $Q$: $W_t^Q = W_t + \lambda t$ is a Brownian motion, and $dS_t = rS_t\,dt + \sigma S_t\,dW_t^Q$.

**a) VaR definition.**

**c) VaR drawback solved by ES.**

**d) Physical vs risk-neutral dynamics.**

No. VaR uses the **physical measure $P$** with drift $\mu$ (real-world stock dynamics). Option pricing uses the **risk-neutral measure $Q$** with drift $r$. The physical measure captures actual risk probabilities; the risk-neutral measure is used for arbitrage-free pricing.

### Q5 Solution â€” Mastery: Barings Collapse

**a) Short history.**

Nick Leeson was a trader at Barings' Singapore branch (1992â€“1995). He engaged in massive unauthorized futures and options trading on the Nikkei 225 and Japanese government bonds. He hid losses in a secret "88888" error account, fabricating reports to London. By early 1995, cumulative hidden losses exceeded Â£800 million.

**b) The option combination.**

Leeson **sold (shorted) straddles**: short a call and short a put with the same strike $K$ and maturity $T$.

Payoff: $Y = -(S_T-K)^+ - (K-S_T)^+ = -|S_T - K|$

This is an inverted V-shape: maximum payoff is 0 at $S_T = K$, with losses increasing as $S_T$ moves away from $K$ in either direction. The losses are **unbounded**.

**c) What went wrong.**

By selling straddles, Leeson collected premiums at time 0. If the Nikkei stayed near $K$, the options would expire worthless and he would keep the premium. He was betting on **low volatility** and a stable market.

The **Great Hanshin (Kobe) earthquake** struck on January 17, 1995, causing the Nikkei to drop dramatically. The short straddle position suffered enormous losses from the short put side. Rather than cutting losses, Leeson doubled down by buying more futures, compounding the losses. Total losses reached Â£827 million, exceeding Barings' entire capital, causing the bank to collapse on February 26, 1995.

**d) Risk measure introduced.**

**Value at Risk (VaR)** was formalised and adopted industry-wide shortly after, through the Basel Committee's market risk amendment (1996).

VaR for confidence level $\alpha$ and horizon $H$ is the loss $q$ such that $P(L_H < q) = \alpha$. It represents the **maximum expected loss** over period $H$ with probability $\alpha$. Had proper VaR limits been in place and monitored, Leeson's massive accumulated exposure would have been flagged and prevented.

**c) Zero-coupon bond (notional $N$, maturity $T$, constant $r$).**

Value at $H$: $Ne^{-r(T-H)}$. Value at 0: $Ne^{-rT}$.

$L_H = Ne^{-rT} - Ne^{-r(T-H)} = Ne^{-rT}(1-e^{rH})$ â€” **deterministic** (no randomness).

VaR equals $L_H$ itself. If $r > 0$: $L_H < 0$ (guaranteed gain), so VaR is trivially negative.

**d) $P$ vs $Q$ for VaR.**

VaR uses the **physical measure $P$** (drift $\mu$) because it measures real-world risk. Option pricing uses $Q$ (drift $r$). Using $Q$ for VaR would give the "risk-neutral loss distribution," which doesn't represent actual probabilities of loss.
