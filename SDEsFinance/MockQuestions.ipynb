## Mock Exam 1

### Problem 1 — SDEs: $dX = m\,dt + \sigma X\,dW$

Consider the SDE $dX_t = m\,dt + \sigma X_t\,dW_t$, $X_0 = x_0$ where $m \in \mathbb{R}$, $\sigma > 0$ and $x_0 \in \mathbb{R}$ deterministic.

a) Prove that the SDE admits a unique solution.

b) For the solution $X$, find $\mathbb{E}(X_T)$ and $\text{Var}(X_T)$ for any given $T > 0$.

c) If $Y = X_t^3$, find the SDE satisfied by $Y$. Write down the drift, diffusion coefficient and initial condition for the $Y$ SDE.

### Problem 2 — Straddle Option Pricing (Black Scholes)

Consider a stock market where the stock price $S$ follows $dS_t = rS_t\,dt + \sigma S_t\,dW_t$ under the risk neutral measure $Q$, with initial stock price $s_0 > 0$ deterministic. The risk free rate $r$ is a non-negative deterministic constant. Consider a straddle payoff on $S$, with maturity $T$ and strike $K = S_0 e^{rT}$:
$$Y = (S_T - K)^+ + (K - S_T)^+$$

a) Make a plot of the payoff as a function of $S_T$. Explain what kind of investor would buy this payoff. What would the investor rely on, to make money, when buying this product?

b) Calculate the price of the straddle at time 0. You can use the formula for a call option in Black-Scholes without deriving it, if you remember it, but derive the put from the call using parity.

c) How does the straddle price above change with $\sigma$? In particular, calculate the Vega of the straddle, namely $\text{Vega} = \frac{\partial V}{\partial \sigma}$, and discuss your findings.

### Problem 3 — Straddle Option Pricing (Displaced Diffusion)

Consider a stock market where the stock price $S$ follows the displaced diffusion (DD) dynamics
$$dS_t = rS_t\,dt + \sigma(S_t - \alpha e^{rt})\,dW_t, \quad s_0,$$
where $\alpha$ is a deterministic shift, under the risk neutral measure $Q$, with initial stock price $s_0 > 0$ deterministic. Consider a straddle payoff with maturity $T$ and strike $K = S_0 e^{rT}$:
$$Y = (S_T - K)^+ + (K - S_T)^+$$

a) Compute the straddle price at time 0 in the DD model.

b) Check that in the limit case $\alpha = 0$ you get back the Black Scholes price of a straddle.

c) Can you comment if, ceteris paribus, adding the shift $\alpha$ increases or decreases the Straddle price compared to the pure Black Scholes case? More generally, what is the impact of $\alpha$ on the straddle price? How is the price sensitive to $\alpha$?

### Problem 4 — Risk Measures: VaR for a stock position

Consider the dynamics of an equity asset price $S$ in the Black and Scholes model, under both probability measures $P$ (the Physical or Historical measure) and $Q$ (the risk neutral measure).

a) Define Value at Risk (VaR) for a time horizon $H$ with confidence level $\alpha$ for a general portfolio.

b) Compute VaR for horizon $H$ and confidence level $\alpha$ for a portfolio with $N$ units of equity, where the equity price follows the Black Scholes process above.

c) Explain at least one drawback of VaR as a risk measure.

d) Is the equity dynamics you used for VaR the same you would have used to price an equity call option in Black Scholes?

## Mock Exam 2

### Problem 1 — SDEs: $dX = m(X-a)\,dt + \sigma(X-a)\,dW$

Consider the Ito SDE
$$dX_t = m(X_t - a)\,dt + \sigma(X_t - a)\,dW_t, \quad x_0$$
where $a$ and $m$ are deterministic constants, $\sigma > 0$ and $x_0$ is deterministic. We assume $x_0 > a$.

a) Does this equation admit a unique global solution?

b) Write the equation in Stratonovich form.

c) If the equation admits solutions, find a solution. Extra points if you give two possible ways to get a solution.

d) Compute the following probability for the solution you found in point c): $P\{X_t > x_0\}$.

### Problem 2 — Option pricing Black Scholes (2 stocks)

Consider a portfolio with a call option on a first stock $S^{(1)}$ with strike $K_1$ and a put option on a second stock $S^{(2)}$ with strike $K_2$, both options with maturity $T$. The final payoff is
$$Y = (S^{(1)}_T - K_1)^+ + (K_2 - S^{(2)}_T)^+.$$
The risk-free rate $r > 0$ is a positive deterministic constant. Assume both stocks follow Black Scholes under $P$:
$$dS^{(1)}_t = \mu_1 S^{(1)}_t\,dt + \sigma_1 S^{(1)}\,dW^{(1)}_t, \quad dS^{(2)}_t = \mu_2 S^{(2)}_t\,dt + \sigma_2 S^{(2)}\,dW^{(2)}_t, \quad dW_1 dW_2 = \rho\,dt.$$

a) Write the stocks dynamics under the risk neutral measure $Q$.

b) Calculate the price $V_{BS}$ of the portfolio at time 0. You can use the formula for a call option in Black Scholes without deriving it, but for the put option derive it with put-call parity from the call.

c) Does the price $V_{BS}$ of the portfolio depend on the correlation $\rho$? Elaborate and provide intuition on your answer. Also, calculate the sensitivity $\frac{\partial V_{BS}}{\partial \rho}$.

d) How is the portfolio price sensitive to the volatility of the second stock? Calculate $\frac{\partial V_{BS}}{\partial \sigma_2}$. Take the special case of at-the-money-forward options, namely $K_1 = s_0^{(1)} e^{rT}$ and $K_2 = s_0^{(2)} e^{rT}$, and find the sensitivity in this special case.

### Problem 3 — Option pricing with smile (2 stocks)

Consider a portfolio long a call option on a first stock $S^{(1)}$ with strike $K_1$ and short a put option on a second stock $S^{(2)}$ with strike $K_2$, both options with maturity $T$. The final payoff is
$$Y = (S^{(1)}_T - K_1)^+ - (K_2 - S^{(2)}_T)^+.$$
The risk-free rate $r \geq 0$. We assume the strikes are at-the-money-forward: $K_{1,2} = S^{(1,2)}_0 e^{rT}$.

a) Assume that the market volatility smile curve is roughly constant for the first stock, and is decreasing for the second one. Choose suitable models for the first and second stock that are consistent with this pattern.

b) With the chosen models, price the portfolio at time 0.

c) If the market smile pattern for the second stock had been V shaped, what smile model would you have chosen for the second stock? You are not requested to calculate the price, only to discuss the model.

### Problem 4 — Risk measures (2 stocks & Bachelier model)

Consider two stocks in a market with zero interest rate $r = 0$. We assume both stocks follow a Bachelier model under the measure $P$:
$$dS^{(1)}_t = \mu_1\,dt + \sigma_1\,dW^{(1)}_t, \quad dS^{(2)}_t = \mu_2\,dt + \sigma_2\,dW^{(2)}_t, \quad dW_1 dW_2 = \rho\,dt$$

a) Consider a portfolio long an amount $N_1$ of stock $S^{(1)}$, short an amount $N_2$ of stock $S^{(2)}$, both with maturity $T$, and long a zero coupon bond with maturity $T$ and notional $N_B$. The portfolio payoff at maturity $T$ is
$$Y = N_1 S^{(1)}_T - N_2 S^{(2)}_T + N_B.$$
Compute the value at risk of this portfolio when $\rho = 0$ (i.e. the two Brownian motions are independent), for a risk horizon $H < T$ at a confidence level $\alpha$.

b) Thinking of the Barings collapse, describe a situation that could put the bank at serious risk when trading this portfolio. Analyze: large $N_1$ with small $N_2, N_B$; large $N_2$ with small $N_1, N_B$; large $N_B$ with small $N_1, N_2$.

### Problem 2 — Option pricing: no arbitrage & put-call parity

Consider the Black and Scholes basic economy given by a bank account and a stock:
$$dB_t = rB_t\,dt, \quad B_0 = 1, \quad dS_t = \mu S_t\,dt + \sigma S_t\,dW_t, \quad S_0 = 1$$
Consider two at-the-money-forward options with strike $K = S_0 e^{rT}$: a call option and a put option with maturity $T$.

Put-call parity tells us $\text{CallPrice}_0 = \text{PutPrice}_0$ at the ATM-forward strike.

a) Show that if put-call parity is violated at time 0: $\text{CallPrice}_0 = \text{PutPrice}_0 + X$ for a positive amount $X > 0$, even if the condition holds later for every $t > 0$, then one has arbitrage.

b) Does the existence of arbitrage in point a) require the Black Scholes model or is it more general?

c) If put-call parity is violated at time 0 in the opposite direction: $\text{CallPrice}_0 = \text{PutPrice}_0 - X$ with positive $X$, but again holds at times $t > 0$, show that we still have arbitrage.

### Problem 3 — Smile modeling: butterfly spread in BS & Bachelier

All options have maturity $T$.

a) Write the payoff of this product and draw it as a function of $S_T$.

b) Who would buy this product? What are the views on the stock for a client buying this product?

c) Compute the price of this product in a Black Scholes market with $dS_t = \mu S_t\,dt + \sigma S_t\,dW_t$, $S_0 = s_0$ with $r = 0$.

d) Compute the butterfly delta, namely the sensitivity of the butterfly price with respect to $S_0$.

e) Compute the butterfly price in a Bachelier model $dS_t = v\,dW_t^Q$, $S_0 = s_0$.

f) Compute the delta of the butterfly price in the Bachelier model.

g) Consider the limit situation when $X \downarrow 0$. Deduce intuitively the price of the butterfly by looking at the plot of the payoff. Check with a calculation whether the BS and Bachelier prices confirm this intuition.

### Problem 4 — Risk measures: long bond & short forward in B&S
