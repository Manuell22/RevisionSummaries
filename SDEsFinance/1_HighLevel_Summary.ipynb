## What This Course Is About

This course connects **stochastic calculus** (the mathematics of random, continuous-time processes) with **financial modelling** (pricing derivatives, measuring risk, and simulating market scenarios). You will:

1. Build the probability toolkit needed to define and work with Brownian motion and SDEs.
2. Learn Itô calculus — the chain rule for stochastic processes — and solve important SDEs analytically.
3. Use those SDEs to derive the **Black–Scholes** option-pricing framework and extend it to models that capture the **volatility smile**.
4. Quantify financial risk via **Value-at-Risk (VaR)** and **Expected Shortfall (ES)**.
5. Simulate SDEs numerically when no closed-form solution exists (**Euler–Maruyama** scheme).

The course is split into **four parts**, plus 7 mock exams with worked solutions, mastery questions, a group coursework, and Python code appendices.

## Part 1 — Probability, Stochastic Processes & SDEs

### 1.1 Probability Foundations (refresher)

| Concept | One-liner |
|---|---|
| Sample space $\Omega$ | Set of all possible outcomes of a random experiment |
| $\sigma$-field $\mathcal{F}$ | Collection of events we can assign probabilities to |
| Probability measure $P$ | Function $P: \mathcal{F} \to [0,1]$ satisfying Kolmogorov axioms |
| Random variable $X$ | Measurable map $X : \Omega \to \mathbb{R}$ |
| CDF / PDF | $F_X(x) = P(X \le x)$; density $f_X$ s.t. $F_X(x) = \int_{-\infty}^{x} f_X(u)\,du$ |

Key distributions you must know: **Uniform**, **Normal / Gaussian** $N(\mu,\sigma^2)$, **Exponential**, **Lognormal**, and the **Multivariate Normal**.

Important results:
- **Moment Generating Function (MGF):** $M_X(u) = E[e^{uX}]$ — uniquely determines a distribution and is handy for computing moments.
- **Central Limit Theorem (CLT):** The average of $n$ i.i.d. random variables converges in distribution to a Gaussian as $n \to \infty$.
- **Modes of convergence:** almost sure $\Rightarrow$ in probability $\Rightarrow$ in law; $L^p$ convergence $\Rightarrow$ in probability.

### 1.2 Brownian Motion

A **standard Brownian motion** $W = (W_t)_{t \ge 0}$ is the fundamental building block:
- $W_0 = 0$.
- **Independent increments:** $W_t - W_s$ is independent of the past for $s < t$.
- **Gaussian increments:** $W_t - W_s \sim N(0, t-s)$.
- **Continuous paths** (but nowhere differentiable, and of **unbounded variation** — this is why ordinary calculus breaks down).

### 1.3 Stochastic Differential Equations

An SDE has the generic form:

$$dX_t = \mu(t, X_t)\,dt + \sigma(t, X_t)\,dW_t, \quad X_0 = x_0$$

- $\mu$ = **drift** (deterministic pull)
- $\sigma$ = **diffusion coefficient** (random shock intensity; in financial models such as GBM, also called **volatility**)
- $dW_t$ = Brownian "noise"

**Existence & uniqueness** of a solution is guaranteed under **Lipschitz continuity** and **linear growth** conditions on $\mu$ and $\sigma$. When these fail (e.g., fractional powers, quadratic growth), solutions may not be unique or may **explode in finite time**.

### 1.4 Itô's Formula (the stochastic chain rule)

$\varphi(t, x)$ is a **deterministic function** of time and a dummy variable $x$. Itô's formula tells you the SDE satisfied by the new process $Y_t = \varphi(t, X_t)$, where $X_t$ is substituted in for $x$:

$$d\varphi = \frac{\partial \varphi}{\partial t}dt + \frac{\partial \varphi}{\partial x}dX_t + \frac{1}{2}\frac{\partial^2 \varphi}{\partial x^2}(dX_t)^2$$

with the **quadratic-variation rules** $dW \cdot dW = dt$, $dt \cdot dW = 0$, $dt \cdot dt = 0$.

Expanding $(dX_t)^2 = \sigma^2\,dt$ (lower-order terms vanish), the correction term becomes $\tfrac{1}{2}\frac{\partial^2\varphi}{\partial x^2}\sigma^2\,dt$ — a pure $dt$ term. This extra $\tfrac{1}{2}\varphi''\sigma^2\,dt$ is the key difference from ordinary calculus.

**Stratonovich calculus** is an alternative convention where the ordinary chain rule holds — useful for solving SDEs by separation of variables. The Itô-to-Stratonovich conversion adds a correction term: $\circ\,dW = dW + \frac{1}{2}\sigma'\sigma\,dt$.

### 1.5 Landmark SDE Solutions

**Solving an SDE** means finding an explicit formula $X_t = f(t, W_t, x_0)$ — a deterministic expression for $X_t$ in terms of $t$, the Brownian path $W_t$, and the initial condition $x_0$. The standard strategy is to guess an ansatz $\varphi(t, W_t)$, apply Itô's formula, and match coefficients.

| SDE | Name | Solution / Distribution |
|---|---|---|
| $dS = \mu S\,dt + \sigma S\,dW$ | **Geometric Brownian Motion (GBM)** | $S_T = S_0\exp\!\bigl((\mu - \tfrac12\sigma^2)T + \sigma W_T\bigr)$ — lognormal |
| $dX = k(\theta - X)\,dt + \sigma\,dW$ | **Ornstein–Uhlenbeck / Vasicek** | Gaussian, mean-reverting to $\theta$ |
| $dX = k(\theta - X)\,dt + \sigma\sqrt{X}\,dW$ | **CIR / Square-root process** | Non-central $\chi^2$, stays positive (under Feller condition) |
| $dX = m\,dt + \sigma X\,dW$ | **Linear drift + multiplicative noise** | Solved via integrating factor; $E[X_t]$, $\text{Var}(X_t)$ derivable |
| $dX = \tfrac12 X\,dt + \sqrt{1+X^2}\,dW$ | **Sinh SDE** | $X_t = \sinh(\sinh^{-1}(x_0) + W_t)$ |
| $dX = v^2X(1+X^2)\,dt + v(1+X^2)\,dW$ | **Explosion SDE** | $X_t = \tan(\arctan(x_0) + vW_t)$ — explodes in finite time (linear growth fails!) |

### 1.6 Key Theorems

- **Product rule for SDEs:** $d(XY) = X\,dY + Y\,dX + dX\,dY$ (the cross-term matters).
- **Correlated Brownian motions:** If $\rho = \text{Corr}(W^1, W^2)$, then $dW^1 dW^2 = \rho\,dt$.
- **Martingales:** A process with zero drift is a **martingale** — its best forecast for the future is its current value: $E[X_t \mid \mathcal{F}_s] = X_s$.
- **Feynman–Kac:** Connects a PDE to an SDE expectation — lets you compute option prices either way.
- **Girsanov's Theorem:** Tells you how to **change the probability measure** (and thereby the drift of an SDE) while keeping the diffusion coefficient the same. This is the mathematical engine behind risk-neutral pricing.

## Part 2 — SDEs for Option Pricing

### 2.1 The Black–Scholes Economy

Two assets:
- **Risk-free bank account:** $B_t = e^{rt}$ (deterministic growth at rate $r$).
- **Stock:** $dS_t = \mu S_t\,dt + \sigma S_t\,dW_t$ (GBM under the real-world measure $P$).

Ideal-world assumptions: no transaction costs, no dividends (initially), continuous trading, short-selling allowed, constant $r$ and $\sigma$.

### 2.2 Option Pricing — The Big Idea

A **European call** pays $(S_T - K)^+$ at maturity $T$; a **put** pays $(K - S_T)^+$.

**Two equivalent routes to the price:**

1. **PDE approach:** construct a self-financing hedging portfolio, apply Itô's formula, eliminate risk $\Rightarrow$ the **Black–Scholes PDE**:
$$\frac{\partial V}{\partial t} + rS\frac{\partial V}{\partial S} + \tfrac12 \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} = rV$$

2. **Risk-neutral (Feynman–Kac) approach:** use Girsanov to switch from $P$ to the **risk-neutral measure** $Q$ (under which the stock drift becomes $r$), then the price is a discounted expectation:
$$V_0 = e^{-rT}\,E^Q\!\bigl[\text{payoff}(S_T)\bigr]$$

### 2.3 The Black–Scholes Formula

$$\boxed{C = S_0\,\Phi(d_1) - K e^{-rT}\Phi(d_2)}$$

where

$$d_{1,2} = \frac{\ln(S_0/K) + (r \pm \tfrac12\sigma^2)T}{\sigma\sqrt{T}}$$

- **Put price** via put–call parity: $P = C - S_0 + Ke^{-rT}$.
- **Delta:** $\Delta_{\text{call}} = \Phi(d_1)$, $\Delta_{\text{put}} = \Phi(d_1) - 1$.
- **Put–call parity** is model-free: $C - P = S_0 - Ke^{-rT}$ (proven by no-arbitrage). Violation $\Rightarrow$ arbitrage.

### 2.4 Option Strategies & Exotic Payoffs

The course covers several **option strategies** (combinations of calls/puts):

| Strategy | Payoff | Investor view |
|---|---|---|
| **Straddle** (long) | $\|S_T - K\|$ | Expects high volatility |
| **Risk Reversal** | $(S_T - K_1)^+ - (K_2 - S_T)^+$ | Directional bet |
| **Bull Call Spread** | $(S_T - K_1)^+ - (S_T - K_2)^+$, $K_1 < K_2$ | Moderate bullish |
| **Bear Call Spread** | $(S_T - K_2)^+ - (S_T - K_1)^+$, $K_1 > K_2$ | Moderate bearish |
| **Butterfly Spread** | $(S_T - L)^+ - 2(S_T - M)^+ + (S_T - H)^+$ | Low volatility |
| **Calendar Spread** | Call($T_1$) $-$ Call($T_2$), $T_1 > T_2$ | Time-value play |

**Binary/digital options:**
- **Cash-or-Nothing (CNO):** pays $K$ if $S_T > K$, else $0$. Price in BS: $Ke^{-rT}\Phi(d_2)$.
- **Asset-or-Nothing (ANO):** pays $S_T$ if $S_T > K$, else $0$. Price in BS: $S_0\Phi(d_1)$.

### 2.5 The Greeks

| Greek | Measures sensitivity to | Key fact |
|---|---|---|
| $\Delta$ | Stock price $S$ | Hedge ratio; $\Delta_{\text{call}} = \Phi(d_1)$ |
| $\Gamma$ | Second-order in $S$ (convexity) | Always positive for long options |
| $\Theta$ | Time decay | Usually negative for long options |
| Vega ($\mathcal{V}$) | Volatility $\sigma$ | $\mathcal{V} = S_0\phi(d_1)\sqrt{T}$; always positive |
| $\rho$ | Interest rate $r$ | $\rho_{\text{call}} = KTe^{-rT}\Phi(d_2)$ |

### 2.6 Fundamental Theorems of Asset Pricing & Numeraires

1. **No arbitrage** $\iff$ there exists an **equivalent martingale measure** $Q$.
2. **Market completeness** (every claim can be hedged) $\iff$ $Q$ is **unique**.

In the BS model, $Q$ is unique $\Rightarrow$ the market is complete and every European payoff has a unique price.

A **numeraire** is any positive-value asset used for discounting. The bank account $B_t$ is the standard numeraire; the stock $S_t$ can also be a numeraire (leading to a different measure). Changing numeraire is equivalent to changing measure.

**Forward price** of a stock: $F(0,T) = S_0 e^{rT}$ — this is model-independent (no-arbitrage argument).

### 2.7 Dividends Extension

For a stock paying a continuous dividend yield $q$, replace $S_0$ with $S_0 e^{-qT}$ and use $(r-q)$ in $d_1, d_2$:
$$C = S_0 e^{-qT}\Phi(d_1) - Ke^{-rT}\Phi(d_2), \quad d_{1,2} = \frac{\ln(S_0/K) + (r - q \pm \sigma^2/2)T}{\sigma\sqrt{T}}$$

## Part 2 (cont.) — The Volatility Smile

### Why Do We Need More Than Black–Scholes?

If BS were perfectly correct, **implied volatility** (the $\sigma$ you plug into BS to match a market price) would be the same for every strike $K$. In reality it varies — this is the **volatility smile/skew**.

The course covers four alternative models that generalise the GBM diffusion to produce different smile shapes:

| Model | Stock SDE (under $Q$) | Smile shape | Key idea |
|---|---|---|---|
| **Bachelier** | $dS = \sigma\,dW$ (normal, $r=0$) | Downward smile | Additive (absolute) volatility |
| **Displaced Diffusion (DD)** | $dX = rX\,dt + \sigma X\,dW$, $S = \alpha e^{rt} + X$ | Skew (direction depends on sign of $\alpha$) | Shifts the lognormal distribution |
| **CEV** | $dS = rS\,dt + \nu S^\gamma\,dW$ | Skew controlled by $\gamma$ | Power-law local volatility |
| **Mixture Diffusion (MDD)** | Weighted average of BS prices at different $\sigma_i$ | U-shaped smile | Density is a mixture of lognormals |

You can also **combine** DD with MDD (**Shifted MDD**) for even richer smile shapes (asymmetric smile with curvature).

**Calibration in practice:** Given market implied vols $\nu_{\text{mkt}}(K_i)$, find model parameters $\beta$ minimising $\sum_i [V_{\text{BS}}(\nu_{\text{mkt}}) - V_{\text{Model}}(\beta)]^2$. This is a non-linear optimisation problem typically solved numerically.

**Local vol vs. Stochastic vol:** The models above are *local volatility* (volatility is a function of $S$ and $t$). They all suffer from **future smile flattening** — the predicted smile for future dates flattens to flat as the density converges to a single lognormal. *Stochastic volatility* models (e.g., **Heston**) let volatility follow its own SDE, preserving the smile into the future but introducing **incomplete markets** (no unique $Q$).

## Part 3 — Risk Measures

Here we switch from pricing (under $Q$) to **risk measurement** (under the real-world measure $P$).

### 3.1 Setup

Given a portfolio whose value at a future horizon $H$ is $V_H$, define the **loss**
$$L_H = V_0 - V_H$$
(positive $L_H$ means you lost money).

To find the distribution of $L_H$:
1. **Simulate** the underlying(s) under $P$ from today to $H$ (using the real-world drift $\mu$).
2. **Price** the portfolio at $H$ using the risk-neutral ($Q$) formulas (e.g., BS with remaining maturity $T-H$).
3. Compute $L_H = V_0 - V_H$ for each scenario.

### 3.2 Value-at-Risk (VaR)

$$\text{VaR}_{H,\alpha} = \alpha\text{-percentile of the loss distribution}$$

i.e., $P(L_H \le \text{VaR}) = \alpha$ (typically $\alpha = 95\%$ or $99\%$).

**Drawbacks:** VaR tells you *nothing* about how bad losses can get beyond the threshold, and it is **not sub-additive** (diversification can appear to *increase* risk).

### 3.3 Expected Shortfall (ES)

$$\text{ES}_{H,\alpha} = E[L_H \mid L_H > \text{VaR}_{H,\alpha}]$$

ES is the **average loss in the tail** beyond VaR. It is always $\ge$ VaR and **is sub-additive**, making it a **coherent** risk measure.

### 3.4 Coherent Risk Measures

A risk measure $\rho$ is *coherent* if it satisfies four axioms:
1. **Translation invariance:** Adding cash $c$ reduces risk by $c$.
2. **Sub-additivity:** $\rho(X + Y) \le \rho(X) + \rho(Y)$ — diversification doesn't increase risk.
3. **Positive homogeneity:** $\rho(\lambda X) = \lambda \rho(X)$ for $\lambda > 0$.
4. **Monotonicity:** If $X \le Y$ in every scenario, then $\rho(X) \ge \rho(Y)$.

**VaR fails** sub-additivity. **ES satisfies** all four.

**Limitation:** Homogeneity ignores **liquidity risk** — a position 100× larger doesn't just scale the risk, it moves markets.

### 3.5 Historical Context

- **Barings Bank (1995):** Nick Leeson's short straddle positions amplified losses after the Kobe earthquake — risk management was non-existent.
- **LTCM (1998):** Model failure and extreme leverage; VaR alone is insufficient.
- **2008 Financial Crisis:** Systemic risk, CDOs, model mis-specification.
- **Artzner et al. (1999):** Formalised coherent risk measures and proposed ES as the replacement for VaR.
- **Basel III/IV:** Regulators shifted from VaR to ES as the primary regulatory risk measure.

### 3.6 Practical Workflow (Monte Carlo)

1. Simulate $N$ scenarios of $S_H$ under $P$ (using GBM one-shot: $S_H = S_0 e^{(\mu - \sigma^2/2)H + \sigma\sqrt{H}\,Z}$, $Z \sim N(0,1)$).
2. For each scenario, price the portfolio at $H$ under $Q$ (e.g., BS formula with remaining maturity $T-H$).
3. Compute $L_H^{(j)} = V_0 - V_H^{(j)}$.
4. Sort losses; the $\alpha$-quantile is VaR; the mean of losses above VaR is ES.

**Time series analysis:** Use historical log-returns $r_i = \ln(S_{i+1}/S_i)$ to estimate $\mu$ and $\sigma$ under $P$. QQ plots check normality. Historical vol $\ne$ implied vol (different measures!).

## Part 4 — Numerical Solution of SDEs

### The Euler–Maruyama Scheme

When an SDE has no closed-form solution, approximate it on a discrete time grid $0 = t_0 < t_1 < \cdots < t_n = T$ with step $\delta = \Delta t$:

$$X_{t_{i+1}} = X_{t_i} + \mu(t_i, X_{t_i})\,\delta + \sigma(t_i, X_{t_i})\,\sqrt{\delta}\;N_i$$

where $N_i \sim N(0,1)$ are independent draws.

- **Converges** under the same Lipschitz / linear-growth conditions that guarantee existence of the SDE solution.
- **Strong convergence** of order $\tfrac12$: $E|X_T^{\Delta t} - X_T| \le C\,(\Delta t)^{1/2}$.
- Compare with "one-shot" simulation whenever a closed-form solution is available (e.g., GBM) to validate your scheme.
- **Vectorize** in Python for efficiency (100k+ scenarios feasible).

Typical exam use: simulate an SDE forward, then compute risk measures on the terminal distribution.
