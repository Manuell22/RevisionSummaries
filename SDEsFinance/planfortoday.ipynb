## STUDY BLOCK 1 — 10:30 to 11:20
### Pages 206–221 | Fundamental Theorems of No-Arbitrage + Martingale Framework

**Content to cover:**
- **No-Arbitrage: Main Steps I–IV** (pp. 206–209): The connecting thread between (i) self-financing PDE, (ii) Feynman-Kac expectation, and (iii) Girsanov measure change.
- **First & Second Fundamental Theorems** (pp. 210–214):
  - *First FT*: $\exists$ martingale measure $\mathbb{Q}$ $\Leftrightarrow$ no arbitrage (NFLVR).
  - *Second FT*: Complete arbitrage-free market $\Leftrightarrow$ $\exists!$ martingale measure.
  - Pricing theorem: $\exists\, \mathbb{Q}$ ⟹ price = $\mathbb{Q}$-discounted expectation.
  - Complete vs incomplete markets; $n$-dimensional extension.
- **The Martingale Approach** (pp. 215–221):
  - Why martingales? Zero-drift $\Leftrightarrow$ fair game.
  - **Proof that $S_t/B_t$ is a martingale under $\mathbb{Q}$**: compute $d(e^{-rt}S_t)$ using $dS_t = rS_t dt + \sigma S_t dW^\mathbb{Q}_t$, show drift = 0.
  - **General argument**: any model $dS_t = r S_t dt + v(t,S_t) dW^\mathbb{Q}_t$ (any local vol) has $S/B$ as a martingale under $\mathbb{Q}$.
  - **Pricing formula** (Eq. 29): $V_t = \mathbb{E}^\mathbb{Q}_t[e^{-r(T-t)}f(S_T)]$ for any $v(t,S_t)$.
  - Numeraire: the bank account $B_t$ is the canonical choice; others possible.
  - No need to know $\mu$: risk-neutral world.
  - Weak point: uniqueness of $\varphi$ requires full completeness theory.

**Key things to understand deeply (not just read):**
1. Why the drift under $\mathbb{Q}$ is *always* $rS_t$ regardless of the model for $dS$ — because $S/B$ must be a martingale.
2. Why the Black-Scholes market is complete (one asset, one Brownian motion).
3. In an incomplete market, there are *many* martingale measures — attainable claim prices are still unique, but non-attainable ones are not.

## STUDY BLOCK 2 — 11:30 to 12:20
### Pages 222–241 | Zero-Coupon Bonds, Forward Contracts, Put-Call Parity, Dynamic Hedging

**Content to cover:**
- **Zero-Coupon Bonds I–III** (pp. 222–224):
  - Default-free ZCB with notional 1: $P(t,T) = \mathbb{E}^\mathbb{Q}_t[e^{-r(T-t)}] = e^{-r(T-t)}$ (constant $r$).
  - Stochastic $r$ (Vasicek/CIR): $P(t,T) = \mathbb{E}^\mathbb{Q}_t[D(t,T)]$ where $D(t,T)=e^{-\int_t^T r_u du}$ is the stochastic discount factor.
- **Forward Contracts I–V** (pp. 225–229):
  - Payoff $S_T - K$; price $= S_t - e^{-r(T-t)}K$ — **model-independent** (only needs $S/B$ martingale).
  - Forward price $F_{t,T} = S_t e^{r(T-t)}$: value of $K$ making VFWD = 0.
  - **$F_{t,T}$ is a martingale under $\mathbb{Q}$**: $dF_{t,T} = \sigma F_{t,T} dW^\mathbb{Q}$.
  - Under $\mathbb{P}$: $dF_{t,T} = (\mu-r)F_{t,T}dt + \sigma F_{t,T}dW^\mathbb{P}$.
  - ATMF: $K = F_{0,T} = S_0 e^{rT}$.
- **Put-Call Parity & Put Price I–IV** (pp. 230–233):
  - Key identity: $(S_T-K)^+ - (K-S_T)^+ = S_T - K$ ⟹ **Call - Put = Forward Price** (model-independent).
  - BS put formula: $V^{\text{PUT}}_{\text{BS}} = Ke^{-rT}\Phi(-d_2) - S_0\Phi(-d_1)$.
  - Delta of put: $\Delta^{\text{PUT}} = -\Phi(-d_1)$ (always negative).
- **Dynamic Hedging I–IV** (pp. 234–237):
  - Bank sold a call, received $V_0$. Sets up self-financing $\Delta$-hedge: holds $\Delta_t = \Phi(d_1(t))$ shares.
  - Rebalances continuously; terminal value $V_T = (S_T-K)^+$ by construction.
  - **Delta hedging** is only possible in complete markets.
  - Perfect hedge ⟹ bank makes zero net gain (all $V_0$ goes into the hedge).
- **Incomplete Markets I–III** (pp. 238–241):
  - Metatheorem: complete market iff #assets = #Brownian motions.
  - Heston: 2 Brownian motions ($W^S$, $W^V$), but only $S_t$ is tradeable ⟹ incomplete under $S$ alone.
  - Fix: add a specific option $\bar{C}_t$ as a second traded asset.
  - Real markets: always incomplete (credit risk, liquidity, operational risk).
  - Discrete rebalancing introduces hedging error vs. the ideal continuous case.

**Key things to understand deeply:**
1. Why the forward contract price is model-independent — it only needs $\mathbb{E}^\mathbb{Q}[S_T/B_T] = S_t/B_t$.
2. How to derive the put formula from the call formula and put-call parity.
3. Why a Heston model requires a second traded asset to be complete.

## STUDY BLOCK 3 — 12:30 to 13:20
### Pages 242–254 | The Greeks: Delta, Theta, Gamma, Vega, Rho

**Content to cover:**
- **Delta I–V** (pp. 242–245):
  - $\Delta(t) = \partial V(t)/\partial S_t$. For BS call: $\Delta = \Phi(d_1(t))$.
  - **Critical point**: always differentiate keeping $K$ general, then substitute. Mixing $K=S_0$ *before* differentiation gives the wrong joint sensitivity.
  - Correct ATM: $\Delta_0 = \Phi\!\left(\frac{(r+\sigma^2/2)\sqrt{T}}{\sigma}\right)$.
- **Theta I** (pp. 245–246):
  - $\Theta_t = -\partial V/\partial(T-t) = \partial V/\partial t$.
  - BS call: $\Theta^{\text{call}} = -\frac{S_t\varphi(d_1)\sigma}{2\sqrt{T-t}} - rKe^{-r(T-t)}\Phi(d_2)$.
  - Always negative for a call: holding without hedging loses value over time.
- **Gamma I** (pp. 246–247):
  - $\Gamma_t = \partial^2 V/\partial S^2$. BS call: $\Gamma^{\text{call}} = \varphi(d_1)/(S_t\sigma\sqrt{T-t})$.
  - Always positive for a call.
  - **Theta-Gamma-Delta Ito equation** (read from Ito's formula):
$$dV = \Theta_t dt + \tfrac{1}{2}\Gamma_t \sigma^2 S_t^2 dt + \Delta_t dS_t$$
  - Gamma offsets Theta if the market moves enough in $S$.
- **Vega I–III** (pp. 248–252):
  - $\nu_t = \partial V/\partial \sigma$.
  - Detailed derivation leading to: $\text{VegaCall}_{\text{BS}} = S_0\varphi(d_1)\sqrt{T}$ (uses the Greeks Lemma, Eq. 20).
  - Always positive: call value always increases with $\sigma$ in BS.
  - By put-call parity: $\text{VegaCall} = \text{VegaPut}$ (model-independent!).
- **Rho** (p. 252–253):
  - $\rho_t = \partial V/\partial r$. BS call: $\rho^{\text{call}} = TKe^{-rT}\Phi(d_2)>0$.
  - Put/call Theta, Gamma, Rho: all derived from put-call parity once call greeks known.
- **Higher-Order Greeks** (p. 253–254):
  - Speed ($\partial^3V/\partial S^3$), Vanna, Charm, Vomma, etc.
  - Key message: high-order greeks require high smoothness of $V$ — not always realistic. Use carefully, especially when computed numerically.

**Key things to understand deeply:**
1. The correct way to compute a Greek for ATM options (differentiate first, substitute $K=S_0$ after).
2. How the Ito equation for $dV$ links $\Delta$, $\Theta$, and $\Gamma$, and why in continuous time the $\Delta$-hedge is perfect.
3. Why $\text{Vega}_\text{call} = \text{Vega}_\text{put}$: this follows from differentiating put-call parity with respect to $\sigma$.

## STUDY BLOCK 4 — 14:05 to 14:55
### Pages 255–284 | Volatility Smile: Introduction, Bachelier Model, Displaced Diffusion

**Content to cover:**
- **Why the Smile Exists** (pp. 255–263):
  - BS model predicts *one* $\nu$ for all strikes $K$ (volatility is a property of $S$, not of the option).
  - In practice: different market prices for different $K$ require different implied vols $\nu(K)$.
  - The curve $K \mapsto \nu(K)$ is the **volatility smile**. For a correct model it should be flat — it isn't.
  - Traders use BS as a *metric* (not a model) to quote prices as implied vols.
  - **Algorithm to generate a model smile**: compute $V_\text{Model}(K) = \mathbb{E}^\mathbb{Q}[e^{-rT}(S_T-K)^+]$, then invert BS formula for $\nu(K)$.
  - Any model with $dS_t = rS_t dt + \sigma(t,S_t)S_t dW^\mathbb{Q}_t$ is arbitrage-free (drift = $rS_t$).
  - Terminology: absolute vol $b(t,S)$ vs relative/return vol $\sigma(t,S)$.
- **Bachelier Model I–XI** (pp. 265–275):
  - Historical context: 1900, pre-no-arbitrage theory, arithmetic Brownian motion.
  - Modern form: set $r=0$ so ABM stays ABM under $\mathbb{Q}$. **BaM:** $dS_t = \sigma dW_t$.
  - $S_T = s_0 + \sigma\sqrt{T}\,N(0,1)$.
  - **Call price derivation** (key integral):
$$V_{\text{BaM}} = (s_0-K)\Phi\!\left(\frac{s_0-K}{\sigma\sqrt{T}}\right) + \sigma\sqrt{T}\,\varphi\!\left(\frac{s_0-K}{\sigma\sqrt{T}}\right)$$
  - Smile shape: monotonically **decreasing** (return vol $\sigma/S$ decreases with $S$).
  - Can also be used for *forward prices* $F_{t,T}$ (already martingales, no $r=0$ needed).
  - **Bachelier Delta**: $\partial V_{\text{BaM}}/\partial s_0 = \Phi\!\left(\frac{s_0-K}{\sigma\sqrt{T}}\right)$ (same form as BS but with normal instead of lognormal argument).
- **Displaced Diffusion Model I–VI** (pp. 276–281):
  - **DDM Def:** $S_t = \alpha e^{rt} + X_t$ where $X_t$ is Black-Scholes with vol $\sigma$.
  - **DDM SDE:** $dS_t = rS_t dt + \sigma(S_t - \alpha e^{rt})dW_t$ (drift = $rS_t$ ✓).
  - $\alpha=0$ recovers BS exactly.
  - **Call price:** $V_{\text{DDM}} = V_{\text{BS}}(0, S_0-\alpha,\, K-\alpha e^{rT},\, T,\, \sigma,\, r)$.
  - Smile: monotonically decreasing for $\alpha<0$, increasing for $\alpha>0$.
  - Effect of $\alpha$: moves the smile curve up/down and tilts it.
  - **Cannot** generate U-shaped smiles.

**Key things to understand deeply:**
1. Why any model with $\mu^\mathbb{Q}(t,S) = rS$ is arbitrage-free (from the martingale/numeraire argument of Block 1).
2. The Bachelier call price integral — be able to follow the steps from $\mathbb{E}[(\sigma\sqrt{T}N - y)^+]$ to the closed form.
3. How the DDM shifts and tilts the smile: what $\alpha$ controls.

**Active recall prompt (do at 14:55):** State the Bachelier call formula. How does the DDM price relate to BS? Why can DDM only produce monotone smiles?

## STUDY BLOCK 5 — 15:05 to 16:05
### Pages 285–309 | CEV Model, Smile Intuition, Mixture Diffusion Dynamics

**Content to cover:**
- **CEV Model I–III** (pp. 282–284):
  - $dS_t = rS_t dt + \nu S_t^\gamma dW_t$, $\gamma\in(0,1]$.
  - $\gamma=1/2$: Feller square-root process; $\gamma=1$: Black-Scholes (flat smile).
  - $\gamma<1$: $S=0$ is absorbing boundary ⟹ can model default risk.
  - Smile: monotonically **decreasing** for $\gamma\in(0,1)$, due to declining return vol $\nu S^{\gamma-1}$.
  - CEV vs DDM: CEV keeps $S\geq 0$, DDM allows $S<0$ for $\alpha<0$ — prefer CEV if non-negativity matters; otherwise use DDM (simpler formulas).
- **Anticipating Smile Shape from Return Volatility I–VI** (pp. 285–290):
  - General principle: compare the model's **return volatility** against BS's constant $\nu$.
  - If model return vol $\downarrow$ as $S\uparrow$ ⟹ implied vol $\nu(K)$ must $\downarrow$ as $K\uparrow$ ⟹ **decreasing smile**.
  - If model return vol $\uparrow$ as $S\uparrow$ ⟹ **increasing smile**.
  - **Bachelier**: return vol = $\sigma/S$ — decreasing ⟹ decreasing smile.
  - **CEV ($\gamma<1$)**: return vol = $\nu S^{\gamma-1}$ — decreasing ⟹ decreasing smile.
  - **DDM ($\alpha<0$)**: return vol = $\sigma(1-\alpha e^{rt}/S_t)$ — decreasing ⟹ decreasing; **DDM ($\alpha>0$)**: increasing ⟹ increasing smile.
  - Note: this is a rule of thumb / qualitative argument, not a rigorous proof.
- **Mixture Diffusion Dynamics I–IX** (pp. 291–300):
  - **Motivation:** Keep lognormal density (traders' preference) but *mix* $N$ of them.
  - Density: $p_{S_t}(y) = \sum_{i=1}^N \lambda_i p^{\text{lognormal}}_{t,\sigma_i}(y)$, $\sum\lambda_i=1$, $\lambda_i>0$.
  - Return vol: $\sigma_{\text{mix}}(t,y)^2 = \sum_i \Lambda_i(t,y)\sigma_i^2$ where $\Lambda_i = \lambda_i p_{i,t}(y)/\sum_j\lambda_j p_{j,t}(y)$ (weighted average of $\sigma_i^2$).
  - **Key analytical tractability** (MDD call price formula):
$$V^{\text{Call}}_{\text{mix}} = \sum_{i=1}^N \lambda_i\, V^{\text{Call}}_{\text{BS}}(0, S_0, K, T, \sigma_i, r)$$
  - Same linear combination applies to **all greeks** (linearity of differentiation).
  - **Smile shape:** minimum at $S_0 e^{rT}$ (at-the-money forward) — proven in B.&Mercurio papers.
  - **Calibration:** choose $\lambda_i$, $\sigma_i$ to match market smile; can fit most market patterns with $N=2$ or $N=3$.
  - SDE: unique strong solution exists if $\sigma_i(t)$ time-dependent sharing $\sigma_0$ initially.
  - No-arbitrage check: $\mathbb{E}^\mathbb{Q}[S_t] = S_0e^{rt}$ ✓ (drift is $rS_t$).

**Key things to understand deeply:**
1. The closed-form tractability of MDD — the call price is *exactly* a weighted average of BS call prices. This is why traders prefer this model.
2. The smile minimum at the at-the-money forward — this is a property of mixing symmetric (lognormal) distributions.
3. Why return volatility direction predicts smile slope — this is intuition, not a theorem.

## STUDY BLOCK 6 — 16:25 to 17:15
### Pages 310–336 | Shifted MDD, Local Vol Flattening, Heston, Context & Crises

**Content to cover:**
- **Shifted Mixture Dynamics I–IX** (pp. 301–309):
  - **Motivation:** MDD smile minimum is always at $S_0e^{rT}$ — not suitable for equity smiles with minimum elsewhere.
  - **Definition:** $S_t = s_0\alpha e^{rt} + X_t$ where $X_t$ follows MDD with initial condition $x_0 = s_0(1-\alpha)$.
  - **Call price:**
$$V^{\text{Call}}_{\text{shift-mix}} = \sum_{i=1}^N \lambda_i\, V^{\text{Call}}_{\text{BS}}(0,\, \underbrace{s_0(1-\alpha)}_{\tilde{S}_0},\, \underbrace{K - s_0\alpha e^{rT}}_{\tilde{K}},\, T,\, \sigma_i,\, r)$$
  - Shift $\alpha$ moves the smile minimum away from $S_0 e^{rT}$.
  - Calibration examples: MIB30 equity (2000), USD/Euro FX (2001).
- **Local Volatility & Future Smile Flattening I–V** (pp. 310–314):
  - **Local vol definition:** $\sigma(t,S_t)$ is deterministic *given* $\mathcal{F}_t^{W^S}$, but random from time 0 (since $S_t$ is random).
  - **Problem:** Future smile $K\mapsto\nu(T_1,K,T;s_1)$ computed conditional on $S_{T_1}=s_1$ is *flatter* than the initial smile $K\mapsto\nu(0,K,T;s_0)$.
  - Reason: all randomness in vol comes from $S_t$ itself — no independent vol driver.
  - Practical consequence: traders in FX/Equity avoid local vol models in front-office pricing; used more in risk management or model validation.
- **Heston Stochastic Volatility I–II** (pp. 315–316):
  - Two-factor model under $\mathbb{Q}$:
$$dS_t = rS_t dt + \sqrt{V_t}\,S_t\,dW^S_t,\qquad dV_t = k(\theta - V_t)dt + \sigma_V\sqrt{V_t}\,dW^V_t$$
$$dW^S\,dW^V = \rho\,dt$$
  - $V_t$ is CIR process: mean-reverts to $\theta$ with speed $k$, vol-of-vol $\sigma_V$.
  - New randomness in $\sqrt{V_t}$ is *not* $\mathcal{F}^{W^S}$-measurable ⟹ vol has a random life of its own ⟹ no future smile flattening.
  - Incomplete market: need $S_t$ + a traded option $\bar{C}_t$ to hedge both $W^S$ and $W^V$.
  - Other SVM: Hull-White, SABR.
- **Derivatives Markets & Crises** (pp. 317–336):
  - Notional outstanding: ~$708\times 10^{12}$ USD (2011), $\approx 9\times$ world GDP. Historical origin: Thales (~580 BC).
  - BS assumptions (know them all): short selling allowed, divisible, no transaction costs/dividends/default, $r$ constant, continuous trading, complete markets.
  - **Barings 1995:** Nick Leeson, short straddle $Y = -|S_T-K|$, Kobe earthquake caused large $S$ moves ⟹ catastrophic losses. Led to VaR requirement.
  - **LTCM 1998** (irony: Merton & Scholes on board), Parmalat, 2007–2008 subprime/credit crisis.
  - Introduction of CVA (Credit Valuation Adjustment) after 2008; valuation adjustments (XVAs) proliferated.

**Key things to understand deeply:**
1. How the shift $\alpha$ in Shifted MDD maps to a shifted BS formula — it's the same trick as DDM.
2. Why local vol has a flat future smile but stochastic vol doesn't — link to whether vol has independent randomness.
3. The short straddle payoff $Y = -|S_T - K|$ and why it profits from low vol but loses catastrophically from large moves.

**Active recall prompt (do at 17:15):** Compare local vol and stochastic vol models: why does one flatten the smile and the other doesn't? Write the Heston model equations.

### Pages 337–381 | Part 3: Risk Measures — Time Series, VaR, Expected Shortfall

**Content to cover:**
- **Financial Time Series Under $\mathbb{P}$** (pp. 339–359):
  - Log-returns $\log(S_{t_{i+1}}/S_{t_i}) \sim \mathcal{N}((\mu-\tfrac{1}{2}\sigma^2)\delta,\, \sigma^2\delta)$ in BS.
  - Empirical check: skewness and excess kurtosis both = 0 for Gaussian.
  - S&P500 5y (2017–2022): skew $= -1.13$, excess kurtosis $= 21.46$ — **fat tails, asymmetric**. Historical vol annualised = 19.25%. Implies log-returns are NOT normal.
  - S&P500 1y (2021–2022): skew $= 0.29$, excess kurtosis $= 0.65$ — closer to normal.
  - Implied vol (from options, $\mathbb{Q}$) generally $>$ historical vol ($\mathbb{P}$): vol risk premium.
  - **QQ plot:** plots empirical quantiles vs standard normal. Straight line = normal; S-curve or wings = fat tails.
  - MDD under $\mathbb{P}$ (drift $\mu S_t dt$) can replicate fat tails and high kurtosis.
- **Value at Risk I–IX** (pp. 361–369):
  - Portfolio loss at horizon $H$: $L_H = \text{Portfolio}_0 - \text{Portfolio}_H$.
  - $\text{VaR}_{H,\alpha}$: the $\alpha$-percentile of $L_H$ under $\mathbb{P}$.
$$\mathbb{P}[L_H < \text{VaR}_{H,\alpha}] = \alpha$$
  - Loss distribution uses **two measures**: simulate scenarios under $\mathbb{P}$ up to $H$, then price portfolio under $\mathbb{Q}$ in each scenario.
  - Example: simulate $S_H$ under $\mathbb{P}$; price call with remaining life $T-H$ using BS (a $\mathbb{Q}$-expectation) in each scenario.
- **VaR Drawbacks I–III** (pp. 370–374):
  - **Drawback 1:** VaR is tail-blind — same 99% VaR can hide a tiny loss vs. a catastrophic one beyond the percentile.
  - **Drawback 2:** VaR is **not sub-additive** in general: $\text{VaR}(P_1+P_2) \not\leq \text{VaR}(P_1)+\text{VaR}(P_2)$.
- **Expected Shortfall Definition I–II** (pp. 376–377):
  - $\text{ES}_{H,\alpha} = \mathbb{E}^\mathbb{P}[L_H \mid L_H > \text{VaR}_{H,\alpha}]$.
  - Always $\text{ES} > \text{VaR}$ by definition.
  - Derivation: $\text{ES} = \mathbb{E}^\mathbb{P}[L_H\mathbf{1}_{L_H>\text{VaR}}]/(1-\alpha)$.
  - Also called: CVaR, AVaR, ETL.
  - **Sub-additive** ✓ (fixes drawback 2).
  - Partial fix for drawback 1: looks at tail expectation, but not full tail structure.
- **ES Drawbacks I–IV** (pp. 378–381):
  - Same tail-structure blindness as VaR (just an expectation).
  - **Homogeneity drawback:** $\text{VaR}(kP) = k\,\text{VaR}(P)$ and similarly for ES — **ignores liquidity risk and market impact**. Selling $10^6$ shares vs 1 share is not $10^6\times$ more risky — it's much worse due to price impact.

**Key things to understand deeply:**
1. The interplay of $\mathbb{P}$ and $\mathbb{Q}$ in VaR computation — simulate under $\mathbb{P}$, price under $\mathbb{Q}$.
2. Why VaR is non-sub-additive (draw two loss distributions with the same 99% VaR but very different tails).
3. Why ES is an improvement but still insufficient for full risk understanding.

**Content to cover:**

  - **Key result for VaR/ES vs $\rho$:**

| $\rho$ | VaR | ES | ES$-$VaR |
|--------|-----|----|---------|
| $-1$ | 30.9 | 34.6 | 3.7 |
| $0$ | 26.1 | 29.5 | 3.4 |
| $+1$ | 17.4 | 18.0 | 0.6 |

  - **Intuition:** When $\rho=-1$: if $S^{(1)}\downarrow$ (call loses), then $S^{(2)}\uparrow$ (put loses too). Both legs lose simultaneously ⟹ large loss. When $\rho=+1$: if $S^{(1)}\downarrow$ (call loses), $S^{(2)}\downarrow$ (put gains). The two positions partially offset. More positive correlation ⟹ more natural hedging ⟹ lower risk.
  - ES$-$VaR largest at $\rho=-1$: deepest tail; smallest at $\rho=+1$: very thin tail.
  - General lesson: correlation has enormous (up to 78–92%) impact on VaR/ES.
- **Part 4: Euler-Maruyama Scheme** (pp. 451–469):
  - **Motivation:** Smile models (CEV, MDD, Heston) often have no closed-form $S_T$ distribution ⟹ need numerical simulation of the SDE path.
  - **General Euler Scheme:** Replace differentials with increments on time grid $t_0<t_1<\ldots<t_n=T$, $\delta = \Delta t$:
$$X_{t_{i+1}} = X_{t_i} + \mu(t_i, X_{t_i})\Delta t + \sigma(t_i, X_{t_i})\sqrt{\delta}\,N_i(0,1)$$
  - **Convergence:** Strong convergence of order $1/2$: $\mathbb{E}|X^{\Delta t}_T - X_T| \leq C(T)(\Delta t)^{1/2}$.
  - Requires Lipschitz + linear growth conditions.
  - **GBM Example** (pp. 457–463): $dS_t = \mu S_t dt + \sigma S_t dW_t$.
    - Euler: $S_{t_{i+1}} = S_{t_i}(1 + \mu\Delta t + \sigma\sqrt{\delta}N_i)$.
    - Compare vs. one-shot exact: $S_T = S_0\exp((\mu-\sigma^2/2)T + \sigma\sqrt{T}N)$.
    - Results: both give nearly identical histograms; log-return mean, std, skew, kurtosis match theory.
    - Naive (naive loop) vs vectorized (matrix operations): both accurate, vectorized faster.
  - **SDE $dX_t = m\,dt + \sigma X_t\,dW_t$** (pp. 463–468):
    - No exact closed form for the SDE solution (only $\mathbb{E}[X_T]$ and $\text{Var}[X_T]$ are known).
    - Euler: $X_{t_{i+1}} = X_{t_i} + m\Delta t + \sigma X_{t_i}\sqrt{\delta}N_i$.
    - Interesting qualitative result: with $x_0=0$, the sign of $X_t$ matches the sign of $m$ for all $t$ (proven via comparison theorem).

**Key things to understand deeply:**
1. Why $\rho=+1$ gives the lowest VaR for a (call + put) portfolio — the two options naturally offset each other.
2. The Euler scheme: it is just discretising $dX = \mu\,dt + \sigma\,dW$ as a finite recurrence. Why it converges to order $1/2$.
3. Why the GBM example validates the Euler scheme — comparing histogram and moments against the known lognormal.

**Content to cover:**
- **$dX_t = m\,dt + \sigma X_t\,dW_t$ — sign analysis** (pp. 468–469):
  - Result: for $X_0=0$ and $m>0$: $X_t>0$ for all $t$; $m<0$: $X_t<0$; $m=0$: $X_t=0$.
  - This is supported by Euler simulation and provable via Yamada's comparison theorem (1973), which is beyond course scope.
  - Intuition: at $X=0$ the diffusion term vanishes, so only the drift $m$ determines the initial direction.
- **Complex SDE study** (pp. 470–483): $dX_t = \left(-3kX_t + 3X_t^{1/3}\sigma^2\right)dt + 3\sigma X_t^{2/3}\,dW_t$, $X_0=x_0$
  - This is an exam-level worked example — trace through all parts:
  - **(a) Existence/Uniqueness check:** Cubic root $x^{1/3}$ is NOT Lipschitz at $x=0$ ⟹ the sufficient conditions theorem does **not** apply.
  - **(b) Euler scheme** (pp. 473–477): Apply standard Euler update; $n=10^6$ scenarios. Histogram of $X_1$: right-skewed, positive excess kurtosis. (Skew $\approx 0.66$, kurt $\approx 0.65$).
  - **(c) Solve analytically** (pp. 477–480): Let $Y_t = X_t^{1/3}$. Apply Ito's formula *formally* (warning: cubic root not twice differentiable at 0):
$$dY_t = -kY_t\,dt + \sigma\,dW_t$$
  This is an Ornstein-Uhlenbeck (OU) process! Solve:
$$Y_t = y_0 e^{-kt} + \sigma\int_0^t e^{-k(t-u)}dW_u \quad \Rightarrow \quad X_t = Y_t^3$$
  - **(d) Non-uniqueness at $x_0=0$** (p. 480): If $X_0=0$, then $X_t\equiv 0$ also satisfies the SDE (RHS vanishes at $X=0$). So there are two solutions: $X_t=0$ and $X_t = Y_t^3$ with $y_0=0$. ⟹ SDE does not have a unique strong solution for $x_0=0$, consistent with (a).
  - **(e) One-shot simulation** (pp. 480–481): Since $Y_T \sim \mathcal{N}\!\left(y_0 e^{-kT},\, \frac{\sigma^2}{2k}(1-e^{-2kT})\right)$, simulate $Y_T$ in one step and compute $X_T = Y_T^3$:
$$X_T^j = \left(y_0 e^{-kT} + \sqrt{\frac{\sigma^2}{2k}(1-e^{-2kT})}\, N^j\right)^3$$
  - **(f) Comparison** (pp. 482–483): Euler and one-shot densities match visually — validates the Euler simulation and the analytical solution.
