Let $z = \cos\theta + i\sin\theta$ and

$$P = \frac{1}{\sqrt{2}} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix},$$

where $\bar{z}$ represents the complex conjugate of $z$.

## Part (a)

**Question:** Show that $P$ is unitary by showing clearly that $P^*P = I$, where $P^*$ is the conjugate transpose of $P$ and $I$ is the $2 \times 2$ identity matrix.

### Solution

We are given:

$$P = \frac{1}{\sqrt{2}} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix}$$

where $z = \cos\theta + i\sin\theta$.

**Key property:** Since $z = \cos\theta + i\sin\theta = e^{i\theta}$, we have $|z|^2 = z\bar{z} = 1$.

**Step 1: Find $P^*$ (the conjugate transpose of $P$).**

First, we take the complex conjugate of each entry of $P$:

$$\bar{P} = \frac{1}{\sqrt{2}} \begin{bmatrix} \overline{z} & \overline{\bar{z}} \\ \overline{iz} & \overline{-i\bar{z}} \end{bmatrix}$$

Now we simplify each entry:
- $\overline{z} = \bar{z}$ (by definition)
- $\overline{\bar{z}} = z$ (conjugate of conjugate gives back the original)
- $\overline{iz} = \bar{i}\bar{z} = -i\bar{z}$
- $\overline{-i\bar{z}} = \overline{(-i)}\overline{(\bar{z})} = (i)(z) = iz$

So:

$$\bar{P} = \frac{1}{\sqrt{2}} \begin{bmatrix} \bar{z} & z \\ -i\bar{z} & iz \end{bmatrix}$$

**Step 2: Transpose $\bar{P}$ to get $P^*$.**

$$P^* = \bar{P}^T = \frac{1}{\sqrt{2}} \begin{bmatrix} \bar{z} & -i\bar{z} \\ z & iz \end{bmatrix}$$

**Step 3: Compute $P^*P$.**

$$P^*P = \frac{1}{\sqrt{2}} \begin{bmatrix} \bar{z} & -i\bar{z} \\ z & iz \end{bmatrix} \cdot \frac{1}{\sqrt{2}} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix}$$

$$= \frac{1}{2} \begin{bmatrix} \bar{z} & -i\bar{z} \\ z & iz \end{bmatrix} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix}$$

**Step 4: Compute each entry of the product.**

**Entry (1,1):**

$$(P^*P)_{11} = \bar{z} \cdot z + (-i\bar{z}) \cdot (iz) = \bar{z}z + (-i)(i)\bar{z}z = |z|^2 + (-i^2)|z|^2$$

Since $i^2 = -1$:

$$= |z|^2 + |z|^2 = 1 + 1 = 2$$

**Entry (1,2):**

$$(P^*P)_{12} = \bar{z} \cdot \bar{z} + (-i\bar{z}) \cdot (-i\bar{z}) = \bar{z}^2 + i^2\bar{z}^2 = \bar{z}^2 + (-1)\bar{z}^2 = \bar{z}^2 - \bar{z}^2 = 0$$

**Entry (2,1):**

$$(P^*P)_{21} = z \cdot z + (iz) \cdot (iz) = z^2 + i^2z^2 = z^2 + (-1)z^2 = z^2 - z^2 = 0$$

**Entry (2,2):**

$$(P^*P)_{22} = z \cdot \bar{z} + (iz)(-i\bar{z}) = z\bar{z} + (-i^2)z\bar{z} = |z|^2 + (-(-1))|z|^2$$

$$= |z|^2 + |z|^2 = 1 + 1 = 2$$

**Step 5: Putting it all together.**

$$P^*P = \frac{1}{2} \begin{bmatrix} 2 & 0 \\ 0 & 2 \end{bmatrix} = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = I$$

Therefore $P^*P = I$, which shows that $P$ is **unitary**. $\square$

```python
import numpy as np

# Numerical verification of Part (a) for a specific theta
theta = np.pi / 4  # choose an arbitrary theta
z = np.cos(theta) + 1j * np.sin(theta)
zbar = np.conj(z)

P = (1 / np.sqrt(2)) * np.array([[z, zbar],
                                   [1j * z, -1j * zbar]])

P_star = np.conj(P).T

product = P_star @ P
print("P* P =")
print(np.round(product, 10))
print("\nThis is the identity matrix, confirming P is unitary.")
```

## Part (b)

**Question:** Use the previous result to determine $a$, $b$ and $c$ so that

$$P = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & a \\ b & c \end{bmatrix}$$

is unitary.

### Solution

From Part (a), we showed that the matrix

$$P = \frac{1}{\sqrt{2}} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix}$$

is unitary for **any** $z = \cos\theta + i\sin\theta$.

**Step 1: Match the (1,1) entry to find $\theta$.**

Comparing the given matrix with the general form, the (1,1) entry gives us:

$$z = -1$$

Since $z = \cos\theta + i\sin\theta$, we need:

$$\cos\theta + i\sin\theta = -1$$

Comparing real and imaginary parts:
- Real part: $\cos\theta = -1$
- Imaginary part: $\sin\theta = 0$

Therefore $\theta = \pi$.

**Step 2: Compute $z$ and $\bar{z}$.**

With $\theta = \pi$:

$$z = \cos\pi + i\sin\pi = -1 + i(0) = -1$$

$$\bar{z} = \cos\pi - i\sin\pi = -1 - i(0) = -1$$

So $z = \bar{z} = -1$ (both are real).

**Step 3: Find $a$, $b$, and $c$.**

Using the general form $P = \frac{1}{\sqrt{2}} \begin{bmatrix} z & \bar{z} \\ iz & -i\bar{z} \end{bmatrix}$:

- **$a$** corresponds to the (1,2) entry: $a = \bar{z} = -1$
- **$b$** corresponds to the (2,1) entry: $b = iz = i(-1) = -i$
- **$c$** corresponds to the (2,2) entry: $c = -i\bar{z} = -i(-1) = i$

Therefore:

$$\boxed{a = -1, \quad b = -i, \quad c = i}$$

and the unitary matrix is:

$$P = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & -1 \\ -i & i \end{bmatrix}$$

```python
# Numerical verification of Part (b)
P_b = (1 / np.sqrt(2)) * np.array([[-1, -1],
                                     [-1j,  1j]])

P_b_star = np.conj(P_b).T

product_b = P_b_star @ P_b
print("P* P =")
print(np.round(product_b, 10))
print("\nThis confirms P is unitary with a = -1, b = -i, c = i.")
```

## Part (c)

**Question:** A matrix $A$ is said to be Hermitian if $A^* = A$. Let $A$ be a $2 \times 2$ Hermitian matrix of the form:

$$A = \begin{bmatrix} \frac{1}{2} & x + iy \\ x - iy & \frac{1}{2} \end{bmatrix}$$

### Part (c)(i)

**Question:** Given that $\mathbf{v}_1 = \begin{bmatrix} -1 \\ -i \end{bmatrix}$ is an eigenvector of $A$ associated with eigenvalue $\lambda_1 = 2$, find $x$ and $y$.

### Solution

**Step 1: Write the eigenvalue equation.**

Since $\mathbf{v}_1$ is an eigenvector of $A$ with eigenvalue $\lambda_1 = 2$, we have:

$$A\mathbf{v}_1 = \lambda_1 \mathbf{v}_1$$

$$\begin{bmatrix} \frac{1}{2} & x + iy \\ x - iy & \frac{1}{2} \end{bmatrix} \begin{bmatrix} -1 \\ -i \end{bmatrix} = 2 \begin{bmatrix} -1 \\ -i \end{bmatrix}$$

**Step 2: Compute the left-hand side.**

**Row 1:**

$$\frac{1}{2} \cdot (-1) + (x + iy) \cdot (-i) = -\frac{1}{2} + (-ix - i^2 y) = -\frac{1}{2} + (-ix + y)$$

since $i^2 = -1$, so $-i^2 y = y$.

$$= \left(-\frac{1}{2} + y\right) - ix$$

**Row 2:**

$$(x - iy)(-1) + \frac{1}{2}(-i) = -x + iy - \frac{i}{2}$$

$$= -x + \left(y - \frac{1}{2}\right)i$$

**Step 3: Compute the right-hand side.**

$$2 \begin{bmatrix} -1 \\ -i \end{bmatrix} = \begin{bmatrix} -2 \\ -2i \end{bmatrix}$$

**Step 4: Equate and solve.**

**From Row 1:** $\left(-\frac{1}{2} + y\right) - ix = -2$

Comparing real and imaginary parts:
- Real: $-\frac{1}{2} + y = -2 \implies y = -2 + \frac{1}{2} = -\frac{3}{2}$
- Imaginary: $-x = 0 \implies x = 0$

**From Row 2:** $-x + \left(y - \frac{1}{2}\right)i = -2i$

Comparing real and imaginary parts:
- Real: $-x = 0 \implies x = 0$ ✓ (consistent)
- Imaginary: $y - \frac{1}{2} = -2 \implies y = -2 + \frac{1}{2} = -\frac{3}{2}$ ✓ (consistent)

Therefore:

$$\boxed{x = 0, \quad y = -\frac{3}{2}}$$

So the matrix $A$ is:

$$A = \begin{bmatrix} \frac{1}{2} & 0 + i\left(-\frac{3}{2}\right) \\ 0 - i\left(-\frac{3}{2}\right) & \frac{1}{2} \end{bmatrix} = \begin{bmatrix} \frac{1}{2} & -\frac{3}{2}i \\ \frac{3}{2}i & \frac{1}{2} \end{bmatrix}$$

```python
# Numerical verification of Part (c)(i)
x = 0
y = -3/2

A = np.array([[1/2, x + 1j*y],
              [x - 1j*y, 1/2]])

v1 = np.array([[-1], [-1j]])
lam1 = 2

print("A =")
print(A)
print("\nA @ v1 =")
print(A @ v1)
print("\nlambda_1 * v1 =")
print(lam1 * v1)
print("\nThese are equal, confirming x = 0 and y = -3/2.")
```

### Part (c)(ii)

**Question:** Find the other eigenvalue of $A$ and a corresponding eigenvector.

### Solution

**Step 1: Find $\lambda_2$ using the trace.**

For any $2 \times 2$ matrix, the sum of the eigenvalues equals the trace:

$$\lambda_1 + \lambda_2 = \text{tr}(A) = \frac{1}{2} + \frac{1}{2} = 1$$

Since $\lambda_1 = 2$:

$$2 + \lambda_2 = 1 \implies \boxed{\lambda_2 = -1}$$

**Step 2: Verify using the determinant.**

The product of the eigenvalues equals the determinant:

$$\det(A) = \frac{1}{2} \cdot \frac{1}{2} - \left(-\frac{3}{2}i\right)\left(\frac{3}{2}i\right) = \frac{1}{4} - \left(-\frac{9}{4}i^2\right) = \frac{1}{4} - \frac{9}{4} = -\frac{8}{4} = -2$$

Check: $\lambda_1 \cdot \lambda_2 = 2 \cdot (-1) = -2$ ✓

**Step 3: Find the eigenvector $\mathbf{v}_2$ corresponding to $\lambda_2 = -1$.**

We solve $(A - \lambda_2 I)\mathbf{v}_2 = \mathbf{0}$:

$$A - (-1)I = A + I = \begin{bmatrix} \frac{1}{2} + 1 & -\frac{3}{2}i \\ \frac{3}{2}i & \frac{1}{2} + 1 \end{bmatrix} = \begin{bmatrix} \frac{3}{2} & -\frac{3}{2}i \\ \frac{3}{2}i & \frac{3}{2} \end{bmatrix}$$

**Step 4: Solve the homogeneous system.**

From the first row:

$$\frac{3}{2} v_1 - \frac{3}{2}i \, v_2 = 0$$

Dividing both sides by $\frac{3}{2}$:

$$v_1 - i \, v_2 = 0$$

$$v_1 = i \, v_2$$

Let $v_2 = -i$ (a convenient choice), then $v_1 = i(-i) = -i^2 = 1$.

Therefore:

$$\boxed{\mathbf{v}_2 = \begin{bmatrix} 1 \\ -i \end{bmatrix}}$$

**Step 5: Verify this is an eigenvector.**

$$A\mathbf{v}_2 = \begin{bmatrix} \frac{1}{2} & -\frac{3}{2}i \\ \frac{3}{2}i & \frac{1}{2} \end{bmatrix} \begin{bmatrix} 1 \\ -i \end{bmatrix}$$

**Row 1:** $\frac{1}{2}(1) + \left(-\frac{3}{2}i\right)(-i) = \frac{1}{2} + \frac{3}{2}i^2 = \frac{1}{2} - \frac{3}{2} = -1$

**Row 2:** $\frac{3}{2}i(1) + \frac{1}{2}(-i) = \frac{3}{2}i - \frac{1}{2}i = i$

So $A\mathbf{v}_2 = \begin{bmatrix} -1 \\ i \end{bmatrix} = (-1)\begin{bmatrix} 1 \\ -i \end{bmatrix} = \lambda_2 \mathbf{v}_2$ ✓

```python
# Numerical verification of Part (c)(ii)
v2 = np.array([[1], [-1j]])
lam2 = -1

print("A @ v2 =")
print(A @ v2)
print("\nlambda_2 * v2 =")
print(lam2 * v2)
print("\nThese are equal, confirming v2 is an eigenvector with eigenvalue -1.")

# Also verify with numpy's eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nNumpy eigenvalues:", eigenvalues)
```

### Part (c)(iii)

**Question:** Find a unitary matrix $U$ and a diagonal matrix $\Lambda$ such that $A = U\Lambda U^*$.

### Solution

**Step 1: Normalise the eigenvectors.**

To build a unitary matrix $U$, we need the eigenvectors to be orthonormal.

**Normalising $\mathbf{v}_1$:**

$$\|\mathbf{v}_1\| = \sqrt{|{-1}|^2 + |{-i}|^2} = \sqrt{1 + 1} = \sqrt{2}$$

$$\hat{\mathbf{v}}_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 \\ -i \end{bmatrix}$$

**Normalising $\mathbf{v}_2$:**

$$\|\mathbf{v}_2\| = \sqrt{|1|^2 + |{-i}|^2} = \sqrt{1 + 1} = \sqrt{2}$$

$$\hat{\mathbf{v}}_2 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\ -i \end{bmatrix}$$

**Step 2: Verify orthogonality of the normalised eigenvectors.**

$$\hat{\mathbf{v}}_1^* \hat{\mathbf{v}}_2 = \frac{1}{2} \begin{bmatrix} -1 & i \end{bmatrix} \begin{bmatrix} 1 \\ -i \end{bmatrix} = \frac{1}{2}\left[(-1)(1) + (i)(-i)\right] = \frac{1}{2}\left[-1 - i^2\right] = \frac{1}{2}[-1 + 1] = 0$$

The eigenvectors are indeed orthogonal. ✓

**Step 3: Construct the unitary matrix $U$.**

The columns of $U$ are the normalised eigenvectors:

$$U = \begin{bmatrix} \hat{\mathbf{v}}_1 & \hat{\mathbf{v}}_2 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & 1 \\ -i & -i \end{bmatrix}$$

**Step 4: Construct the diagonal matrix $\Lambda$.**

$$\Lambda = \begin{bmatrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{bmatrix} = \begin{bmatrix} 2 & 0 \\ 0 & -1 \end{bmatrix}$$

**Step 5: Verify that $U$ is unitary, i.e., $U^* U = I$.**

$$U^* = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & i \\ 1 & i \end{bmatrix}$$

$$U^* U = \frac{1}{2} \begin{bmatrix} -1 & i \\ 1 & i \end{bmatrix} \begin{bmatrix} -1 & 1 \\ -i & -i \end{bmatrix}$$

**Entry (1,1):** $(-1)(-1) + (i)(-i) = 1 - i^2 = 1 + 1 = 2$

**Entry (1,2):** $(-1)(1) + (i)(-i) = -1 - i^2 = -1 + 1 = 0$

**Entry (2,1):** $(1)(-1) + (i)(-i) = -1 - i^2 = -1 + 1 = 0$

**Entry (2,2):** $(1)(1) + (i)(-i) = 1 - i^2 = 1 + 1 = 2$

$$U^*U = \frac{1}{2}\begin{bmatrix} 2 & 0 \\ 0 & 2 \end{bmatrix} = I \quad \checkmark$$

**Step 6: Verify $A = U\Lambda U^*$.**

First compute $U\Lambda$:

$$U\Lambda = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & 1 \\ -i & -i \end{bmatrix} \begin{bmatrix} 2 & 0 \\ 0 & -1 \end{bmatrix} = \frac{1}{\sqrt{2}} \begin{bmatrix} -2 & -1 \\ -2i & i \end{bmatrix}$$

Now compute $(U\Lambda)U^*$:

$$U\Lambda U^* = \frac{1}{\sqrt{2}} \begin{bmatrix} -2 & -1 \\ -2i & i \end{bmatrix} \cdot \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & i \\ 1 & i \end{bmatrix}$$

$$= \frac{1}{2} \begin{bmatrix} (-2)(-1) + (-1)(1) & (-2)(i) + (-1)(i) \\ (-2i)(-1) + (i)(1) & (-2i)(i) + (i)(i) \end{bmatrix}$$

$$= \frac{1}{2} \begin{bmatrix} 2 - 1 & -2i - i \\ 2i + i & -2i^2 + i^2 \end{bmatrix}$$

$$= \frac{1}{2} \begin{bmatrix} 1 & -3i \\ 3i & 2 + (-1) \end{bmatrix}$$

Wait — let me redo the (2,2) entry more carefully:

$(-2i)(i) + (i)(i) = -2i^2 + i^2 = -2(-1) + (-1) = 2 - 1 = 1$

$$= \frac{1}{2} \begin{bmatrix} 1 & -3i \\ 3i & 1 \end{bmatrix} = \begin{bmatrix} \frac{1}{2} & -\frac{3}{2}i \\ \frac{3}{2}i & \frac{1}{2} \end{bmatrix} = A \quad \checkmark$$

**Final Answer for Part (c)(iii):**

$$\boxed{U = \frac{1}{\sqrt{2}} \begin{bmatrix} -1 & 1 \\ -i & -i \end{bmatrix}, \qquad \Lambda = \begin{bmatrix} 2 & 0 \\ 0 & -1 \end{bmatrix}}$$

and we have verified that $A = U\Lambda U^*$. $\square$

```python
# Numerical verification of Part (c)(iii)
U = (1 / np.sqrt(2)) * np.array([[-1, 1],
                                   [-1j, -1j]])

Lambda = np.array([[2, 0],
                    [0, -1]])

U_star = np.conj(U).T

# Check U is unitary
print("U* U =")
print(np.round(U_star @ U, 10))

# Check A = U Lambda U*
A_reconstructed = U @ Lambda @ U_star
print("\nU Lambda U* =")
print(np.round(A_reconstructed, 10))

print("\nOriginal A =")
print(A)

print("\nA = U Lambda U*?", np.allclose(A, A_reconstructed))
```
