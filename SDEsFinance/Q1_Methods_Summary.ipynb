# Q1 Methods Summary — SDEs

This notebook summarises the **method used to solve each Mock Exam Problem 1**, including where the integrating factor technique appears.

---

## Key Toolbox

| Tool | When to use |
|---|---|
| **Lipschitz + linear growth** | Existence & uniqueness |
| **Stratonovich conversion + separation of variables** | Explicit solution when diffusion is multiplicative |
| **Integrating factor** ($e^{\int a(s)ds}$) | Explicit solution for linear SDEs — avoids Stratonovich |
| **Substitution** ($Y = X - a$, $Y = X^n$, etc.) | Reduces nonlinear SDE to a known form |
| **Itô isometry / zero mean of Itô integrals** | Taking $\mathbb{E}[\cdot]$ through integrals |
| **ODE for $m(t) = \mathbb{E}[X_t]$** | Computing the mean when direct calculation is hard |
| **Itô's formula on $X_t^2$** | Computing $\mathbb{E}[X_t^2]$ and hence variance |

## Mock 1 — $dX = m\,dt + \sigma X\,dW$, $X_0 = x_0$

### a) Existence & uniqueness
Drift $\mu = m$ (constant), diffusion $\sigma(x) = \sigma x$.
- **Lipschitz:** $|m-m| + |\sigma x - \sigma y| = \sigma|x-y|$ ✓
- **Linear growth:** $|m| + \sigma|x| \leq \max(|m|,\sigma)(1+|x|)$ ✓

### b) Mean and variance
Integrate the SDE: $X_T = x_0 + mT + \sigma\int_0^T X_t\,dW_t$.

**Mean:** Take $\mathbb{E}[\cdot]$, use $\mathbb{E}[\text{Itô integral}] = 0$:
$$\mathbb{E}[X_T] = x_0 + mT$$

**Variance:** Apply Itô to $X_t^2$, take $\mathbb{E}$, get a linear ODE for $m_2(t) = \mathbb{E}[X_t^2]$:
$$\frac{dm_2}{dt} = 2m(x_0+mt) + \sigma^2 m_2, \quad m_2(0) = x_0^2$$
Solve the ODE, then $\text{Var}(X_T) = m_2(T) - (x_0+mT)^2$.

### c) SDE for $Y = X_t^3$
Apply Itô with $\varphi(x) = x^3$: $\varphi' = 3x^2$, $\varphi'' = 6x$.
$$dY_t = 3X_t^2(m + \sigma^2 X_t)\,dt + 3\sigma X_t^3\,dW_t$$

## Mock 2 — $dX = m(X-a)\,dt + \sigma(X-a)\,dW$, $X_0 = x_0 > a$

### a) Existence & uniqueness
Both drift and diffusion are linear in $x$: Lipschitz with $K = |m|+\sigma$, linear growth holds. ✓

### b) Stratonovich form
Correction: $\frac{1}{2}\sigma(x-a)\cdot\sigma = \frac{\sigma^2}{2}(x-a)$.
Stratonovich drift $= m(x-a) - \frac{\sigma^2}{2}(x-a) = (m-\sigma^2/2)(x-a)$.
$$dX_t = \left(m-\frac{\sigma^2}{2}\right)(X_t-a)\,dt + \sigma(X_t-a)\circ dW_t$$

### c) Explicit solution
**Method 1 — Substitution:** Let $Y_t = X_t - a$. Then $dY = mY\,dt + \sigma Y\,dW$ (GBM!).
$$\boxed{X_t = a + (x_0-a)\exp\!\left[(m-\tfrac{1}{2}\sigma^2)t + \sigma W_t\right]}$$

**Method 2 — Stratonovich separation:** From Stratonovich form, $\frac{dX}{X-a} = (m-\sigma^2/2)dt + \sigma\circ dW_t$. Integrate both sides to get $\ln(X_t-a) - \ln(x_0-a) = (m-\sigma^2/2)t + \sigma W_t$.

### d) Probability $P\{X_t > x_0\}$
$X_t > x_0 \iff (m-\sigma^2/2)t + \sigma W_t > 0 \iff W_t > -(m-\sigma^2/2)t/\sigma$.
Since $W_t \sim N(0,t)$:
$$P\{X_t > x_0\} = \Phi\!\left(\frac{(m-\sigma^2/2)\sqrt{t}}{\sigma}\right)$$

## Mock 3 — $dX = \frac{1}{3}X^{1/3}\,dt + X^{2/3}\,dW$, $X_0 = x_0$

### a) Explicit solution (Stratonovich + separation)
Itô→Stratonovich correction: $\frac{1}{2}x^{2/3}\cdot\frac{2}{3}x^{-1/3} = \frac{1}{3}x^{1/3}$.
Stratonovich drift $= \frac{1}{3}x^{1/3} - \frac{1}{3}x^{1/3} = 0$.
$$dX_t = X_t^{2/3}\circ dW_t$$
Separate: $X^{-2/3}\,dX = \circ\,dW_t$. Integrate: $3X_t^{1/3} - 3x_0^{1/3} = W_t$.
$$\boxed{X_t = \left(x_0^{1/3} + \frac{W_t}{3}\right)^3}$$

### b) Verification
Compute $dX_t$ from the explicit solution using Itô's formula applied to $f(w) = (x_0^{1/3} + w/3)^3$.

### c) Non-uniqueness when $X_0 = 0$
- **Solution 1:** $X_t = (W_t/3)^3$ (from above with $x_0=0$)
- **Solution 2:** $X_t \equiv 0$ (constant — check: $dX = 0$, RHS $= 0$ ✓)

Non-uniqueness is expected because $x^{1/3}$ and $x^{2/3}$ are **not Lipschitz at $x=0$** (derivatives blow up).

## Mock 4 — $dX = v^2X(1+X^2)\,dt + v(1+X^2)\,dW$, $X_0 = x_0$

### a) Explicit solution (Stratonovich + separation)
Correction: $\frac{1}{2}v(1+x^2)\cdot 2vx = v^2x(1+x^2)$.
Stratonovich drift $= v^2x(1+x^2) - v^2x(1+x^2) = 0$.
$$dX_t = v(1+X_t^2)\circ dW_t$$
Separate: $\frac{dX}{1+X^2} = v\circ dW_t$. Integrate: $\arctan(X_t) - \arctan(x_0) = vW_t$.
$$\boxed{X_t = \tan(\arctan(x_0) + vW_t)}$$

### b) Verification
Let $Z_t = \arctan(x_0) + vW_t$, $X_t = \tan(Z_t)$. Apply Itô to $\tan(Z_t)$:
$\frac{d}{dz}\tan z = 1+\tan^2z = 1+X^2$, $\frac{d^2}{dz^2}\tan z = 2\tan z(1+\tan^2 z) = 2X(1+X^2)$.
$$dX = (1+X^2)v\,dW + \frac{1}{2}\cdot 2X(1+X^2)v^2\,dt = v(1+X^2)\,dW + v^2X(1+X^2)\,dt\quad\checkmark$$

### c) Problem with the solution
The tangent has vertical asymptotes at $vW_s = \pm\pi/2 + n\pi$. Brownian motion is unbounded, so the solution **explodes in finite time with positive probability**. The linear growth condition fails (cubic drift, quadratic diffusion). The global existence theorem does **not** apply.

## Mock 5 — OU: $dX = (b(t)-a(t)X)\,dt + \sigma(t)\,dW$, $X_0 = x_0$

### a) Existence & uniqueness
Drift $\mu(t,x) = b(t)-a(t)x$ linear in $x$; diffusion $\sigma(t,x)=\sigma(t)$ independent of $x$.
- **Lipschitz:** $|a(t)||x-y| \leq K|x-y|$ ✓
- **Linear growth:** $|b(t)| + |a(t)||x| + |\sigma(t)| \leq 2K(1+|x|)$ ✓

### b) Solution via Stratonovich
Correction $= \frac{1}{2}\sigma(t)\cdot 0 = 0$ (diffusion independent of $x$), so Itô = Stratonovich.
**Integrating factor:** Set $Y_t = e^{\int_0^t a(s)ds}X_t$. By Itô:
$$dY_t = e^{\int_0^t a(s)ds}\left[a(t)X_t\,dt + dX_t\right] = e^{\int_0^t a(s)ds}b(t)\,dt + e^{\int_0^t a(s)ds}\sigma(t)\,dW_t$$
Integrate:
$$\boxed{X_T = e^{-\int_0^T a(s)ds}\left[x_0 + \int_0^T e^{\int_0^t a(s)ds}b(t)\,dt + \int_0^T e^{\int_0^t a(s)ds}\sigma(t)\,dW_t\right]}$$

> **This is where the integrating factor is used. $\star$**

### c) Same solution without Stratonovich
Same integrating factor approach but motivated differently: multiply the SDE by $e^{\int_0^t a(s)ds}$ and recognise the left-hand side as $d(e^{\int_0^t a}X_t)$ via Itô's product rule.

### d) Expected value
Take $\mathbb{E}$, use $\mathbb{E}[\text{Itô integral}] = 0$:
$$\mathbb{E}[X_T] = e^{-\int_0^T a(s)ds}\left[x_0 + \int_0^T e^{\int_0^t a(s)ds}b(t)\,dt\right]$$

### e) Is $X_t$ Gaussian?
Yes — $X_t$ is a linear function of Itô integrals of deterministic functions, which are Gaussian. Linear transforms of Gaussians are Gaussian.

## Mock 6 — $dX = \frac{X}{2}\,dt + \sqrt{1+X^2}\,dW$, $X_0 = x_0$

### a) Existence & uniqueness
- **Lipschitz:** $|x/2-y/2| + |\sqrt{1+x^2}-\sqrt{1+y^2}| \leq \frac{1}{2}|x-y| + |x-y| = \frac{3}{2}|x-y|$ ✓
  (Key step: $|\sqrt{1+x^2}-\sqrt{1+y^2}| \leq |x-y|$ by squaring both sides.)
- **Linear growth:** $|x/2| + \sqrt{1+x^2} \leq \frac{3}{2}(1+|x|)$ ✓

### b) Solution via Stratonovich
Correction: $\frac{1}{2}\sqrt{1+x^2}\cdot\frac{x}{\sqrt{1+x^2}} = \frac{x}{2}$.
Stratonovich drift $= \frac{x}{2} - \frac{x}{2} = 0$.
$$dX_t = \sqrt{1+X_t^2}\circ dW_t$$
Separate: $\frac{dX}{\sqrt{1+X^2}} = \circ\,dW_t$. Integrate using $\int\frac{dx}{\sqrt{1+x^2}} = \sinh^{-1}(x)$:
$$\sinh^{-1}(X_t) - \sinh^{-1}(x_0) = W_t$$
$$\boxed{X_t = \sinh(\sinh^{-1}(x_0) + W_t)}$$

### c) Expected value
From the Itô SDE in integral form: $\mathbb{E}[X_t] = x_0 + \int_0^t \frac{1}{2}\mathbb{E}[X_s]\,ds$.
Set $m_t = \mathbb{E}[X_t]$: $\frac{dm_t}{dt} = \frac{1}{2}m_t$, $m_0 = x_0$.
$$\boxed{\mathbb{E}[X_t] = x_0 e^{t/2}}$$

## Mock 7 — CIR: $dX = k(\theta-X)\,dt + \sigma\sqrt{X}\,dW$, $X_0 = x_0$

(Existence & uniqueness given; $2k\theta > \sigma^2$ ensures $X_t > 0$ a.s.)

### a) Expected value $\mathbb{E}[X_T]$
Integrate, take $\mathbb{E}$, use $\mathbb{E}[\text{Itô integral}] = 0$:
$$\mathbb{E}[X_T] = x_0 + \int_0^T k(\theta - \mathbb{E}[X_t])\,dt$$
ODE for $m(t) = \mathbb{E}[X_t]$: $\frac{dm}{dt} = k(\theta - m)$, $m(0) = x_0$.
$$\boxed{\mathbb{E}[X_T] = x_0 e^{-kT} + \theta(1-e^{-kT})}$$
$\lim_{T\to\infty}\mathbb{E}[X_T] = \theta$ — justifies $\theta$ as the **long-term mean**.

### b) Itô differential of $X_t^2$
Apply Itô with $f(x) = x^2$: $f' = 2x$, $f'' = 2$.
$$d(X_t^2) = 2X_t\,dX_t + (dX_t)^2 = [(2k\theta+\sigma^2)X_t - 2kX_t^2]\,dt + 2\sigma X_t^{3/2}\,dW_t$$

### c) Variance
Let $m_2(t) = \mathbb{E}[X_t^2]$. Taking $\mathbb{E}$ of the above:
$$\frac{dm_2}{dt} = (2k\theta+\sigma^2)m(t) - 2k\,m_2(t)$$
Linear ODE with known input $m(t)$. Solve, then:
$$\text{Var}(X_T) = m_2(T) - m(T)^2 = x_0\frac{\sigma^2}{k}(e^{-kT}-e^{-2kT}) + \theta\frac{\sigma^2}{2k}(1-e^{-kT})^2$$

## Where the Integrating Factor Appears

The integrating factor technique is used in **Mock 5** (OU process) and in the **2024 past paper Q1** (OU process).

**It is never needed for Mocks 1–4, 6, 7** — those are solved by Stratonovich separation or substitution.

### When to use it
Use the integrating factor when the SDE has the form:
$$dX_t = (b(t) - a(t)X_t)\,dt + \sigma(t)\,dW_t$$
i.e. drift is **affine in $X$**, diffusion is **independent of $X$**.

### The recipe
1. Set $Y_t = e^{\int_0^t a(s)ds}\,X_t$ (the integrating factor is $e^{\int_0^t a(s)ds}$)
2. Apply Itô's product rule to $dY_t$: the $X$-terms cancel, leaving $dY_t = e^{\int_0^t a}b(t)\,dt + e^{\int_0^t a}\sigma(t)\,dW_t$
3. Integrate from $0$ to $T$ and substitute back $X_t = e^{-\int_0^t a}Y_t$

## Stratonovich Correction Formula

To convert $dX = \mu(X)\,dt + \sigma(X)\,dW$ to Stratonovich:
$$dX = \underbrace{\left(\mu(X) - \frac{1}{2}\sigma(X)\sigma'(X)\right)}_{\text{Stratonovich drift}}dt + \sigma(X)\circ dW$$

The correction **vanishes** (Stratonovich drift = 0 for the $dt$ term entirely) in Mocks 3, 4, 6 — enabling pure separation of variables with $\circ dW$ on the right.

## Existence & Uniqueness Template

For $dX = \mu(t,X)\,dt + \sigma(t,X)\,dW$:
1. **Lipschitz:** $|\mu(t,x)-\mu(t,y)| + |\sigma(t,x)-\sigma(t,y)| \leq K|x-y|$
2. **Linear growth:** $|\mu(t,x)| + |\sigma(t,x)| \leq K'(1+|x|)$

Both satisfied → unique global strong solution.
