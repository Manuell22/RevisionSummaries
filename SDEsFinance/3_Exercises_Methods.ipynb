## A. SDE Problem Methods (Problem 1 in every exam)

### A1. Checking Existence & Uniqueness

**Recipe:**
1. Identify $\mu(x)$ and $\sigma(x)$ from the SDE $dX = \mu(X)\,dt + \sigma(X)\,dW$.
2. Check **Lipschitz:** $|\sigma(x) - \sigma(y)| \le K|x-y|$ and similarly for $\mu$.
   - Polynomial: Lipschitz iff the function is globally differentiable with bounded derivative.
   - $\sigma(x) = x^{1/3}$: unbounded derivative at $0$ → NOT Lipschitz.
   - $\sigma(x) = 1 + x^2$: Lipschitz on bounded sets, but not globally.
3. Check **linear growth:** $|\sigma(x)|^2 \le C(1+|x|^2)$.
   - $\sigma(x) = v(1+x^2)$: $|\sigma|^2 \sim x^4$ → violates linear growth → **explosion** possible.
   - $\sigma(x) = \sigma\sqrt{x}$: satisfies linear growth.

### A2. Computing $E[X_t]$ and $\text{Var}(X_t)$ via Itô

**Recipe (without solving the SDE):**
1. Apply Itô to $f(X) = X^2$: $dX^2 = 2X\,dX + (dX)^2 = (2X\mu + \sigma^2)\,dt + 2X\sigma\,dW$.
2. Take expectations: $\frac{d}{dt}E[X^2_t] = 2E[X_t\mu(X_t)] + E[\sigma^2(X_t)]$ (the $dW$ term has zero expectation).
3. Similarly, from $dX$: $\frac{d}{dt}E[X_t] = E[\mu(X_t)]$.
4. If $\mu$ and $\sigma$ are linear in $X$, these become **closed ODEs** you can solve explicitly.
5. $\text{Var}(X_t) = E[X^2_t] - (E[X_t])^2$.

**Example (CIR):** $dX = k(\theta - X)dt + \sigma\sqrt{X}\,dW$.
- $E[X_t]$: $\frac{d}{dt}E[X] = k(\theta - E[X])$ → $E[X_t] = \theta + (X_0 - \theta)e^{-kt}$.
- $E[X^2_t]$: Apply Itô to $X^2$, use $dX\cdot dX = \sigma^2 X\,dt$, solve the resulting ODE.

### A3. Solving SDEs — Three Methods

**Method 1: Guess & Verify (Itô)**
- Guess $X_t = g(W_t, t)$ (e.g., $e^{(\cdots)t + (\cdots)W_t}$ for GBM).
- Apply Itô to $g$, match drift and diffusion with the original SDE.
- Works for: GBM, OU.

**Method 2: Stratonovich Conversion + Separation**
- Convert Itô SDE to Stratonovich: $dX = \tilde\mu\,dt + \sigma(X) \circ dW$ where $\tilde\mu = \mu - \frac{1}{2}\sigma\sigma'$.
- If $\tilde\mu = 0$: $\frac{dX}{\sigma(X)} = \circ\,dW$ → integrate both sides. 
- If $\tilde\mu \ne 0$: $\frac{dX}{\sigma(X)} = c\,dt + \circ\,dW$ → still separable.
- Works for: cubic-root, explosion (tangent), sinh SDEs.

**Method 3: Integrating Factor**
- For $dX = m\,dt + \sigma X\,dW$: multiply by $e^{-\sigma W_t + \sigma^2 t/2}$.
- The product $Y_t = X_t e^{-\sigma W_t + \sigma^2 t/2}$ satisfies $dY = m\,e^{-\sigma W + \sigma^2 t/2}\,dt$.
- Integrate: $Y_T - Y_0 = m\int_0^T e^{-\sigma W_s + \sigma^2 s/2}\,ds$.

## B. Option Pricing Methods (Problem 2 in every exam)

### B1. Pricing Vanilla Options in BS

**Call:** $C = S_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)$, $d_{1,2} = \frac{\ln(S_0/K) + (r \pm \sigma^2/2)T}{\sigma\sqrt{T}}$.

**Put:** $P = Ke^{-rT}\Phi(-d_2) - S_0\Phi(-d_1)$ or via PCP: $P = C - S_0 + Ke^{-rT}$.

### B2. Pricing Strategies — Combine Building Blocks

| Strategy | Price in BS |
|---|---|
| **Straddle** $(K)$ | $C(K) + P(K)$ |
| **Short straddle** | $-(C(K) + P(K))$ |
| **Risk reversal** $(K_1, K_2)$ | $C(K_1) - P(K_2)$ |
| **Bull call spread** $(K_1 < K_2)$ | $C(K_1) - C(K_2)$ |
| **Bear call spread** $(K_1 < K_2)$ | $C(K_2) - C(K_1)$ |
| **Butterfly** $(L < M < H)$ | $C(L) - 2C(M) + C(H)$ |
| **Calendar spread** ($T_1 > T_2$, same $K$) | $C(S_0, K, T_1) - C(S_0, K, T_2)$ |

### B3. Computing Strategy Greeks

**Delta of a portfolio** = sum of component deltas.
- $\Delta_{\text{straddle}} = \Delta_C + \Delta_P = \Phi(d_1) + (\Phi(d_1) - 1) = 2\Phi(d_1) - 1$.
- $\Delta_{\text{bull call}} = \Phi(d_1(K_1)) - \Phi(d_1(K_2))$ (always positive).
- **Vega** of a portfolio: sum of component vegas. All vanilla long positions have positive vega; short positions have negative vega.

### B4. Pricing Binary Options

- **Cash-or-Nothing** (pays $K$ if $S_T > K$): $V_0 = Ke^{-rT}\Phi(d_2)$.
- **Asset-or-Nothing** (pays $S_T$ if $S_T > K$): $V_0 = S_0\Phi(d_1)$.
- Delta of CNO: $\Delta_{\text{CNO}} = \frac{Ke^{-rT}\phi(d_2)}{S_0\sigma\sqrt{T}}$.

### B5. Pricing Under Two Correlated Stocks

For two stocks $S_1, S_2$ with correlated GBMs:
- Apply Itô product rule with $dW^1 dW^2 = \rho\,dt$.
- Portfolio value involves both stocks → compute joint distribution.
- PCP still holds for each stock individually.

### B6. Arbitrage Proof via Put–Call Parity

**When asked "prove there exists an arbitrage":**
1. Compute $C - P$ from given prices.
2. Compute $S_0 - Ke^{-rT}$ from market data.
3. If they differ, construct a replicating portfolio:
   - If $C - P > S_0 - Ke^{-rT}$: sell call, buy put, buy stock, borrow $Ke^{-rT}$.
   - If $C - P < S_0 - Ke^{-rT}$: buy call, sell put, sell stock, lend $Ke^{-rT}$.
4. Show the portfolio costs nothing at maturity but yields positive cash at $t=0$.

## C. Smile Model Methods (Problem 3 in every exam)

### C1. Bachelier Pricing

**Call:** $C_{\text{Bach}} = (S_0 - K)\Phi(d) + \sigma_N\sqrt{T}\,\phi(d)$ where $d = \frac{S_0 - K}{\sigma_N\sqrt{T}}$.

**Put:** $P_{\text{Bach}} = (K - S_0)\Phi(-d) + \sigma_N\sqrt{T}\,\phi(-d)$.

**Delta:** $\Delta_{\text{call}} = \Phi(d)$, $\Delta_{\text{put}} = \Phi(d) - 1$.

**Symmetry property:** $\phi(d)$ is symmetric around $d=0$ (i.e., $S_0 = K$). This means:
- **Butterfly delta in Bachelier = 0** when strikes are symmetric around $S_0$ (because the two $\Phi$ terms from the wings cancel exactly).
- **Straddle delta** at $K = S_0$: $2\Phi(0) - 1 = 0$.

### C2. Displaced Diffusion (DD) Pricing

**Setup:** $S_t = \alpha e^{rt} + X_t$ where $dX = rX\,dt + \sigma X\,dW$.

**Call price:** Reduce to a BS call on $X$ with shifted strike:
$$C_{\text{DD}} = C_{\text{BS}}(X_0,\; K - \alpha e^{rT},\; T,\; r,\; \sigma)$$
where $X_0 = S_0 - \alpha$.

**Constraint:** Need $K - \alpha e^{rT} > 0$ for the call to make sense, and $X_0 > 0$.

### C3. Mixture Diffusion (MDD) Pricing

$$C_{\text{MDD}} = \sum_{i=1}^n \lambda_i\, C_{\text{BS}}(S_0, K, T, r, \sigma_i), \quad \sum \lambda_i = 1$$

Similarly for binary options:
$$\text{CNO}_{\text{MDD}} = \sum \lambda_i\, \text{CNO}_{\text{BS}}(\sigma_i), \quad \text{ANO}_{\text{MDD}} = \sum \lambda_i\, \text{ANO}_{\text{BS}}(\sigma_i)$$

**Delta:** $\Delta_{\text{MDD}} = \sum \lambda_i\, \Delta_{\text{BS}}(\sigma_i) = \sum \lambda_i\, \Phi(d_1(\sigma_i))$.

### C4. Shifted MDD / DD + MDD

Combine displacement $\alpha$ with mixture:
$$C_{\text{SMDD}} = \sum \lambda_i\, C_{\text{BS}}(X_0,\; K-\alpha e^{rT},\; T,\; r,\; \sigma_i)$$

Produces **asymmetric smile with curvature** — the most flexible local vol model in the course.

### C5. Smile Shape Discussion

When asked to "describe the smile shape":
- **BS (GBM):** Flat implied vol → no smile.
- **Bachelier:** Downward smile (higher implied vol at ATM, lower at wings in BS terms).
- **DD ($\alpha > 0$):** Downward skew. **DD ($\alpha < 0$):** Upward skew.
- **MDD:** Symmetric U-shape.
- **Shifted MDD ($\alpha \ne 0$):** Asymmetric smile.
- **All local vol models** suffer from **future smile flattening**.

## D. Risk Measures Methods (Problem 4 in every exam)

### D1. General Recipe

1. **Price portfolio today** ($t=0$) using $Q$-formulas: $V_0$.
2. **Simulate $S_H$** under $P$: $S_H = S_0 e^{(\mu-\sigma^2/2)H + \sigma\sqrt{H}Z}$, $Z\sim N(0,1)$.
3. **Reprice at $H$** under $Q$: $V_H = V(S_H, T-H, r, \sigma)$.
4. **Loss:** $L_H = V_0 - V_H$.
5. **VaR:** $\alpha$-quantile of $L_H$. **ES:** $E[L_H \mid L_H > \text{VaR}]$.

### D2. Single Stock (No Options)

If portfolio = long stock: $V_H = S_H$, $V_0 = S_0$.
$$L_H = S_0 - S_H = S_0(1 - e^{(\mu-\sigma^2/2)H + \sigma\sqrt{H}Z})$$
- $L_H$ is monotone decreasing in $Z$.
- VaR at level $\alpha$: use $Z = -\Phi^{-1}(\alpha)$ (loss quantile corresponds to stock dropping).
- Analytical formulas exist for GBM stock.

### D3. Option Portfolios (Monte Carlo)

For portfolios with optionality (straddle, spread, etc.), $V_H$ is a non-linear function of $S_H$ → no closed form for VaR/ES. Use Monte Carlo:
- $N \ge 100{,}000$ scenarios.
- **Vega analysis:** Repeat with $\sigma \pm \Delta\sigma$ to see how risk changes.

### D4. Portfolio with Bachelier Stock

If the stock follows Bachelier: $S_H = S_0 + \sigma_N\sqrt{H}\,Z$, $Z \sim N(0,1)$.
- Loss is $L_H = V_0 - V_H$ where $V_H$ uses Bachelier pricing formulas.
- If the loss is **normally distributed** (e.g., long stock in Bachelier): 
  - $\text{VaR}_\alpha = \mu_L + \sigma_L \Phi^{-1}(\alpha)$
  - $\text{ES}_\alpha = \mu_L + \sigma_L \frac{\phi(\Phi^{-1}(\alpha))}{1-\alpha}$

### D5. Forward Portfolios & Deterministic Losses

**Key insight:** A forward contract has payoff $S_T - K$ — fully linear in $S_T$. But forward VALUE at horizon $H$ is $V_H = S_H - Ke^{-r(T-H)}$, which depends on $S_H$ (random under $P$).

However, if the portfolio is a **bond + forward** combination such that all randomness cancels (e.g., short bond pays back at $H$, forward expires at $T > H$), the loss can be **deterministic**:
- If $L_H$ is a constant $c$: $\text{VaR}$ at any level $< 1$ equals $c$, but $\text{ES}$ is undefined ($P(L > \text{VaR}) = 0$ if $L$ is a point mass).
- This is a known edge case tested in Exam 7.

### D6. Bond + Stock (Different Maturities)

Common setup: long bond maturing at $H$, long stock valued at $H$.
- Bond: $V_0^{\text{bond}} = e^{-rH} \cdot \text{face}$, at $H$ the bond returns face value (deterministic).
- Stock: $V_H^{\text{stock}} = S_H$ (random).
- Total loss: $L_H = (V_0^{\text{bond}} + V_0^{\text{stock}}) - (\text{face} + S_H)$.
- Only the stock component contributes randomness.

## F. Common Exam Patterns & Tricks

### F1. SDE Problems — What to Expect

Every Exam Problem 1 gives you an SDE and asks some subset of:
- (a) State conditions for existence/uniqueness. Do they hold?
- (b) Find $E[X_t]$ (usually via Itô on $X$, take expectations).
- (c) Find $\text{Var}(X_t)$ (via Itô on $X^2$, take expectations, subtract $(E[X])^2$).
- (d) Solve the SDE explicitly.
- (e) If conditions fail, what can go wrong?

**SDEs that have appeared:**
| Exam | SDE | Solution Method |
|---|---|---|
| 1 | $dX = m\,dt + \sigma X\,dW$ | Integrating factor |
| 2 | $dX = m(X-a)\,dt + \sigma(X-a)\,dW$ | Substitution $Y = X-a$, then GBM |
| 3 | $dX = \frac{\sigma^2}{6}X^{-1/3}dt + \sigma X^{1/3}dW$ | Stratonovich separation |
| 4 | $dX = v^2X(1+X^2)dt + v(1+X^2)dW$ | Stratonovich → tangent (EXPLOSION) |
| 5 | $dX = (b(t)-a(t)X)dt + \sigma(t)dW$ | Time-dependent OU (integrating factor) |
| 6 | $dX = \frac{1}{2}Xdt + \sqrt{1+X^2}dW$ | Stratonovich → sinh |
| 7 | $dX = k(\theta-X)dt + \sigma\sqrt{X}dW$ | CIR (moments via Itô on $X^2$) |

### F2. Pricing Problems — What to Expect

Every Exam Problem 2 asks you to price an option strategy in BS and compute Greeks.

**Strategies that have appeared:** straddle (E1), two-stock portfolio (E2), binary options ANO/CNO (E3), butterfly (E4), risk reversal (E5), bull call spread (E6), calendar spread (E7), arbitrage proof (E4).

**Key tricks:**
- Always check if put–call parity simplifies things (e.g., $|S-K| = C + P$ for straddle).
- For portfolios of calls/puts: price = linear combination of BS prices, delta = linear combination of BS deltas.
- Short positions: negate everything.

### F3. Smile Problems — What to Expect

Every Exam Problem 3 asks you to price the *same* (or similar) strategy under a non-BS model.

**Models that have appeared:** DD (E1, E3, E5), Bachelier (E2, E4, E6, E7), MDD (E3), selection/comparison (E2).

**Key tricks:**
- Bachelier call: $C = (S_0-K)\Phi(d) + \sigma_N\sqrt{T}\phi(d)$, $d = (S_0-K)/(\sigma_N\sqrt{T})$.
- DD: shift strike by $\alpha e^{rT}$, use $X_0 = S_0 - \alpha$, apply BS to the shifted quantities.
- MDD: weighted average of BS formulas.
- Binary options under MDD: weighted average of BS binary prices.
- Bachelier delta is symmetric around ATM — exploited for butterfly.

### F4. Risk Measure Problems — What to Expect

Every Exam Problem 4 asks for VaR/ES of a portfolio over horizon $H < T$.

**Portfolios that have appeared:** single stock (E1), short straddle in Bachelier (E2), stock in Bachelier (E3), bond + short forward (E4), bond + stock (E5), short bond + long forward (E6), forward-only (E7).

**Key tricks:**
- **GBM stock:** loss involves lognormal → no closed-form VaR for option portfolios → Monte Carlo.
- **Bachelier stock:** loss is normal → analytical VaR/ES formulas.
- **Forward portfolio:** loss can be deterministic → edge case.
- **Vega analysis:** always asked — how do VaR/ES change with $\sigma$?
- **Sub-additivity check:** often asked — give example where VaR fails.

## H. Exam Strategy Tips

1. **Problem 1 (SDE):** Always start by writing down $\mu(x)$ and $\sigma(x)$ explicitly. Check Lipschitz & linear growth. For moments: Itô on $X^n$, take expectations. For solving: try Stratonovich if the SDE looks non-standard.

2. **Problem 2 (BS pricing):** Write the payoff as a combination of calls/puts. Price each using BS. Delta = $\sum \Delta_i$. Check with PCP when possible.

3. **Problem 3 (Smile):** Identify which model is being used. Apply the appropriate formula (Bachelier/DD/MDD). For DD: shift the strike. For MDD: weighted average. For Bachelier: remember $r=0$ and the symmetry properties.

4. **Problem 4 (Risk):** Set up $V_0$, $V_H(S_H)$, $L_H = V_0 - V_H$. Identify if loss is monotone in $Z$ (can find VaR analytically) or non-monotone (need Monte Carlo). Always discuss vega sensitivity and coherence.

5. **Time management:** Each problem is worth 25%. Don't spend too long on the SDE — move on and collect marks from the pricing/risk problems which are more formulaic.
