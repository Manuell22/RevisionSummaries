## PART 1: Probability, Stochastic Processes & SDEs

### 1. Probability Spaces & Random Variables

The triple $(\Omega, \mathcal{F}, P)$:
- $\Omega$ = sample space (set of all possible outcomes).
- $\mathcal{F}$ = $\sigma$-algebra (collection of events we can measure). Must be closed under complements and countable unions.
- $P$ = probability measure satisfying: $P(\Omega)=1$, $P(\emptyset)=0$, countable additivity.

A **random variable** $X: \Omega \to \mathbb{R}$ assigns a number to each outcome. It must be *measurable* w.r.t. $\mathcal{F}$.

**Discrete vs Continuous RVs:**
- Discrete: $P(X = x_i)$ given by a probability mass function.
- Continuous: described by a PDF $f_X(x)$ with $P(a \le X \le b) = \int_a^b f_X(x)\,dx$.

The **CDF** $F_X(x) = P(X \le x)$ always exists and uniquely characterises the distribution.

### 2. Key Distributions

| Distribution | PDF / PMF | Mean | Variance | MGF $M_X(u)$ |
|---|---|---|---|---|
| **Uniform** $U(a,b)$ | $\frac{1}{b-a}$ on $[a,b]$ | $\frac{a+b}{2}$ | $\frac{(b-a)^2}{12}$ | $\frac{e^{ub}-e^{ua}}{u(b-a)}$ |
| **Normal** $N(\mu,\sigma^2)$ | $\frac{1}{\sigma\sqrt{2\pi}}e^{-(x-\mu)^2/(2\sigma^2)}$ | $\mu$ | $\sigma^2$ | $e^{\mu u + \sigma^2 u^2/2}$ |
| **Exponential** $\text{Exp}(\lambda)$ | $\lambda e^{-\lambda x}$, $x \ge 0$ | $1/\lambda$ | $1/\lambda^2$ | $\frac{\lambda}{\lambda - u}$ |
| **Lognormal** | If $\ln X \sim N(\mu,\sigma^2)$ | $e^{\mu+\sigma^2/2}$ | $(e^{\sigma^2}-1)e^{2\mu+\sigma^2}$ | — |

**Lognormal** is central to finance: if $S$ follows GBM, $S_T$ is lognormal. Its density is right-skewed and positive — suitable for stock prices.

**Multivariate Normal:** $(X_1, \ldots, X_n) \sim N(\boldsymbol{\mu}, \boldsymbol{\Sigma})$. Fully characterised by the mean vector and covariance matrix. If jointly normal, **uncorrelated $\iff$ independent** (this is special to the Gaussian!).

### 3. Moment Generating Functions, Independence & Convergence

**MGF:** $M_X(u) = E[e^{uX}]$. If it exists in a neighbourhood of $0$, it uniquely determines the distribution and gives moments:
$$E[X^n] = M_X^{(n)}(0)$$

**Independence:** $X, Y$ independent $\iff$ $f_{X,Y}(x,y) = f_X(x) f_Y(y)$ $\iff$ $M_{X+Y}(u) = M_X(u) M_Y(u)$.

**Central Limit Theorem (CLT):** Let $X_1, \ldots, X_n$ be i.i.d. with mean $\mu$, variance $\sigma^2$. Then:
$$\frac{\bar{X}_n - \mu}{\sigma/\sqrt{n}} \xrightarrow{d} N(0,1)$$

This is why the Gaussian appears everywhere — sums of many small independent effects are approximately normal.

**Convergence hierarchy:**
1. **Almost sure** (a.s.) $\Rightarrow$ **In probability** $\Rightarrow$ **In distribution (law)**
2. **$L^p$ convergence** $\Rightarrow$ **In probability**

These are not equivalent in the reverse direction. Convergence in distribution is the weakest; a.s. is the strongest.

### 4. Brownian Motion

**Definition:** A stochastic process $W = (W_t)_{t \ge 0}$ is a standard Brownian motion if:
1. $W_0 = 0$.
2. $W_t - W_s \sim N(0, t-s)$ for $0 \le s < t$.
3. Increments over non-overlapping intervals are independent.
4. Paths $t \mapsto W_t(\omega)$ are continuous.

**Key properties:**
- $E[W_t] = 0$, $\text{Var}(W_t) = t$, $\text{Cov}(W_s, W_t) = \min(s,t)$.
- Paths are **nowhere differentiable** a.s. — $W$ is incredibly rough.
- **Unbounded variation:** $\sum |W_{t_{i+1}} - W_{t_i}| \to \infty$ on any interval. This means you cannot use Riemann–Stieltjes integration naively.
- **Quadratic variation:** $\sum (W_{t_{i+1}} - W_{t_i})^2 \to t$ (deterministic!). This is the source of the extra $\frac{1}{2}$ term in Itô's formula.

**Filtration:** $\mathcal{F}_t$ = information available up to time $t$ (generated by $\{W_s : 0 \le s \le t\}$). A process is **adapted** if $X_t$ is $\mathcal{F}_t$-measurable (it only uses information available at time $t$).

### 5. Stochastic Integrals & Itô Calculus

**Itô integral:** $\int_0^T f(t,\omega)\,dW_t$ is defined as a limit of sums where $f$ is evaluated at the **left endpoint**:
$$\sum f(t_i)(W_{t_{i+1}} - W_{t_i})$$

Key properties:
- $E\!\left[\int_0^T f\,dW\right] = 0$ (zero-mean).
- **Itô isometry:** $E\!\left[\left(\int_0^T f\,dW\right)^2\right] = \int_0^T E[f^2]\,dt$.
- The Itô integral is a **martingale**.

**Itô's Formula** (1-D): Let $X_t$ satisfy $dX = \mu\,dt + \sigma\,dW$. For $\varphi(t,x) \in C^{1,2}$:

$$d\varphi = \left(\frac{\partial \varphi}{\partial t} + \mu\frac{\partial \varphi}{\partial x} + \frac{\sigma^2}{2}\frac{\partial^2 \varphi}{\partial x^2}\right)dt + \sigma\frac{\partial \varphi}{\partial x}\,dW$$

The extra $\frac{\sigma^2}{2}\varphi''$ term arises from $(dW)^2 = dt$.

**Quadratic variation rules:**
$$dW \cdot dW = dt, \quad dt \cdot dW = 0, \quad dt \cdot dt = 0$$

**Product rule for SDEs:** $d(XY) = X\,dY + Y\,dX + dX\,dY$. The cross-term $dX\,dY$ is non-zero when both $X$ and $Y$ have Brownian components.

### 6. SDEs — Existence, Uniqueness & Solution Methods

**Generic SDE:** $dX_t = \mu(t,X_t)\,dt + \sigma(t,X_t)\,dW_t$, $X_0 = x_0$.

**Existence & Uniqueness Theorem:** A unique strong solution exists if $\mu, \sigma$ are:
1. **Lipschitz continuous** in $x$: $|\mu(t,x) - \mu(t,y)| + |\sigma(t,x) - \sigma(t,y)| \le K|x - y|$.
2. **Linear growth** in $x$: $|\mu(t,x)|^2 + |\sigma(t,x)|^2 \le C(1 + |x|^2)$.

When these fail — as with $\sigma \propto x^{1/3}$ (non-Lipschitz) or $\sigma \propto (1+x^2)$ (super-linear growth) — pathological behaviour can occur:
- **Non-uniqueness** of solutions.
- **Finite-time explosion:** $X_t \to \pm\infty$ in finite time (e.g., the tangent SDE).

**Solution strategies:**
1. **Guess and verify** using Itô's formula (e.g., GBM: guess $X = e^{f(W_t, t)}$).
2. **Stratonovich conversion:** convert to $\circ\,dW$, use ordinary chain rule / separation of variables, then convert back.
3. **Integrating factor** (for $dX = m\,dt + \sigma X\,dW$): multiply by $e^{-\sigma W_t + \sigma^2 t/2}$.

**Computing moments without solving the SDE:**
- Apply Itô's formula to $X^2$ (or $X^n$) to get $dX^2 = \ldots\,dt + \ldots\,dW$.
- Take expectations: the $dW$ term vanishes $\Rightarrow$ ODE for $E[X^2_t]$.
- $\text{Var}(X_t) = E[X^2_t] - (E[X_t])^2$.

### 7. Geometric Brownian Motion (GBM) — The Finance Workhorse

$$dS_t = \mu S_t\,dt + \sigma S_t\,dW_t$$

**Solution:** Apply Itô to $\ln S$:
$$d\ln S = \left(\mu - \tfrac12\sigma^2\right)dt + \sigma\,dW$$
$$\Rightarrow S_T = S_0 \exp\!\left((\mu - \tfrac12\sigma^2)T + \sigma W_T\right)$$

- $\ln S_T \sim N\!\left(\ln S_0 + (\mu-\sigma^2/2)T,\;\sigma^2 T\right)$ — so $S_T$ is **lognormal**.
- $E[S_T] = S_0 e^{\mu T}$, $\text{Var}(S_T) = S_0^2 e^{2\mu T}(e^{\sigma^2 T} - 1)$.
- The $-\frac{1}{2}\sigma^2$ correction in the exponent is the **Itô correction** — often tested in exams.

**Under $Q$ (risk-neutral):** Replace $\mu \to r$: $dS = rS\,dt + \sigma S\,dW^Q$.

### 8. Ornstein–Uhlenbeck (Vasicek) & CIR Processes

**OU / Vasicek:** $dX_t = k(\theta - X_t)\,dt + \sigma\,dW_t$
- **Mean-reverting** to $\theta$ with speed $k$.
- Solution: $X_t = \theta + (X_0 - \theta)e^{-kt} + \sigma \int_0^t e^{-k(t-s)}\,dW_s$.
- $X_t$ is Gaussian: $E[X_t] = \theta + (X_0 - \theta)e^{-kt}$, $\text{Var}(X_t) = \frac{\sigma^2}{2k}(1-e^{-2kt})$.
- Long-run: $X_t \xrightarrow{d} N(\theta, \sigma^2/(2k))$.
- **Can go negative** — fine for interest rates in some models, but not for stock prices.

**Time-dependent OU:** $dX_t = (b(t) - a(t)X_t)\,dt + \sigma(t)\,dW_t$ — same structure, but with time-varying coefficients. Solved similarly via integrating factor.

**CIR (Cox–Ingersoll–Ross):** $dX_t = k(\theta - X_t)\,dt + \sigma\sqrt{X_t}\,dW_t$
- Mean-reverting AND stays **non-negative** (under Feller condition $2k\theta \ge \sigma^2$).
- Distribution: non-central $\chi^2$.
- $E[X_t] = \theta + (X_0-\theta)e^{-kt}$, same as OU. But variance differs.
- **Moments via Itô:** Apply Itô to $X^2$ to derive $E[X^2_t]$ and hence $\text{Var}(X_t)$.

### 9. More SDE Types (Exam Favourites)

**$dX = m\,dt + \sigma X\,dW$ (linear drift, multiplicative noise):**
- Solved by **integrating factor**: multiply by $e^{-\sigma W_t + \sigma^2 t/2}$.
- Apply Itô to find $E[X_t]$ and $\text{Var}(X_t)$ by applying Itô to $X^2$, $X^3$ etc.
- Appears in Exam 1.

**Cubic-root SDE:** $dX = \frac{\sigma^2}{6X^{1/3}}\,dt + \sigma X^{1/3}\,dW$
- Diffusion coefficient $\sigma(x) = \sigma x^{1/3}$ is NOT Lipschitz at $x=0$.
- Solved via Stratonovich: convert $\to$ separate variables $\to X_t^{2/3} = X_0^{2/3} + \frac{2}{3}\sigma W_t$.
- Appears in Exams 3, 5.

**Explosion SDE:** $dX = v^2 X(1+X^2)\,dt + v(1+X^2)\,dW$
- Diffusion $\sigma(x) = v(1+x^2)$ violates **linear growth**.
- Solution: $X_t = \tan(\arctan(X_0) + vW_t)$.
- **Explodes in finite time** — $X_t \to \pm\infty$ when $\arctan(X_0) + vW_t$ hits $\pm\pi/2$.
- Object lesson: without linear growth, solutions can blow up. Appears in Exam 4.

**Sinh SDE:** $dX = \frac{1}{2}X\,dt + \sqrt{1+X^2}\,dW$
- Solution: $X_t = \sinh(\sinh^{-1}(X_0) + W_t)$.
- Found via Stratonovich conversion + separation. Appears in Exam 6.

### 10. Correlated Brownian Motions

If $W^1, W^2$ are BMs with $\text{Corr}(dW^1, dW^2) = \rho$, then:
$$dW^1 \cdot dW^2 = \rho\,dt$$

Construction: Given independent BMs $Z^1, Z^2$, set $W^1 = Z^1$, $W^2 = \rho Z^1 + \sqrt{1-\rho^2}\,Z^2$.

**Multi-stock problems:** Two stocks with correlated GBMs:
$$dS_i = \mu_i S_i\,dt + \sigma_i S_i\,dW^i, \quad i=1,2$$
To find $d(S_1 S_2)$ or $d(S_1/S_2)$, use the product rule with $dW^1 dW^2 = \rho\,dt$ — important in exams.

### 11. Martingales, Girsanov & Feynman–Kac

**Martingale:** An adapted process $M_t$ with $E[M_t \mid \mathcal{F}_s] = M_s$ for $s < t$. In SDE language: **zero drift** $\Rightarrow$ martingale. Any process of the form $\int_0^t f\,dW$ is a martingale (assuming integrability).

**Girsanov's Theorem:** Under mild conditions, we can define a new probability measure $Q$ such that:
$$\tilde{W}_t = W_t + \int_0^t \gamma_s\,ds$$
is a BM under $Q$. The change of measure is via the **Radon–Nikodym derivative**:
$$\frac{dQ}{dP} = \exp\!\left(-\int_0^T \gamma_s\,dW_s - \frac{1}{2}\int_0^T \gamma_s^2\,ds\right)$$

**Intuition:** Think of a biased coin. Under $P$, heads has probability $p$; under $Q$, probability $q$. The Radon–Nikodym derivative $\frac{dQ}{dP}$ re-weights each outcome. Girsanov does the same for continuous processes — it changes the drift without touching the diffusion.

**Financial application:** Under $P$, $dS = \mu S\,dt + \sigma S\,dW^P$. Set $\gamma = (\mu-r)/\sigma$ (market price of risk). Then under $Q$: $dS = rS\,dt + \sigma S\,dW^Q$ — the drift becomes $r$.

**Feynman–Kac Theorem:** The PDE
$$\frac{\partial u}{\partial t} + \mu\frac{\partial u}{\partial x} + \frac{\sigma^2}{2}\frac{\partial^2 u}{\partial x^2} - ru = 0, \quad u(T,x) = g(x)$$
has solution $u(t,x) = e^{-r(T-t)}E^Q[g(X_T) \mid X_t = x]$ where $dX = \mu\,dt + \sigma\,dW^Q$.

This connects the BS PDE to the risk-neutral pricing formula.

## PART 2: Option Pricing & The Volatility Smile

### 12. The Black–Scholes Economy & Self-Financing Portfolios

**Two assets:**
- Bank account: $dB = rB\,dt \Rightarrow B_t = e^{rt}$.
- Stock: $dS = \mu S\,dt + \sigma S\,dW$ (GBM under $P$).

**Assumptions:** Continuous trading, no transaction costs, no dividends (initially), constant $r, \sigma$, short-selling allowed.

**Self-financing portfolio:** $\Pi_t = a_t S_t + b_t B_t$ with $d\Pi = a_t\,dS + b_t\,dB$ — changes in value come only from changes in asset prices, not from injecting/withdrawing cash.

**Contingent claim:** A payoff $g(S_T)$ at maturity $T$. The price $V_t = V(t, S_t)$ must satisfy the **replication**: find $a_t, b_t$ such that $\Pi_T = g(S_T)$, then $V_t = \Pi_t$ (no-arbitrage).

### 13. Deriving the Black–Scholes PDE

1. Let $V = V(t,S)$. Apply Itô: $dV = \left(\frac{\partial V}{\partial t} + \mu S \frac{\partial V}{\partial S} + \frac{\sigma^2 S^2}{2}\frac{\partial^2 V}{\partial S^2}\right)dt + \sigma S\frac{\partial V}{\partial S}\,dW$.
2. Form portfolio $\Pi = V - \Delta S$ where $\Delta = \frac{\partial V}{\partial S}$.
3. Then $d\Pi = \left(\frac{\partial V}{\partial t} + \frac{\sigma^2 S^2}{2}\frac{\partial^2 V}{\partial S^2}\right)dt$ — the $dW$ term is gone (risk eliminated).
4. No arbitrage $\Rightarrow$ $d\Pi = r\Pi\,dt$.
5. Substituting $\Pi = V - \Delta S$ and equating:

$$\boxed{\frac{\partial V}{\partial t} + rS\frac{\partial V}{\partial S} + \frac{\sigma^2 S^2}{2}\frac{\partial^2 V}{\partial S^2} = rV}$$

**Crucial observation:** $\mu$ (real-world drift) does **not** appear — the price is preference-free!

### 14. The Black–Scholes Formula & Greeks

**Call price:** $C(S_0, K, T, r, \sigma) = S_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)$

$$d_1 = \frac{\ln(S_0/K) + (r + \sigma^2/2)T}{\sigma\sqrt{T}}, \quad d_2 = d_1 - \sigma\sqrt{T}$$

**Put price** (via put–call parity): $P = Ke^{-rT}\Phi(-d_2) - S_0\Phi(-d_1)$.

**Greeks:**

| Greek | Formula (call) | Interpretation |
|---|---|---|
| $\Delta$ | $\Phi(d_1)$ | Shares to hold for delta-hedge |
| $\Gamma$ | $\frac{\phi(d_1)}{S_0 \sigma\sqrt{T}}$ | Curvature — how fast $\Delta$ changes |
| $\Theta$ | $-\frac{S_0\phi(d_1)\sigma}{2\sqrt{T}} - rKe^{-rT}\Phi(d_2)$ | Time decay |
| Vega | $S_0\phi(d_1)\sqrt{T}$ | Sensitivity to vol; always $> 0$ |
| $\rho$ | $KTe^{-rT}\Phi(d_2)$ | Sensitivity to interest rate |

where $\phi(x) = \frac{1}{\sqrt{2\pi}}e^{-x^2/2}$ is the standard normal PDF.

**BS PDE in terms of Greeks:** $\Theta + rS\Delta + \frac{1}{2}\sigma^2 S^2 \Gamma = rV$.

### 15. Put–Call Parity, Forwards & Arbitrage

**Put–call parity:** $C - P = S_0 - Ke^{-rT}$ (model-independent — holds for any model, not just BS).

**Proof:** $(S_T - K)^+ - (K - S_T)^+ = S_T - K$ always. By no-arbitrage, price of LHS = price of RHS = $S_0 - Ke^{-rT}$.

**Arbitrage proof pattern** (exam favourite): If observed prices violate PCP:
1. Compute $C - P$ and $S_0 - Ke^{-rT}$.
2. If $C - P > S_0 - Ke^{-rT}$: sell call, buy put, buy stock, borrow $Ke^{-rT}$ → riskless profit.
3. If $C - P < S_0 - Ke^{-rT}$: reverse all positions.
4. Show the terminal payoff is zero and the initial cash flow is strictly positive.

**Forward contract:** Agreement to buy at price $K$ at time $T$. Payoff = $S_T - K$.
- **Forward price:** $F(0,T) = S_0 e^{rT}$ — this is model-independent (no-arbitrage).
- Forward value at $t=0$: $V_0 = S_0 - Ke^{-rT}$ (can be positive, negative, or zero).
- **Zero cost** when $K = F(0,T)$.

### 16. Option Strategies in Detail

**Straddle** (long call + long put, same strike):
- Payoff: $|S_T - K|$. Price: $C(K) + P(K)$. Delta: $2\Phi(d_1) - 1$.
- **Short straddle** is the reverse — profits if $S$ stays near $K$ (sell volatility).

**Risk Reversal** (long call $K_1$ + short put $K_2$, $K_1 > K_2$):
- Payoff: $(S_T - K_1)^+ - (K_2 - S_T)^+$. Bullish directional bet with limited downside.
- **Bear Risk Reversal** reverses the signs (coursework topic).

**Bull Call Spread** (long call $K_1$ + short call $K_2$, $K_1 < K_2$):
- Payoff capped between $0$ and $K_2 - K_1$. Limited upside, limited downside.

**Bear Call Spread** (short call $K_1$ + long call $K_2$, $K_1 < K_2$):
- The reverse of bull call spread. Profits if $S_T$ stays below $K_1$.

**Butterfly Spread** (long call $L$ + long call $H$ + short 2 calls $M$, $L < M < H$, $M = (L+H)/2$):
- Payoff is tent-shaped, peaks at $M$. Bet on low volatility.
- In Bachelier model: $\Delta_{\text{butterfly}} = 0$ by symmetry of the normal density!

**Calendar Spread** (long call maturity $T_1$ + short call maturity $T_2$, $T_1 > T_2$, same strike):
- Not a pure terminal payoff — at $T_2$, the short call expires and the long call still has time value.
- Value at $T_2$ depends on where $S_{T_2}$ is. Delta analysis is non-trivial.

### 17. Binary/Digital Options

**Cash-or-Nothing (CNO):** Pays fixed amount $K$ if $S_T > K^*$ (some barrier), else $0$.
- BS price: $Ke^{-rT}\Phi(d_2)$.
- Under MDD (Mixture Diffusion): weighted average of BS CNO prices at different vols.
- Delta of CNO: $\frac{Ke^{-rT}\phi(d_2)}{\sigma S_0 \sqrt{T}}$.

**Asset-or-Nothing (ANO):** Pays $S_T$ if $S_T > K$, else $0$.
- BS price: $S_0\Phi(d_1)$.
- Note: $\text{Call} = \text{ANO}(K) - K \cdot e^{-rT} \cdot \frac{\text{CNO}(K)}{K} = S_0\Phi(d_1) - Ke^{-rT}\Phi(d_2)$ — recovering BS!

### 18. Numeraires & FTAP

A **numeraire** is any strictly positive asset used to denominate prices. The bank account $B_t = e^{rt}$ is the standard one.

**FTAP in numeraire language:**
1. No arbitrage $\iff$ $\exists$ equivalent measure $Q^N$ under which every asset price divided by the numeraire $N$ is a martingale.
2. Market is complete $\iff$ $Q^N$ is unique.

Any positive-value asset can be a numeraire (stock, bond, forward price). Changing numeraire = changing measure.

### 19. Implied Volatility & the Smile

**Implied volatility** $\sigma_{\text{imp}}(K, T)$: the $\sigma$ you plug into the BS formula to reproduce the market price of an option with strike $K$ and maturity $T$.

If BS were exact, $\sigma_{\text{imp}}$ would be the same for all $K$ — a flat line. In practice:
- **Smile:** U-shape ($\sigma_{\text{imp}}$ higher for deep ITM and OTM).
- **Skew:** monotonically decreasing (equity markets post-1987 crash).

The smile tells us the market "knows" the BS model is wrong — the true distribution has fatter tails / asymmetry.

### 20. Bachelier Model

$$dS_t = \sigma_N\,dW_t \quad (r = 0)$$

- $S_T \sim N(S_0, \sigma_N^2 T)$ — **normal** distribution (can go negative!).
- **Call price:** $C_{\text{Bach}} = (S_0 - K)\Phi\!\left(\frac{S_0 - K}{\sigma_N\sqrt{T}}\right) + \sigma_N\sqrt{T}\,\phi\!\left(\frac{S_0 - K}{\sigma_N\sqrt{T}}\right)$
- Delta: $\Delta = \Phi\!\left(\frac{S_0 - K}{\sigma_N\sqrt{T}}\right)$
- Produces a **downward-sloping** implied vol smile (when expressed in BS implied vol terms).
- Historically the first option pricing model (Louis Bachelier, 1900 — before Black–Scholes!).

### 21. Displaced Diffusion (DD) Model

Introduce auxiliary $X_t$ with $dX = rX\,dt + \sigma X\,dW$, then set:
$$S_t = \alpha e^{rt} + X_t$$

- $\alpha$ is a **shift parameter**. $S_t = \alpha e^{rt} + X_t$ where $X$ is GBM.
- If $\alpha > 0$: distribution shifts right, skew goes one way. If $\alpha < 0$: the reverse.
- Call price: re-express the BS formula with shifted arguments.
- $S_T$ given $X_T$ is lognormal shifted by $\alpha e^{rT}$.
- DD produces a **skew** (tilted smile) rather than a symmetric smile.

### 22. CEV (Constant Elasticity of Variance)

$$dS_t = rS_t\,dt + \nu S_t^\gamma\,dW_t$$

- $\gamma = 1$: BS. $\gamma < 1$: downward skew (leverage effect). $\gamma > 1$: upward skew.
- Local volatility: $\sigma_{\text{loc}}(S) = \nu S^{\gamma-1}$.
- No simple closed-form for the call price (involves non-central $\chi^2$ distribution).

### 23. Mixture Diffusion Model (MDD)

The density of $S_T$ is a **mixture of lognormals**:
$$f_{\text{MDD}}(x) = \sum_{i=1}^{n} \lambda_i\, f_{\text{LN}}(x;\, r - \sigma_i^2/2,\, \sigma_i^2 T)$$
with $\sum \lambda_i = 1$, $\lambda_i > 0$.

Equivalently, the call price is a **weighted average of BS prices**:
$$C_{\text{MDD}} = \sum_{i=1}^n \lambda_i\, C_{\text{BS}}(S_0, K, T, r, \sigma_i)$$

- Produces a **U-shaped (symmetric) smile**.
- Can be combined with DD to create **Shifted MDD** — asymmetric smile with curvature.

### 24. Calibration in Practice

**Problem:** Given observed market implied vols $\sigma_{\text{mkt}}(K_i)$ for strikes $K_1, \ldots, K_m$, find model parameters $\beta$ that best fit.

**Objective function:**
$$\min_\beta \sum_{i=1}^m \left[V_{\text{BS}}(\sigma_{\text{mkt}}(K_i)) - V_{\text{Model}}(\beta; K_i)\right]^2$$

This is a non-linear least-squares problem. For MDD with 2 components: $\beta = (\lambda, \sigma_1, \sigma_2)$ — 3 parameters. For Shifted MDD: add $\alpha$ — 4 parameters.

**Practical notes:**
- More parameters = better fit, but risk of overfitting.
- Need good initial guesses; multiple local minima possible.
- After calibration: use the model to price exotic options or compute Greeks.

### 25. Future Smile Flattening & Stochastic Vol

**The problem with local vol models** (Bachelier, DD, CEV, MDD, Shifted MDD):
- Calibrate today's smile perfectly? ✓
- Predict tomorrow's smile correctly? ✗

The predicted future smile **flattens** — as $t \to T$, the implied vol surface becomes flat. This is because the mixture density converges to a single lognormal.

**Stochastic volatility** (e.g., Heston: $dv_t = \kappa(\theta - v_t)\,dt + \xi\sqrt{v_t}\,dW^v_t$ with $\text{Corr}(dW^S, dW^v) = \rho$):
- Volatility itself is random → smile is preserved into the future.
- But this introduces **incomplete markets** — $Q$ is not unique, so there's no unique price.
- Need additional assumptions (e.g., specify market price of volatility risk).

### 26. Dividends

For continuous dividend yield $q$:
- Stock accumulates dividends continuously: $dS = (\mu - q)S\,dt + \sigma S\,dW$ under $P$.
- Under $Q$: $dS = (r-q)S\,dt + \sigma S\,dW^Q$.
- Modified BS formula: $C = S_0 e^{-qT}\Phi(d_1) - Ke^{-rT}\Phi(d_2)$ with $d_{1,2} = \frac{\ln(S_0/K) + (r-q \pm \sigma^2/2)T}{\sigma\sqrt{T}}$.
- Forward price: $F(0,T) = S_0 e^{(r-q)T}$.

## PART 3: Risk Measures & Time Series

### 27. Time Series Analysis Under $P$

**Log-returns:** $r_i = \ln(S_{i+1}/S_i)$.

If $S$ follows GBM with parameters $(\mu, \sigma)$, then $r_i \sim N((\mu - \sigma^2/2)\Delta t,\; \sigma^2 \Delta t)$ i.i.d.

**Estimation from data:**
- $\hat{\mu}_{\text{ann}} = \frac{1}{\Delta t}\bar{r} + \frac{\hat\sigma^2}{2}$ (annualised drift).
- $\hat{\sigma}_{\text{ann}} = \frac{s_r}{\sqrt{\Delta t}}$ where $s_r$ is the sample std of daily log-returns.

**Checking normality:**
- **QQ plot:** Plot quantiles of the data against theoretical normal quantiles. Departures from a straight line indicate non-normality (fat tails, skewness).
- **Historical vol $\ne$ implied vol:** Historical vol is estimated under $P$; implied vol is backed out under $Q$. They measure different things.

### 28. Value-at-Risk (VaR)

**Definition:** Given a portfolio loss $L_H = V_0 - V_H$ over horizon $H$:
$$\text{VaR}_{H,\alpha} = \inf\{l : P(L_H \le l) \ge \alpha\}$$
i.e., the $\alpha$-quantile of the loss distribution ($\alpha$ typically 95% or 99%).

**For a single stock** under GBM: if portfolio = long 1 share,
$$L_H = S_0 - S_H = S_0\left(1 - e^{(\mu-\sigma^2/2)H + \sigma\sqrt{H}\,Z}\right)$$
VaR at level $\alpha$: use $Z = \Phi^{-1}(\alpha)$ (since loss is monotone decreasing in $Z$, actually $Z = \Phi^{-1}(1-\alpha)$ for the loss quantile — be careful with sign conventions!).

**For options portfolios:** No closed form. Use Monte Carlo:
1. Simulate $S_H$ under $P$.
2. Reprice the portfolio at $H$ using BS (with remaining maturity $T-H$).
3. Compute $L_H^{(j)}$, sort, find quantile.

**Drawbacks:**
- Tells you nothing about losses **beyond** VaR.
- **Not sub-additive:** $\text{VaR}(A+B)$ can exceed $\text{VaR}(A) + \text{VaR}(B)$ — diversification can appear to increase risk!

### 29. Expected Shortfall (ES) & Coherent Risk Measures

**Expected Shortfall:**
$$\text{ES}_{H,\alpha} = E[L_H \mid L_H > \text{VaR}_{H,\alpha}]$$

This is the **average of losses in the tail** beyond VaR. Always $\ge$ VaR.

**Coherent risk measures** (Artzner et al. 1999) satisfy:
1. **Translation invariance:** $\rho(X + c) = \rho(X) - c$ (adding cash reduces risk).
2. **Sub-additivity:** $\rho(X+Y) \le \rho(X) + \rho(Y)$ (diversification doesn't increase risk).
3. **Positive homogeneity:** $\rho(\lambda X) = \lambda \rho(X)$ for $\lambda > 0$.
4. **Monotonicity:** If $X \le Y$ a.s. then $\rho(X) \ge \rho(Y)$ (worse outcomes = higher risk).

| Measure | Sub-additive? | Coherent? |
|---|---|---|
| VaR | NO | NO |
| ES | YES | YES |

**Limitation of homogeneity:** Real-world risk doesn't scale linearly — large positions move markets (**liquidity risk**).

**For Bachelier/normal losses:** If $L \sim N(\mu_L, \sigma_L^2)$:
- $\text{VaR}_\alpha = \mu_L + \sigma_L \Phi^{-1}(\alpha)$
- $\text{ES}_\alpha = \mu_L + \sigma_L \frac{\phi(\Phi^{-1}(\alpha))}{1-\alpha}$

### 30. Monte Carlo Workflow for Risk Measures

**Step-by-step (e.g., short straddle on one stock):**

1. **Today:** Price portfolio at $t=0$:
   - $V_0 = -(C_0 + P_0)$ for a short straddle (negative because you sold).
   - Use BS formula with full maturity $T$.

2. **Simulate under $P$:** Generate $N$ scenarios:
   $$S_H^{(j)} = S_0 \exp\!\left((\mu - \tfrac12\sigma^2)H + \sigma\sqrt{H}\,Z^{(j)}\right), \quad Z^{(j)} \sim N(0,1)$$

3. **Reprice at $H$ under $Q$:** For each $S_H^{(j)}$, compute:
   $$V_H^{(j)} = -(C_{\text{BS}}(S_H^{(j)}, K, T-H, r, \sigma) + P_{\text{BS}}(S_H^{(j)}, K, T-H, r, \sigma))$$

4. **Compute losses:** $L_H^{(j)} = V_0 - V_H^{(j)}$.

5. **Sort and extract:**
   - VaR = $\alpha$-percentile of sorted $\{L_H^{(j)}\}$.
   - ES = mean of all $L_H^{(j)}$ exceeding VaR.

**Special cases in exams:**
- **Forward portfolios:** Loss may be deterministic if the portfolio is self-financing with no optionality → VaR = ES = that deterministic loss (or undefined if $P(L > \text{VaR}) = 0$).
- **Bond + stock (different maturities):** Bond matures at $H$ → repaid at par at horizon.

### 31. Vega Sensitivity & Historical Context

**Vega analysis:** How do VaR and ES change when $\sigma$ changes?
- Propagate $\sigma$ into both the simulation (under $P$: $\sigma$ affects $S_H$ distribution) and the repricing (under $Q$: $\sigma$ enters BS formula).
- For a short straddle: $\uparrow\sigma \Rightarrow$ wider $S_H$ distribution + higher option prices = both effects increase risk.
- For strategies with mixed long/short positions: vega can partially cancel.

**Historical milestones:**
- **1900:** Bachelier's thesis — first mathematical model of asset prices.
- **1973:** Black–Scholes–Merton formula.
- **1987:** Black Monday crash → volatility smile/skew appears.
- **1995:** **Barings Bank collapse** — Nick Leeson's short straddles on Nikkei 225. Kobe earthquake caused huge losses. No risk management. ES would have flagged the tail risk.
- **1998:** LTCM — model overconfidence + extreme leverage.
- **1999:** Artzner et al. formalise coherent risk measures.
- **2008:** Financial crisis — systemic risk, CDO model failures.
- **Basel III/IV:** Shift from VaR to ES as the primary regulatory risk measure.

## PART 4: Numerical Solution of SDEs

### 32. Euler–Maruyama Scheme

For $dX = \mu(t,X)\,dt + \sigma(t,X)\,dW$ with no closed-form solution, discretise on grid $t_0 < t_1 < \cdots < t_n$ with $\delta = t_{i+1} - t_i$:

$$X_{t_{i+1}} = X_{t_i} + \mu(t_i, X_{t_i})\delta + \sigma(t_i, X_{t_i})\sqrt{\delta}\;N_i, \quad N_i \sim N(0,1) \text{ i.i.d.}$$

**Convergence:**
- Requires Lipschitz + linear growth on $\mu, \sigma$ (same conditions as SDE existence).
- **Strong convergence** order $\frac{1}{2}$: $E|X_T^{(\delta)} - X_T| = O(\delta^{1/2})$.
- Halving $\delta$ reduces error by factor $1/\sqrt{2}$ — not very fast, but robust.

**Validation strategy:** When a closed-form solution exists (e.g., GBM), compare Euler–Maruyama with the "one-shot" exact formula using the **same Brownian path** $W_T = \sum \sqrt{\delta}\,N_i$:
- One-shot GBM: $S_T = S_0 \exp((\mu - \sigma^2/2)T + \sigma W_T)$.
- As $\delta \to 0$, the Euler paths should converge to the exact path.

**Practical implementation (Python):**
- **Naive:** for-loop over time steps → slow.
- **Vectorised:** pre-generate all $N_i^{(j)}$ in a matrix, then use numpy array operations → 100x+ faster. Essential for $N > 10000$ scenarios.

### 33. Euler–Maruyama Examples

**GBM:** $dS = \mu S\,dt + \sigma S\,dW$
$$S_{i+1} = S_i + \mu S_i \delta + \sigma S_i \sqrt{\delta}\,N_i = S_i(1 + \mu\delta + \sigma\sqrt{\delta}\,N_i)$$

**$dX = m\,dt + \sigma X\,dW$:**
$$X_{i+1} = X_i + m\delta + \sigma X_i\sqrt{\delta}\,N_i$$

**Cubic-root SDE:** $dX = \frac{\sigma^2}{6}X^{-1/3}\,dt + \sigma X^{1/3}\,dW$:
$$X_{i+1} = X_i + \frac{\sigma^2}{6}X_i^{-1/3}\delta + \sigma X_i^{1/3}\sqrt{\delta}\,N_i$$
(care needed at $X \approx 0$ — numerics can be unstable).

**Using Euler for risk measures:** When the loss at horizon $H$ involves an SDE with no closed form:
1. Run Euler from $0$ to $H$ to get $X_H$.
2. Reprice at $H$ using whatever model applies.
3. Feed into VaR/ES calculation.

## Appendices — Key Topics

### A. Stratonovich Calculus (Solution Technique)

**Itô ↔ Stratonovich conversion:** Given $dX = \mu\,dt + \sigma(X)\,dW$ (Itô), the equivalent Stratonovich SDE is:
$$dX = \left(\mu - \frac{1}{2}\sigma\sigma'\right)dt + \sigma(X) \circ dW$$
where $\sigma' = \frac{\partial \sigma}{\partial X}$.

**Why Stratonovich?** The ordinary chain rule applies: $d\varphi(X) = \varphi'(X) \circ dX$. This lets you use **separation of variables** just like in ODE theory:
$$\frac{dX}{\sigma(X)} = \ldots\,dt + \circ\,dW \quad \Rightarrow \quad \int \frac{dX}{\sigma(X)} = \ldots$$

Used extensively in exams for cubic-root, explosion, and sinh SDEs.

### B. Girsanov Intuition (Discrete Analogy)

Consider tossing a coin: $P(\text{heads}) = p$, $P(\text{tails}) = 1-p$.  
Define $Q$ by $Q(\text{heads}) = q$, $Q(\text{tails}) = 1-q$.

The Radon–Nikodym derivative $\frac{dQ}{dP}(\omega)$ re-weights each outcome:
- $\frac{dQ}{dP}(\text{heads}) = q/p$, $\frac{dQ}{dP}(\text{tails}) = (1-q)/(1-p)$.

Girsanov extends this to continuous paths: $\frac{dQ}{dP} = \mathcal{E}(-\gamma \cdot W)$ where $\gamma = (\mu-r)/\sigma$ is the market price of risk.

### C. Solving $dX = m\,dt + \sigma X\,dW$ (Integrating Factor)

Multiply both sides by $e^{-\sigma W_t + \sigma^2 t/2}$ (the "integrating factor"):
$$d\!\left(X_t e^{-\sigma W_t + \sigma^2 t/2}\right) = m\,e^{-\sigma W_t + \sigma^2 t/2}\,dt$$

The LHS has no $dW$ term → integrate both sides in $dt$ → solve for $X_t$. From this, derive $E[X_t]$ and higher moments.
