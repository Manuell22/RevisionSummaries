# Finite Elements — Exercises & Methods

## A. Deriving Weak Forms from Strong PDEs

**Recipe:**
1. Start with strong PDE: $-\nabla^2u=f$ (or $u-\nabla^2u=f$, etc.)
2. Multiply by test function $v\in V$, integrate over $\Omega$
3. Apply integration by parts / Green's identity on each triangle:
   $$-\int_K v\nabla^2u\,dx = \int_K\nabla v\cdot\nabla u\,dx - \int_{\partial K}v\frac{\partial u}{\partial n}\,dS$$
4. Sum over elements; interior edges cancel ($n^-=-n^+$)
5. Apply boundary conditions:
   - **Dirichlet** ($u=g$ on $\Gamma_D$): enforce via trial space $V=\{v:v|_{\Gamma_D}=0\}$; test functions vanish, boundary integral disappears
   - **Neumann** ($\partial u/\partial n=g_N$ on $\Gamma_N$): boundary integral becomes $\int_{\Gamma_N}v\,g_N\,dS$, absorbed into $F(v)$
6. Result: find $u\in V$ s.t. $b(u,v)=F(v)$ $\forall v\in V$

**Example: Poisson with mixed BCs**
- $b(u,v)=\int_\Omega\nabla u\cdot\nabla v\,dx$
- $F(v)=\int_\Omega fv\,dx + \int_{\Gamma_N}g_Nv\,dS$

**Example: Helmholtz**
- $b(u,v)=\int_\Omega(uv+\nabla u\cdot\nabla v)\,dx = (u,v)_{H^1}$
- $F(v)=\int_\Omega fv\,dx$

## B. Verifying Unisolvence / Constructing Nodal Bases

**General strategy (Dual Condition Lemma):**
Show that $N_i(v)=0$ for all $i$ $\Rightarrow$ $v\equiv0$.

**1D approach (degree $k$):** $v$ vanishes at $k+1$ points $\Rightarrow$ $v=0$ by fundamental theorem of algebra.

**2D — P1 (linear on triangle):**
1. Let $\Pi_1=\{x:L_1(x)=0\}$ be line through $z_2,z_3$
2. $v|_{\Pi_1}$ is linear and vanishes at 2 points $\Rightarrow$ $v|_{\Pi_1}=0$ everywhere on $\Pi_1$
3. Hyperplane lemma: $v=cL_1$
4. $v(z_1)=cL_1(z_1)=0$ and $L_1(z_1)\neq0$ $\Rightarrow$ $c=0$

**2D — P2 (quadratic on triangle):**
1. 3 nodes on edge $\Pi_1$ $\Rightarrow$ $v|_{\Pi_1}$ is degree 2 with 3 zeros $\Rightarrow$ $v=L_1Q_1$, $Q_1$ linear
2. On line $\Pi_2$: $Q_1|_{\Pi_2}$ vanishes at 2 midpoints $\Rightarrow$ $Q_1=cL_2$
3. So $v=cL_1L_2$; at edge 3 midpoint: $0=v=cL_1(m_{12})L_2(m_{12})\neq0$ $\Rightarrow$ $c=0$

**Vandermonde approach:** Construct $V_{ij}=N_j(\psi_i)$, show $\det V\neq0$. Then $\phi_i=\sum_j\mu_{ij}\psi_j$ with $\mu=V^{-1}$.

**Constructing nodal basis (1D Lagrange):**
$$\phi_i(x)=\prod_{j\neq i}\frac{x-x_j}{x_i-x_j}$$

## C. Interpolation Error Estimates

**Goal:** Bound $\|u-\mathcal{I}_hu\|_{H^i}$ for $u\in H^{k+1}$.

**Key result** (Lagrange, degree $k$):
$$\|u-\mathcal{I}_hu\|_{H^i(\Omega)} \leq Ch^{k+1-i}|u|_{H^{k+1}(\Omega)}$$

**Proof programme:**
1. **Bound $\mathcal{I}_{K_1}$** on diameter-1 triangle: $\|\mathcal{I}_{K_1}u\|_{H^k}\leq C\|u\|_{H^k}$ (triangle inequality + Sobolev)
2. **Local error on unit triangle**: $|u-\mathcal{I}_{K_1}u|_{H^i}\leq C|u|_{H^{k+1}}$
   - Subtract $Q_{k,B}u$ (Taylor polynomial): $\mathcal{I}_KQ=Q$ (projection), so $\mathcal{I}_K(u-Q)-(u-Q)$
3. **Scale to diameter $d$**: change of variables $x=dy$ gives factor $d^{k+1-i}$
4. **Sum over elements**: $\sum_K d_K^{2(k+1-i)} \leq h^{2(k+1-i)}$ (max element diameter)

**Important technical steps:**
- Taylor error: $\|D^\beta(f-Q_{k,B}f)\|_{L^2}\leq C\frac{|\Omega|^{1/2}}{|B|^{1/2}}d^{k+1-|\beta|}|f|_{H^{k+1}}$
- Need minimum aspect ratio $>0$ (no degenerate elements)
- Sobolev's inequality: need $k>n/2$ to evaluate point values

## D. Proving Continuity & Coercivity of Bilinear Forms

**Continuity** ($|b(u,v)|\leq M\|u\|\|v\|$):
- Poisson: $|\int\nabla u\cdot\nabla v|\leq|u|_{H^1}|v|_{H^1}\leq\|u\|_{H^1}\|v\|_{H^1}$ (Schwarz)
- Helmholtz: $|(u,v)_{H^1}|\leq\|u\|_{H^1}\|v\|_{H^1}$ (Schwarz), $M=1$
- Stokes $a(u,v)$: $|a(u,v)|\leq 2\mu\|\varepsilon(u)\|\|\varepsilon(v)\|\leq 2\mu\|u\|_{H^1}\|v\|_{H^1}$

**Coercivity** ($b(u,u)\geq\gamma\|u\|^2$):
- Helmholtz: $b(u,u)=\|u\|_{H^1}^2$, $\gamma=1$
- Poisson Neumann (on $\bar{V}_h=\{u:\bar{u}=0\}$): use mean estimate $\|u\|_{L^2}\leq C|u|_{H^1}$ to get $\|u\|_{H^1}^2\leq(1+C^2)|u|_{H^1}^2=(1+C^2)b(u,u)$ so $\gamma=1/(1+C^2)$
- Poisson Dirichlet: use Poincaré $\|u\|_{L^2}\leq C|u|_{H^1}$ for $u\in\mathring{H}^1$
- Stokes on $Z$: $a(u,u)=2\mu\|\varepsilon(u)\|^2_{L^2}\geq\mu\|\nabla u\|^2=\mu|u|^2_{H^1}$ (Korn's) $\geq\gamma\|u\|_{H^1}^2$ (Poincaré)

**Pattern:** Use Schwarz for continuity, Poincaré/mean estimate to promote seminorm to full norm for coercivity.

## E. Applying Lax-Milgram & Céa's Lemma

### Lax-Milgram Checklist
1. Identify $V$ (Hilbert space), $b(\cdot,\cdot)$ (bilinear), $F(\cdot)$ (linear functional)
2. Prove $b$ is **continuous**: find $M$
3. Prove $b$ is **coercive**: find $\gamma$ (check if $h$-dependent!)
4. Prove $F$ is **bounded**: $|F(v)|\leq\|F\|_{V'}\|v\|_V$
5. Conclude: $\exists!$ $u\in V$ with $b(u,v)=F(v)$ and $\|u\|\leq\frac{1}{\gamma}\|F\|_{V'}$

### Céa's Lemma Application
With $u$ (exact) and $u_h$ (FE):
$$\|u-u_h\|_V \leq \frac{M}{\gamma}\inf_{v\in V_h}\|u-v\|_V$$

**Standard choice**: $v=\mathcal{I}_hu$ (interpolant):
$$\|u-u_h\|_{H^1} \leq \frac{M}{\gamma}\|u-\mathcal{I}_hu\|_{H^1} \leq \frac{M}{\gamma}Ch^k|u|_{H^{k+1}}$$

**Stability check:** If $\gamma=\gamma_h\to 0$ as $h\to 0$, the scheme is unstable!

## F. Aubin-Nitsche $L^2$ Estimate

**When to use:** You have $H^1$ convergence and want $L^2$ convergence (gains $+1$ power of $h$).

**Requirements:** Elliptic regularity on convex domain: $|w|_{H^2}\leq C\|g\|_{L^2}$ for dual problem.

**Recipe:**
1. Define $e=u-u_h$, let $w$ solve the dual: $b(v,w)=(v,e)_{L^2}$ $\forall v$
2. Choose $v=e$: $\|e\|_{L^2}^2 = b(e,w)$
3. Use Galerkin orthogonality: $b(e,w) = b(e,w-\mathcal{I}_hw)$
4. Continuity: $\leq M\|e\|_{H^1}\|w-\mathcal{I}_hw\|_{H^1}$
5. Interpolation: $\|w-\mathcal{I}_hw\|_{H^1}\leq Ch|w|_{H^2}$
6. Elliptic regularity: $|w|_{H^2}\leq C\|e\|_{L^2}$
7. Combine: $\|e\|_{L^2}^2\leq CMh\|e\|_{H^1}\|e\|_{L^2}$ $\Rightarrow$ $\|e\|_{L^2}\leq Ch\|e\|_{H^1}$
8. Substitute $H^1$ rate: $\|e\|_{L^2}\leq Ch^{k+1}|u|_{H^{k+1}}$

## G. Stokes & Mixed Problems (Brezzi Framework)

**Recipe for showing well-posedness:**
1. Identify the saddle-point structure: $a(u,v)+b(v,p)=F(v)$, $b(u,q)=G(q)$
2. Define kernel $Z=\{v:b(v,q)=0\;\forall q\}$
3. Prove **coercivity of $a$ on $Z$**: often uses Korn + Poincaré
4. Prove **inf-sup for $b$**: continuous version usually given, use Fortin for discrete

**Fortin's trick:**
1. Construct bounded $\Pi_h:V\to V_h$ with $b(v-\Pi_hv,q_h)=0$ $\forall q_h\in Q_h$
2. Then: $\sup_{v_h}\frac{b(v_h,q_h)}{\|v_h\|}\geq\frac{b(\Pi_hv,q_h)}{\|\Pi_hv\|}=\frac{b(v,q_h)}{\|\Pi_hv\|}\geq\frac{\beta\|q_h\|}{C_\Pi}$

**MINI element** ($P1+B3/P1$):
- Velocity: $(P1)^d$ enriched with cubic bubble $b_K=\lambda_1\lambda_2\lambda_3$
- Pressure: $P1$
- Fortin operator: $\Pi_h = \mathcal{I}_{h,1}v + \alpha_K b_K$ where $\alpha_K$ chosen to satisfy $b(v-\Pi_hv,q_h)=0$

**Taylor-Hood** ($P2/P1$):
- Velocity: $(P2)^d$, Pressure: $P1$
- Inf-sup proved via macro-element technique or Fortin operator

**P1-P1 is unstable:** dim$(Q_h)>$ dim$(V_h)$ effectively $\Rightarrow$ $B$ not surjective $\Rightarrow$ inf-sup fails

## H. Key Results to Prove

| Result | Proof strategy |
|--------|---------------|
| P1 unisolvence | Restrict to edge $\to$ 2 zeros on linear $\to$ hyperplane lemma $\to$ vertex condition |
| P2 unisolvence | 3+ zeros per edge $\to$ iterated hyperplane lemma $\to$ midpoint condition |
| Weak form from strong | Multiply, integrate, IBP, interior edges cancel, BCs |
| Strong form from weak | Test with $C_0^\infty$ to get PDE, test boundary to get Neumann BC |
| Riesz representation | $u=$ limit of $L(e_n)e_n$ (ONB), parallelogram law |
| Lax-Milgram | Fix $u$; $v\mapsto b(u,v)$ is bounded linear $\to$ Riesz $\to$ contraction map |
| Mean estimate | $u-\bar{u}=$ function of $Du$ via change of variables $\Sigma:K\to K$ |
| Trace theorem | Construct Argyris $\xi$ with $\xi|_{\partial\Omega}=v^2$, FE divergence theorem |
| Céa's lemma | Galerkin orthogonality + coercivity/continuity (3-line proof) |
| Taylor error | Change of variables integral bound, Sobolev's inequality |
| Scaling argument | Change of variables $x=dy$, collect powers of $d$ |
| Aubin-Nitsche | Dual problem + Galerkin orthogonality + elliptic regularity |
| Brezzi's theorem | Surjectivity of $B$ (inf-sup) + Lax-Milgram on kernel $Z$ |
| Fortin's trick | $b(\Pi_hv,q_h)=b(v,q_h)$ $\to$ transfer continuous inf-sup to discrete |

## I. Formulae Cheat Sheet

### Norms & Spaces
$$\|u\|_{L^2}=\left(\int|u|^2\right)^{1/2}, \quad |u|_{H^k}=\left(\sum_{|\alpha|=k}\int|D^\alpha u|^2\right)^{1/2}, \quad \|u\|_{H^k}=\left(\sum_{j=0}^{k}|u|_{H^j}^2\right)^{1/2}$$

### Convergence Rates (degree $k$ Lagrange)
| Norm | Error | Rate |
|------|-------|------|
| $H^1$ | $\|u-u_h\|_{H^1}\leq Ch^k|u|_{H^{k+1}}$ | $O(h^k)$ |
| $L^2$ | $\|u-u_h\|_{L^2}\leq Ch^{k+1}|u|_{H^{k+1}}$ | $O(h^{k+1})$ |
| Interpolation | $\|u-\mathcal{I}_hu\|_{H^i}\leq Ch^{k+1-i}|u|_{H^{k+1}}$ | $O(h^{k+1-i})$ |

**P1**: $O(h)$ in $H^1$, $O(h^2)$ in $L^2$.  
**P2**: $O(h^2)$ in $H^1$, $O(h^3)$ in $L^2$.

### Key Constants
| Symbol | Meaning |
|--------|--------|
| $M$ | Continuity constant of $b$ |
| $\gamma$ | Coercivity constant of $b$ |
| $\beta$ | Inf-sup constant |
| $C_\Pi$ | Fortin operator bound |
| $h$ | Maximum element diameter |

### Inequalities
- **Schwarz**: $|(u,v)|\leq\|u\|\|v\|$
- **Poincaré**: $\|u\|_{L^2}\leq C|u|_{H^1}$ for $u\in\mathring{H}^1$
- **Mean estimate**: $\|u-\bar{u}\|_{L^2}\leq C|u|_{H^1}$
- **Trace**: $\|u\|_{L^2(\partial\Omega)}\leq C\|u\|_{H^1(\Omega)}$
- **Sobolev**: $k>n/2$ $\Rightarrow$ $\|u\|_{L^\infty}\leq C\|u\|_{H^k}$
- **Korn**: $\|\varepsilon(u)\|^2\geq\frac{1}{2}|u|_{H^1}^2$ on $\mathring{H}^1$

### Stokes
$$a(u,v)=2\mu\int\varepsilon(u):\varepsilon(v), \quad b(v,q)=-\int q\,\nabla\cdot v$$
$$\varepsilon(u)=\tfrac{1}{2}(\nabla u+\nabla u^T)$$

## J. Exam Strategy

### Five-Step Convergence Analysis Programme
1. **Well-posedness of strong form** — classical PDE theory / elliptic regularity
2. **Derive weak form** — IBP, BCs, identify $b$, $F$, $V$
3. **Show equivalence** strong $\iff$ weak
4. **Lax-Milgram / Brezzi** — continuity, coercivity (+ inf-sup for mixed)
5. **Céa + interpolation** — choose best approximation, apply error estimates

### Common Question Types
- Derive weak form for a given PDE with given BCs
- Prove unisolvence for a given element
- Compute convergence rates for given element degree
- Verify Brezzi conditions for a given element pair
- Prove a key inequality (Poincaré, trace, mean estimate)

### Key Things to Remember
- Interior edges cancel in weak form derivation (opposite normals)
- Neumann BCs are *natural* (built into $F$); Dirichlet are *essential* (built into $V$)
- Lax-Milgram needs coercivity on the **whole space**; Brezzi needs it on the **kernel** only
- P1-P1 unstable for Stokes; need enrichment (bubble) or higher order (Taylor-Hood)
- $L^2$ rate = $H^1$ rate $+1$ (from Aubin-Nitsche), requires elliptic regularity
- When proving coercivity for Poisson, you need an inequality to promote $|\cdot|_{H^1}$ to $\|\cdot\|_{H^1}$
- Fortin: construct $\Pi_h$ preserving $b$, then discrete inf-sup follows
