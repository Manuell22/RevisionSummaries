## Conceptual Flow

```
Strong PDE (e.g. -∇²u = f)
   │  multiply by test function v, integrate by parts
   ▼
Variational Problem: find u ∈ V s.t. b(u,v)=F(v) ∀v ∈ V
   │  restrict to V_h ⊂ V (Galerkin)
   ▼
FE Approximation: find u_h ∈ V_h s.t. b(u_h,v)=F(v) ∀v ∈ V_h
   │  expand in nodal basis → matrix system Ku = f
   ▼
Solve linear system
   │
   ▼
Error analysis:
   Lax-Milgram (existence) → Céa's lemma (best approx) → Interpolation error (h^k rates)
   │
   ├── H¹ error: ||u-u_h||_{H¹} ≤ Ch^k |u|_{H^{k+1}}      [Céa + interpolation]
   └── L² error: ||u-u_h||_{L²} ≤ Ch^{k+1}|u|_{H^{k+1}}   [Aubin-Nitsche duality]
```

## Core Theorems & Lemmas

| Result | Statement (short) | Where Used |
|--------|-------------------|------------|
| **Unisolvence (Dual condition)** | $\mathcal{N}$ determines $\mathcal{P}$ $\iff$ Vandermonde $V$ invertible $\iff$ $N_i(v)=0\;\forall i\Rightarrow v\equiv 0$ | Verifying FE well-defined |
| **Sobolev's inequality** | $k>n/2\Rightarrow\|u\|_{L^\infty}\leq C\|u\|_{H^k}$ | Bounding pointwise from integrals |
| **Schwarz inequality** | $|(u,v)|\leq\|u\|\|v\|$ | Continuity proofs |
| **Riesz representation** | $\forall L\in H'\;\exists u\in H$ s.t. $L(v)=(u,v)$ | Dual space structure |
| **Lax-Milgram** | Continuous + coercive $b$ + continuous $F$ $\Rightarrow$ unique $u$, $\|u\|\leq\frac{1}{\gamma}\|F\|_{V'}$ | Existence & uniqueness of FE solutions |
| **Céa's lemma** | $\|u-u_h\|\leq\frac{M}{\gamma}\min_{v\in V_h}\|u-v\|$ | Reduces error to best approximation |
| **Mean estimate** | $\|u-\bar{u}\|_{L^2}\leq C|u|_{H^1}$ | Coercivity for Neumann Poisson |
| **Trace theorem** | $\|u\|_{L^2(\partial\Omega)}\leq C\|u\|_{H^1(\Omega)}$ | Boundary terms control |
| **Poincaré inequality** | $\|u\|_{L^2}\leq C|u|_{H^1}$ for $u|_{\Gamma_0}=0$ | Coercivity for Dirichlet Poisson |
| **Elliptic regularity** | $|w|_{H^2}\leq C\|f\|_{L^2}$ on convex domains | Aubin-Nitsche trick |
| **Brezzi's theorem** | Coercivity on kernel $Z$ + inf-sup $\Rightarrow$ unique $(u,p)$ + stability bounds | Mixed problems (Stokes) |
| **Fortin's trick** | $\exists\Pi_h:V\to V_h$ preserving $b$ + bounded $\Rightarrow$ discrete inf-sup | Proving inf-sup for element pairs |

## Important Definitions

| Term | Definition |
|------|----------|
| Triangulation $\mathcal{T}$ | Non-overlapping triangles covering $\bar{\Omega}$, no hanging nodes |
| Ciarlet element | Triple $(K,\mathcal{P},\mathcal{N})$: domain, shape functions, nodal variables (basis for $\mathcal{P}'$) |
| Nodal basis | $\{\phi_j\}$ s.t. $N_i(\phi_j)=\delta_{ij}$ |
| Unisolvence | $\mathcal{N}$ determines $\mathcal{P}$ (only $v\equiv0$ satisfies $N_i(v)=0\;\forall i$) |
| Pk element | Lagrange element of degree $k$ on triangles |
| $C^m$ geometric decomposition | Nodal vars assigned to mesh entities s.t. $C^m$ continuity enforced between elements |
| Bilinear form | $b:V\times V\to\mathbb{R}$, linear in each argument |
| Continuous b.f. | $|b(u,v)|\leq M\|u\|\|v\|$ |
| Coercive b.f. | $b(u,u)\geq\gamma\|u\|^2$ |
| Hilbert space | Complete inner product space |
| Weak derivative | $\int_\Omega\phi D_w^\alpha f\,dx=(-1)^{|\alpha|}\int_\Omega D^\alpha\phi\,f\,dx$ $\forall\phi\in C_0^\infty$ |
| Sobolev space $H^k$ | $\{u\in L^1_{loc}: \|u\|_{H^k}<\infty\}$ |
| Galerkin approximation | Restrict variational problem to $V_h\subset V$ |
| Inf-sup condition | $\inf_q\sup_v b(v,q)/(\|v\|\|q\|)\geq\beta>0$ |

## Exam Notes

### Convergence Rates to Remember
- **Pk in $H^1$**: error $\sim O(h^k)$
- **Pk in $L^2$**: error $\sim O(h^{k+1})$ (one order better, via Aubin-Nitsche)
- Interpolation error: $\|\mathcal{I}_hu-u\|_{H^i}\leq Ch^{k+1-i}|u|_{H^{k+1}}$

### Analysis Pipeline (5-step programme)
1. Derive weak/variational form of PDE (multiply by test fn, integrate by parts)
2. Choose FE space $V_h\subset H^k$ ($C^0$ for $H^1$, $C^1$ for $H^2$)
3. Galerkin approximation: restrict trial & test to $V_h$
4. Show continuity + coercivity → Lax-Milgram → existence & stability
5. Céa's lemma + interpolation error → convergence rate

### L6 Stokes (Mastery)
- Not coercive globally → use Brezzi's theorem instead of Lax-Milgram
- Need inf-sup condition: not all $V_h\times Q_h$ pairs work (P1-P1 fails!)
- Working pairs: MINI (P1+B3 / P1), Taylor-Hood (P2 / P1)
- Fortin's trick: construct $\Pi_h$ to verify discrete inf-sup

## The Module at a Glance — An Intuitive Walkthrough

This module tells one continuous story: **how to approximately solve PDEs on complicated domains, and how to rigorously prove the approximation is good.** Each lecture builds a layer of the theory, and the whole arc can be understood as six chapters of a single argument.

---

### L1 — Introduction: *"Turn a PDE into a linear system"*

We start with a PDE we can't solve exactly — Poisson's equation $-\nabla^2 u = f$ on a square. The key move is *not* to try to satisfy the equation pointwise (our approximate solutions won't be smooth enough for that). Instead, we multiply both sides by a test function $v$, integrate over the domain, and use integration by parts to shift one derivative from $u$ onto $v$. This produces the **weak form**: $\int \nabla v \cdot \nabla u = \int v f$, which only requires first derivatives — something piecewise-linear functions can handle.

We chop the domain into triangles (a **triangulation**), build the space $V_h$ of continuous piecewise-linear functions on that mesh, and demand the weak form holds for every $v \in V_h$. Expanding the unknown $u_h$ in a basis of "tent functions" (one per node, peaking at 1 and falling to 0 at neighbours) turns the whole thing into a sparse linear system $K\mathbf{u} = \mathbf{f}$ that a computer can solve. This lecture is the practical blueprint — everything that follows makes it rigorous.

---

### L2 — Finite Element Spaces: *"How to build the building blocks"*

L1 used P1 (linear) elements on triangles, but there's a whole zoo of finite elements. This lecture gives the **abstract recipe**: Ciarlet's triple $(K, \mathcal{P}, \mathcal{N})$ — a shape, a polynomial space, and a set of degrees of freedom (DOFs) that uniquely determine a function in that space. The DOFs are measurements like "value at vertex $z_i$" or "average over edge $e_j$."

The central concept is **unisolvence**: the DOFs must be rich enough that the only polynomial giving zero on all of them is the zero polynomial. This is what makes the element well-posed — it guarantees you can reconstruct a unique polynomial from measured data. The Vandermonde matrix $V_{ij} = N_j(\psi_i)$ is the computational face of this: unisolvence $\iff$ $V$ is invertible.

We see concrete examples — P1 (linear on triangles, 3 DOFs), P2 (quadratic, 6 DOFs), Hermite (cubics with gradient data, 10 DOFs), Argyris (quintics with Hessian data, 21 DOFs for $C^1$). The lecture then explains how to glue local elements into a **global space**: assign each DOF to a mesh entity (vertex, edge, or cell), and identify shared DOFs across adjacent elements to enforce continuity. Lagrange elements share vertex/edge values → $C^0$; Argyris shares derivatives too → $C^1$.

---

### L3 — Interpolation: *"How well can polynomials approximate smooth functions?"*

Now we ask a purely approximation-theoretic question: given a smooth function $u$, how close is its interpolant $\mathcal{I}_h u$ (the element-by-element polynomial that matches $u$ at the DOFs) to $u$ itself?

The answer is the **interpolation error estimate** $\|\mathcal{I}_h u - u\|_{H^i} \leq C h^{k+1-i} |u|_{H^{k+1}}$. In words: using degree-$k$ polynomials on a mesh of size $h$, the error in the $H^i$ norm shrinks like $h^{k+1-i}$. Halving the mesh size with P1 elements cuts $H^1$ error by half and $L^2$ error by a factor of four.

The proof works in four clean steps: (1) bound the interpolator on a reference triangle of diameter 1, (2) show the interpolation *error* on that reference triangle is controlled by the $(k+1)$-th Sobolev seminorm (using the averaged Taylor polynomial as an intermediary — the interpolator reproduces polynomials exactly, so only the non-polynomial part of $u$ contributes to error), (3) scale everything to a triangle of diameter $d$ by a change of variables, collecting powers of $d$, and (4) sum over all elements, replacing $d$ by $h$.

This lecture is purely about approximation — it says nothing about PDEs yet. But it provides the ammunition Céa's lemma will need in L5.

---

### L4 — Solvability & Stability: *"Does the discrete problem even have a solution?"*

Before asking whether $u_h$ converges to $u$, we should ask: does $u_h$ exist? Is it unique? Is it well-behaved as $h \to 0$?

The answer comes from abstract functional analysis. We formalise the weak form as: find $u \in V$ such that $b(u,v) = F(v)$ for all $v \in V$, where $b$ is a bilinear form and $F$ is a linear functional. The **Lax-Milgram theorem** says: if $b$ is *continuous* ($|b(u,v)| \leq M\|u\|\|v\|$, it doesn't blow up) and *coercive* ($b(u,u) \geq \gamma\|u\|^2$, it's bounded below), then there is exactly one solution $u$ and $\|u\| \leq \frac{1}{\gamma}\|F\|$.

Continuity says the bilinear form is under control; coercivity says it can't collapse to zero for nonzero inputs. Together, they make $b$ behave like an inner product (up to constants). For the Helmholtz equation $u - \nabla^2 u = f$, the bilinear form literally *is* the $H^1$ inner product, so $M = \gamma = 1$ — the perfect case. For Poisson, coercivity requires more work: you need the Poincaré inequality (Dirichlet BCs) or the mean estimate (Neumann BCs) to control $\|u\|_{L^2}$ by $|u|_{H^1}$.

**Stability** means the coercivity constant $\gamma_h$ stays positive as $h \to 0$ — the discrete problems don't degenerate. This is what guarantees uniformly bounded solutions.

---

### L5 — Convergence: *"The FE solution is nearly the best approximation"*

This is the payoff lecture. We now prove the FE solution $u_h$ actually converges to the true solution $u$, and quantify the rate.

First, we formalise **weak derivatives** and **Sobolev spaces** $H^k$ — the proper functional-analytic setting where everything lives. The key insight is that the FE derivative (piecewise, defined on each element) coincides with the weak derivative for continuous FE functions, so the FE world sits cleanly inside the Sobolev world.

The centrepiece is **Céa's lemma**: $\|u - u_h\|_V \leq \frac{M}{\gamma}\min_{v \in V_h}\|u - v\|_V$. The FE error is within a constant factor of the *best possible* approximation in $V_h$. This decouples the PDE analysis (Lax-Milgram, coercivity) from the approximation theory (how well can $V_h$ approximate $u$?). The proof uses **Galerkin orthogonality** — the error $u - u_h$ is $b$-orthogonal to $V_h$, which is the variational analogue of orthogonal projection.

Plugging in the interpolation estimate from L3 gives:
- **$H^1$ error**: $O(h^k)$ for degree-$k$ elements
- **$L^2$ error**: $O(h^{k+1})$ via the **Aubin-Nitsche trick** — a clever duality argument using an adjoint problem that squeezes out one extra power of $h$. This is the most elegant result in the module: the $L^2$ error is always one order better than $H^1$, for free.

---

### L6 — Stokes: *"What happens when Lax-Milgram breaks down"*

The final lecture is the mastery topic, and it's where everything gets harder. The Stokes equations model viscous incompressible flow: find velocity $u$ and pressure $p$ satisfying $-2\mu\nabla\cdot\varepsilon(u) + \nabla p = f$ with $\nabla \cdot u = 0$. The incompressibility constraint $\nabla \cdot u = 0$ couples $u$ and $p$, and pressure acts as a Lagrange multiplier enforcing it.

The combined variational form is a **saddle-point problem**, and it fails to be coercive — pressure appears linearly but never quadratically, so you can always find a direction in $(u,p)$-space where the form vanishes. Lax-Milgram is dead; we need new tools.

**Brezzi's theorem** replaces Lax-Milgram: require (1) coercivity of the viscous term $a(u,v)$ only on the divergence-free subspace (where the constraint is already satisfied), and (2) the **inf-sup condition** for the divergence coupling $b(v,q)$ — every nonzero pressure must be "visible" to some velocity through $b$. If both hold, there is a unique solution with stability bounds.

The dramatic twist: unlike coercivity, the inf-sup condition does **not** automatically pass from $V \times Q$ to $V_h \times Q_h$. A bad choice of element pair can have $\beta_h \to 0$, producing spurious pressure oscillations. The simplest pairing P1-P1 (linear velocity, linear pressure) **fails** — there are too many pressure DOFs, and some pressure modes are invisible to the velocity space. The fix is to either enrich velocity (MINI element: P1 + cubic bubbles) or increase its degree (Taylor-Hood: P2 velocity / P1 pressure). **Fortin's trick** provides a practical tool to verify discrete inf-sup: construct a bounded projection $\Pi_h: V \to V_h$ that preserves the divergence pairing.

---

### The Big Picture

The module moves from **construction** (L1–L2: build the discrete spaces) through **approximation** (L3: how well can polynomials fit smooth functions?) to **analysis** (L4–L5: prove existence, uniqueness, and convergence rates) and finally **extension** (L6: what changes for coupled/constrained systems?). The entire theory rests on three pillars:

1. **Variational formulation** — recast the PDE so it only requires weak derivatives
2. **Polynomial approximation** — FE spaces can approximate $H^{k+1}$ functions at rate $h^{k+1-i}$ in $H^i$
3. **Stability** — Lax-Milgram (or Brezzi) ensures the discrete solve doesn't ruin the approximation

If any one of these fails — wrong weak form, bad element choice, loss of coercivity or inf-sup — the method breaks down. The course teaches you to build all three correctly.
