## Conceptual Flow

```
Strong PDE (e.g. -∇²u = f)
   │  multiply by test function v, integrate by parts
   ▼
Variational Problem: find u ∈ V s.t. b(u,v)=F(v) ∀v ∈ V
   │  restrict to V_h ⊂ V (Galerkin)
   ▼
FE Approximation: find u_h ∈ V_h s.t. b(u_h,v)=F(v) ∀v ∈ V_h
   │  expand in nodal basis → matrix system Ku = f
   ▼
Solve linear system
   │
   ▼
Error analysis:
   Lax-Milgram (existence) → Céa's lemma (best approx) → Interpolation error (h^k rates)
   │
   ├── H¹ error: ||u-u_h||_{H¹} ≤ Ch^k |u|_{H^{k+1}}      [Céa + interpolation]
   └── L² error: ||u-u_h||_{L²} ≤ Ch^{k+1}|u|_{H^{k+1}}   [Aubin-Nitsche duality]
```

## Key Equations

| Equation | Formula |
|----------|--------|
| Poisson (strong) | $-\nabla^2 u = f$ |
| Poisson (weak) | $\int_\Omega \nabla v\cdot\nabla u_h\,dx = \int_\Omega vf\,dx$ |
| Helmholtz (weak) | $\int_\Omega uv + \nabla u\cdot\nabla v\,dx = \int_\Omega vf\,dx$ |
| Stiffness matrix | $K_{ij}=\int_\Omega \nabla\phi_i\cdot\nabla\phi_j\,dx$ |
| Lagrange basis (1D) | $\phi_i(x)=\prod_{j\neq i}\frac{x-x_j}{x_i-x_j}$ |
| Vandermonde | $V_{ij}=N_j(\psi_i)$, $\mu=V^{-1}$ gives nodal basis |
| $H^k$ seminorm | $|u|_{H^k}^2=\sum_{|\alpha|=k}\int_\Omega|D^\alpha u|^2\,dx$ |
| $H^k$ norm | $\|u\|_{H^k}^2=\sum_{i=0}^k|u|_{H^i}^2$ |
| Sobolev inequality | $\|u\|_{L^\infty}\leq C\|u\|_{H^k}$ for $k>n/2$ |
| Taylor poly error | $\|D^\beta(f-Q_{k,B}f)\|_{L^2}\leq C\frac{|\Omega|^{1/2}}{|B|^{1/2}}d^{k+1-|\beta|}|f|_{H^{k+1}}$ |
| Interpolation error (local) | $|\mathcal{I}_Ku-u|_{H^i(K)}\leq C_Kd^{k+1-i}|u|_{H^{k+1}(K)}$ |
| Interpolation error (global) | $\|\mathcal{I}_hu-u\|_{H^i}\leq Ch^{k+1-i}|u|_{H^{k+1}}$ |
| Lax-Milgram bound | $\|u\|_V\leq\frac{1}{\gamma}\|F\|_{V'}$ |
| Céa's lemma | $\|u-u_h\|_V\leq\frac{M}{\gamma}\min_{v\in V_h}\|u-v\|_V$ |
| $H^1$ convergence (Pk) | $\|u_h-u\|_{H^1}\leq Ch^k\|u\|_{H^{k+1}}$ |
| $L^2$ convergence (Pk) | $\|u_h-u\|_{L^2}\leq Ch^{k+1}\|u\|_{H^{k+1}}$ |
| Inf-sup condition | $\inf_{q\in Q}\sup_{v\in V}\frac{b(v,q)}{\|v\|_V\|q\|_Q}\geq\beta$ |
| Stokes (strong) | $-2\mu\nabla\cdot\varepsilon(u)+\nabla p=f$, $\nabla\cdot u=0$ |

## Core Theorems & Lemmas

| Result | Statement (short) | Where Used |
|--------|-------------------|------------|
| **Unisolvence (Dual condition)** | $\mathcal{N}$ determines $\mathcal{P}$ $\iff$ Vandermonde $V$ invertible $\iff$ $N_i(v)=0\;\forall i\Rightarrow v\equiv 0$ | Verifying FE well-defined |
| **Sobolev's inequality** | $k>n/2\Rightarrow\|u\|_{L^\infty}\leq C\|u\|_{H^k}$ | Bounding pointwise from integrals |
| **Schwarz inequality** | $|(u,v)|\leq\|u\|\|v\|$ | Continuity proofs |
| **Riesz representation** | $\forall L\in H'\;\exists u\in H$ s.t. $L(v)=(u,v)$ | Dual space structure |
| **Lax-Milgram** | Continuous + coercive $b$ + continuous $F$ $\Rightarrow$ unique $u$, $\|u\|\leq\frac{1}{\gamma}\|F\|_{V'}$ | Existence & uniqueness of FE solutions |
| **Céa's lemma** | $\|u-u_h\|\leq\frac{M}{\gamma}\min_{v\in V_h}\|u-v\|$ | Reduces error to best approximation |
| **Mean estimate** | $\|u-\bar{u}\|_{L^2}\leq C|u|_{H^1}$ | Coercivity for Neumann Poisson |
| **Trace theorem** | $\|u\|_{L^2(\partial\Omega)}\leq C\|u\|_{H^1(\Omega)}$ | Boundary terms control |
| **Poincaré inequality** | $\|u\|_{L^2}\leq C|u|_{H^1}$ for $u|_{\Gamma_0}=0$ | Coercivity for Dirichlet Poisson |
| **Elliptic regularity** | $|w|_{H^2}\leq C\|f\|_{L^2}$ on convex domains | Aubin-Nitsche trick |
| **Brezzi's theorem** | Coercivity on kernel $Z$ + inf-sup $\Rightarrow$ unique $(u,p)$ + stability bounds | Mixed problems (Stokes) |
| **Fortin's trick** | $\exists\Pi_h:V\to V_h$ preserving $b$ + bounded $\Rightarrow$ discrete inf-sup | Proving inf-sup for element pairs |

## Important Definitions

| Term | Definition |
|------|----------|
| Triangulation $\mathcal{T}$ | Non-overlapping triangles covering $\bar{\Omega}$, no hanging nodes |
| Ciarlet element | Triple $(K,\mathcal{P},\mathcal{N})$: domain, shape functions, nodal variables (basis for $\mathcal{P}'$) |
| Nodal basis | $\{\phi_j\}$ s.t. $N_i(\phi_j)=\delta_{ij}$ |
| Unisolvence | $\mathcal{N}$ determines $\mathcal{P}$ (only $v\equiv0$ satisfies $N_i(v)=0\;\forall i$) |
| Pk element | Lagrange element of degree $k$ on triangles |
| $C^m$ geometric decomposition | Nodal vars assigned to mesh entities s.t. $C^m$ continuity enforced between elements |
| Bilinear form | $b:V\times V\to\mathbb{R}$, linear in each argument |
| Continuous b.f. | $|b(u,v)|\leq M\|u\|\|v\|$ |
| Coercive b.f. | $b(u,u)\geq\gamma\|u\|^2$ |
| Hilbert space | Complete inner product space |
| Weak derivative | $\int_\Omega\phi D_w^\alpha f\,dx=(-1)^{|\alpha|}\int_\Omega D^\alpha\phi\,f\,dx$ $\forall\phi\in C_0^\infty$ |
| Sobolev space $H^k$ | $\{u\in L^1_{loc}: \|u\|_{H^k}<\infty\}$ |
| Galerkin approximation | Restrict variational problem to $V_h\subset V$ |
| Inf-sup condition | $\inf_q\sup_v b(v,q)/(\|v\|\|q\|)\geq\beta>0$ |

## Exam Notes

### Convergence Rates to Remember
- **Pk in $H^1$**: error $\sim O(h^k)$
- **Pk in $L^2$**: error $\sim O(h^{k+1})$ (one order better, via Aubin-Nitsche)
- Interpolation error: $\|\mathcal{I}_hu-u\|_{H^i}\leq Ch^{k+1-i}|u|_{H^{k+1}}$

### Analysis Pipeline (5-step programme)
1. Derive weak/variational form of PDE (multiply by test fn, integrate by parts)
2. Choose FE space $V_h\subset H^k$ ($C^0$ for $H^1$, $C^1$ for $H^2$)
3. Galerkin approximation: restrict trial & test to $V_h$
4. Show continuity + coercivity → Lax-Milgram → existence & stability
5. Céa's lemma + interpolation error → convergence rate

### L6 Stokes (Mastery)
- Not coercive globally → use Brezzi's theorem instead of Lax-Milgram
- Need inf-sup condition: not all $V_h\times Q_h$ pairs work (P1-P1 fails!)
- Working pairs: MINI (P1+B3 / P1), Taylor-Hood (P2 / P1)
- Fortin's trick: construct $\Pi_h$ to verify discrete inf-sup
