## Block 1 — 12:00–13:00 — L1: Introduction to Finite Elements (~600 lines)

**Lecture notes:** `LecNotes/L1_introduction.txt`

### Poisson's equation (strong form)

$$-\nabla^2 u = f \quad \text{in } \Omega, \qquad u = g \text{ on } \Gamma_D, \qquad \nabla u \cdot n = h \text{ on } \Gamma_N$$

**Key definitions to note:**
- **Triangulation** of a polygonal domain $\Omega$: non-overlapping triangles covering $\Omega$, shared whole edges/vertices
- **P1 FE space:** $V_h = \{v \in C^0(\Omega) : v|_K \in P_1 \;\forall K \in \mathcal{T}_h\}$
- $\mathring{V}_h = \{v \in V_h : v|_{\Gamma_D} = 0\}$ (strong Dirichlet BCs)

### L² spaces and FE derivative

- $L^2(\Omega) = \{f : \int_\Omega |f|^2 < \infty\}$ — functions equal a.e. are identified
- **FE partial derivative:** defined as the $L^2$ function that equals the classical derivative on each triangle interior
- Key: FE derivative exists for $C^0$ piecewise smooth functions even though classical derivative may not exist on edges

### Weak formulation of Poisson

Integration by parts on each triangle → interior edge integrals cancel for $C^0$ functions:

$$\text{Find } u_h \in V_h \text{ s.t. } \int_\Omega \nabla u_h \cdot \nabla v_h = \int_\Omega f v_h + \int_{\Gamma_N} h v_h \quad \forall v_h \in \mathring{V}_h$$

### Practical implementation

- **Nodal basis:** $\phi_i(z_j) = \delta_{ij}$ for mesh vertices $z_j$
- Expand $u_h = \sum_j u_j \phi_j$ → **stiffness matrix** $K_{ij} = \int_\Omega \nabla\phi_i \cdot \nabla\phi_j$, **load vector** $f_i = \int_\Omega f\phi_i + \int_{\Gamma_N} h\phi_i$
- Assembly: $K = \sum_K K^{(K)}$ element-by-element; $K$ is **sparse** (bandwidth $\sim \sqrt{N}$ in 2D)

### 🧠 Key things to understand deeply
1. Why do interior edge integrals cancel in the integration by parts? (Continuity of $C^0$ FE functions + opposite normals)
2. Why is $K$ sparse? (Nodal basis functions have small support)
3. How does the strong form relate to the FE weak form? (Test with $v_h$, integrate by parts)

## Block 2 — 13:10–14:10 — L2 §§1–4: Ciarlet's FE, Vandermonde, 2D/3D Lagrange (~555 lines)

**Lecture notes:** `LecNotes/L2_fespaces.txt`, lines 4–575

---

### Ciarlet's finite element

**Definition (Ciarlet triple):** A finite element is $(K, \mathcal{P}, \mathcal{N})$ where:
- $K$ = cell (triangle, tetrahedron, etc.)
- $\mathcal{P}$ = polynomial space on $K$ (e.g. $P_k(K)$, dim $= n$)
- $\mathcal{N} = \{N_1, \ldots, N_n\}$ = nodal variables (functionals on $\mathcal{P}$, elements of dual space)

**Nodal basis:** $\{\psi_1, \ldots, \psi_n\} \subset \mathcal{P}$ such that $N_i(\psi_j) = \delta_{ij}$.

### 1D Lagrange element of degree $k$

- $K = [0, 1]$, $\mathcal{P} = P_k$, nodes at $x_i = i/k$, $i = 0, \ldots, k$
- $N_i(f) = f(x_i)$ (point evaluation)
- Nodal basis = Lagrange polynomials: $\psi_i(x) = \prod_{j \neq i} \frac{x - x_j}{x_i - x_j}$

### Vandermonde matrix and unisolvence

- **Vandermonde matrix** $V_{ij} = p_j(x_i)$ where $\{p_j\}$ is a basis of $\mathcal{P}$ and $\{x_i\}$ are nodes
- **Lemma:** Nodal basis coefficients come from $V^{-1}$
- **Unisolvence:** $\mathcal{N}$ determines elements of $\mathcal{P}$ uniquely $\iff$ $V$ is invertible $\iff$ ($N_i(f) = 0 \;\forall i \implies f = 0$)

### 2D/3D Lagrange elements

- **Lemma (polynomial vanishing on hyperplane):** If $p \in P_k$ vanishes on a hyperplane $H$, then $p = \ell_H \cdot q$ where $\ell_H$ is the affine function defining $H$ and $q \in P_{k-1}$
- **P1 on triangle:** 3 nodes (vertices), $\dim P_1 = 3$; unisolvence by vanishing argument
- **P2 on triangle:** 6 nodes (vertices + edge midpoints), $\dim P_2 = 6$; unisolvence by repeated vanishing on edges
- **Pk on triangle:** $\binom{k+2}{2}$ nodes (equally spaced lattice points); unisolvence by induction on $k$

### 🧠 Key things to understand deeply
1. What are the three components of Ciarlet's triple and what role does each play?
2. Why does unisolvence reduce to invertibility of the Vandermonde matrix?
3. The inductive vanishing argument for $P_k$ unisolvence — understand the logic, not just the result

## Block 3 — 14:10–15:10 — L2 §§5–6 + L3 §1: Exotic Elements, Global Continuity, Interpolation Operators (~435 lines)

**Lecture notes:** `L2_fespaces.txt` lines 575–915 + `L3_interpolation.txt` lines 4–75

---

### Exotic elements (L2 §5)

- **Cubic Hermite element:** $(K, P_3, \mathcal{N})$ where nodes include function values AND derivative values at vertices. Gives $C^1$ global spaces.
- **Quintic Argyris element:** $(K, P_5, \mathcal{N})$ — nodes include values, all 1st and 2nd derivatives at vertices + normal derivative at edge midpoints. Gives $C^2$ global spaces for 4th-order problems.

**Key contrast:** Lagrange elements use only point evaluation nodes; Hermite/Argyris use derivative evaluations → higher global smoothness.

### Global continuity (L2 §6)

**Definition (FE space):** Given triangulation $\mathcal{T}_h$ and local elements $(K, \mathcal{P}_K, \mathcal{N}_K)$, the FE space is $V_h = \{v : v|_K \in \mathcal{P}_K\; \forall K,\; \text{+ continuity conditions}\}$.

**Geometric decomposition (key concept):**
- **Local geometric decomposition:** Assign each DOF to a sub-entity of $K$ (vertex, edge, face, interior)
- **$C^m$ geometric decomposition:** DOFs on entity $e$ depend only on $f|_e$ → functions sharing entity DOF values are $C^m$ across that entity
- **Global $C^m$ decomposition** → **$C^m$ FE space**

**Discontinuous FE space ($P_k$ DG):** No continuity across edges. All DOFs are interior. Used in DG methods.

**Lemma:** If the local elements admit a $C^m$ geometric decomposition, then the global FE space is $C^m$.

### Interpolation operators (L3 §1)

- **Local interpolator:** $\mathcal{I}_K : C(K) \to \mathcal{P}_K$ defined by $N_i(\mathcal{I}_K f) = N_i(f)$ for all nodal variables
- Properties: linear, preserves polynomials ($\mathcal{I}_K p = p$ for $p \in \mathcal{P}_K$), **projection** ($\mathcal{I}_K^2 = \mathcal{I}_K$)
- **Global interpolator:** $\mathcal{I}_h : C(\Omega) \to V_h$ defined element-by-element; well-defined if FE space is $C^0$

### 🧠 Key things to understand deeply
1. Why does the geometric decomposition determine global smoothness?
2. Why is $\mathcal{I}_K$ a projection? (Because $\mathcal{I}_K p = p$ for $p \in \mathcal{P}_K$)
3. What goes wrong with the global interpolator if the space is not $C^0$?

1. What extra nodal variables does Hermite use compared to Lagrange?
2. State the definition of a $C^m$ geometric decomposition.

## Block 4 — 15:30–16:30 — L3 §§2–4: H^k Norms, Averaged Taylor Polynomials, Interpolation Errors (~735 lines)

**Lecture notes:** `LecNotes/L3_interpolation.txt`, lines 75–810

---

### Measuring interpolation errors: H^k norms (L3 §2)

**Multi-index notation:** $\alpha = (\alpha_1, \ldots, \alpha_d)$, $|\alpha| = \sum \alpha_i$, $D^\alpha = \partial^{|\alpha|}/\partial x_1^{\alpha_1}\cdots\partial x_d^{\alpha_d}$

**Sobolev seminorm and norm:**
$$|f|_{H^k(\Omega)} = \left(\sum_{|\alpha|=k} \|D^\alpha f\|_{L^2}^2\right)^{1/2}, \qquad \|f\|_{H^k(\Omega)} = \left(\sum_{j=0}^k |f|_{H^j}^2\right)^{1/2}$$

**Theorem (Sobolev's inequality):** For $k > d/2$, $H^k(\Omega) \hookrightarrow C(\bar\Omega)$: every $H^k$ function is continuous and $\|f\|_\infty \leq C\|f\|_{H^k}$.

### Averaged Taylor polynomials (L3 §3)

**Definition:** $Q_{k,B}f(x) = \frac{1}{|B|}\int_B T_k f(x; y)\,dy$ where $T_k f(x;y) = \sum_{|\alpha|\leq k} \frac{D^\alpha f(y)}{\alpha!}(x-y)^\alpha$ is the Taylor polynomial centred at $y$, averaged over a ball $B$.

**Theorem (Taylor error estimate):** For $f \in H^{k+1}(\Omega)$, triangle $K$ with diameter $d_K$, inscribed ball $B_K$:

$$|f - Q_{k,B_K}f|_{H^i(K)} \leq C\left(\frac{d_K}{\rho_K}\right)^d d_K^{k+1-i} |f|_{H^{k+1}(K)}, \quad 0 \leq i \leq k+1$$

where $\rho_K$ = radius of inscribed ball. The ratio $d_K/\rho_K$ measures **shape regularity**.

### Local and global interpolation errors (L3 §4)

**Key results (chain of lemmas):**
1. **Bound on $\|\mathcal{I}_{K_1}u\|_{H^k}$:** interpolator on unit element is bounded
2. **Interpolation error on unit triangle:** $|u - \mathcal{I}_{K_1}u|_{H^i} \leq C|u|_{H^{k+1}}$
3. **Scaling argument:** map to reference element, scale back → factor of $d_K^{k+1-i}$

**Theorem (Global interpolation error):** For $u \in H^{k+1}(\Omega)$, Lagrange $P_k$ interpolation on a shape-regular mesh:

$$\|u - \mathcal{I}_h u\|_{H^i(\Omega)} \leq Ch^{k+1-i}|u|_{H^{k+1}(\Omega)}, \quad 0 \leq i \leq k$$

**This is the central approximation result** — it says $P_k$ interpolation converges at rate $O(h^{k+1})$ in $L^2$ and $O(h^k)$ in $H^1$.

### 🧠 Key things to understand deeply
1. The scaling argument: why does mapping to the reference element and back give the $d_K^{k+1-i}$ factor?
2. Why does shape regularity ($d_K/\rho_K$ bounded) matter? (Degenerate triangles → bad constants)
3. The global interpolation error rate: $O(h^{k+1-i})$ in $H^i$ — how does increasing $k$ help?

## Block 5 — 16:40–17:40 — L4 §§1–3: Hilbert Spaces, Riesz Representation, Lax-Milgram (~362 lines)

**Lecture notes:** `LecNotes/L4_feprobs.txt`, lines 6–395

---

### Hilbert spaces (L4 §1)

**Definitions (chain):**
- **Inner product space:** vector space + bilinear form $\langle \cdot, \cdot \rangle$ (symmetric, positive definite, linear)
- **Norm:** $\|u\| = \langle u, u \rangle^{1/2}$
- **Cauchy sequence:** $\|u_n - u_m\| \to 0$ as $n, m \to \infty$
- **Complete space:** every Cauchy sequence converges
- **Hilbert space** = complete inner product space

**Key examples:**
- $L^2(\Omega)$ with $\langle f, g \rangle = \int_\Omega fg$
- $H^1(\Omega)$ with $\langle f, g \rangle_{H^1} = \int_\Omega fg + \int_\Omega \nabla f \cdot \nabla g$

**Theorem (Schwarz inequality):** $|\langle u, v \rangle| \leq \|u\|\,\|v\|$

### Linear forms on Hilbert spaces (L4 §2)

- **Bounded/continuous linear functional:** $F \in V'$ with $|F(v)| \leq C\|v\|$ for all $v$
- **Lemma:** Continuous $\iff$ bounded (for linear functionals)
- **Dual norm:** $\|F\|_{V'} = \sup_{v \neq 0} \frac{|F(v)|}{\|v\|}$

**Theorem (Riesz representation):** For every $F \in V'$ (continuous linear functional on Hilbert space $V$), there exists a unique $u \in V$ such that $F(v) = \langle u, v \rangle$ for all $v$, and $\|u\| = \|F\|_{V'}$.

### Variational problems and Lax-Milgram (L4 §3)

**Definition (Linear variational problem):** Find $u \in V$ such that $a(u, v) = F(v)$ for all $v \in V$.

- $a: V \times V \to \mathbb{R}$ is a **bilinear form**
- **Continuous:** $|a(u, v)| \leq M\|u\|\,\|v\|$
- **Coercive:** $a(u, u) \geq \alpha\|u\|^2$ for some $\alpha > 0$

**Theorem (Lax-Milgram):** If $a$ is continuous and coercive on Hilbert space $V$, and $F \in V'$, then the variational problem has a **unique solution** $u$ with stability: $\|u\| \leq \frac{1}{\alpha}\|F\|_{V'}$.

**Examples of variational problems:**
- Helmholtz: $a(u,v) = \int_\Omega \nabla u \cdot \nabla v + \int_\Omega uv$, $F(v) = \int_\Omega fv$
- Poisson (Dirichlet): $a(u,v) = \int_\Omega \nabla u \cdot \nabla v$, same $F$

### 🧠 Key things to understand deeply
1. Why is coercivity the key condition? (It gives the lower bound → uniqueness + stability)
2. How does Lax-Milgram generalise Riesz? (Riesz = Lax-Milgram with $a = \langle \cdot, \cdot \rangle$)
3. For Poisson with pure Dirichlet: why is $a(u,v) = \int \nabla u \cdot \nabla v$ coercive on $\mathring{H}^1$ but NOT on $H^1$?

### Active recall prompts
1. State the Lax-Milgram theorem precisely. What are the hypotheses and conclusions?

## Block 6 — 18:05–19:05 — L4 §4: Solvability and Stability of FE Discretisations (~775 lines)

**Lecture notes:** `LecNotes/L4_feprobs.txt`, lines 395–1170

---

### Overview

This is the heaviest section in L4 — it applies Lax-Milgram to prove well-posedness of specific FE problems. Read the theorem statements and proof sketches; skim the exercises.

### Helmholtz solvability

$a(u,v) = \int_\Omega \nabla u \cdot \nabla v + \int_\Omega uv$ on $H^1$: coercivity is immediate ($a(u,u) = \|u\|_{H^1}^2$), so Lax-Milgram gives existence + uniqueness on any subspace.

### Poisson with pure Neumann

**Problem:** $a(u,v) = \int_\Omega \nabla u \cdot \nabla v$ is NOT coercive on $H^1$ (constants have $a(c,c) = 0$).

**Lemma (Mean estimate for FE spaces):** For $v_h \in V_h$:
$$\|v_h\|_{L^2}^2 \leq C\left(|\Omega|\,\bar{v}_h^2 + h^2|v_h|_{H^1}^2\right) \quad \text{where } \bar{v}_h = \frac{1}{|\Omega|}\int_\Omega v_h$$

**Theorem:** If we constrain $\bar{u}_h = 0$, then $a$ is coercive on the subspace → unique solution (up to constant).

### FE divergence theorem and trace theorem

- **FE divergence theorem:** $\int_\Omega \nabla \cdot F_h = \int_{\partial\Omega} F_h \cdot n$ for piecewise smooth $F_h$ (true because interior edge contributions cancel)
- **Trace theorem:** For $v_h \in V_h$ continuous, $\|v_h\|_{L^2(\partial\Omega)}^2 \leq C(h^{-1}\|v_h\|_{L^2(\Omega)}^2 + h|v_h|_{H^1(\Omega)}^2)$

### Poisson with partial Dirichlet BCs

On $\mathring{V}_h = \{v_h \in V_h : v_h|_{\Gamma_D} = 0\}$ where $|\Gamma_D| > 0$:
- **Poincaré-Friedrichs:** $\|v_h\|_{L^2} \leq C|v_h|_{H^1}$ for all $v_h \in \mathring{V}_h$
- → $a(u,v) = \int \nabla u \cdot \nabla v$ is coercive on $\mathring{V}_h$ → Lax-Milgram applies

### 🧠 Key things to understand deeply
1. Why is Helmholtz "easy" but Poisson is "hard"? (The $\int uv$ term gives coercivity for free)
2. The mean estimate: why does it need both $\bar{v}_h$ and $|v_h|_{H^1}$?
3. Why does having a Dirichlet boundary ($|\Gamma_D|>0$) restore coercivity? (Poincaré-Friedrichs kills the kernel)

### Active recall prompts
1. Why can't you solve Poisson with pure Neumann without fixing the mean?
2. State the Poincaré-Friedrichs inequality and explain when it applies.

## Block 7 — 19:25–20:25 — L5 §§1–4: Weak Derivatives, Sobolev Spaces, Galerkin, Céa's Lemma (~404 lines)

**Lecture notes:** `LecNotes/L5_convergence.txt`, lines 6–427

---

### Weak derivatives (L5 §1)

**Motivation:** FE functions may not have classical derivatives everywhere (e.g. kinks at element boundaries). Need a weaker notion.

**Definition (Weak derivative):** $g \in L^1_{loc}(\Omega)$ is the weak $\alpha$-th derivative of $f$ if:
$$\int_\Omega g\,\phi = (-1)^{|\alpha|}\int_\Omega f\,D^\alpha\phi \qquad \forall \phi \in C_0^\infty(\Omega)$$

**Key results:**
- FE derivative of a $C^0$ FE function = its weak derivative
- Classical derivative (where it exists) = weak derivative
- Weak derivatives are unique (up to a.e.)

### Sobolev spaces (L5 §2)

**Definition:** $H^k(\Omega) = \{f \in L^2(\Omega) : D^\alpha f \in L^2(\Omega)\; \forall |\alpha| \leq k\}$ (weak derivatives)

**Lemma:** $H^k(\Omega)$ is a Hilbert space with inner product $\langle f, g \rangle_{H^k} = \sum_{|\alpha|\leq k}\int_\Omega D^\alpha f \cdot D^\alpha g$.

**Theorem (H = W):** $C^\infty(\bar\Omega) \cap H^k$ is dense in $H^k$ — can approximate any Sobolev function by smooth functions.

**Theorem (Sobolev's inequality):** $k > d/2 \implies H^k(\Omega) \hookrightarrow C(\bar\Omega)$, i.e. $\|f\|_\infty \leq C\|f\|_{H^k}$.

### Variational formulations of PDEs (L5 §3)

- Helmholtz on $H^1(\Omega)$: well-posed by Lax-Milgram (same as L4 but now on infinite-dimensional $H^1$, not just $V_h$)
- **Trace:** For $f \in H^1(\Omega)$, the restriction $f|_{\partial\Omega}$ is well-defined in $L^2(\partial\Omega)$
- Integration by parts generalises to $H^2 \times H^1$: if $u \in H^2$, then $-\int \nabla^2 u \cdot v = \int \nabla u \cdot \nabla v - \int_{\partial\Omega} \frac{\partial u}{\partial n} v$
- **Strong → variational → strong:** equivalence of formulations for smooth solutions

### Galerkin approximation and Céa's lemma (L5 §4)

**Definition (Galerkin approximation):** Given variational problem on $V$: find $u \in V$ s.t. $a(u,v) = F(v)\;\forall v \in V$. The Galerkin approximation is: find $u_h \in V_h \subset V$ s.t. $a(u_h, v_h) = F(v_h)\;\forall v_h \in V_h$.

**Galerkin orthogonality:** $a(u - u_h, v_h) = 0$ for all $v_h \in V_h$.

**Theorem (Céa's lemma):** If $a$ is continuous ($M$) and coercive ($\alpha$), then:
$$\|u - u_h\| \leq \frac{M}{\alpha} \inf_{v_h \in V_h} \|u - v_h\|$$

**Meaning:** The FE solution is quasi-optimal — it's within a factor $M/\alpha$ of the best approximation in $V_h$.

### 🧠 Key things to understand deeply
1. Why do we need weak derivatives? (FE functions aren't classically differentiable everywhere)
2. Céa's lemma is the bridge: it converts approximation theory (interpolation error) into convergence (FE error)
3. Galerkin orthogonality — understand the geometry (projection in the $a$-inner product)

### Active recall prompts
1. Define the weak derivative. Why is it a generalisation of the classical derivative?
2. State Céa's lemma. Why does it reduce convergence of FE to approximation theory?

### H^k interpolation for Sobolev functions (L5 §5)

Same interpolation error results as L3, but now for $u \in H^{k+1}(\Omega)$ (weak derivatives), not just smooth $u$. The averaged Taylor polynomial machinery extends to Sobolev functions.

$$\|u - \mathcal{I}_h u\|_{H^i(\Omega)} \leq Ch^{k+1-i}|u|_{H^{k+1}(\Omega)}$$

### ⭐ Convergence of FE for Helmholtz (L5 §6)

**Combine Céa + interpolation:**

$$\|u - u_h\|_{H^1} \leq \frac{M}{\alpha}\|u - \mathcal{I}_h u\|_{H^1} \leq Ch^k|u|_{H^{k+1}}$$

So $P_k$ FE converges at rate $O(h^k)$ in $H^1$ for Helmholtz.

### ⭐ Elliptic regularity

**Theorem:** If $\Omega$ is convex (or has smooth boundary) and $f \in L^2$, then the solution $u$ of Helmholtz satisfies $u \in H^2$ with $\|u\|_{H^2} \leq C\|f\|_{L^2}$.

### ⭐ Aubin-Nitsche trick ($L^2$ error estimate)

**Theorem:** Under elliptic regularity:
$$\|u - u_h\|_{L^2} \leq Ch^{k+1}|u|_{H^{k+1}}$$

**Proof idea (duality argument):** Define auxiliary problem $a(w, v) = (u-u_h, v)_{L^2}$ for all $v$. By regularity, $w \in H^2$. Then:
$$\|u-u_h\|_{L^2}^2 = a(w, u-u_h) = a(w - \mathcal{I}_h w, u-u_h) \leq M\|w - \mathcal{I}_h w\|_{H^1}\|u-u_h\|_{H^1}$$

The $\|w - \mathcal{I}_h w\|_{H^1} \leq Ch|w|_{H^2} \leq Ch\|u-u_h\|_{L^2}$ gives the extra factor of $h$.

**Summary:** $P_k$ FE gives $O(h^k)$ in $H^1$ and $O(h^{k+1})$ in $L^2$ — **one order higher in $L^2$**.

### Stokes equation introduction (L6 §§1–2)

**Strong form:** $-2\mu\nabla\cdot\varepsilon(u) + \nabla p = f$, $\nabla\cdot u = 0$ in $\Omega$, $u = 0$ on $\partial\Omega$.

- $\varepsilon(u) = \frac{1}{2}(\nabla u + \nabla u^\top)$ = symmetric strain rate
- Incompressibility: $\nabla\cdot u = 0$
- Pressure $p$ unique only up to a constant → work in $\mathring{L}^2 = \{q : \int q = 0\}$

**Variational form:** Find $(u, p) \in V \times Q$ s.t.:
$$a(u, v) + b(v, p) = F(v) \quad \forall v \in V, \qquad b(u, q) = 0 \quad \forall q \in Q$$

where $a(u,v) = 2\mu\int_\Omega \varepsilon(u):\varepsilon(v)$, $b(v,q) = -\int_\Omega q\,\nabla\cdot v$.

**Key issue:** This is a **saddle-point problem** — Lax-Milgram does not directly apply. Need Brezzi's theorem (next session).

### Inf-sup condition (L6 §3)

**Definition:** Bilinear form $b: V \times Q \to \mathbb{R}$ satisfies the **inf-sup condition** if:
$$\inf_{q \in Q\setminus\{0\}} \sup_{v \in V\setminus\{0\}} \frac{b(v,q)}{\|v\|_V\|q\|_Q} \geq \beta > 0$$

**Operator interpretation:** Define $B: V \to Q'$ by $(Bv)(q) = b(v,q)$ and $B^*: Q \to V'$ by $(B^*q)(v) = b(v,q)$.
- Inf-sup $\iff$ $B^*$ is injective with closed range
- Equivalent to: $\|B^*q\|_{V'} \geq \beta\|q\|_Q$ for all $q$
