## What This Course Is About

This course builds the mathematical foundations of **game theory** — the study of strategic decision-making among rational agents. You will:

1. Learn to model strategic interactions as **games** in normal and extensive form.
2. Master the concepts of **dominance**, **best response**, and **Nash equilibrium** in pure strategies.
3. Extend to **mixed (randomised) strategies** and prove **Nash's Theorem** (every finite game has an equilibrium).
4. Study **zero-sum games**, the **Minimax Theorem**, and optimal play in strictly competitive settings.
5. Explore **cooperative game theory** and the **Nash Bargaining Solution**.
6. Analyse **congestion/traffic games**, potential functions, and the **Price of Anarchy**.
7. Solve **combinatorial games** (Nim) using Nim sums, the Mex Rule, and game equivalence.
8. *(Masters only)* Study **game trees**, backward induction, imperfect information, and **subgame-perfect equilibria**.

The course has **7 chapters** (+ Ch 8 for Masters), 7 problem sets, a group coursework (10%), and a 1.5-hour exam (90%).

## Part 1 — Foundations: Dominance, Best Response & Equilibria (Ch 1–2)

### 1.1 What Is a Game?

| Element | Description |
|---|---|
| **Players** | The decision-makers (at least 1) |
| **Moves / Strategies** | The actions available to each player |
| **Payoffs** | The outcome (utility) for each player given all choices |

Games are displayed in **normal (strategic) form** — a payoff matrix — or **extensive (tree) form**.

**Key assumptions:** Players are **rational and greedy**, rules are fixed, no cheating.

### 1.2 Key Definitions

| Concept | One-liner |
|---|---|
| **Pure strategy** | A deterministic choice of action in every game position |
| **Strictly dominated** | Strategy $a$ is strictly dominated by $a'$ if $g(a,b) < g(a',b)$ for **all** opponent strategies $b$ |
| **Weakly dominated** | $g(a,b) \le g(a',b)$ for all $b$, strict for at least one |
| **Best response** | Strategy giving the highest payoff given the opponent's choice |
| **Nash equilibrium** | A profile where **every** player's strategy is a best response to the others |

### 1.3 Key Results

- **Iterated deletion of dominated strategies** preserves equilibria (Prop 2.14–2.15). If it reduces to a single profile → unique equilibrium (**dominance solvable**).
- A strategy that is **never a best response** is dominated (Prop 2.12).
- Deleting **weakly** dominated strategies may lose equilibria — be cautious.

### 1.4 Landmark Games

| Game | Key Concept Illustrated |
|---|---|
| Everybody Bids | Iterated deletion → all bid 0 |
| Prisoner's Dilemma | Strict dominance; individual incentive vs common good |
| Dating Game (Battle of Sexes) | Multiple equilibria; coordination |
| Cournot Duopoly | Continuous strategies; best-response functions; cartel vs equilibrium |
| Matching Pennies | No pure-strategy equilibrium |
| Rock-Paper-Scissors | No pure-strategy equilibrium |

## Part 2 — Mixed Strategies & Nash's Theorem (Ch 3)

### 2.1 Mixed Strategies

A **mixed strategy** $\alpha = (p_1, \ldots, p_n)$ assigns probabilities to pure strategies. Expected payoff:
$$g_A(\alpha, \beta) = \sum_i \sum_j g_A(a_i, b_j)\, p_i q_j$$

### 2.2 Finding Mixed Equilibria

- A player mixes **only** when **indifferent** between the pure strategies in the mix.
- **Equaliser strategy**: makes the opponent indifferent among their pure strategies.
- The equilibrium mixing probability depends on the **opponent's payoffs**, not your own.

**Key technique:** Set expected payoffs equal across the opponent's pure strategies and solve.

### 2.3 Nash's Theorem (Thm 3.23)

> **Every finite game has at least one (mixed) equilibrium.**

Proved via **Brouwer's Fixed Point Theorem**: define a continuous map on the compact convex set of mixed strategy pairs; a fixed point is an equilibrium.

### 2.4 Additional Tools

| Tool | Use |
|---|---|
| **Upper envelope method** | Graph expected payoffs vs opponent's mixing parameter; best response = upper envelope |
| **Subgame checking** | For $m \times n$ games, check all $k \times k$ subgames for equilibria |
| **Degenerate games** | Games where some mixed strategy leaves opponent with > $k$ best responses; can yield infinite families of equilibria |

## Part 3 — Zero-Sum Games & the Minimax Theorem (Ch 4)

In a **zero-sum game**, $g_A + g_B = 0$ for every outcome. Only $g_A$ is displayed.

### 3.1 Max-Min and Min-Max

| Concept | Definition |
|---|---|
| **Max-min** strategy $\hat{\alpha}$ | Maximises the **worst-case** payoff: $\max_\alpha \min_\beta g(\alpha,\beta)$ |
| **Min-max** strategy $\hat{\beta}$ | Minimises the **best-case** payoff for the opponent: $\min_\beta \max_\alpha g(\alpha,\beta)$ |

### 3.2 The Minimax Theorem (Von Neumann, 1928)

$$\boxed{\max_\alpha \min_\beta g(\alpha,\beta) = v = \min_\beta \max_\alpha g(\alpha,\beta)}$$

The common value $v$ is the **value of the game**. At equilibrium, max-min = min-max.

### 3.3 Solving Zero-Sum Games

- **Lower envelope diagram**: plot $g(\alpha, b_j)$ for each pure strategy $b_j$; the lower envelope gives B's best response; A maximises.<br>
- Can freely delete **weakly dominated** and **payoff equivalent** strategies without changing the value (Prop 4.30).
- Equilibrium iff max-min = min-max (Prop 4.28).

## Part 4 — Cooperative Games & Bargaining (Ch 5)

When players can form **binding agreements**, they can achieve outcomes impossible under non-cooperative play.

### 4.1 Key Concepts

| Concept | Definition |
|---|---|
| **Security (threat) level** $t_i$ | Max-min payoff for player $i$ — guaranteed minimum |
| **Threat point** $(t_A, t_B)$ | The pair of security levels — disagreement outcome |
| **Bargaining set** $S$ | Convex hull of achievable payoff pairs with $(x,y) \ge (t_A, t_B)$ |
| **Pareto-optimal** | No player can be made better off without making another worse off |

### 4.2 The Nash Bargaining Solution (Thm 5.33)

Under five axioms (feasibility, Pareto-optimality, scale invariance, symmetry, independence of irrelevant alternatives):

$$\boxed{\text{Maximise } (x - t_A)(y - t_B) \text{ over } (x,y) \in S}$$

This is the unique solution satisfying all five axioms.

## Part 5 — Congestion Games & Price of Anarchy (Ch 6)

### 5.1 Setup

- $N$ users route through a network; each edge has a cost function $c_e(x)$ (increasing in flow $x$).
- Each user picks a path minimising their own cost (**selfish routing**).

### 5.2 Key Results

| Result | Statement |
|---|---|
| **Existence** (Thm 6.38) | Every congestion game has an equilibrium (via potential function $\phi(f) = \sum_e \sum_{k=1}^{f_e} c_e(k)$) |
| **Price of Anarchy** | $\text{POA} = \frac{\text{worst equilibrium cost}}{\text{social optimum cost}}$ |
| $\text{POA} \le 5/2$ | Atomic flow, affine costs (Prop 6.40) |
| $\text{POA} \le 4/3$ | Splittable flow, affine costs (Prop 6.41) |

### 5.3 Key Examples

- **Pigou network**: all users take the "cheap" route → POA = 4/3.
- **Braess paradox**: adding a zero-cost edge can **increase** equilibrium cost — more capacity can hurt!

## Part 6 — Combinatorial Games (Ch 7)

Two-player, perfect information, sequential, no chance, normal play convention (last to move wins).

### 6.1 Nim and the Nim Sum

- **Nim**: piles of tokens; remove ≥ 1 from one pile; last to move wins.
- **Bouton's Theorem (1901)**: A Nim position is a **losing position** iff the **Nim sum** (XOR of all pile sizes in binary) equals zero.

### 6.2 Game Equivalence and the Mex Rule

| Concept | Definition |
|---|---|
| **Game sum** $G + H$ | Play in one component, leave the other unchanged |
| **Equivalence** $G \equiv H$ | $G + J$ losing $\iff$ $H + J$ losing for all $J$ |
| **Copycat principle** | $G + G \equiv 0$ for any impartial game |
| **Nim value** (Sprague–Grundy) | $G \equiv {*}m$ where $m = \text{mex}(\{\text{Nim values of options}\})$ |
| **Mex rule** (Thm 7.66) | Every impartial game is equivalent to a Nim pile of size = mex of its options' Nim values |

### 6.3 Proof Tools

- **Top-down induction** (Thm 7.51): if $P(x)$ holds whenever $P(y)$ holds for all simpler $y$, then $P$ holds everywhere.
- All proofs in this chapter use top-down induction on game complexity.
