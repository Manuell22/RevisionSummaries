## A. Finding Pure-Strategy Equilibria (Ch 2)

### Recipe

1. Write the game in normal form (payoff matrix).
2. **Circle** A's best-response payoff in each **column**.
3. **Box** B's best-response payoff in each **row**.
4. Any cell with **both** a circle and a box is a pure-strategy Nash equilibrium.

### Iterated Deletion of Dominated Strategies

1. Scan for **strictly dominated** strategies (a row/column that is always worse than another).
2. Delete it. Repeat until no more dominated strategies exist.
3. If a single cell remains → **unique equilibrium** (dominance solvable).
4. ⚠️ If deleting **weakly** dominated strategies, note that equilibria may be lost.

### Common Pitfalls

- In the Prisoner's Dilemma, $(c,c)$ is NOT an equilibrium — both can deviate to $d$.
- Weak dominance ≠ strict dominance. Don't claim uniqueness from weak deletion alone.
- A strategy may be dominated by a **mixed** strategy even if not by any pure strategy (e.g. shooting middle in penalty kicks).

## B. Finding Mixed-Strategy Equilibria (Ch 3)

### Recipe for $2 \times 2$ Games

1. Let $\alpha = (p, 1-p)$ and $\beta = (q, 1-q)$.
2. **Using A's payoffs**, set $g_A(a_1, \beta) = g_A(a_2, \beta)$ → solve for $q$.
3. **Using B's payoffs**, set $g_B(\alpha, b_1) = g_B(\alpha, b_2)$ → solve for $p$.
4. Verify $0 \le p, q \le 1$.
5. The mixed equilibrium is $(\alpha^*, \beta^*) = ((p, 1-p), (q, 1-q))$.

**Key principle**: Player A's mixing probability $p$ is determined from **B's payoff equations**, and vice versa.

### Recipe for $m \times 2$ Games — Upper Envelope Method

1. Let $\beta = (q, 1-q)$.
2. Compute $g_A(a_i, \beta)$ for each $a_i$ (linear in $q$).
3. **Plot** all lines on the same axes (horizontal: $q$, vertical: expected payoff).
4. Trace the **upper envelope** (highest line for each $q$).
5. Check all **intersection points on the envelope** and **endpoints** ($q=0, q=1$):
   - At intersections: A mixes between the two intersecting strategies. Solve for $p$ from B's indifference.
   - At endpoints: pure strategy equilibria.
6. For each candidate, verify that excluded strategies don't give a higher payoff.

### Recipe for $m \times n$ Games — Subgame Method

1. Enumerate all $k \times k$ subgames ($k \le \min(m,n)$).
2. Find equilibria of each subgame.
3. Check each subgame equilibrium against **all excluded** strategies: if $g_A(a_{\text{excl}}, \beta^*) > g_A(\alpha^*, \beta^*)$, discard it.

### Degenerate Games

If $p = 0$ or $p = 1$ in a mixed equilibrium:
- Only need the **inequality** (that the played strategy is ≥ the unplayed), not equality.
- Can produce **infinite families** of equilibria parameterised by the opponent's mixing probability.

## C. Zero-Sum Games — Value & Optimal Strategies (Ch 4)

### Recipe

1. Check if the game is zero-sum (only one payoff per cell, or $g_A + g_B = 0$).
2. First try to **simplify**: delete weakly dominated and payoff-equivalent strategies (Prop 4.30 — safe in zero-sum).
3. **Lower envelope method** (for $2 \times m$):
   - Let $\alpha = (p, 1-p)$. Plot $g(\alpha, b_j)$ vs $p$ for each pure $b_j$.
   - The **lower envelope** = B's best response.
   - The **maximum of the lower envelope** = max-min payoff = **value** $v$.
   - The $p$ achieving this max is A's max-min (equilibrium) strategy.
   - At that point, determine B's mixing from the active constraints.
4. **Algebraic method** (any size):
   - Compute $g(\alpha, b_j)$ for each pure $b_j$. Set $\min_j g(\alpha, b_j) = v$; maximise over $\alpha$.
   - Similarly for B.
   - If max-min = min-max → value found.

### Quick Checks

- **Saddle point**: a cell that is simultaneously the minimum in its row and maximum in its column → pure-strategy equilibrium, value = that cell.
- In zero-sum games, all equilibria give the **same** value $v$ to A.

## D. Cooperative Games — Nash Bargaining (Ch 5)

### Recipe

1. **Compute threat point** $(t_A, t_B)$:
   - $t_A = \max_\alpha \min_\beta g_A(\alpha, \beta)$ (A's max-min payoff using A's own payoffs).
   - $t_B = \max_\beta \min_\alpha g_B(\alpha, \beta)$ (B's max-min payoff using B's own payoffs).
2. **Plot** pure-strategy payoff pairs and draw their **convex hull**. Restrict to $x \ge t_A, y \ge t_B$ → **bargaining set** $S$.
3. Identify the **Pareto-optimal frontier** (upper-right boundary of $S$).
4. **Maximise** the Nash product $(x - t_A)(y - t_B)$ over each segment of the frontier:
   - Parameterise the segment (e.g. $y = f(x)$).
   - Differentiate and solve, or complete the square.
   - If maximiser is outside the segment, the answer is at the nearer endpoint.
5. Compare across segments; the global max gives the **Nash bargaining solution**.

### Common Pitfalls

- The threat point is computed as if the game were **non-cooperative** (max-min).
- The convex hull in cooperative play includes **all** convex combinations (not just the restricted payoff set of non-cooperative play).
- The solution may require a **joint mixed strategy** (not always a pure strategy pair).

## E. Congestion Games & Price of Anarchy (Ch 6)

### Finding Equilibria

1. Identify all paths from origin to destination.
2. Set up cost equations as functions of user counts.
3. **Atomic flow**: an equilibrium occurs when no single user can reduce their cost by switching paths. Check integer allocations where costs across paths differ by at most 1.
4. **Splittable flow**: set all path costs equal; solve the resulting system.

### Social Optimum

- Write average cost as a function of flow variables.
- Minimise (complete the square or use calculus).
- For atomic flow, check integer values near the continuous optimum.

### Price of Anarchy

$$\text{POA} = \frac{\text{Worst equilibrium average cost}}{\text{Social optimum average cost}}$$

Bounds for affine costs $c(x) = ax + b$:
- Atomic: $\text{POA} \le 5/2$
- Splittable: $\text{POA} \le 4/3$

### Braess Paradox Checklist

If asked "does adding edge X improve things?": compute equilibria and social optima for both networks and compare.

## F. Combinatorial Games — Nim & the Mex Rule (Ch 7)

### Is a Nim Position Winning or Losing?

1. Write all pile sizes in **binary**.
2. Compute the **Nim sum** (XOR each column).
3. Nim sum = 0 → **losing**; Nim sum ≠ 0 → **winning**.

### Finding a Winning Move in Nim

1. Find the leftmost '1' column in the Nim sum.
2. Choose any pile that has a '1' in that column.
3. **Flip** the digits of that pile at every '1' column of the Nim sum.
4. Reduce the pile to the new (smaller) value → this leaves a zero (losing) position.

### Mex Rule for General Impartial Games

1. Compute the Nim value of every **option** of the current position (recursively).
2. The Nim value of the current position = $\operatorname{mex}(\{\text{Nim values of options}\})$.
3. Position is losing iff Nim value = 0.

### Game Sums

If a game decomposes as $G_1 + G_2 + \cdots + G_k$:
1. Find the Nim value of each component.
2. The overall Nim value = Nim sum (XOR) of all component values.
3. **Copycat principle**: $G + G \equiv 0$.

### Rook/Queen Move Games

- **Rook** at $(i,j)$: equivalent to $*i + *j$. Losing iff $i = j$.
- **Queen** at $(i,j)$: compute Nim values recursively (extra diagonal moves). Consult the Nim value table.
- **Multiple pieces**: game = sum of individual games; XOR Nim values.

## G. Game Trees & Imperfect Information (Ch 8, Masters)

### Backward Induction Recipe

1. Start at terminal nodes. Work upward.
2. At each decision node, the player picks the move maximising their payoff.
3. At chance nodes, replace with expected payoffs.
4. If multiple moves give equal payoff, track all paths (multiple arrows).

### Strategic Form of a Game Tree

1. List all (reduced) strategies for each player.
2. For each strategy profile, trace through the tree to compute expected payoffs.
3. Find equilibria of this matrix game.
4. Check which are **subgame-perfect** (SPE).

### Imperfect Information

1. Identify **information sets** (connected by dashed lines).
2. A strategy specifies a move per **information set** (not per node).
3. Write the strategic form. Find mixed equilibria if needed.
4. Interpret as **behaviour strategies** (randomise at each information set when reached).
5. For SPE: check every **subgame** (subtree with complete information sets). Ensure equilibrium holds in each.

### Reduced Strategies

- Replace moves at unreachable decision nodes with $*$.
- For SPE, must specify optimal moves even at unreachable nodes (fill in the $*$).
