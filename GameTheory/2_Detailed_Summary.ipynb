## Chapter 1 — Prelude

### 1.1 Core Definitions

| # | Definition |
|---|---|
| **Def 1.1** | A **move** is the action a player makes on their turn to progress from one game position to the next. |
| **Def 1.2** | An **outcome** is the final result (win/lose/draw or payoff vector) once the game has been played. |
| **Def 1.3** | A **strategy** is a *complete* description of all moves in *every* position, including responses to random/opponent moves. It is a program to play the game mechanically. |
| **Def 1.4** | A **pure strategy** is a strategy involving no self-imposed randomness. |
| **Def 1.5** | A game is **finite** if all players have finitely many pure strategies; otherwise it is **infinite**. |

### 1.2 Principles of Mathematical Game Theory

1. **Values**: Players are **rational and greedy** — they maximise their own payoff only.
2. **Rules**: Games have fixed, specified moves — no "thinking outside the box".
3. **No Cheating**: The game state is manipulated only through legal moves.

### 1.3 Elements of a Game

- **Players**: number and identity.
- **Moves**: all possible moves at each point, for all players, and their order.
- **Payoffs**: outcomes for all players for every combination of moves.

### 1.4 Classification of Games

- Number of players (including single-player vs nature).
- Simultaneous vs sequential moves.
- Complete vs incomplete information.
- Zero-sum (constant-sum) vs general-sum.
- Communication allowed or not.
- Cooperative vs non-cooperative.

### 1.5 Displaying Games

- **Normal (strategic) form**: payoff matrix/table.
- **Extensive (tree) form**: game tree with decision nodes.

## Chapter 2 — Dominance, Best Response & Equilibria

### 2.1 Notation

- **$N$-player**: Players $1, 2, \ldots, N$; player $i$ has strategy set $S_i$, typical strategy $s_i$. Payoff: $g_i(s_1, \ldots, s_N)$.
- **Two-player**: Players A and B; $A_s = \{a_1, a_2, \ldots\}$, $B_s = \{b_1, b_2, \ldots\}$. Payoffs: $g_A(a_i, b_j)$, $g_B(a_i, b_j)$.

### 2.2 Dominated Strategies

**Def 2.6 (Strictly dominated, 2-player).** Strategy $a$ is **strictly dominated** by $a'$ if:
$$g_A(a, b) < g_A(a', b) \quad \forall\, b \in B_s.$$

**Def 2.7 (Strictly dominated, N-player).** Strategy $s_i$ is strictly dominated by $s_i'$ if:
$$g_i(s_1, \ldots, s_i, \ldots, s_N) < g_i(s_1, \ldots, s_i', \ldots, s_N) \quad \forall\, s_j \in S_j,\, j \ne i.$$

**Def 2.8 (Weakly dominated, 2-player).** Strategy $a$ is **weakly dominated** by $a'$ if:
$$g_A(a, b) \le g_A(a', b) \quad \forall\, b,$$
with at least one strict inequality.

**Def 2.9 (Weakly dominated, N-player).** Analogous extension.

**Def 2.10 (Payoff equivalence, N-player).** $s_i$ and $s_i'$ are **payoff equivalent** if:
$$g_i(\ldots, s_i, \ldots) = g_i(\ldots, s_i', \ldots) \quad \forall\, s_j.$$

### 2.3 Best Response

**Def 2.11.** Strategy $s_i$ is a **best response** to the opponents' strategies $(s_1, \ldots, s_{i-1}, s_{i+1}, \ldots, s_N)$ if:
$$g_i(\ldots, s_i, \ldots) \ge g_i(\ldots, s_i', \ldots) \quad \forall\, s_i' \in S_i.$$

**Prop 2.12.** A strategy that is **never a best response** to any opponent strategy combination is a dominated strategy.  
*(Note: may be dominated by a mixed strategy even if not by any pure strategy.)*

### 2.4 Equilibria (Nash Equilibrium)

**Def 2.13.** An **equilibrium** (Nash equilibrium) is a strategy profile $(s_1^*, \ldots, s_N^*)$ where **every** player's strategy is a best response to the others. No player has incentive to unilaterally deviate.

**Finding pure equilibria in 2-player games**: Circle A's best-response payoffs (per column), box B's best-response payoffs (per row). Where circle meets box → equilibrium.

### 2.5 Key Example Games

| Game | Key Feature |
|---|---|
| **Prisoner's Dilemma** | Defect strictly dominates cooperate for both; $(d,d)$ is unique equilibrium but $(c,c)$ is better for both |
| **Defence of Rome** | Weak dominance for Hannibal; equilibrium at $(a_1, b_1)$ |
| **Dating Game** | Two pure equilibria $(B,B)$ and $(T,T)$; coordination game |
| **Cournot Duopoly** | Continuous strategies; best response functions; cartel vs equilibrium |
| **Matching Pennies** | No pure equilibrium |
| **Rock-Paper-Scissors** | No pure equilibrium |

### 2.6 Iterated Deletion of Dominated Strategies

**Prop 2.14.** Let $s_i'$ weakly dominate or be payoff equivalent to $s_i$ in game $G$. Let $G'$ be $G$ with $s_i$ removed. Then:
- (a) any equilibrium of $G'$ is an equilibrium of $G$;
- (b) if $s_i'$ *strictly* dominates $s_i$, then $G$ and $G'$ have the **same** equilibria.

⚠️ Deleting **weakly** dominated strategies may lose equilibria.

**Prop 2.15 (Dominance Solvability).** If iterated deletion of dominated strategies reduces the game to a single strategy profile $s$, then $s$ is the **unique** equilibrium of $G$.

- Order of deletion does not matter.
- Multiple dominated strategies may be removed simultaneously.
- Partial deletion also preserves equilibria (useful for simplifying games).

### 2.7 Cournot Duopoly — Infinite Version

Strategies: $x, y \in [0, 12]$. Payoffs:
$$g_A(x,y) = x(12 - x - y), \quad g_B(x,y) = y(12 - x - y).$$

**Best responses**:
$$x^* = 6 - \tfrac{y}{2}, \quad y^* = 6 - \tfrac{x}{2}.$$

Solving: unique equilibrium at $x = y = 4$, payoffs $(16, 16)$.  
Cartel outcome: $x = y = 3$, payoffs $(18, 18)$ — not an equilibrium.

## Chapter 3 — Mixed Equilibria

### 3.1 Mixed Strategies

**Def 3.16.** A **mixed strategy** $\alpha = (p_1, \ldots, p_n)$ is a probability distribution over pure strategies, with $p_i \ge 0$ and $\sum_i p_i = 1$.

**Def 3.17 (Expected payoff, finite).** If $\alpha = (p_1, \ldots, p_n)$ and $\beta = (q_1, \ldots, q_m)$:
$$g_A(\alpha, \beta) = \sum_{i=1}^{n} \sum_{j=1}^{m} g_A(a_i, b_j)\, p_i q_j.$$

For infinite strategy sets, replace sums with integrals over probability densities.

**Def 3.18 / 3.19 (Mixed equilibrium).** A pair $(\alpha^*, \beta^*)$ is in **mixed equilibrium** if:
$$g_A(\alpha^*, \beta^*) \ge g_A(\alpha, \beta^*) \;\forall\, \alpha, \quad g_B(\alpha^*, \beta^*) \ge g_B(\alpha^*, \beta) \;\forall\, \beta.$$

All previous definitions (dominance, best response) extend naturally to mixed strategies.

### 3.2 Key Results for Finding Mixed Equilibria

**Prop 3.20.** The maximum expected payoff over **all** mixed strategies equals the maximum over **pure** strategies only:
$$\max_{\alpha \in \bar{A}_s} g_A(\alpha, \beta^*) = \max_{a \in A_s} g_A(a, \beta^*).$$

→ You only need to check pure strategies to find the best response.

**Def 3.21 (Equaliser strategy).** $\alpha^*$ is an **equaliser** for A (using B's payoffs) if:
$$g_B(\alpha^*, b) = c \quad \forall\, b \in B_s.$$

**Prop 3.22.** If $\alpha^*$ is an equaliser for A (using B's payoffs) and $\beta^*$ is an equaliser for B (using A's payoffs), then $(\alpha^*, \beta^*)$ is an equilibrium.

### Key Insight

A player mixes **only** over pure strategies that give **equal** (maximal) expected payoff. The equilibrium mixing probability is determined by the **opponent's** payoffs, not one's own.

### 3.3 Nash's Theorem

**Theorem 3.23 (Nash, 1951).**
> **Every finite game has at least one (mixed) equilibrium.**

**Proof sketch:**
1. Let $C = \bar{A}_s \times \bar{B}_s$ (compact, convex set of mixed strategy pairs).
2. Construct $f: C \to C$ where $f(\alpha, \beta) = (\alpha^*, \beta^*)$ by:
   - $\chi_i(\alpha, \beta) = \max\{0,\, g_A(a_i, \beta) - g_A(\alpha, \beta)\}$ (excess payoff of pure $a_i$)
   - Update: $p_i^* = \frac{p_i + \chi_i}{1 + \sum_k \chi_k}$ (renormalise)
3. Show: $(\alpha, \beta)$ is an equilibrium $\iff$ $f(\alpha, \beta) = (\alpha, \beta)$ (fixed point).
4. By **Brouwer's Fixed Point Theorem**, continuous $f: C \to C$ on compact convex $C$ has a fixed point. $\square$

### 3.4 Finding Equilibria: Subgame Method

For an $m \times n$ game where $m > 2$, $n = 2$:
1. Enumerate all $\binom{m}{2}$ subgames of size $2 \times 2$.
2. Solve each subgame for its equilibria.
3. **Check**: for each subgame equilibrium $(\alpha, \beta)$, verify that **no excluded pure strategy** gives a higher payoff than $g_A(\alpha, \beta)$.
4. Collect surviving equilibria as equilibria of the full game.

⚠️ In an $m \times 2$ game, A generally cannot mix over all $m$ strategies (only 1 free parameter $q$).

### 3.5 The Upper Envelope Method

When one player (B) has exactly 2 pure strategies with $\beta = (q, 1-q)$:
1. Plot $g_A(a_i, \beta)$ vs $q$ for each pure $a_i$ → straight lines.
2. The **upper envelope** = B's best response boundary.
3. Check **intersection points** on the upper envelope and **endpoints** ($q=0, q=1$) for equilibria.

Faster than subgame enumeration for large $m$.

### 3.6 Degenerate Games

**Def 3.24.** A two-player game is **degenerate** if some mixed strategy assigns positive probability to exactly $k$ pure strategies so that the other player has **more than $k$** pure best responses.

Consequence: can produce **infinite families** of equilibria. When a strategy has probability 0 in equilibrium, we only need the **inequality** $g_A(a_i, \beta) \ge g_A(a_j, \beta)$ rather than equality.

### 3.7 Geometry of Games

The **payoff set** $S$ for a $2 \times 2$ game lies within the convex hull of the 4 pure-strategy payoff points. Players choosing independently means $S$ is generally a strict subset of the convex hull, with curved boundaries between diagonally opposite payoff points.

## Chapter 4 — Zero-Sum Games

### 4.1 Setup

In a **zero-sum** game: $g_A(\alpha, \beta) + g_B(\alpha, \beta) = 0$ for all strategy pairs. Write $g(\alpha, \beta) = g_A(\alpha, \beta)$; only one payoff per cell.

### 4.2 Max-Min and Min-Max

**Def 4.25.** A **max-min strategy** $\hat{\alpha}$ satisfies:
$$\min_{\beta} g(\hat{\alpha}, \beta) = \max_{\alpha} \min_{\beta} g(\alpha, \beta).$$

**Def 4.26.** A **min-max strategy** $\hat{\beta}$ satisfies:
$$\max_{\alpha} g(\alpha, \hat{\beta}) = \min_{\beta} \max_{\alpha} g(\alpha, \beta).$$

**Prop 4.27.** Can replace $\min_\beta / \max_\alpha$ over mixed strategies with $\min_b / \max_a$ over pure strategies.

**Key inequality:**
$$\max_\alpha \min_\beta g(\alpha, \beta) \le \min_\beta \max_\alpha g(\alpha, \beta).$$

### 4.3 Equilibria ↔ Max-Min/Min-Max

**Prop 4.28.** In a finite zero-sum game, $(\hat{\alpha}, \hat{\beta})$ is an equilibrium **iff**:
1. $\hat{\alpha}$ is max-min,
2. $\hat{\beta}$ is min-max, and
3. $\max_\alpha \min_\beta g = \min_\beta \max_\alpha g$.

### 4.4 Lower Envelope Diagram

Plot $g(\alpha, b_j)$ vs $p$ for each pure $b_j$. The **lower envelope** gives B's best response cost. Player A maximises the lower envelope → **max-min strategy**. The maximum point gives the **value of the game**.

### 4.5 Minimax Theorem

**Theorem 4.29 (Von Neumann, 1928).**
$$\boxed{\max_\alpha \min_\beta g(\alpha, \beta) = v = \min_\beta \max_\alpha g(\alpha, \beta)}$$

where $v$ is the **value of the game** (unique).

**Proof:** By induction on $n + m$ (game size). Key step: if some row inequality $g(a_k, \beta) < v$ is strict, delete row $k$ to get $G'$ with value $v'$. Show $v' = v$ by contradiction (convex combination $\beta^* = (1-\varepsilon)\beta + \varepsilon\beta'$ would reduce $v$ below its minimum).

### 4.6 Consequences

- Equilibrium payoff is **unique** ($v$ to A, $-v$ to B).
- Equilibrium strategies are max-min/min-max and independent of opponent's play.
- **Prop 4.30**: Can delete **weakly dominated** and **payoff equivalent** strategies in zero-sum games without changing the value. Any equilibrium of the reduced game is an equilibrium of the original.

## Chapter 5 — Cooperative Games

### 5.1 Bargaining Set

**Def 5.31.** The **bargaining set** $S$ is the convex hull of all payoff pairs $(g_A(a_i,b_j), g_B(a_i,b_j))$ restricted to:
$$x \ge t_A, \quad y \ge t_B,$$
where $t_A = \max_\alpha \min_\beta g_A(\alpha, \beta)$ (**security level / threat level** of A), similarly $t_B$.

$(t_A, t_B)$ is the **threat point** — the disagreement outcome.

In cooperative play, the **entire** convex hull is achievable (via joint mixed strategies with binding agreements), unlike non-cooperative play where the payoff set is a subset.

### 5.2 Bargaining Axioms

**Def 5.32.** A **Nash bargaining solution** $N(S) = (X,Y)$ satisfies:

| Axiom | Statement |
|---|---|
| **(a) Feasibility** | $(X,Y) \in S$ |
| **(b) Pareto-optimality** | No $(x,y) \in S$ with $x \ge X, y \ge Y$ except $(X,Y)$ itself |
| **(c) Scale invariance** | Solution unchanged by affine rescaling of payoffs |
| **(d) Symmetry** | If $t_A = t_B$ and $S$ is symmetric, then $X = Y$ |
| **(e) Independence of irrelevant alternatives** | If $S \subset T$ with same threat point, then $N(T) \notin S$ or $N(T) = N(S)$ |

### 5.3 Nash Bargaining Solution

**Theorem 5.33.** Under axioms (a)–(e), every bargaining set $S$ containing a point with $x > t_A, y > t_B$ has a **unique** solution:

$$\boxed{N(S) = \arg\max_{(x,y) \in S} (x - t_A)(y - t_B)}$$

**Proof sketch:**
1. Translate so threat point is at origin. Rescale so maximiser is at $(1,1)$.
2. Show $S'' \subseteq T = \{x \ge 0, y \ge 0, x+y \le 2\}$. By symmetry of $T$, $N(T) = (1,1)$.
3. By independence of irrelevant alternatives, $N(S'') = (1,1)$.
4. Uniqueness: any other point on hyperbola $xy=1$ would create interior points with larger product. $\square$

**Method**: Maximise $(x - t_A)(y - t_B)$ over each segment of the Pareto-optimal frontier; check boundary constraints.

## Chapter 6 — Congestion Games

### 6.1 Definitions

**Def 6.34 (Congestion network).** Directed graph with:
- Finite set of nodes, finite set of directed edges.
- Each edge $e$ has a **weakly increasing** cost function $c_e(x)$.

**Def 6.35 (Congestion game).** $N$ users, each with origin $O_i$ and destination $D_i$. Strategy $P_i$ = path from $O_i$ to $D_i$. Flow on edge $e$: $f_e = |\{i : e \in P_i\}|$. Cost to user $i$:
$$\text{Cost}_i(P_i) = \sum_{e \in P_i} c_e(f_e).$$

**Def 6.36 (Best response).** $P_i$ is a best response if:
$$\sum_{e \in P_i} c_e(f_e) \le \sum_{e \in P_i \cap Q_i} c_e(f_e) + \sum_{e \in Q_i \setminus P_i} c_e(f_e + 1) \quad \forall\, Q_i.$$

**Def 6.37 (Equilibrium).** All $N$ strategies are mutual best responses.

**Types of flow**: **Atomic** (discrete users) vs **Splittable** (continuum of mass).

### 6.2 Existence of Equilibria

**Theorem 6.38.** Every congestion game has at least one equilibrium.

**Proof:** Define the **potential function**:
$$\phi(f) = \sum_e \left( c_e(1) + c_e(2) + \cdots + c_e(f_e) \right).$$

Key property: if user $i$ switches from $P_i$ to $Q_i$, the change in $\phi$ equals the change in cost to user $i$:
$$\phi(f^*) - \phi(f) = \sum_{e \in Q_i} c_e(f_e^*) - \sum_{e \in P_i} c_e(f_e).$$

A flow minimising $\phi$ is therefore an equilibrium (no user can reduce their cost). $\square$

A game with such a potential is called a **potential game**.

### 6.3 Price of Anarchy

**Def 6.39.**
$$\text{POA} = \frac{\text{Worst equilibrium average cost per user}}{\text{Social optimum average cost per user}}.$$

**Prop 6.40.** For **atomic** flow with affine costs $c(x) = ax + b$: $\text{POA} \le 5/2$.

**Prop 6.41.** For **splittable** flow with affine costs: $\text{POA} \le 4/3$.

### 6.4 Key Examples

| Example | Key Finding |
|---|---|
| **Pigou network** | 100 users; equilibrium cost 99–100; social optimum 75; POA = 4/3 |
| **Braess paradox** | Adding a zero-cost edge **increases** equilibrium cost from 150 to ~200 |
| **Imperial lunch** | Multiple equilibria; POA ≈ 1.03–1.04 |

## Chapter 7 — Combinatorial Games

### 7.1 Setup

Two-player, perfect information, sequential, no chance moves, ending condition (play must terminate), **normal play convention** (last to move wins).

**Def 7.42 (Option).** A position reachable by a single move.

**Impartial**: available moves don't depend on whose turn it is (e.g. Nim).  
**Partizan**: moves depend on whose turn it is (e.g. Chess).

### 7.2 Winning and Losing Positions

**Prop 7.43.** A position is **losing** iff all its options are winning. A position is **winning** iff at least one option is losing (moving there is a **winning move**).

This is a recursive definition, grounded by: a position with **no options** is losing (vacuously all options are winning).

### 7.3 Nim and Bouton's Theorem

**Nim**: Piles of tokens; remove ≥1 from one pile; last to move wins.

**Nim sum**: Write pile sizes in binary; add **mod 2 without carry** (XOR each column).

**Prop 7.44 (Bouton, 1901).** A Nim position is **losing** iff the Nim sum = 0 ("zero position").

**Proof**: 
- (a) Any move from a zero position changes one row → changes a column parity → non-zero.
- (b) From non-zero: find leftmost '1' column in Nim sum; pick a pile using that power of 2; flip digits at all '1' columns of the Nim sum → reduces that pile → zero position.

### 7.4 Two Nim Piles

- 0 piles: losing (no moves).
- 1 pile: winning (take all).
- 2 piles: winning iff unequal (equalise to create losing position).

### 7.5 Partial Orders and Top-Down Induction

**Def 7.45.** Partial order: transitive + reflexive + antisymmetric. Total order: add totality.

**Def 7.48.** $H \le G$ ($H$ **simpler than** $G$): there is a sequence of moves from $G$ to $H$.

**Prop 7.49.** $\le$ is a partial order on games.

**Prop 7.50.** Every non-empty subset of games has a **minimal element** (by ending condition).

**Theorem 7.51 (Top-Down Induction).**  
If $P(x)$ holds whenever $P(y)$ holds for all $y <  x$, then $P(x)$ holds for all $x \in S$.

$$\bigl(\forall x : (\forall y < x : P(y)) \Rightarrow P(x)\bigr) \;\Rightarrow\; \forall x : P(x).$$

Base case is vacuously true for minimal elements.

### 7.6 Game Sums and Equivalence

**Def 7.52 (Game sum).** $G + H$ has options: $\{G_i + H\} \cup \{G + H_j\}$ (move in one component, leave the other).

**Prop 7.53.** $+$ is commutative, associative, with zero element $0$.

**Def 7.54 (Equivalence).** $G \equiv H$ iff for **any** game $J$: $G + J$ is losing $\iff$ $H + J$ is losing.

**Lemma 7.55.** $\equiv$ is an equivalence relation.

**Prop 7.56.** Two Nim piles are equivalent iff they have **equal size**.

**Prop 7.57.** $G$ is losing $\iff$ $G \equiv 0$.

**Corollary 7.58.** Any two losing games are equivalent.

**Lemma 7.59.** $G \equiv H \Rightarrow G + K \equiv H + K$.

**Lemma 7.60.** If $J$ is losing: $G + J \equiv G$.

**Prop 7.61 (Copycat Principle).** $G + G \equiv 0$ for any impartial game $G$.  
*Proof:* Any option $G' + G$ has counter-move to $G' + G' \equiv 0$ (losing by induction).

**Lemma 7.62.** $G \equiv H$ iff $G + H \equiv 0$.

### 7.7 Nim Values and the Mex Rule

**Def 7.63.** $*n$ = single Nim pile of size $n$. Options: $*0, *1, \ldots, *(n-1)$.

**Def 7.64.** If $G \equiv *m$, then $m$ is the **Nim value** (Sprague–Grundy value) of $G$.

**Def 7.65 (Mex).** $\operatorname{mex}(S) = \min\{k \in \mathbb{N} : k \notin S\}$.

**Theorem 7.66 (Mex Rule).** Every impartial game $G$ has Nim value $m = \operatorname{mex}(\{$Nim values of options of $G\})$, i.e. $G \equiv *m$.

**Proof:** Show $G + *m \equiv 0$ by showing all options are winning. For option $G + *h$ ($h < m$): since $h \in S$, some option $H$ of $G$ has $H \equiv *h$, counter-move to $H + *h \equiv 0$. For option $H + *m$: since $s_H \ne m$, either $s_H < m$ (counter to $H + *s_H \equiv 0$) or $s_H > m$ (by induction, $H$ has an option $H' \equiv *m$, counter to $H' + *m \equiv 0$). $\square$

### 7.8 Nim Sums of Piles

**Def 7.67.** If $*k \equiv *m + *n$, then $k = m \oplus n$ (the **Nim sum**).

**Theorem 7.68.** Write $n = 2^a + 2^b + 2^c + \cdots$ (binary). Then:
$$*n \equiv *(2^a) + *(2^b) + *(2^c) + \cdots$$

**Computing Nim sums**: Write numbers in binary, XOR column by column (cancel pairs), sum remaining powers of 2.

**Example**: $4 \oplus 5 \oplus 9 \oplus 14 = 4 \oplus (4+1) \oplus (8+1) \oplus (8+4+2) = 2 \oplus 4 = 6$.

### 7.9 Example Games

| Game | Key Insight |
|---|---|
| **Rook-move** | Rook at $(i,j)$ ≡ $*i + *j$; losing iff $i=j$ (diagonal) |
| **Queen-move** | Extra diagonal move; Nim values computed recursively via mex |
| **Multiple rooks** | Game sum of individual games; use Nim sums |

## Chapter 8 — Game Trees & Imperfect Information *(Masters only)*

### Part I: Perfect Information

**Def 8.69.** A **pure strategy** in a game tree specifies a move for **every** decision node of that player.

**Def 8.70.** The **strategic form** of a game tree: set of strategies for each player + expected payoffs for every profile.

**Backward Induction**: Analyse from the terminal nodes upward. Each player picks the move maximising their payoff at each node. At chance nodes, compute expected payoffs.

**Def 8.71 (Reduced strategy).** Omit moves at decision nodes unreachable due to own earlier choices (replace with $*$).

**Def 8.72 (Subgame, perfect info).** Any subtree of the game tree.

**Def 8.73 (SPE).** A strategy profile giving an equilibrium in **every** subgame.

**Theorem 8.74.** Backward induction leads to an SPE. → Game trees with perfect information have pure-strategy SPE.

### Part II: Imperfect Information

**Information sets**: Connected nodes (dashed lines) where a player doesn't know which node they're at. Same moves available at all nodes within a set.

**Def 8.75.** A strategy in a game tree of imperfect information specifies a move for every **information set** of the player.

**Behaviour strategies**: Randomised choice of move at each information set, performed when that set is reached (more natural than deciding all randomness at the start).

**Perfect recall**: Players remember all their previous moves and all information learned. We always assume this.

**Def 8.76 (Subgame, imperfect info).** A subtree where every information set is **fully inside** or **fully outside**.

**Def 8.77 (SPE, imperfect info).** A profile of behaviour strategies defining an equilibrium in every subgame.

**Theorem 8.78.** Any game tree with imperfect information and perfect recall has an SPE in behaviour strategies.
