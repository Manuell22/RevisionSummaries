## Key Equation Box

### Binomial Model (one-period)
$$\tilde{p} = \frac{1+r-d}{u-d}, \qquad \tilde{q} = 1-\tilde{p}$$

$$V_0 = \tilde{p}\,V_1(H) + \tilde{q}\,V_1(T) \quad \text{(RNPF)}$$

$$\Delta = \frac{V_1(H)-V_1(T)}{S_1(H)-S_1(T)} \quad \text{(delta-hedge)}$$

### No-Arbitrage Condition (binomial)
$$d < 1+r < u \quad (\text{one-period}) \qquad D < 1+R < U \quad (\text{multi-period})$$

### Multi-Period RNPF
$$V_n = E^Q[V_{n+1}|F_n], \qquad V_0 = E^Q[V_N]$$

### American Options
$$V_N = I_N^+, \quad V_n = \max(I_n,\, C_n), \quad C_n = \tilde{p}\,V_{n+1}(H) + \tilde{q}\,V_{n+1}(T)$$
$$\tau^* = \inf\{n \le N : I_n = V_n\}$$

### Call Option (binomial, closed form)
$$c(n,x) = \frac{1}{(1+r)^{N-n}} \sum_{j=0}^{N-n} \binom{N-n}{j} \tilde{p}^j (1-\tilde{p})^{N-n-j} (x u^j d^{N-n-j} - K)^+$$

## Algorithmic / Method Toolkit

| Method | When to Use |
|--------|-------------|
| **FM algorithm** | Finding arbitrage (system of linear inequalities) or price bounds (LP) |
| **Replication** | Solve for (G₀,…) s.t. portfolio payoff = derivative payoff |
| **Delta-hedging formula** | Quick pricing in binomial: Δ = (V_up − V_down)/(S_up − S_down) |
| **RNPF** | Price = E^Q[discounted payoff]; use in one-period and multi-period |
| **Backward induction** | Multi-period: start at VN, work back using Vₙ = p̃ Vₙ₊₁(H) + q̃ Vₙ₊₁(T) |
| **American option algorithm** | Vₙ = max(intrinsic, continuation) at each node |
| **Markov pricing** | If W is Q-Markov and V_N = f(W_N), then Vₙ = fₙ(Wₙ) by backward induction |
| **Independence lemma** | E[f(X,Y)|G] = g(X) where g(x) = E[f(x,Y)], when Y indep of G |
