# Options Pricing — Exercises & Methods

Organised by problem type. For each type: method/recipe, key formulas, common pitfalls.

---
## A. Checking No-Arbitrage in the Binomial Model

### Recipe
1. Identify $u, d, r$ (or $U_n, D_n, R_n$ for multi-period)
2. Check: $d < 1+r < u$
   - If violated → construct explicit arbitrage:
     - $1+r \ge u$: put everything in bond
     - $1+r \le d$: borrow from bond, buy stock
3. For multi-period: check at **every node** (Theorem 87)

### Common Pitfall
- Forgetting that $u, d$ are **gross** return factors ($S_1 = S_0 u$ or $S_0 d$), not percentage returns
- In multi-period, $U_n, D_n, R_n$ can depend on $\omega$ and $n$

## B. Finding Arbitrage with FM Algorithm

### Setup
Given payoff matrix, find portfolio $G$ s.t. $G \cdot \Delta\bar{S}(\omega) \ge 0$ for all $\omega$, with strict inequality somewhere.

### FM Algorithm Steps
1. Write system of linear inequalities $Ax \le b$ (or $Ax > 0$)
2. **Eliminate** one variable at a time:
   - Separate inequalities by sign of coefficient of chosen variable
   - Combine +ve with −ve to eliminate; keep zero-coefficient ones
3. Repeat until 0 variables remain
4. **If** final system is consistent → back-substitute for solution
5. **If** contradictory → no solution (no arbitrage)

### Tips
- Choose variable that minimizes number of new inequalities
- Keep track carefully: # new inequalities = (# positive) × (# negative) + (# zero)
- For finding arbitrage, set up as: $G \cdot \Delta\bar{S}(\omega_i) \ge 0$ for all $i$, plus $\sum_i G \cdot \Delta\bar{S}(\omega_i) > 0$

## C. Pricing by Replication (One-Period)

### Recipe
Given derivative payoff $X = (X(\omega_1), \ldots, X(\omega_M))$, find portfolio $G$ replicating $X$:

1. Solve: $G \cdot S_1(\omega_i) = X(\omega_i)$ for each $\omega_i$
2. Price = initial cost $= G \cdot S_0$

### Binomial Model (quick)
$$\Delta = G_0^1 = \frac{X(H) - X(T)}{S_1(H) - S_1(T)}, \qquad G_0^0 = \frac{X(H) - \Delta \cdot S_1(H)}{B_1}$$
$$\text{Price} = G_0^0 B_0 + G_0^1 S_0$$

### Alternatively via RNPF
$$V_0 = \frac{\tilde{p}\,X(H) + \tilde{q}\,X(T)}{1+r}, \qquad \tilde{p} = \frac{1+r-d}{u-d}$$

## D. Finding Price Bounds (Incomplete Markets)

When derivative is **not replicable**, AFP is an open interval $(d(X), u(X))$.

### Method 1: LP + FM Algorithm
1. Upper bound: maximize $E^Q[\bar{X}_1]$ over $Q \in M$
2. Lower bound: minimize $E^Q[\bar{X}_1]$ over $Q \in M$
3. Reformulate as LP, solve with FM algorithm

### Method 2: Via EMMs (Corollary 70–71)
$$\text{AFP}(X) = \{E^Q[\bar{X}_1] : Q \in M\}$$
- Find all EMMs $Q$ (i.e. all $Q \sim P$ with $E^Q[\bar{S}_1] = \bar{S}_0$)
- Compute $E^Q[\bar{X}_1]$ for extreme EMMs → gives bounds

### Worked Example Pattern
Call option: $u(X) = 5/12$, $d(X) = 3/8$ from FM on LP formulation.

## E. Multi-Period Backward Induction Pricing

### Recipe
1. Draw the **binomial tree** for $S$ (or for $(S, W)$ if exotic)
2. Compute risk-neutral probabilities: $\tilde{p}_n = \frac{(1+R_n)-D_n}{U_n - D_n}$
3. Set terminal: $V_N = $ payoff
4. **Backward step** ($n = N-1, \ldots, 0$):
$$V_n = \tilde{p}_n \cdot V_{n+1}(\cdot, H) + \tilde{q}_n \cdot V_{n+1}(\cdot, T)$$
5. **Hedging** at each node:
$$G_n = \frac{V_{n+1}(H) - V_{n+1}(T)}{S_{n+1}(H) - S_{n+1}(T)}$$

### Constant coefficients
- $\tilde{p}$ is the same at every node → simplifies greatly
- Tree is **recombinant**: only $n+1$ nodes at time $n$

### Common Exam Pattern
- $N = 2$ or $N = 3$ period tree, compute all node prices, then read off $V_0$ and hedging strategy

## F. American Option Pricing

### Recipe
1. Compute the **European** price tree by backward induction (gives continuation values $C_n$)
2. At each node, compare: intrinsic value $I_n$ vs continuation $C_n$
$$V_n = \max(I_n,\, C_n)$$
3. Terminal: $V_N = I_N^+ = (\text{payoff})^+$
4. **Optimal exercise boundary:** exercise at first time $I_n = V_n$ (i.e. intrinsic = option value)
5. $\tau^* = \inf\{n \le N : I_n = V_n\}$

### Example: American Put ($I_n = K - S_n$)
- Deep in-the-money: early exercise may be optimal
- Always: American value ≥ European value

### Pitfalls
- Don't forget: $V_N = I_N^+ = \max(I_N, 0)$, NOT just $I_N$
- The continuation value uses the **option** values $V_{n+1}$, not the European values

## G. Markov Pricing (Path-Dependent Options)

### When to Use
Payoff depends on path (running max, min, average) → can't just use $S_n$ alone.

### Recipe
1. **Choose** Markov process $W = (S, C)$ where $C$ captures path-dependence (e.g. $C = M = \max S$)
2. **Verify** $W_{n+1} = h_n(W_n, X_{n+1})$ with $X_{n+1}$ indep. of $F_n$ → Corollary 135 gives Markov
3. **Terminal:** $V_N = f_N(W_N)$ (write payoff as function of $W$)
4. **Backward induction on pricing function:**
$$f_n(w) = \tilde{p}\, f_{n+1}(h_n(w,H)) + \tilde{q}\, f_{n+1}(h_n(w,T))$$
5. **Hedging function:**
$$g_n(w) = \frac{f_{n+1}(h_n(w,H)) - f_{n+1}(h_n(w,T))}{s_{n+1}(h_n(w,H)) - s_{n+1}(h_n(w,T))}$$

### Proving a Process is NOT Markov
Find time $n$ and two paths $\omega', \omega''$ s.t.:
- $W_n(\omega') = W_n(\omega'')$ (same present value)
- But $\text{Law}(W_{n+1}|\omega') \ne \text{Law}(W_{n+1}|\omega'')$ (different futures)

Example: $M_n = \max_{k \le n} S_k$ at nodes $TH$ and $TT$ — same $M_2 = 4$ but different conditional distributions.

### Key Insight
- $M$ alone is **not** Markov, but $(S, M)$ **is** Markov
- Choose $W$ to take **fewest values** for computational efficiency
- Only need to evaluate $f_n$ at Im$(W_n)$ — linear growth, not exponential

## H. Proving Key Theorems

### Theorems the Exam May Ask You to Prove

| Theorem | Strategy |
|---------|----------|
| **Thm 24** (Binomial NA) | Two cases: $1+r \ge u$ → bond dominates; $1+r \le d$ → stock dominates. Converse: $V_1 = G(S_1 - (1+r)S_0)$ takes both signs |
| **Thm 40** (NA ⟺ Domination) | (⟸) Domination + arb ⟹ price ≥ 0 but cost = 0, contradiction. (⟹) Contrapositive, construct arb from domination failure |
| **Thm 65** (1st FTAP) | (⟸) EMM + arb → $E^Q[V_1] = 0$ but $V_1 \ge 0, > 0$ somewhere → contradiction. (⟹) Separation theorem: separate cone of achievable payoffs from positive orthant |
| **Prop 46** | Replicable → unique price by law of one price. Not replicable → AFP is open interval, use LP bounds |
| **Thm 87** (Multi-period decomposition) | Show multi-period arb decomposes into one-period arb at some node |
| **Thm 117** (Multi-period FTAP) | (⟸) $E^Q[V_N^{0,G}] = \sum E^Q[G_n(\bar{S}_{n+1}-\bar{S}_n)] = 0$ by martingale. (⟹) Binomial: Thm 88 → $\tilde{P} \in (0,1)$ → Lemma 114 |

### Proofs You Should Know Cold
1. Theorem 24 (short, clean)
2. Theorem 65 ⟸ direction (3 lines)
3. Theorem 117 ⟸ direction (uses martingale property)

| One-period RNPF | $V_0 = \frac{\tilde{p} V_1(H) + \tilde{q} V_1(T)}{1+r}$ |
| Multi-period RNPF | $V_n = E^Q[V_{n+1}|F_n]$, $V_0 = E^Q[V_N]$ |
| American option | $V_n = \max(I_n, C_n)$, $C_n = \tilde{p} V_{n+1}(H) + \tilde{q} V_{n+1}(T)$ |
| Optimal exercise | $\tau^* = \inf\{n: I_n = V_n\}$ |
| Call (closed form) | $c(n,x) = (1+r)^{-(N-n)} \sum_j \binom{N-n}{j} \tilde{p}^j \tilde{q}^{N-n-j} (xu^j d^{N-n-j} - K)^+$ |

| NA condition | $d < 1+r < u$ (binomial) |
| 1st FTAP | NA ⟺ $\exists$ EMM |
| 2nd FTAP | Complete ⟺ unique EMM |

## J. Exam Strategy

### Question Type Catalogue

| Topic | Typical Question |
|-------|------------------|
| Ch 1 | Replicate a forward/binary, explain hedging vs speculation |
| Ch 2 (NA) | Check NA condition, find explicit arbitrage if violated |
| Ch 2 (FM) | Apply FM algorithm to find arbitrage or compute price bounds |
| Ch 2 (Pricing) | Price option in one-period model via replication or RNPF |
| Ch 2 (FTAP) | Prove/state 1st or 2nd FTAP; find all EMMs |
| Ch 3 (Backward) | Price European option on 2–3 period tree; find hedging strategy |
| Ch 3 (American) | Price American put/call; find optimal exercise boundary |
| Ch 3 (Exotic) | Barrier, chooser, lookback — set up backward induction |
| Ch 3 (Markov) | Prove process (not) Markov; set up Markov pricing for path-dependent option |
| Ch 3 (Cond. Exp.) | Compute conditional expectation; verify tower/independence properties |

### Time Management
- Tree-drawing and backward induction are mechanical but time-consuming → practice speed
- FM algorithm: careful bookkeeping saves re-work
- Proof questions (FTAP, NA): memorize the ⟸ direction (shorter); know the ⟹ structure
- Markov pricing: setting up $W$, $h_n$, $f_N$ correctly is the hard part; backward step is then routine
