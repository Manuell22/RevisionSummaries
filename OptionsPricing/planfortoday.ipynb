## Block 1 — Chapter 1 + §2.1–2.5: Derivatives, Replication & Pricing Basics (14:45–15:50, 65 min)

**Sections:** §1.1–1.3, §2.1–2.5 (~530 lines)  
**Coverage:** Week 1 + start of Week 2

### Content to cover:
- **§1.1 The Forward Contract:**
  - Definition: agreement to buy asset at forward price $K$ at expiry $T$; costs nothing at time 0
  - Payoff: $F_T = P_T - K$ — value derives from the underlying $P_T$
  - Purpose: easyJet transfers fuel price risk to bank; hedging vs speculating
  - Bank hedges by trading with other counterparties (e.g. Koch Industries)
- **§1.2 The Binary Option:**
  - Payoff: $B_T = 100 \cdot \mathbf{1}_{\{P_T \geq 59\}}$, where $\mathbf{1}_{\{x \geq b\}} = 1$ if $x \geq b$, else $0$
  - Binary option is in the money if $P_T \geq K$, out of the money otherwise
  - Cost of option tied to probability of prediction coming true (Remark 3)
- **§1.3 Speculation:**
  - Comparing P&L of binary options vs buying gold directly — options magnify gains *and* losses
  - Binary options banned in many jurisdictions as gambling; Barings Bank collapse (1995)
- **§2.1 Pricing by replication (Principle 5 & 6):**
  - **Principle 5 (Law of One Price):** same payoff at $T$ $\Rightarrow$ same value at all previous times
  - **Principle 6 (Pricing via replication):** derivative price = value of replicating portfolio
  - Forward price derivation: $K = L = P_0(1+r)$ (borrow to buy underlying)
  - **Hedging = $-$ Replicating;** short = sold, long = bought, writer = seller
- **§2.2 Pricing a binary option:**
  - Model: $\Omega = \{g, b\}$, $P_T(g) = 61$, $P_T(b) = 56$, $P_0 = 58$, $r = 0$
  - Replicating portfolio: solve $\begin{cases} x + 3h = 100 \\ x - 2h = 0 \end{cases}$ giving $h = 20$, $x = 40$
  - **Replication depends on the model** (Remark 8): changing values changes the price; with 3 states, option may not be replicable
- **§2.3–2.4 Dependence on probabilities & model uncertainty:**
  - Prices depend on model, not just on probabilities — different models $\Rightarrow$ different prices
  - Model risk (Knightian uncertainty): we don't know the true probability distribution
  - "All models are wrong, but some are useful" (George Box)
- **§2.5 What can be used as underlying:**
  - Commodities, bonds, stocks, indices, FX, other derivatives, weather (but NA theory needs traded underlying)

### Key things to understand deeply:
- The replicating portfolio system of equations: set up and solve for $(x, h)$ — this is the core computational skill
- Why the forward contract has a model-independent hedge but the binary option does not
- Law of One Price is the foundation — everything else builds on it
- Model dependence is the Achilles' heel: small model changes can cause large price changes

## Block 2 — §2.6–2.11: Money Markets, No-Arbitrage Principle & Dividends (16:00–17:05, 65 min)

**Sections:** §2.6–2.11 (~481 lines)  
**Coverage:** Week 2 (continued) + Week 3 start

### Content to cover:
- **§2.6 The bank account and the bond:**
  - Bond value: $B_0 = 1$ (normalisation), discrete time $B_{t+1} = B_t(1 + R_t)$, continuous time $dB_t = B_t R_t \, dt$
  - Constant rate: $B_t = (1+r)^t$ (discrete) or $B_t = e^{rt}$ (continuous)
  - Interest rate can be negative ($R_t > -1$ always assumed); real-world examples (Europe/Japan 2008–2023)
  - In reality: no single "interest rate" — many contracts, depends on creditworthiness
- **§2.7 Unspoken modelling assumptions (Remarks 12–15):**
  - **Information:** all participants have same info (no insiders)
  - **Linearity:** buying $h$ shares costs $hS_0$ — no price impact, no transaction costs
  - **Frictionless markets:** no bid-ask spread, no taxes, no margin requirements
  - **Short-selling:** can sell what you don't own (borrow shares, sell, return later)
  - **Divisibility:** can buy fractional shares ($h \in \mathbb{R}$, not just $h \in \mathbb{Z}$)
- **§2.8 Justification of the law of one price:**
  - Exchange rate example: if $A \neq B$, arbitrageurs borrow & trade, law of demand/supply restores equilibrium
  - Arbitrage $\Rightarrow$ market not at equilibrium $\Rightarrow$ forces push prices back
  - Transaction costs can eliminate apparent arbitrage (Remark 16)
- **§2.9 The no-arbitrage principle:**
  - **Definition 17 (Arbitrage):** portfolio $L$ with $V_0^L = 0$, $V_T^L \geq 0$ a.s., and $P(V_T^L > 0) > 0$
  - **Definition 18:** "a.s." = almost surely = probability 1
  - Remark 19: NA $\iff$ any 0-capital portfolio with $V_T \geq 0$ a.s. has $V_T = 0$ a.s.
  - **Binomial model:** $S_1(H) = uS_0$, $S_1(T) = dS_0$; portfolio final wealth $W = S_1 - S_0(1+r)$
  - **Theorem 24:** Binomial model is arbitrage-free $\iff$ $d < 1 + r < u$
- **§2.10 Prices of liquid and illiquid goods:**
  - Liquid goods priced by market (law of demand/supply); illiquid goods need other methods
  - Option pricing: price *illiquid* derivatives based on *liquid* underlying using NA theory
- **§2.11 Dividends:**
  - Cum-dividend $V = S + D$; must use cum-dividend price for arbitrage checks
  - Forward price with dividends: $F = S_0(1+r) - D$
  - Short-seller owes dividend compensation to lender

### Key things to understand deeply:
- **Theorem 24 is fundamental:** $d < 1+r < u$ — the interest rate must sit between the down and up factors
- The formal definition of arbitrage (Def 17) — make sure you can state it precisely with the a.s. conditions
- The modelling assumptions are *necessary simplifications*, not truths — know what each one means
- Dividends change the forward price formula: $F = S_0(1+r) - D$, not $F = S_0(1+r)$

## Block 3 — §2.12–2.15 (Part 1): Discounting, Finite Probability Spaces & FM Algorithm Intro (17:15–18:20, 65 min)

**Sections:** §2.12–2.15 first half (~537 lines)  
**Coverage:** Week 3 (continued)

### Content to cover:
- **§2.12 Discounting and Numeraire:**
  - Discounted value: $\bar{W}_t := W_t / B_t$ — "present value" or "real terms"
  - Portfolio value in discounted terms: $\bar{V}_t^{x,h} = x + h \cdot (\bar{S}_t - S_0)$ (eq. 5)
  - **Key advantage:** in discounted terms $r = 0$ effectively; simplifies all formulas
  - Gains from trade: $\bar{V}_t^{x,h} - \bar{V}_0^{x,h} = h \cdot (\bar{S}_t - S_0)$ (eq. 6) — increment of discounted price × strategy
  - Continuous-time preview: gains become stochastic integral $\int_0^s H_t \cdot d\bar{S}_t$
  - **Numeraire:** any portfolio with strictly positive value; bond is standard; choice doesn't affect qualitative properties (arbitrage, replicability)
- **§2.13 Finite Probability Spaces:**
  - $\Omega = \{\omega_i : i = 1, \ldots, n\}$ with $P(\{\omega_i\}) > 0$ for all $i$ (w.l.o.g.)
  - Random variables identified with vectors in $\mathbb{R}^n$; random vectors with matrices
  - Probability $P$ identified with positive vector $p \in \mathbb{R}_+^n$ with $\sum_i p_i = 1$
  - All pricing/hedging problems reduce to **linear algebra** on finite prob. spaces
  - W.l.o.g. $S_1$ is injective: any r.v. is a function of $S_1$, i.e. any r.v. is a derivative payoff (Remark 27)
  - Complete models on finite spaces have $|\Omega| \leq m + 1$ (Remark 28)
- **§2.14 How to find arbitrage (linear algebra formulation):**
  - $W := \text{span}\{S_1^j - S_0^j : j = 1, \ldots, m\}$ = set of discounted payoffs replicable at cost 0 (eq. 7)
  - **Arbitrage payoffs** = $W \cap (\mathbb{R}_+^n \setminus \{0\})$ (eq. 8)
  - NA $\iff$ $W \cap \mathbb{R}_+^n = \{0\}$ — a vector space meets the positive orthant only at the origin
  - Need a method to check if a polyhedron is empty $\Rightarrow$ enter the FM algorithm
- **§2.15 The Fourier-Motzkin algorithm (introduction):**
  - For equalities $Ax = b$: eliminate variables one at a time (standard Gaussian-style)
  - Example 29: solve $2x + 3y = 6$, $x + 3y = 1$ by eliminating $y$ then solving for $x$
  - **Generalisation to inequalities** $Ax \geq b$: partition rows into $I_>, I_<, I_=$ based on coefficient sign of eliminated variable
  - New system (eq. 16): pair every upper bound with every lower bound $\Rightarrow$ one fewer variable
  - Solution recovery: solve reduced system, then pick $z \in [\max_{j \in I_>} \ldots, \min_{i \in I_<} \ldots]$ (eq. 17)

### Key things to understand deeply:
- Discounting is not just bookkeeping — it fundamentally simplifies the math (gains = $h \cdot \Delta\bar{S}$)
- The identification "random variable ↔ vector in $\mathbb{R}^n$" turns probability into linear algebra
- $W \cap \mathbb{R}_+^n = \{0\}$ is the geometric characterisation of no-arbitrage — visualise it
- The FM algorithm is the key computational tool for the next several lectures — master the elimination procedure

## Block 4 — §2.15 (Part 2) + §2.16: FM Algorithm Mastery & Finding Arbitrage (18:40–19:45, 65 min)

**Sections:** §2.15 second half + §2.16 (~482 lines)  
**Coverage:** Week 3 (end) + Week 4 start

### Content to cover:
- **§2.15 FM algorithm (continued) — Example 31:**
  - System of 5 inequalities in 3 variables: full worked elimination
  - Step 1: eliminate $x_3$ — partition into $I_> = \{1,2\}$, $I_< = \{3,4\}$, $I_= = \{5\}$
  - Pair every lower bound with every upper bound: $|I_<| \times |I_>| = 4$ new inequalities + those from $I_=$
  - Step 2: eliminate $x_2$ — reduce to single inequality in $x_1$: $x_1 \geq 24/11$
  - **Solution recovery:** choose $x_1 \geq 24/11$, then $x_2$ from (eq. 20), then $x_3$ from interval (eq. 21)
  - Finding one specific solution: e.g. $x_1 = 3, x_2 = 10, x_3 = -1$
- **Remark 32:** FM doesn't give a clean description of the solution set (unlike simplex)
- **Remark 33 (Geometric interpretation):**
  - Eliminating variables = computing projections: $\pi_y(P) = \{y : \exists z \text{ s.t. } (y,z) \in P\}$
  - Each FM iteration computes $\pi_{n-1}^n(P)$ and shows it's a polyhedron
  - Projection of a polyhedron is a polyhedron
- **Remark 34:** FM is simple but slow — number of inequalities can explode; simplex is faster in practice
- **§2.16 How to find arbitrage with the FM algorithm:**
  - **Example 35 (Trinomial model):** $r = 1/9$, two stocks, $n = 3$ states
  - Compute discounted increments: $\bar{S}_1^1 - S_0^1 = (1, 1, -1)^T$, $\bar{S}_1^2 - S_0^2 = (2, -2, -2)^T$
  - $W = \{h_1(1,1,-1)^T + h_2(2,-2,-2)^T : h \in \mathbb{R}^2\}$
  - Check $W \cap \mathbb{R}_+^3 = \{0\}$: set up system $w_1 \geq 0, w_2 \geq 0, w_3 \geq 0$ with $w = h_1 \bar{\Delta S}^1 + h_2 \bar{\Delta S}^2$
  - Substitute, eliminate variables using FM — determine whether solutions exist with $w \neq 0$
  - If arbitrage exists: read off the arbitrage portfolio $(h_1, h_2)$ from FM output

### Key things to understand deeply:
- The FM algorithm in action: be able to do a full 3-variable elimination by hand
- The number of inequalities grows as $|I_<| \times |I_>|$ per step — this is why FM is impractical for large problems
- Geometric picture: FM = iterated projection of a polyhedron
- For arbitrage detection: translate $W \cap \mathbb{R}_+^n \setminus \{0\} \neq \emptyset$ into a system of inequalities, then apply FM

## Block 5 — §2.16 (end) + §2.17: Arbitrage Detection & Domination Principles (19:55–21:00, 65 min)

**Sections:** §2.16 completion + §2.17 (~473 lines)  
**Coverage:** Week 4

### Content to cover:
- **§2.16 (completion) — Trinomial model example wrap-up:**
  - Complete the FM elimination: determine whether $W \cap \mathbb{R}_+^3 \setminus \{0\} \neq \emptyset$
  - If arbitrage exists: identify the arbitrage portfolio — e.g. $h_2 < 0$, $h_1 = -2h_2$
  - Verify: check that the portfolio gives non-negative payoff in all states and strictly positive in at least one
- **§2.17 The no-arbitrage and the domination principles:**
  - **Principle 36 (Weak Domination Principle):** if $V_T^L \leq V_T^M$ a.s., then $V_t^L \leq V_t^M$ for all $t < T$
  - **Super/sub-replication:** $U$ super-replicates $X$ if $V_T^U \geq X_T$ a.s.; $D$ sub-replicates if $V_T^D \leq X_T$ a.s.
  - **Principle 37 (Pricing via super/sub-replication):** derivative price must lie between sub- and super-replicating costs
  - **Price bounds:** $u(X) := \inf\{V_0^U : \bar{V}_T^U \geq \bar{X}_T\}$, $d(X) := \sup\{V_0^D : \bar{V}_T^D \leq \bar{X}_T\}$ (eqs. 25–26)
  - If $X$ is replicable: $d(X) = u(X) = $ replication price (interval collapses to a point)
  - **Principle 38 (Domination Principle):** strict domination + law of one price combined
  - **Theorem 40 (the big equivalence):** In the linear one-period model, the following are equivalent:
    1. Domination principle holds
    2. Strict Domination principle holds
    3. No arbitrage
  - **Full proof of Theorem 40:** $(1 \Rightarrow 2)$ trivial; $(2 \Rightarrow 3)$ arbitrage violates strict domination; $(3 \Rightarrow 1)$ by contradiction — construct portfolio $N = (0, h-g) = (M-L) + (x-y, 0)$ and show $N$ is an arbitrage
  - **Uniform arbitrage:** portfolio $L$ with $V_0^L = 0$ and $V_T^L \geq c > 0$ a.s.
  - **Theorem 42:** Weak Domination $\iff$ no uniform arbitrage; Domination $\Rightarrow$ Weak Domination $\Rightarrow$ Law of One Price (reverse implications fail)

### Key things to understand deeply:
- The hierarchy: **Domination $\Rightarrow$ Weak Domination $\Rightarrow$ Law of One Price** (but not the reverse!)
- **Theorem 40 is the key structural result:** NA = Domination Principle in linear one-period models
- The proof technique of Theorem 40: constructing the arbitrage portfolio $N$ from the violation — understand this construction
- Price bounds $d(X) \leq u(X)$: for incomplete markets, there's an interval of no-arbitrage prices, not a unique one
