## Chapter 1 — What Are Derivatives?

A **derivative** (a.k.a. contingent claim) is a financial instrument whose payoff depends on an underlying asset.

### Forward Contracts (Example 1)
- Agreement to buy/sell asset at future date $T$ for strike $K$
- Payoff at maturity: $S_T - K$ (long), $K - S_T$ (short)
- No cost to enter (chosen so fair at inception)

### Binary (Digital) Options (Example 2)
- Cash binary: pays 1 if $S_T \ge K$, else 0
- Asset binary: pays $S_T$ if $S_T \ge K$, else 0

### Speculation vs Hedging
- Derivatives can be used for **speculation** (P&L comparison: option vs underlying) or **hedging** (reducing risk)

### Fundamental Principles
- **Principle 5 (Law of One Price):** Portfolios with identical payoffs have the same price
- **Principle 6 (Pricing by Replication):** If portfolio $G$ replicates derivative $X$, then price of $X$ = cost of $G$
- **Remark 7:** Hedging = −Replicating (opposite sign portfolio)

## Chapter 2 — One-Period Models

### 2.1 Setup and Modelling Assumptions

**Model:** Two times $t=0,1$. Bond $B$ and stock $S$ on finite probability space $(\Omega, \mathcal{A}, P)$.

- **Bank account/Bond:** $B_t = B_0(1+r)^t$ (discrete), $B_t = B_0 e^{rt}$ (continuous)
- **Remark 12:** We use all available information (no insider info, no partial knowledge)
- **Remark 13 (Linearity):** Cost of $\alpha$ units of asset = $\alpha \times$ price of one unit
- **Remark 14–15 (Short-selling):** Allowed — sell what you don't own by borrowing
- **Underlying types:** Commodities, bonds, stocks, indices, FX, other derivatives

### 2.2 Arbitrage

**Definition 17 (Arbitrage):** A self-financing portfolio (starting from 0 wealth) $G$ is an arbitrage if:
$$V_1^{0,G} \ge 0 \text{ a.s.} \quad \text{and} \quad P(V_1^{0,G} > 0) > 0$$

**Definition 18 (a.s.):** An event holds almost surely (a.s.) if it occurs with probability 1.

**Remark 19–22:** Discussion of the subtlety of a.s. vs surely; model-dependent nature of arbitrage.

### 2.3 Binomial Model — No-Arbitrage Condition

**Example 23 (Binomial Model):** $\Omega = \{H,T\}$, with $S_1(H) = S_0 u$, $S_1(T) = S_0 d$.

**Theorem 24:** The binomial model is arbitrage-free if and only if
$$d < 1+r < u$$

*Proof sketch:*
- If $1+r \ge u$: invest everything in bond (arbitrage)
- If $1+r \le d$: borrow from bond, buy stock (arbitrage)
- If $d < 1+r < u$: any portfolio $V_1 = G_0(S_1 - (1+r)S_0)$ takes both signs → no arbitrage

### 2.4 Discounting, Numeraire, Dividends

- **Remark 25 (Discounting):** Work with $\bar{S}_t = S_t / B_t$ (discounted price)
- **Remark 26 (Numeraire):** $B$ serves as numeraire; all prices quoted relative to $B$
- **Dividends:** Bring asset $S$ forward by subtracting dividend flow; liquid vs illiquid goods distinction

### 2.5 Finite Probability Spaces

- **Remark 27–28:** $\Omega = \{\omega_1, \ldots, \omega_M\}$, $P(\omega_i) = p_i > 0$
- Portfolio $G = (G_0^0, G_0^1, \ldots, G_0^d)$ investing in bond + $d$ stocks
- Payoff at time 1: $V_1^{0,G}(\omega) = G_0 \cdot (\bar{S}_1(\omega) - \bar{S}_0)$

### 2.6 Finding Arbitrage — FM Algorithm

**Problem:** Given system $Ax \le b$ (or $Ax > 0$), find a solution or prove none exists.

**Fourier–Motzkin (FM) Algorithm:**
1. Pick a variable $x_k$ to eliminate
2. Classify inequalities into: $x_k$ has +ve coeff, −ve coeff, or zero coeff
3. Form new system by combining each +ve inequality with each −ve inequality to cancel $x_k$; keep zero-coeff inequalities
4. Repeat until no variables remain
5. Check final numerical inequalities; if consistent → back-substitute to find solution; if contradictory → no solution

**Application:** Finding arbitrage = finding $G$ s.t. $G \cdot (\bar{S}_1(\omega) - \bar{S}_0) \ge 0$ for all $\omega$ and $> 0$ for some $\omega$. Reformulate as system of linear inequalities and apply FM.

**Examples 29, 31, 35:** Worked FM algorithm applications for arbitrage detection.

### 2.7 Domination Principles

**Principle 36 (Weak Domination):** If $X \ge Y$ a.s. then $\text{price}(X) \ge \text{price}(Y)$

**Principle 37 (Strict Domination):** If additionally $P(X > Y) > 0$ then $\text{price}(X) > \text{price}(Y)$

**Principle 38:** If the market is arbitrage-free and $X \ge 0$ a.s. then $\text{price}(X) \ge 0$

**Theorem 40:** The market is arbitrage-free ⟺ the Domination Principle holds.

*Proof sketch:*
- (⟸) If domination holds, a portfolio worth ≥0 with some >0 must have non-negative initial cost, so no free lunch
- (⟹) If NA but domination fails, there exists a derivative priced wrongly → construct arbitrage (contradiction)

### 2.8 Arbitrage-Free Prices and Replicability

**Definition 44 (AFP):** The set of arbitrage-free prices for derivative $X$:
$$\text{AFP}(X) = \{p \in \mathbb{R} : \text{market } + \text{derivative at price } p \text{ is NA}\}$$

**Lemma 45:** If market is NA, then AFP(X) is a non-empty interval.

**Proposition 46:**
- If $X$ is **replicable**: AFP(X) = singleton (unique fair price = replication cost)
- If $X$ is **not replicable**: AFP(X) = open interval $(d(X), u(X))$

*Proof:* Replicable → law of one price forces unique value. Not replicable → price bounds from LP, strict inequalities because you can't perfectly offset.

**Definition 48 (Complete Market):** Every derivative is replicable.

### 2.9 Pricing as Linear Programming

**Definition 49:** A **linear program (LP)** is: maximize $c \cdot x$ subject to $Ax \le b$.

**Pricing formulation (eqs 32–34):**
- Upper bound $u(X) = \max\{x_0 \cdot \bar{X}_1 : \text{constraints}\}$
- Lower bound $d(X) = \min\{\ldots\}$

**Theorem 51:** LP duality connects primal (pricing) with dual (EMM characterization).

**Corollary 52:** Price bounds can be computed via FM algorithm on the LP.

**Worked Example:** Call option pricing using FM algorithm → $u(X) = 5/12$, $d(X) = 3/8$.

### 2.10 Binomial Model Pricing Formulas

**Risk-neutral probability:**
$$\tilde{p} = \frac{1+r-d}{u-d}, \qquad \tilde{q} = 1 - \tilde{p} = \frac{u-(1+r)}{u-d}$$

**RNPF (Risk-Neutral Pricing Formula):**
$$V_0 = \tilde{p}\,\bar{V}_1(H) + \tilde{q}\,\bar{V}_1(T) = E^Q[\bar{V}_1]$$

**Delta-Hedging Formula (eq 47):**
$$G_0 = \Delta = \frac{V_1(H) - V_1(T)}{S_1(H) - S_1(T)}$$

**Example 57:** Full worked binomial pricing.

### 2.11 EMM and the Fundamental Theorems

**Definition 60 (EMM):** $Q \sim P$ such that $\bar{S}$ is a $Q$-martingale, i.e. $E^Q[\bar{S}_1] = \bar{S}_0$.

**Sets:** $M = \{Q \sim P : \bar{S} \text{ is } Q\text{-martingale}\}$, $\bar{M} = \{Q \ge 0 : E^Q[\bar{S}_1] = \bar{S}_0\}$

---

**Theorem 65 (1st FTAP):** NA ⟺ $M \ne \emptyset$ (there exists an EMM).

*Proof (finite Ω):*
- (⟸) If $Q \in M$ exists and $G$ is an arbitrage, then $E^Q[V_1^{0,G}] = G_0 \cdot E^Q[\bar{S}_1 - \bar{S}_0] = 0$, but $V_1^{0,G} \ge 0$ Q-a.s. with $Q(V_1^{0,G}>0)>0$ → contradiction
- (⟹) Uses **Separation Theorem 66**: if two convex sets don't intersect, a hyperplane separates them. Construct EMM from the separating functional.

**Theorem 68 (Uniform Arbitrage):** No uniform arbitrage ⟺ $\bar{M} \ne \emptyset$.

---

**Corollary 70–71 (RNPF):**
$$\text{AFP}(X) = \{E^Q[\bar{X}_1] : Q \in M\}$$

Any arbitrage-free price equals discounted expected payoff under some EMM.

---

**Corollary 75:** $X$ is replicable ⟺ $E^Q[\bar{X}_1]$ is the same for all $Q \in M$.

**Corollary 76 (2nd FTAP):** Market is complete ⟺ $|M| = 1$ (unique EMM).

*Consequence:* In binomial model (complete), $\tilde{p}$ uniquely determines pricing.

## Chapter 3 — Multi-Period Models

### 3.1 Information, Filtrations, Measurability

**Lemma 81–82:** A random variable $X$ on finite $\Omega$ is $\mathcal{G}$-measurable iff it is constant on each **atom** of $\mathcal{G}$.

**Definition 83 (σ-algebra of information):** At time $n$, information is the σ-algebra $F_n = \sigma(X_1, \ldots, X_n)$.

**Definition 84 (Filtration):** $\mathbb{F} = (F_n)_{n=0}^N$ with $F_0 \subseteq F_1 \subseteq \cdots \subseteq F_N \subseteq \mathcal{A}$.

An adapted process: $W_n$ is $F_n$-measurable for each $n$.

### 3.2 Self-Financing Portfolios

Portfolio $G = (G_n)_{n=0}^{N-1}$ is **self-financing** if wealth changes come only from price changes:
$$V_n^{x,G} = x + \sum_{k=0}^{n-1} G_k \cdot (\bar{S}_{k+1} - \bar{S}_k) \quad \text{(discrete stochastic integral, eqs 65–70)}$$

No external cash injection/withdrawal between times.

### 3.3 Multi-Period Arbitrage Theorems

**Definition 85 (Multi-period Arbitrage):** As before, self-financing from 0 with $V_N \ge 0$ a.s. and $P(V_N > 0) > 0$.

**Theorem 86:** Multi-period NA ⟺ Domination Principle holds.

**Theorem 87 (Key!):** Multi-period arbitrage exists ⟺ there exists a one-period arbitrage at some node.

*This decomposes the global NA problem into local (one-period) checks.*

**Theorem 88 (Multi-period Binomial NA):** NA ⟺ $D_n < 1 + R_n < U_n$ for all $n$.

With constant parameters: NA ⟺ $d < 1+r < u$.

### 3.4 Backward Induction Pricing

In the multi-period binomial model, price a derivative with payoff $Y_N$ by:

1. Set $V_N = Y_N$
2. For $n = N-1, N-2, \ldots, 0$:
$$V_n(\omega_{(n)}) = \tilde{P}_n(\omega_{(n)})\, V_{n+1}(\omega_{(n)},H) + \tilde{Q}_n(\omega_{(n)})\, V_{n+1}(\omega_{(n)},T)$$

where the risk-neutral transition probabilities are:
$$\tilde{P}_n = \frac{(1+R_n) - D_n}{U_n - D_n}, \qquad \tilde{Q}_n = 1 - \tilde{P}_n$$

**Delta-hedging at each node:**
$$G_n(\omega_{(n)}) = \frac{V_{n+1}(\omega_{(n)},H) - V_{n+1}(\omega_{(n)},T)}{S_{n+1}(\omega_{(n)},H) - S_{n+1}(\omega_{(n)},T)}$$

**Examples 89–90:** Worked multi-period backward induction.

### 3.5 Exotic Options

#### Derivatives Paying Cashflows (eq 78)
Payoff stream $(Y_k)_{k=0}^N$ instead of single terminal payoff. Price = $\sum_{k} E^Q[\bar{Y}_k | F_n]$.

#### Random Maturity / Stopping Times
**Definition 94:** Random variable $\tau : \Omega \to \{0,\ldots,N\}$.

**Definition 95 (Stopping Time):** $\tau$ is a stopping time if $\{\tau \le n\} \in F_n$ for all $n$.
Equivalently, the decision to stop at or before time $n$ depends only on information up to $n$.

**Example 96 (Barrier Option):** Down-and-out call — payoff $(S_N - K)^+$ but only if $S$ never drops below barrier $L$. Maturity is $\tau = \min(N, \inf\{n : S_n \le L\})$.

#### Chooser Option (Example 97)
At time $n$, holder chooses between call and put. Price by backward induction, at choice node take max of call and put values.

### 3.6 American Options

Can be exercised at any time $n \le N$. Intrinsic value $I_n$ at time $n$.

**Pricing Algorithm (eqs 80–82):**
$$V_N = I_N^+ \qquad (\text{terminal payoff})$$
$$C_n = \tilde{p}\,V_{n+1}(H) + \tilde{q}\,V_{n+1}(T) \qquad (\text{continuation value})$$
$$V_n = \max(I_n,\, C_n) \qquad (\text{exercise vs continue})$$

**Optimal exercise time:**
$$\tau^* = \inf\{n \le N : I_n = V_n\}$$

**Example 99:** Worked American put option.

Key insight: American option is always worth $\ge$ European counterpart (extra flexibility).

### 3.7 Conditional Expectation

**Theorem 103:** For finite $\Omega$ and partition atoms $A$ of $\mathcal{G}$:
$$E[X|\mathcal{G}](\omega) = \frac{\sum_{\omega' \in A} X(\omega') P(\omega')}{P(A)} \quad \text{for } \omega \in A$$

**Definition 105:** The **atoms** of $\mathcal{G}$ are the minimal non-empty sets in $\mathcal{G}$.

**Theorem 107:** $E[X|\mathcal{G}]$ is the unique $\mathcal{G}$-measurable r.v. with $E[ZW] = E[XW]$ for all bounded $\mathcal{G}$-meas. $W$.

**Theorem 109:** Conditional expectation always exists (even for infinite $\Omega$).

**Theorem 110 (Kolmogorov characterization):** $Z = E[X|\mathcal{G}]$ iff $Z$ is $\mathcal{G}$-meas. and $E[ZW] = E[XW]$ for all bounded $\mathcal{G}$-meas. $W$.

**Theorem 111:** $E[X|\mathcal{G}]$ is the $L^2$-projection: minimizes $E[(X-C)^2]$ over $\mathcal{G}$-meas. $C$.

### 3.8 Properties of Conditional Expectation

1. **Linearity:** $E[X+Z|\mathcal{G}] = E[X|\mathcal{G}] + E[Z|\mathcal{G}]$
2. **Independence:** If $X$ indep. of $\mathcal{G}$, then $E[X|\mathcal{G}] = E[X]$
3. **Taking out what is known:** If $X$ is $\mathcal{G}$-meas., $E[XZ|\mathcal{G}] = X\,E[Z|\mathcal{G}]$
4. **Tower property:** If $\mathcal{G} \subseteq \mathcal{H}$, $E[E[X|\mathcal{H}]|\mathcal{G}] = E[X|\mathcal{G}]$; in particular $E[E[X|\mathcal{H}]] = E[X]$
5. **Jensen inequality:** If $\varphi$ convex, $E[\varphi(X)|\mathcal{G}] \ge \varphi(E[X|\mathcal{G}])$

### 3.9 RNPF in Multi-Period Binomial Model

**Risk-neutral transition probabilities (eq 91–92):**
$$\tilde{P}_n(\omega) = \frac{(1+R_n) - D_n}{U_n - D_n}, \qquad \tilde{Q}_n = 1 - \tilde{P}_n$$

**RNPF (eq 96):**
$$V_n = E^Q[V_{n+1}|F_n]$$

**Extended RNPF (eq 97):**
$$V_k = E^Q[V_n|F_k] \quad \text{for } 0 \le k \le n \le N$$

**In particular (eq 98):** $V_0 = E^Q[V_N]$.

**Lemma 114:** Bijection between probabilities $Q$ on $\Omega$ and $\mathbb{F}$-adapted transition probability processes $\tilde{P}$ with values in $[0,1]$. Moreover $Q \sim P$ iff $0 < \tilde{P} < 1$.

**Remark 115:** Often easier to use transition probabilities (local) rather than the global $Q$.

### 3.10 The FTAP in the Multi-Period Setting

**Definition 116:**
- **Martingale:** Adapted process with $E[Y_t|F_s] = Y_s$ for $s \le t$
- **Martingale Measure:** $Q$ s.t. $\bar{S}$ is a $Q$-martingale
- **EMM:** Equivalent Martingale Measure ($Q \sim P$)

**Theorem 117 (1st FTAP, multi-period):** NA ⟺ $M(\bar{S}) \ne \emptyset$.

*Proof (⟸):* If $G$ is arbitrage, $E^Q[V_N^{0,G}] = \sum_n E^Q[G_n \cdot (\bar{S}_{n+1} - \bar{S}_n)] = 0$ (martingale property), contradicting $V_N \ge 0, V_N > 0$ somewhere.

*Proof (⟹, binomial):* NA → $D < 1+R < U$ → $\tilde{P}_n \in (0,1)$ → Lemma 114 gives $Q \in M$.

**Corollary 118 (2nd FTAP, multi-period):** Complete ⟺ unique EMM.

### 3.11 Permutation-Invariant Processes and Recombinant Trees

**Problem:** Backward induction on full binary tree has $2^n$ nodes at time $n$ → exponential cost.

**Solution:** If $(B,S)$ is **permutation-invariant** (value depends only on # of H's, not their order), then $S_n$ takes only $n+1$ values → tree is **recombinant** (branches merge).

**Constant-coefficient binomial:** $S_n$ values are $\{S_0 u^k d^{n-k} / (1+r)^n : k=0,\ldots,n\}$.

**Common path-dependent payoffs** $Y_N = f_N(S_N, M_N, L_N, A_N)$ where:
- $M_n = \max_{k \le n} S_k$ (running max)
- $L_n = \min_{k \le n} S_k$ (running min)
- $A_n = \frac{1}{n+1}\sum_{k \le n} S_k$ (running average)

To price efficiently with path dependence, need **Markov property** (§3.17).

### 3.12 Independence

**Definition 119:** Events $A, B$ are independent if $P(A \cap B) = P(A)P(B)$.

**Definition 120 (Mutual independence):** Collection $\{A_i\}$ is independent if every finite sub-collection satisfies the product formula.

**Remark 121:** Pairwise independence ≠ mutual independence.

**Theorem 122:** $(X_j)$ are independent iff $E[\prod f_j(X_j)] = \prod E[f_j(X_j)]$ for all bounded Borel $f_j$.

**Remark 128 (Key criterion):** $(X_i)_{i=1}^N$ with values in $\{H,T\}$ are independent iff $P(X_{k+1}=H|\sigma(X_1,\ldots,X_k))$ is constant for all $k$.

**Theorem 125:** $\mathcal{B}, \mathcal{C}$ independent iff $E[Y|\mathcal{C}] = E[Y]$ for all $\mathcal{B}$-meas. $Y \in L^1$.

### 3.13 Markov Processes

**Definition 130 (Markov Property):**
$$E[f(X_t)|F_s] = E[f(X_t)|X_s] \quad \forall f \text{ bounded Borel}, \; s \le t$$

Equivalent formulations:
- $E[f(X_t)|F_s]$ is $\sigma(X_s)$-measurable (eq 107)
- $\exists g$ Borel s.t. $E[f(X_t)|F_s] = g(X_s)$ (eq 108)
- For $T = \{0,\ldots,N\}$, only need to check $t = s+1$

**Intuition:** Given the present, future is conditionally independent of the past.

**Example 131:** Running max $M_n = \max_{k \le n} S_k$ is **NOT** Markov. Proof: find two nodes with same $M$-value but different conditional law of $M_{n+1}$ (e.g. at $TH$ and $TT$, $M_2=4$ but future distributions differ).

**Corollary 135 (Main tool):** If $W_{k+1} = f_k(W_k, X_{k+1})$ where $X_{k+1}$ is indep. of $F_k$, then $W$ is Markov and:
$$E[f(W_{k+1})|F_k] = g(W_k), \quad g(w) = E[f(f_k(w, X_{k+1}))]$$

**Example 136:** In binomial model with constant coefficients, $S$ is Markov under both $P$ and $Q$ (since $\tilde{P}_n$ is deterministic).

**Independence Lemma (Lemma 132):** If $X$ is $\mathcal{G}$-meas., $Y$ indep. of $\mathcal{G}$, then
$$E[f(X,Y)|\mathcal{G}] = g(X) \quad \text{where } g(x) = E[f(x,Y)]$$

### 3.14 Markov Pricing

If $V_N = f_N(W_N)$ for $Q$-Markov process $W$, then by backward induction:
$$V_n = f_n(W_n) \quad \text{for some } f_n, \; \forall n$$

**Recipe:** From RNPF $V_n = E^Q[V_{n+1}|F_n] = E^Q[f_{n+1}(W_{n+1})|F_n] = f_n(W_n)$.

**Binomial with constant coefficients (eq 105):**
$$f_n(s) = \tilde{p}\,f_{n+1}(su) + \tilde{q}\,f_{n+1}(sd)$$

**Hedging function (eq 112):**
$$g_n(w) = \frac{f_{n+1}(h_n(w,H)) - f_{n+1}(h_n(w,T))}{s_{n+1}(h_n(w,H)) - s_{n+1}(h_n(w,T))}$$

**Remark 137:** Often need $W = (S, C)$ multi-dimensional (e.g. $(S, M)$ for lookback) even if payoff depends only on $C$.

**Remark 138:** Hedging function $G_n = g_n(W_n)$ also computable quickly via Markov property.

Only need to evaluate at $\text{Im}(W_n)$ — at most $n+1$ points per time step (linear, not exponential).

### 3.15 Lookback Put — Worked Markov Pricing Example

**Derivative:** Floating lookback put with payoff $V_N = M_N - S_N$, where $M_n = \max_{i \le n} S_i$.

**Step 1:** Choose Markov process $W = (S, M)$.

**Step 2:** Verify $W_{n+1} = h_n(W_n, X_{n+1})$:
$$h_n(s, m, x) = (s\,q(x),\; m \vee (s\,q(x))) \quad \text{where } q(H)=u,\; q(T)=d$$

**Step 3:** Pricing functions by backward induction:
$$f_N(s,m) = \frac{m-s}{(1+r)^N}$$
$$f_n(s,m) = \tilde{p}\,f_{n+1}(su,\, m \vee su) + \tilde{q}\,f_{n+1}(sd,\, m \vee sd)$$

**Step 4:** Hedging:
$$g_n(s,m) = \frac{f_{n+1}(su,\, m \vee su) - f_{n+1}(sd,\, m \vee sd)}{su - sd}$$

### 3.16 Pricing the Call — Closed-Form Binomial Formula

For call with strike $K$, maturity $N$, in binomial model with constant $u,d,r$:

Since $S$ is $Q$-Markov and $C_N = (S_N - K)^+$, by independence lemma:
$$C_n = c(n, S_n) \quad \text{where}$$
$$c(n, x) = \frac{1}{(1+r)^{N-n}} \sum_{j=0}^{N-n} \binom{N-n}{j} \tilde{p}^j (1-\tilde{p})^{N-n-j} (x\,u^j\,d^{N-n-j} - K)^+$$

**Hedging:**
$$G_n = \frac{c(n+1, S_n u) - c(n+1, S_n d)}{S_n u - S_n d}$$

Properties: $c(n,\cdot)$ is positive, increasing, convex → $G_n \ge 0$ (no short-selling needed to replicate a call).
