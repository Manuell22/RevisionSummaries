## Strategy Overview

The plan is split into two phases:

1. **Phase 1 — Note Reading (Apr 6–17, 12 days):** Read through lecture notes for all 5 remaining modules (half a module per day). Priority order follows exam proximity.
2. **Phase 2 — Focused Revision (Apr 18 → each exam):** Deep study using summary notebooks, past papers, and active recall. Each module gets an intense revision block in the days before its exam.

### Guiding Principles
- **Exam-proximity ordering:** Read the notes for nearest exams first, so you maximise the gap between first read and exam (more time for spaced repetition).
- **Half-module per day** for note reading — keeps it manageable and prevents fatigue.
- **Interleaving in Phase 2:** When revising modules with nearby exams (e.g. FE Thu / OP Fri / GT Mon), alternate between them to strengthen recall.
- **Active recall every evening:** Spend 15–20 min recalling the day's material without notes before bed.
