# Quantum Mechanics 1 — Detailed Summary

## Chapter 1 — Classical Mechanics

### Newton → Hamilton
- Newton's 2nd law: m q̈ = F(q) = −dV/dq
- **Phase space**: (q, p) with p = mq̇; trajectory = curve in phase space
- **Hamilton's canonical equations**:
  - ṗ = −∂H/∂q
  - q̇ = ∂H/∂p
- Standard Hamiltonian: H(p,q) = p²/(2m) + V(q)

### Potential Analysis
- Fixed points at V'(q₀) = 0: minima → stable equilibria, maxima → unstable
- Classically allowed region: E ≥ V(q); forbidden: E < V(q)
- Turning points: E = V(q)

### Examples
- **Free particle**: H = p²/(2m), p = const, q(t) = q₀ + (p₀/m)t
- **Harmonic oscillator**: V = mω²q²/2, solution q(t) = q₀cos(ωt) + (p₀/mω)sin(ωt)

### Poisson Brackets
- Definition: {A,B} = (∂A/∂p)(∂B/∂q) − (∂A/∂q)(∂B/∂p)
- Properties: antisymmetry, linearity, **Jacobi identity**, Leibniz rule
- **Fundamental Poisson bracket**: {p_j, q_k} = δ_{jk}
- Equation of motion: dA/dt = {H,A} + ∂A/∂t
- Cyclic variable: if H independent of q_j then p_j conserved
- **Key analogy**: Poisson bracket ↔ (i/ℏ)×commutator in QM

## Chapter 2 — Schrödinger Dynamics

### Time-Dependent Schrödinger Equation (TDSE)
- iℏ ∂ψ/∂t = Ĥψ, where Ĥ = −(ℏ²/2m)∂²/∂x² + V(x)
- ψ(x,t) = wave function; |ψ(x,t)|² = probability density
- Normalisation: ∫|ψ|²dx = 1
- Hilbert space: ψ ∈ L²(ℝ) (square integrable)

### Probability Current & Conservation
- Current: j = (ℏ/2mi)(ψ*ψ' − ψ'*ψ)
- Continuity equation: ∂|ψ|²/∂t + ∂j/∂x = 0
- Total probability conserved: d/dt ∫|ψ|²dx = 0

### Superposition & Expectation Values
- Linearity of TDSE → superposition principle
- Expectation value: ⟨Ĥ⟩ = ∫ψ*Ĥψ dx

### Time-Independent Schrödinger Equation (TISE)
- Separation of variables ψ(x,t) = T(t)φ(x) → Ĥφ_E = Eφ_E
- T(t) = e^{−iEt/ℏ}
- **Stationary states**: ψ(x,t) = e^{−iEt/ℏ}φ_E(x); |ψ|² = |φ_E|² (time-independent)
- Superpositions of stationary states → interference terms (time-dependent probability)

### Method of Stationary States
1. Solve TISE to get eigenstates {φ_n} and energies {E_n}
2. Expand initial state: ψ(x,0) = Σ c_n φ_n(x) with c_n = ∫φ_n*(x)ψ(x,0)dx
3. Time-evolve: ψ(x,t) = Σ c_n e^{−iE_nt/ℏ} φ_n(x)

### Particle in a Box
- V = 0 for 0 ≤ x ≤ L, V = ∞ otherwise
- E_n = n²π²ℏ²/(2mL²), n = 1,2,3,…
- φ_n(x) = √(2/L) sin(nπx/L)
- **Quantum quench** (half-box example): revival time τ = 16mL²/(ℏπ)

## Chapter 3 — Mathematical Background

### Hilbert Space
- Complex vector space with inner product (·,·)
- **Complete** (every Cauchy sequence converges) → Hilbert space
- **Separable**: admits countable orthonormal basis (QM uses only separable Hilbert spaces)

### Operators
- Linear operator: Â(α|u⟩ + β|v⟩) = αÂ|u⟩ + βÂ|v⟩
- **Adjoint**: ⟨Â†u|v⟩ = ⟨u|Âv⟩
- **Hermitian**: Â† = Â on domain D(Â)
- **Self-adjoint**: Â = Â† and D(Â) = D(Â†) (stronger than Hermitian)

### Theorem 1 (Riesz Representation)
H is anti-isomorphic to its dual H': every continuous linear functional F(ψ) = ⟨f|ψ⟩ for unique f ∈ H.

### Dirac Notation
- **Ket** |ψ⟩ ∈ H, **Bra** ⟨φ| ∈ H', **Bracket** ⟨φ|ψ⟩ = inner product
- Adjoint rules: conjugate scalars, swap bra↔ket, replace operators with adjoints, reverse order
- **Outer product**: |ψ⟩⟨φ| is an operator; |ψ⟩⟨ψ| is a projection operator
- **Resolution of identity**: Σ_n |n⟩⟨n| = Î (when {|n⟩} is an ONB)

### Eigenvalue Theorems (Hermitian Operators)

| Theorem | Statement |
|---------|----------|
| **Thm 2** | Eigenvalues of Hermitian operators are **real** |
| **Thm 3** | Eigenvectors for distinct eigenvalues are **orthogonal** |
| **Thm 4** | ⟨Â⟩ real for all states ⟺ Â Hermitian |
| **Thm 5** | Eigenvectors form a **complete orthonormal basis** |

**Proof sketch (Thm 2)**: Âφ = aφ ⟹ ⟨φ|Â|φ⟩ = a⟨φ|φ⟩. Since ⟨φ|Â|φ⟩ = ⟨Âφ|φ⟩ = a*⟨φ|φ⟩, we get a = a*.

### Vector/Matrix Representation
- In ONB {|n⟩}: |ψ⟩ → column vector (ψ_n), Â → matrix A_{mn} = ⟨m|Â|n⟩
- Change of basis via resolution of identity

## Chapter 4 — The Five Principles of Quantum Mechanics

### Principle 1 (States)
The state of a quantum system is described by a non-zero vector |ψ⟩ in a Hilbert space H. States differing only by a scalar factor represent the same physical state (projective Hilbert space / rays).

### Principle 2 (Observables)
Every measurable physical quantity is associated with a Hermitian operator. Position and momentum satisfy [q̂, p̂] = iℏÎ.

### Principle 3 (Born Rule)
- Measurement outcomes = eigenvalues of the observable
- **Probability**: P(a_j) = ⟨ψ|P̂_j|ψ⟩ / ⟨ψ|ψ⟩ where P̂_j projects onto eigenspace of a_j
- Non-degenerate case: P(a_j) = |⟨φ_j|ψ⟩|²
- Continuous case: P(λ ∈ I) = ∫_I |ψ(λ)|² dλ
- **Projection operators**: P̂² = P̂, P̂ = Σ|φ_n⟩⟨φ_n| (sum over eigenspace)
- **Expectation value**: ⟨Â⟩ = ⟨ψ|Â|ψ⟩/⟨ψ|ψ⟩ = Σ a_j P(a_j)

### Principle 4 (State Collapse)
After measuring eigenvalue a_j, the state collapses to P̂_j|ψ⟩ / √⟨ψ|P̂_j|ψ⟩.

### Principle 5 (Dynamics)
- Schrödinger equation: iℏ|ψ̇⟩ = Ĥ|ψ⟩
- Time-evolution operator: Û(t) = e^{−iĤt/ℏ} (unitary)
- Method of stationary states in Dirac notation

### Theorem 6 (Uncertainty Relation)
For observables Â, B̂ with Ĉ = i[Â,B̂]:

**ΔA · ΔB ≥ ½|⟨Ĉ⟩|**

where ΔA = √(⟨Â²⟩ − ⟨Â⟩²).

**Proof**: Define |f⟩ = (Â−⟨Â⟩)|ψ⟩, |g⟩ = (B̂−⟨B̂⟩)|ψ⟩. Then (ΔA)² = ⟨f|f⟩, (ΔB)² = ⟨g|g⟩. By Cauchy-Schwarz: ⟨f|f⟩⟨g|g⟩ ≥ |⟨f|g⟩|² ≥ |Im⟨f|g⟩|² = ¼|⟨Ĉ⟩|².

Special case: **Heisenberg uncertainty** Δp·Δx ≥ ℏ/2.

### Theorem 7 (Commuting Observables)
[Â,B̂] = 0 ⟹ common eigenvectors exist.

### CSCO (Complete Set of Commuting Observables)
A maximal set of mutually commuting observables whose eigenvalues uniquely label every basis vector.

## Chapter 5 — The Harmonic Oscillator

### Setup
- Ĥ = p̂²/(2m) + mω²q̂²/2 = ℏω(â†â + ½)
- Dimensionless operators: Q̂ = √(mω/ℏ) q̂, P̂ = √(1/(mωℏ)) p̂; [Q̂,P̂] = iÎ

### Ladder Operators
- **Annihilation**: â = (Q̂+iP̂)/√2
- **Creation**: â† = (Q̂−iP̂)/√2
- [â, â†] = Î
- **Number operator**: N̂ = â†â; Ĥ = ℏω(N̂ + ½)
- [N̂, â] = −â, [N̂, â†] = â†

### Theorem 8: Spectrum of N̂
N̂ has spectrum = {0, 1, 2, …} (non-negative integers).

**Proof via three lemmas:**
- **Lemma 1**: Eigenvalues of N̂ ≥ 0; â|λ⟩ = 0 iff λ = 0
- **Lemma 2**: If |λ⟩ is eigenvector with eigenvalue λ ≠ 0, then â|λ⟩ is eigenvector with eigenvalue λ−1
- **Lemma 3**: â†|λ⟩ is eigenvector with eigenvalue λ+1
- Combining: starting from any eigenvalue, lowering must terminate at 0, so eigenvalues = non-negative integers

### Energy Spectrum & Eigenstates
- **E_n = ℏω(n + ½)**, n = 0,1,2,…
- Non-degenerate (proof by contradiction via ladder operators)
- Action of ladder operators:
  - â†|n⟩ = √(n+1)|n+1⟩
  - â|n⟩ = √n|n−1⟩
- Build all states: |n⟩ = (â†)^n / √(n!) |0⟩

### Matrix Representations
- ⟨m|â†|n⟩ = √(n+1) δ_{m,n+1}
- ⟨m|â|n⟩ = √n δ_{m,n−1}
- q̂ and p̂ matrices in HO basis (tridiagonal, eqs 5.21–5.22)

## Chapter 6 — Representations

### Discrete vs Continuous Spectra
- Discrete eigenvalues: normalisable eigenvectors, ⟨φ_n|φ_m⟩ = δ_{nm}, resolution Σ|φ_n⟩⟨φ_n| = Î
- Continuous eigenvalues: **improper eigenvectors** (infinite norm)
  - ⟨φ_ω|φ_{ω'}⟩ = δ(ω−ω') (Dirac delta normalisation)
  - Resolution: ∫|φ_ω⟩⟨φ_ω|dω = Î

### Dirac δ-Distribution
- Defined inside integrals: ∫f(x)δ(x)dx = f(0)
- Gaussian limit: δ(x) = lim_{σ→0} (1/(σ√(2π))) e^{−x²/(2σ²)}
- Fourier: δ(x−ξ) = (1/2π) ∫ e^{ik(x−ξ)} dk

### Position Operator & Its Eigenstates
- Continuous spectrum spanning all ℝ: q̂|x⟩ = x|x⟩
- ⟨x|x'⟩ = δ(x−x')
- Resolution: ∫|x⟩⟨x|dx = Î
- Connection to Hermite polynomials via Mehler's formula

### Position Representation
- **Wave function**: ψ(x) := ⟨x|ψ⟩
- State expansion: |ψ⟩ = ∫ψ(x)|x⟩dx
- Norm: ⟨ψ|ψ⟩ = ∫|ψ(x)|²dx → Hilbert space = L²(ℝ)
- **Position operator**: q̂ψ(x) = xψ(x) (multiplication)
- **Momentum operator**: p̂ψ(x) = −iℏ(∂/∂x)ψ(x) (differentiation)
- Kinetic energy: Ê_kin = −(ℏ²/2m)∂²/∂x²
- **Schrödinger equation in position rep**: iℏψ̇(x,t) = [−(ℏ²/2m)∂²/∂x² + V(x)]ψ(x,t)
- Born rule recovered: P(x ∈ [a,b]) = ∫_a^b |ψ(x)|²dx

### Change of Basis
- Insert resolution of identity in target basis
- HO basis → position: ψ(x) = Σ_n ψ_n φ_n(x), where φ_n(x) = ⟨x|n⟩
- Position → momentum: ψ̃(p) = ⟨p|ψ⟩ = ∫⟨p|x⟩ψ(x)dx (Fourier transform)

## Chapter 7 — Spectral Properties of 1D Quantum Systems

### HO Eigenstates in Position Representation
- φ_n(x) = (mω/πℏ)^{1/4} / √(2^n n!) × H_n(√(mω/ℏ)x) × e^{−mωx²/(2ℏ)}
- Ground state from â|0⟩=0: φ₀(x) = (mω/πℏ)^{1/4} exp(−mωx²/(2ℏ))
- Higher states via (â†)^n in position rep → Hermite polynomials H_n

### General Properties of 1D Eigenfunctions

For Ĥ = −(ℏ²/2m)∂²/∂x² + V(x):

| Property | Statement |
|----------|----------|
| Two types | Bound states (ψ→0 at ±∞) + continuum states (improper, not in L²) |
| Binding potentials | V(x→±∞)→∞ ⟹ only discrete eigenvalues |
| Real-valuedness | Eigenfunctions can always be chosen real |
| Lower bound | No bound state with E < V_min (proof: curvature argument) |
| Expectation bound | ⟨Ĥ⟩ ≥ E₀ (lowest eigenvalue) |
| **Non-degeneracy** | Bound state eigenvalues in 1D are non-degenerate (proof via Wronskian) |
| Zeros | Ground state has no zeros; n-th excited state has n zeros |

**Non-degeneracy proof sketch**: Suppose φ₁,φ₂ same energy. Then d/dx(φ₁'φ₂ − φ₂'φ₁) = 0, so Wronskian = constant. Bound states → Wronskian = 0 at ∞ → constant = 0 → φ₁ ∝ φ₂.

### Parity
- Parity operator: Ŝψ(x) = ψ(−x); eigenvalues s = ±1 (even/odd functions)
- Ŝ² = Î
- [Ĥ,Ŝ] = 0 iff V(x) = V(−x) (symmetric potential)
- Non-degenerate eigenvalues + symmetric potential ⟹ eigenfunctions are either even or odd

### Piecewise Constant Potentials

**Boundary conditions:**
- Finite potential jump: ψ and ψ' continuous
- Infinite potential jump: ψ continuous, ψ' may be discontinuous

**Free particle** (V = V₀ constant):
- φ_E(x) = Ae^{ikx} + Be^{−ikx}, k = √(2m(E−V₀))/ℏ for E > V₀
- φ_E(x) = Ce^{κx} + De^{−κx}, κ = √(2m(V₀−E))/ℏ for E < V₀
- Plane wave interpretation: e^{±ikx} = beam with flux j = ±ℏk/m

**Infinite square well** (V=0 for 0≤x≤L, ∞ otherwise):
- E_n = n²π²ℏ²/(2mL²), φ_n = √(2/L)sin(nπx/L), n=1,2,3,…
- Derived from boundary conditions φ(0) = φ(L) = 0

**Finite square well** (V=0 for |x|≤L/2, V₀ otherwise):
- Bound states for E < V₀ only
- Even solutions: κ = k tan(kL/2)
- Odd solutions: κ = −k cot(kL/2)
- k = √(2mE)/ℏ, κ = √(2m(V₀−E))/ℏ
- Quantisation conditions solved graphically/numerically

## Chapter 8 — Quantum Dynamics

### Heisenberg Equation of Motion
From the TDSE, differentiating ⟨Â⟩ = ⟨ψ|Â|ψ⟩:

**d⟨Â⟩/dt = (i/ℏ)⟨[Ĥ,Â]⟩ + ⟨∂Â/∂t⟩**

Comparison with classical: dA/dt = {H,A} + ∂A/∂t maps Poisson bracket → (i/ℏ)×commutator.

### Ehrenfest Theorem
For Ĥ = p̂²/(2m) + V(q̂):

| Quantum | Classical analogue |
|---------|---------|
| d⟨q̂⟩/dt = ⟨p̂⟩/m | dq/dt = p/m |
| d⟨p̂⟩/dt = −⟨∂V/∂q̂⟩ | dp/dt = −∂V/∂q |

**Key subtlety**: ⟨∂V/∂q̂⟩ ≠ ∂V(⟨q̂⟩)/∂⟨q̂⟩ in general. Equal only for polynomials of degree ≤ 2 (free particle, HO, linear potential). For higher-order V, the Ehrenfest equations are not closed.

**Derivation key step**: [V(q̂), p̂] = iℏ dV(q̂)/dq̂ (computed via position representation on test function).

### Lie Groups and Lie Algebras

**Group**: set G with associative map, identity, inverses.

**Representation**: group homomorphism g → D̂(g) (linear operators on V) with D̂(g₁∘g₂) = D̂(g₁)D̂(g₂).

**Lie group**: continuous group with elements as smooth functions of n real parameters, g(0) = e.

**Generators**: T̂_j = −i ∂D̂/∂x_j |_{x=0}; near identity: D̂(x) ≈ Î + iΣx_jT̂_j.

**Exponential map** (finite parameters): D̂(x) = e^{ixT̂} (from D̂(x) = lim_{m→∞}(1 + ix/m T̂)^m).

**Lie algebra**: vector space with Lie bracket [·,·] that is antisymmetric, bilinear, satisfies Jacobi identity.

**Key examples:**
- {q̂, p̂, Î} with [q̂,p̂] = iℏÎ → translations in phase space
- {â, â†, N̂, Î} → translations + rotations in phase space

**Magnus form**: Û = exp(iΣω_jK̂_j); **Wei-Norman form**: Û = Πⱼ exp(iε_jK̂_j)

### Theorem 9 (Hadamard Lemma)

**e^{sX̂} Ŷ e^{−sX̂} = Ŷ + s[X̂,Ŷ] + (s²/2!)[X̂,[X̂,Ŷ]] + (s³/3!)[X̂,[X̂,[X̂,Ŷ]]] + …**

**Proof**: Define f(s) = e^{sX̂}Ŷe^{−sX̂}. Taylor expand f(s) around s=0. Each derivative brings a commutator: f'(s) = e^{sX̂}[X̂,Ŷ]e^{−sX̂}, f''(s) = e^{sX̂}[X̂,[X̂,Ŷ]]e^{−sX̂}, etc.

**Application** (Exercise 13): e^{iap̂/ℏ}q̂e^{−iap̂/ℏ} = q̂ + a (translation operator).

### SU(2) Group
- SU(2) = {S ∈ ℂ^{2×2} | S†=S⁻¹, det S = 1}
- Lie algebra su(2): [Ĵ_j, Ĵ_k] = iℏ Σ_l ε_{jkl} Ĵ_l
- **Pauli matrices** σ̂₁, σ̂₂, σ̂₃ form a basis (with [σ̂_j, σ̂_k] = 2iΣε_{jkl}σ̂_l)
- Central to angular momentum theory (Chapter 9)

## Chapter 9 — Angular Momentum (Part 1: Definition & Rotations)

### Classical → Quantum Angular Momentum
- Classical: L⃗ = q⃗ × p⃗, components L₁ = q₂p₃−q₃p₂, etc.
- Quantum: L̂ = q̂ × p̂ with [q̂_j, p̂_k] = iℏδ_{jk}

### Defining Commutation Relations
- Orbital: [L̂_j, L̂_k] = iℏ Σ_l ε_{jkl} L̂_l
- **General angular momentum**: any Ĵ with three Hermitian components satisfying [Ĵ_j, Ĵ_k] = iℏ Σ_l ε_{jkl} Ĵ_l
- Total: Ĵ² = Ĵ₁² + Ĵ₂² + Ĵ₃²; [Ĵ², Ĵ_j] = 0 for all j

### Angular Momentum as Generator of Rotations
- Time evolution with Ĥ = L̂_j produces rotations of coordinate system
- Using Hadamard lemma: e^{iθL̂₁/ℏ} q̂₂ e^{−iθL̂₁/ℏ} = q̂₂cos(θ) − q̂₃sin(θ)
- Key commutator: [L̂_j, q̂_k] = iℏ ε_{jkl} q̂_l
- In general: e^{iĴ_jt/ℏ} Ĵ e^{−iĴ_jt/ℏ} = R_j(t) Ĵ (rotation matrix)

### Orbital Angular Momentum Eigenvalues are Integer
- Rewrite L̂₃ = ½(p̂²_x + x̂²) − ½(p̂²_y + ŷ²) = Ĥ_x − Ĥ_y (two HOs with m=1, ω=1)
- Eigenvalues: ℏ(n_x − n_y) where n_x, n_y ∈ {0,1,2,…}
- Therefore eigenvalues of L̂₃ are **integer** multiples of ℏ

## Chapter 9 — Angular Momentum (Part 2: Algebraic Spectrum)

### Setting up the Eigenvalue Problem
- Ĵ² and Ĵ₃ commute → simultaneous eigenstates:
  - Ĵ²|λ,m⟩ = ℏ²λ|λ,m⟩
  - Ĵ₃|λ,m⟩ = ℏm|λ,m⟩
- Question: what are the allowed values of λ and m?

### Ladder Operators
- Ĵ± = Ĵ₁ ± iĴ₂
- Key commutators: [Ĵ₃, Ĵ±] = ±ℏĴ±, [Ĵ₊, Ĵ₋] = 2ℏĴ₃
- Key identities: Ĵ₋Ĵ₊ = Ĵ² − Ĵ₃² − ℏĴ₃, Ĵ₊Ĵ₋ = Ĵ² − Ĵ₃² + ℏĴ₃

### Theorem 10: Spectrum of Angular Momentum
**λ = j(j+1) with 2j ∈ ℕ; for each j, m takes values −j, −j+1, …, j−1, j.**

**Proof structure** (analogous to HO):

- **Lemma 4**: m² ≤ λ (m is bounded). Proof: ⟨Ĵ₁²⟩ + ⟨Ĵ₂²⟩ ≥ 0 ⟹ ℏ²λ ≥ ℏ²m².

- **Lemma 5**: Ĵ₊|λ,m⟩ is eigenvector of Ĵ₃ with eigenvalue ℏ(m+1) (or zero vector).

- **Lemma 6**: Ĵ₋|λ,m⟩ is eigenvector of Ĵ₃ with eigenvalue ℏ(m−1) (or zero vector).

- Maximum m = j: Ĵ₊|λ,j⟩ = 0 → apply Ĵ₋: (Ĵ²−Ĵ₃²−ℏĴ₃)|λ,j⟩ = 0 → **λ = j(j+1)**

- Minimum m = k: Ĵ₋|λ,k⟩ = 0 → apply Ĵ₊: λ = k(k−1). Combined with λ=j(j+1): **k = −j**.

- m runs from −j to j in integer steps → **2j ∈ ℕ** (j = 0, ½, 1, 3/2, 2, …)
- For each j: 2j+1 values of m

### Matrix Elements in Standard Basis |j,m⟩
- Ĵ²|j,m⟩ = ℏ²j(j+1)|j,m⟩
- Ĵ₃|j,m⟩ = ℏm|j,m⟩
- **Ĵ₊|j,m⟩ = ℏ√(j(j+1)−m(m+1)) |j,m+1⟩**
- **Ĵ₋|j,m⟩ = ℏ√(j(j+1)−m(m−1)) |j,m−1⟩**
- Ĵ₁ = (Ĵ₊+Ĵ₋)/2, Ĵ₂ = (Ĵ₊−Ĵ₋)/(2i)
- For j=½: ŝ_j = (ℏ/2)σ̂_j (Pauli matrices)

## Chapter 9 — Angular Momentum (Part 3: Spin)

### Quantum Spin
- Internal degree of freedom (not orbital rotation)
- Fulfils same [Ĵ_j, Ĵ_k] = iℏΣε_{jkl}Ĵ_l commutation relations
- Total spin Ŝ² is a conserved quantity (like mass)
- Can take half-integer values (unlike orbital angular momentum)

### Bosons vs Fermions
- **Bosons**: integer spin (s = 0, 1, 2, …)
- **Fermions**: half-integer spin (s = ½, 3/2, …)

### Spin-½ Representation
- Hilbert space: ℂ² (two-dimensional)
- Ŝ_j = (ℏ/2)σ̂_j where σ̂_j are Pauli matrices
- Pauli matrices + identity form a basis for all 2×2 matrices
- Any two-state quantum system ↔ spin-½

### Stern-Gerlach Experiment (1922)
- Silver atoms through inhomogeneous magnetic field
- Beam splits into two (m = +½, −½) → first measurement of spin
- Silver: total orbital L = 0, angular momentum entirely from electron spin
- Historical: Uhlenbeck & Goudsmit (1925), Pauli matrices, Dirac's relativistic theory

### Applications
- MRI, spintronics, quantum information (qubit = spin-½)
