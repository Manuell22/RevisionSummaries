## 🔴 MUST KNOW — High-frequency exam material

### 1. Time Evolution of a General State

If $|\psi(0)\rangle = \sum_n a_n |\varepsilon_n\rangle$, then:

$$|\psi(t)\rangle = \sum_n a_n e^{-iE_n t/\hbar} |\varepsilon_n\rangle$$

- Coefficients: $a_n = \langle \varepsilon_n | \psi(0) \rangle$
- Only works when $\hat{H}|\varepsilon_n\rangle = E_n|\varepsilon_n\rangle$ (eigenbasis expansion)
- For a single eigenstate $|\varepsilon_n\rangle$: just picks up phase $e^{-iE_n t/\hbar}$ — **stationary state**

### 2. Measurement Probabilities (Born Rule)

For a state $|\psi\rangle = \sum_n a_n |\varepsilon_n\rangle$ and observable $\hat{A}$ with $\hat{A}|\varepsilon_n\rangle = a_n|\varepsilon_n\rangle$:

$$p(a_j) = |\langle \varepsilon_j | \psi \rangle|^2 = |a_j|^2$$

- Expectation value: $\langle \hat{A} \rangle = \langle \psi | \hat{A} | \psi \rangle$
- After measuring $a_j$: state **collapses** to $|\varepsilon_j\rangle$ (non-degenerate case)
- General collapse: $\displaystyle\frac{\hat{P}_j|\psi\rangle}{\sqrt{\langle\psi|\hat{P}_j|\psi\rangle}}$, where $\hat{P}_j = |\varepsilon_j\rangle\langle\varepsilon_j|$

### 3. Harmonic Oscillator — Ladder Operators

**Definitions:**
$$\hat{a} = \frac{1}{\sqrt{2}}(\hat{Q} + i\hat{P}), \qquad \hat{a}^\dagger = \frac{1}{\sqrt{2}}(\hat{Q} - i\hat{P})$$

$$\hat{N} = \hat{a}^\dagger \hat{a}, \qquad \hat{H} = \hbar\omega\!\left(\hat{N} + \tfrac{1}{2}\right)$$

**Commutation relations (MEMORISE ALL):**
$$[\hat{a}, \hat{a}^\dagger] = \hat{I}, \qquad [\hat{N}, \hat{a}] = -\hat{a}, \qquad [\hat{N}, \hat{a}^\dagger] = \hat{a}^\dagger$$

**Action on eigenstates:**
$$\hat{a}|n\rangle = \sqrt{n}\,|n-1\rangle, \qquad \hat{a}^\dagger|n\rangle = \sqrt{n+1}\,|n+1\rangle$$

$$\hat{N}|n\rangle = n|n\rangle, \qquad E_n = \hbar\omega\!\left(n + \tfrac{1}{2}\right), \quad n = 0,1,2,\ldots$$

**Key norm argument (exam proof):**

$\|\hat{a}|\nu\rangle\|^2 = \langle\nu|\hat{a}^\dagger\hat{a}|\nu\rangle = \langle\nu|\hat{N}|\nu\rangle = \nu \geq 0$

Therefore eigenvalues $\nu \geq 0$, and $\hat{a}|0\rangle = 0$ (ground state annihilated by $\hat{a}$).

### 4. Wave Function Integrals

These appear in every wave function question — know them cold.

**Gaussian:**
$$\int_{-\infty}^{\infty} e^{-ax^2}\,dx = \sqrt{\frac{\pi}{a}}, \qquad \text{Re}(a) > 0$$

**Gamma / polynomial × exponential (positive half-line):**
$$\int_0^{\infty} x^n e^{-\lambda x}\,dx = \frac{n!}{\lambda^{n+1}}$$

Special cases:
$$\int_0^\infty e^{-\lambda x}\,dx = \frac{1}{\lambda}, \quad \int_0^\infty x\,e^{-\lambda x}\,dx = \frac{1}{\lambda^2}, \quad \int_0^\infty x^2 e^{-\lambda x}\,dx = \frac{2}{\lambda^3}$$

**Useful trick (from 2023 exam hint):**
$$x^n e^{-\lambda x} = \left(-\frac{d}{d\lambda}\right)^n e^{-\lambda x}$$

### 5. Standard Wave Function Tasks (Recipe)

**Normalisation:** Solve $\int_{-\infty}^{\infty} |\psi(x)|^2 dx = 1$ for the constant $A$.

**Most probable position:** Find $x^*$ where $|\psi(x)|^2$ is maximised. Set $\frac{d}{dx}|\psi(x)|^2 = 0$.

**Expectation value of position:**
$$\langle x \rangle = \int_{-\infty}^{\infty} x\,|\psi(x)|^2\,dx$$

**Expectation value of general observable:**
$$\langle \hat{A} \rangle = \int_{-\infty}^{\infty} \psi^*(x)\,\hat{A}\,\psi(x)\,dx$$

**Probability in interval $[a,b]$:**
$$P(x \in [a,b]) = \int_a^b |\psi(x)|^2\,dx$$

**Note:** The time factor $e^{-i\omega t}$ disappears in $|\psi|^2$ since $|e^{-i\omega t}|^2 = 1$.

### 6. Harmonic Oscillator — Expectation Values & Time Evolution

**Expressing position/momentum via ladder operators:**
$$\hat{Q} = \frac{1}{\sqrt{2}}(\hat{a} + \hat{a}^\dagger), \qquad \hat{P} = \frac{1}{i\sqrt{2}}(\hat{a} - \hat{a}^\dagger)$$

**Expectation value of $\hat{Q}$ in a superposition $|\psi(t)\rangle = \sum_n a_n e^{-iE_nt/\hbar}|n\rangle$:**

Use $\langle m|\hat{Q}|n\rangle = \frac{1}{\sqrt{2}}(\sqrt{n}\,\delta_{m,n-1} + \sqrt{n+1}\,\delta_{m,n+1})$

Only **adjacent** terms $|n\rangle$ and $|n\pm 1\rangle$ contribute — cross terms with other $n$ vanish.

**Typical calculation pattern (2023 Q4c):**
$$\langle \hat{Q}(t)\rangle = \frac{1}{\sqrt{2}}\sum_{n,m} a_m^* a_n\, e^{i(E_m-E_n)t/\hbar}\langle m|\hat{a}+\hat{a}^\dagger|n\rangle$$

### 7. Key Operator Properties

**Hermitian operator $\hat{A} = \hat{A}^\dagger$:**
- Eigenvalues are **real**
- Eigenvectors for distinct eigenvalues are **orthogonal**
- Eigenvectors form a complete basis of the Hilbert space

**Adjoint rules:**
$$(AB)^\dagger = B^\dagger A^\dagger, \qquad (c\hat{A})^\dagger = c^*\hat{A}^\dagger, \qquad (|\phi\rangle\langle\psi|)^\dagger = |\psi\rangle\langle\phi|$$

**Commutator identities (useful for calculations):**
$$[A, BC] = [A,B]C + B[A,C]$$
$$[AB, C] = A[B,C] + [A,C]B$$
$$[\hat{q}, \hat{p}] = i\hbar\hat{I} \quad \text{(canonical commutation relation)}$$

**Unitary operator** $\hat{U}$: $\hat{U}^\dagger = \hat{U}^{-1}$ — preserves inner products (time evolution $e^{-i\hat{H}t/\hbar}$ is unitary).

### 8. Position & Momentum Representation (Ch 6)

**Position eigenstates** $\hat{q}|x\rangle = x|x\rangle$, continuous spectrum:
$$\langle x'|x\rangle = \delta(x-x'), \qquad \int_{-\infty}^\infty |x\rangle\langle x|\,dx = \hat{I}$$

**Wavefunction = coefficient in position basis:**
$$\psi(x) = \langle x|\psi\rangle$$

**Momentum operator in position representation:**
$$\langle x|\hat{p}|\psi\rangle = -i\hbar\frac{\partial}{\partial x}\psi(x)$$

**Schrödinger equation in position representation:**
$$-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}\psi(x) + V(x)\psi(x) = E\psi(x)$$

**Fourier / momentum representation:**
$$\tilde{\psi}(p) = \langle p|\psi\rangle, \qquad \langle x|p\rangle = \frac{1}{\sqrt{2\pi\hbar}}e^{ipx/\hbar}$$

## 🟡 GOOD TO KNOW — Useful but less frequently directly tested

### 9. Resolution of Identity & Completeness

Discrete: $\displaystyle\sum_n |\varepsilon_n\rangle\langle\varepsilon_n| = \hat{I}$

Continuous: $\displaystyle\int d\vartheta\,|\varepsilon(\vartheta)\rangle\langle\varepsilon(\vartheta)| = \hat{I}$

**How to use it:** Insert $\hat{I}$ anywhere to switch basis, e.g.:
$$\langle m|\hat{A}|n\rangle = \int dx\,\langle m|x\rangle\langle x|\hat{A}|n\rangle$$

### 10. Projection Operator

$\hat{P} = \sum_{n \in S}|\varepsilon_n\rangle\langle\varepsilon_n|$ for some subset $S$ of basis vectors.

- $\hat{P}^2 = \hat{P}$ (idempotent) — works **because** eigenvectors are orthogonal
- For single eigenvector: $\hat{P}_j = |\varepsilon_j\rangle\langle\varepsilon_j|$, eigenvalues are $0$ or $1$

### 11. Uncertainty Principle

$$\Delta A \cdot \Delta B \geq \frac{1}{2}|\langle\hat{C}\rangle|, \qquad \hat{C} = i[\hat{A},\hat{B}]$$

Most important instance:
$$\Delta x \cdot \Delta p \geq \frac{\hbar}{2}$$

(since $[\hat{q},\hat{p}] = i\hbar\hat{I}$, so $\hat{C} = i \cdot i\hbar\hat{I} \cdot (-1)$... yields $\hbar$)

**Uncertainty of observable $A$:**
$$(\Delta A)^2 = \langle\hat{A}^2\rangle - \langle\hat{A}\rangle^2$$

### 12. Probability Current (Ch 2)

$$j(x,t) = \frac{\hbar}{2mi}\left(\psi^*\frac{\partial\psi}{\partial x} - \frac{\partial\psi^*}{\partial x}\psi\right)$$

Continuity equation: $\frac{\partial}{\partial t}|\psi|^2 + \frac{\partial j}{\partial x} = 0$ — ensures norm is conserved in time.

### 13. Separation of Variables / Stationary States (Ch 2)

Ansatz $\psi(x,t) = \phi(x)\chi(t)$ inserted into TDSE gives:

$$i\hbar\frac{\dot{\chi}(t)}{\chi(t)} = \frac{\hat{H}\phi(x)}{\phi(x)} = E \quad \text{(constant)}$$

Left side $\Rightarrow$ $\chi(t) = e^{-iEt/\hbar}$

Right side $\Rightarrow$ $\hat{H}\phi(x) = E\phi(x)$ (**TISE**)

**Stationary state:** $|\psi(x,t)|^2 = |\phi(x)|^2$ — time-independent probability density.

**Why $\phi(x)$ treated as constant w.r.t. $t$:** it is defined to depend only on $x$, so $\partial\phi/\partial t = 0$ exactly.

### 14. Dirac Notation Quick Reference

| Symbol | Meaning |
|--------|---------|
| $\vert\psi\rangle$ | Ket — vector in Hilbert space $\mathcal{H}$ |
| $\langle\psi\vert$ | Bra — linear functional in dual space $\mathcal{H}^*$ |
| $\langle\phi\vert\psi\rangle$ | Inner product — a complex number |
| $\vert\phi\rangle\langle\psi\vert$ | Outer product — an operator |
| $\langle\phi\vert\hat{A}\vert\psi\rangle$ | Matrix element of $\hat{A}$ |

**Anti-linear correspondence:** $c\vert\psi\rangle \leftrightarrow c^*\langle\psi\vert$

**Outer product as projector:** $|\varepsilon_j\rangle\langle\varepsilon_j||\psi\rangle = \langle\varepsilon_j|\psi\rangle|\varepsilon_j\rangle$ — projects $|\psi\rangle$ onto $|\varepsilon_j\rangle$ direction.

## 🟢 OPTIONAL — Background knowledge, unlikely to be directly examined

### 15. Proving $\hat{a}|n\rangle$ is an eigenvector of $\hat{N}$ with eigenvalue $n-1$

Use $[\hat{N}, \hat{a}] = -\hat{a}$:

$$\hat{N}(\hat{a}|n\rangle) = (\hat{a}\hat{N} - \hat{a})|n\rangle = \hat{a}(n|n\rangle) - \hat{a}|n\rangle = (n-1)\hat{a}|n\rangle$$

So $\hat{a}|n\rangle$ is an eigenvector of $\hat{N}$ with eigenvalue $n-1$ (or zero vector if $n=0$).

Analogously $\hat{N}(\hat{a}^\dagger|n\rangle) = (n+1)\hat{a}^\dagger|n\rangle$ using $[\hat{N},\hat{a}^\dagger] = \hat{a}^\dagger$.

### 16. Orthogonality of Hermitian Eigenstates (Proof sketch)

For $\hat{A}|\varepsilon_1\rangle = \lambda_1|\varepsilon_1\rangle$ and $\hat{A}|\varepsilon_2\rangle = \lambda_2|\varepsilon_2\rangle$:

1. Take adjoint of first equation: $\langle\varepsilon_1|\hat{A} = \lambda_1^*\langle\varepsilon_1|$ (since $\hat{A}^\dagger = \hat{A}$ and $\lambda_1^* = \lambda_1$ by reality)
2. Multiply on right by $|\varepsilon_2\rangle$: $\langle\varepsilon_1|\hat{A}|\varepsilon_2\rangle = \lambda_1\langle\varepsilon_1|\varepsilon_2\rangle$
3. Also $\langle\varepsilon_1|\hat{A}|\varepsilon_2\rangle = \lambda_2\langle\varepsilon_1|\varepsilon_2\rangle$
4. Subtract: $(\lambda_2 - \lambda_1)\langle\varepsilon_1|\varepsilon_2\rangle = 0$. Since $\lambda_1 \neq \lambda_2$, conclude $\langle\varepsilon_1|\varepsilon_2\rangle = 0$.

### 17. Continuous vs Discrete Spectrum — Key Difference

| | Discrete | Continuous |
|--|---------|-----------|
| Orthonormality | $\langle\varepsilon_n\vert\varepsilon_m\rangle = \delta_{nm}$ | $\langle\varepsilon(\vartheta)\vert\varepsilon(\vartheta')\rangle = \delta(\vartheta-\vartheta')$ |
| Expansion | $\vert\psi\rangle = \sum_n a_n\vert\varepsilon_n\rangle$ | $\vert\psi\rangle = \int d\vartheta\,\psi(\vartheta)\vert\varepsilon(\vartheta)\rangle$ |
| Coefficients | $a_n = \langle\varepsilon_n\vert\psi\rangle$ | $\psi(\vartheta) = \langle\varepsilon(\vartheta)\vert\psi\rangle$ |
| Probability | $p(a_n) = \vert a_n\vert^2$ | $P(\vartheta\in I) = \int_I\vert\psi(\vartheta)\vert^2 d\vartheta$ |
| Eigenstates | Normalisable ($\in\mathcal{H}$) | **Not** normalisable (improper eigenvectors) |
