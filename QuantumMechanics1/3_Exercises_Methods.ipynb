# Quantum Mechanics 1 — Exercises & Methods

## A. Solving Eigenvalue Problems (TISE)

**When**: Asked to find energy levels and eigenstates of a Hamiltonian.

### Recipe — Particle in a Box / Piecewise Constant Potential
1. Write TISE: −(ℏ²/2m)φ'' + V(x)φ = Eφ
2. In each region of constant potential V_j, general solution:
   - E > V_j: φ(x) = Ae^{ikx} + Be^{−ikx}, k = √(2m(E−V_j))/ℏ
   - E < V_j: φ(x) = Ce^{κx} + De^{−κx}, κ = √(2m(V_j−E))/ℏ
3. Apply physical constraints:
   - Bound states: φ → 0 as x → ±∞ (eliminate growing exponentials)
   - Infinite wall: φ = 0 at wall
4. Match boundary conditions at interfaces:
   - Finite jump in V: φ and φ' continuous
   - Infinite jump: φ continuous only
5. Get quantisation condition (transcendental equation) → solve for E

### Recipe — Symmetric Potential (Parity Trick)
If V(x) = V(−x):
- Eigenfunctions are either **even** or **odd** → halve the problem
- Even: φ'(0) = 0 at centre; Odd: φ(0) = 0 at centre
- Separate quantisation conditions for even/odd subsets

### Key Examples
| System | Energies | Wave functions |
|--------|----------|----------------|
| Infinite well [0,L] | n²π²ℏ²/(2mL²) | √(2/L)sin(nπx/L) |
| Finite well [−L/2,L/2] | Even: κ=k·tan(kL/2); Odd: κ=−k·cot(kL/2) | cos/sin inside, exponential tails |
| Harmonic oscillator | ℏω(n+½) | Hermite × Gaussian (Section B below) |

## B. Harmonic Oscillator — Ladder Operator Technique

**When**: HO eigenvalues/eigenstates, matrix elements, or any problem exploiting â,â†.

### Recipe — Proving the Spectrum
1. Define N̂ = â†â with [â,â†] = Î, so [N̂,â] = −â, [N̂,â†] = â†
2. Show eigenvalues ≥ 0: ⟨λ|N̂|λ⟩ = ||â|λ⟩||² ≥ 0
3. â|λ⟩ is eigenstate of N̂ with eigenvalue λ−1 (Lemma 2)
4. â†|λ⟩ is eigenstate of N̂ with eigenvalue λ+1 (Lemma 3)
5. Lowering must terminate: â|0⟩ = 0 → eigenvalues = {0,1,2,…}

### Recipe — Matrix Elements
- â|n⟩ = √n|n−1⟩,  â†|n⟩ = √(n+1)|n+1⟩
- Express q̂, p̂ in terms of â, â†:
  - q̂ = L/√2 (â+â†),  p̂ = i/(L√2) ℏ(â†−â)  [with L=√(ℏ/(mω))]
- Compute ⟨m|q̂^k|n⟩ by expanding in â, â† and using selection rules

### Recipe — Position Representation
- Ground state from â|0⟩ = 0 → ODE: φ₀'(x) = −(mω/ℏ)xφ₀(x) → Gaussian
- Excited states: φ_n(x) ∝ (â†)^n φ₀(x) in position rep → Hermite polynomials

## C. Time Evolution (Method of Stationary States)

**When**: Given ψ(x,0), find ψ(x,t) or expectation values at later times.

### Recipe
1. Find energy eigenstates {φ_n} and eigenvalues {E_n} (solve TISE)
2. Expand initial state: c_n = ⟨φ_n|ψ(0)⟩
3. Time-evolve: |ψ(t)⟩ = Σ c_n e^{−iE_nt/ℏ} |φ_n⟩
4. Compute observables: ⟨Â⟩(t) = Σ_{m,n} c_m* c_n e^{i(E_m−E_n)t/ℏ} ⟨φ_m|Â|φ_n⟩

### Common Pitfalls
- Remember to normalise c_n if initial state isn't normalised
- Stationary states have time-independent |ψ|² but general superpositions don't
- For expectation values, cross terms give oscillations at frequencies (E_m−E_n)/ℏ

### Quantum Quench / Revival
- Sudden change in potential → expand in new eigenbasis
- Revival time: when all phase factors realign (related to ratios of E_n)

## D. Commutator Calculations

**When**: Verifying commutation relations, computing Heisenberg equations, or proving identities.

### Key Rules
- [Â,B̂] = −[B̂,Â] (antisymmetry)
- [Â, B̂Ĉ] = [Â,B̂]Ĉ + B̂[Â,Ĉ] (Leibniz / product rule)
- [Â, B̂+Ĉ] = [Â,B̂] + [Â,Ĉ] (linearity)
- [Â, f(B̂)] — often easiest via position representation on test function

### Fundamental Commutators
| Commutator | Value |
|------------|-------|
| [q̂, p̂] | iℏÎ |
| [â, â†] | Î |
| [N̂, â] | −â |
| [N̂, â†] | â† |
| [Ĵ_j, Ĵ_k] | iℏΣε_{jkl}Ĵ_l |
| [Ĵ₃, Ĵ±] | ±ℏĴ± |
| [Ĵ₊, Ĵ₋] | 2ℏĴ₃ |
| [L̂_j, q̂_k] | iℏε_{jkl}q̂_l |

### Strategy: Position Representation
When stuck, apply [Â, B̂] to a test function ψ(x) using q̂→x, p̂→−iℏ∂/∂x:
- [V(q̂), p̂]ψ = −iℏ[V(x), ∂/∂x]ψ → compute product rule → iℏV'(x)ψ

### Hadamard Lemma Applications
- e^{sX̂}Ŷe^{−sX̂} = Ŷ + s[X̂,Ŷ] + (s²/2!)[X̂,[X̂,Ŷ]] + …
- Compute nested commutators iteratively; look for patterns (often periodic → sin/cos)
- Example: e^{iθĴ₁/ℏ}Ĵ₂e^{−iθĴ₁/ℏ} = Ĵ₂cos(θ) − Ĵ₃sin(θ)

## E. Angular Momentum Problems

**When**: Eigenvalues of Ĵ², Ĵ₃, matrix elements, or rotation-related problems.

### Recipe — Proving the Spectrum (Theorem 10)
1. Define Ĵ± = Ĵ₁ ± iĴ₂ (ladder operators)
2. Show m bounded: m² ≤ λ from ⟨Ĵ₁²⟩+⟨Ĵ₂²⟩ ≥ 0
3. Ĵ₊ raises m by 1, Ĵ₋ lowers m by 1 (Lemmas 5–6)
4. Maximum m=j: Ĵ₊|λ,j⟩=0 → Ĵ₋Ĵ₊|λ,j⟩=0 → λ=j(j+1)
5. Minimum m=k: Ĵ₋|λ,k⟩=0 → k=−j
6. 2j ∈ ℕ; for each j: m=−j,−j+1,…,j (total 2j+1 states)

### Recipe — Matrix Elements
- Ĵ₊|j,m⟩ = ℏ√(j(j+1)−m(m+1))|j,m+1⟩
- Ĵ₋|j,m⟩ = ℏ√(j(j+1)−m(m−1))|j,m−1⟩
- Ĵ₁ = (Ĵ₊+Ĵ₋)/2, Ĵ₂ = (Ĵ₊−Ĵ₋)/(2i)

### Recipe — Spin-½ Matrices
- ŝ_j = (ℏ/2)σ̂_j
- σ̂₁ = [[0,1],[1,0]], σ̂₂ = [[0,−i],[i,0]], σ̂₃ = [[1,0],[0,−1]]
- Basis: |↑⟩ = (1,0)ᵀ (m=+½), |↓⟩ = (0,1)ᵀ (m=−½)

### Rotations
- Angular momentum generates rotations via e^{iθĴ_j/ℏ}
- On position operators: e^{iθL̂_j/ℏ}q̂_k e^{−iθL̂_j/ℏ} = (R_j(θ)q̂)_k
- On angular momentum: e^{iθĴ_j/ℏ}Ĵ_k e^{−iθĴ_j/ℏ} = (R_j(θ)Ĵ)_k

## F. Computing Expectation Values & Uncertainties

**When**: Asked for ⟨Â⟩, ΔA, or to verify the uncertainty relation.

### Recipe — Expectation Value
1. If state in energy eigenbasis: ⟨Â⟩ = Σ|c_n|²⟨n|Â|n⟩ + cross terms
2. If in position rep: ⟨Â⟩ = ∫ψ*(x)Âψ(x)dx
3. For HO: express Â in terms of â, â† and use ⟨m|â|n⟩ = √n δ_{m,n−1}

### Recipe — Uncertainty
1. Compute ⟨Â⟩ and ⟨Â²⟩
2. (ΔA)² = ⟨Â²⟩ − ⟨Â⟩²
3. For uncertainty relation: ΔA·ΔB ≥ ½|⟨Ĉ⟩| where Ĉ = i[Â,B̂]

### Heisenberg Equation Approach
- d⟨Â⟩/dt = (i/ℏ)⟨[Ĥ,Â]⟩ + ⟨∂Â/∂t⟩
- For Ĥ = p̂²/(2m) + V(q̂): use Ehrenfest d⟨q̂⟩/dt = ⟨p̂⟩/m, d⟨p̂⟩/dt = −⟨V'(q̂)⟩

## G. Representation Changes

**When**: Translating between abstract Dirac notation and wave mechanics, or between bases.

### Recipe
1. Insert resolution of identity in the target basis:
   - Position: Î = ∫|x⟩⟨x|dx → ψ(x) = ⟨x|ψ⟩
   - Momentum: Î = ∫|p⟩⟨p|dp → ψ̃(p) = ⟨p|ψ⟩
   - HO basis: Î = Σ_n|n⟩⟨n|
2. Key transformation functions:
   - ⟨x|n⟩ = φ_n(x) (HO eigenfunction)
   - ⟨x|p⟩ = e^{ipx/ℏ}/√(2πℏ) (plane wave) → position↔momentum is Fourier transform
3. Operators in position rep: q̂→x, p̂→−iℏ∂/∂x
4. Operators in momentum rep: p̂→p, q̂→iℏ∂/∂p

## H. Proving Key Theorems

| Theorem | Proof Strategy |
|---------|---------------|
| Thm 2 (real eigenvalues) | Take ⟨φ|Â|φ⟩ = a⟨φ|φ⟩ and ⟨Âφ|φ⟩ = a*⟨φ|φ⟩; use Â†=Â |
| Thm 3 (orthogonality) | Compute ⟨φ_m|Â|φ_n⟩ two ways using Hermiticity; (a_n−a_m)⟨φ_m|φ_n⟩=0 |
| Thm 6 (uncertainty) | Cauchy-Schwarz on |f⟩=(Â−⟨Â⟩)|ψ⟩, |g⟩=(B̂−⟨B̂⟩)|ψ⟩; keep imaginary part |
| Thm 8 (HO spectrum) | Lemmas 1–3 using N̂=â†â; lower bound + ladder structure |
| Non-degeneracy (1D) | Wronskian W(φ₁,φ₂)=const; bound states→W=0→φ₁∝φ₂ |
| Ehrenfest | Apply Heisenberg eq to q̂ and p̂; compute [p̂²/(2m),q̂] and [V(q̂),p̂] |
| Hadamard (Thm 9) | Taylor expand f(s)=e^{sX̂}Ŷe^{−sX̂}; each derivative → nested commutator |
| Thm 10 (ang. mom.) | Lemmas 4–6 using Ĵ±; bounded spectrum + ladder → j(j+1), m=−j..j |

## I. Key Formulae Cheat Sheet

### Fundamentals
- [q̂, p̂] = iℏÎ
- TDSE: iℏ|ψ̇⟩ = Ĥ|ψ⟩
- TISE: Ĥφ_E = Eφ_E
- Û(t) = e^{−iĤt/ℏ}

### Born Rule
- P(a_j) = |⟨φ_j|ψ⟩|² (non-degenerate)
- P(a_j) = ⟨ψ|P̂_j|ψ⟩ (general)
- ⟨Â⟩ = ⟨ψ|Â|ψ⟩ / ⟨ψ|ψ⟩

### Uncertainty
- ΔA·ΔB ≥ ½|⟨i[Â,B̂]⟩|
- Δp·Δx ≥ ℏ/2

### Harmonic Oscillator
- E_n = ℏω(n+½)
- â = (Q̂+iP̂)/√2, â† = (Q̂−iP̂)/√2, [â,â†] = Î
- â|n⟩ = √n|n−1⟩, â†|n⟩ = √(n+1)|n+1⟩
- |n⟩ = (â†)^n/√(n!)|0⟩
- φ₀(x) = (mω/πℏ)^{1/4} exp(−mωx²/(2ℏ))

### Particle in a Box
- E_n = n²π²ℏ²/(2mL²)
- φ_n(x) = √(2/L) sin(nπx/L)

### Representations
- ψ(x) = ⟨x|ψ⟩, ⟨x|x'⟩ = δ(x−x')
- q̂ → x, p̂ → −iℏ∂/∂x
- δ(x) = (1/2π)∫e^{ikx}dk

### Dynamics
- d⟨Â⟩/dt = (i/ℏ)⟨[Ĥ,Â]⟩ + ⟨∂Â/∂t⟩
- Ehrenfest: d⟨q̂⟩/dt = ⟨p̂⟩/m, d⟨p̂⟩/dt = −⟨V'(q̂)⟩

### Angular Momentum
- [Ĵ_j, Ĵ_k] = iℏΣε_{jkl}Ĵ_l
- Ĵ²|j,m⟩ = ℏ²j(j+1)|j,m⟩, Ĵ₃|j,m⟩ = ℏm|j,m⟩
- Ĵ₊|j,m⟩ = ℏ√(j(j+1)−m(m+1))|j,m+1⟩
- Ĵ₋|j,m⟩ = ℏ√(j(j+1)−m(m−1))|j,m−1⟩
- 2j ∈ ℕ, m = −j,…,j (total 2j+1 states)
- Spin-½: ŝ_j = (ℏ/2)σ̂_j

## J. Exam Strategy

### Likely Question Types
1. **Eigenvalue problems**: particle in box, finite well (set up + boundary conditions), HO
2. **Time evolution**: given initial state, find ψ(t) or ⟨Â⟩(t) via method of stationary states
3. **Commutator calculations**: verify [Â,B̂], use to derive Heisenberg equation or uncertainty bound
4. **Ladder operator proofs**: HO spectrum or angular momentum spectrum
5. **Expectation values / uncertainties**: compute ΔA·ΔB and compare to bound
6. **Representation problems**: translate between Dirac notation and position rep
7. **Hadamard lemma applications**: show translation/rotation properties
8. **Spin-½ / Pauli matrix calculations**: matrix elements, measurement probabilities

### Common Mistakes to Avoid
- Forgetting to normalise states after projections/collapses
- Confusing Hermitian (A†=A) with unitary (U†=U⁻¹)
- Wrong commutator signs (especially [N̂,â]=−â not +â)
- Forgetting boundary condition at ∞ for bound states (eliminate growing exponentials)
- Using non-degenerate Born rule for degenerate case (need projector P̂_j, not |φ_j⟩⟨φ_j|)

### Time Management
- Start with eigenvalue/time-evolution questions (most structured)
- Proof questions: state the key lemmas, then fill in details
- For commutator calculations: always check by applying to test function if unsure
- Angular momentum: the algebraic proof follows the HO proof structure closely — draw parallels
