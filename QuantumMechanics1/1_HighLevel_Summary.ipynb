## Exam Notes

- **5 Principles** are the backbone — know them by heart and be able to apply each one
- **Ladder operator proofs** appear for both HO (Ch5) and angular momentum (Ch9) — similar algebraic structure
- **Commutator calculations** are pervasive: [q̂,p̂]=iℏ, angular momentum algebra, Hadamard lemma applications
- **Method of stationary states** is the standard time-evolution technique
- **Piecewise potentials** (Ch7): know boundary conditions, quantisation conditions for finite well, be able to set up and solve
- **Representations** (Ch6): be comfortable translating between abstract Dirac notation and position representation
- **Proofs to know**: Thm 2 (real eigenvalues), Thm 6 (uncertainty relation), Thm 8 (HO spectrum), Thm 10 (angular momentum spectrum), non-degeneracy of 1D bound states, Ehrenfest theorem
- **Key connections**: classical ↔ quantum via Poisson bracket ↔ commutator; Ehrenfest resembles Newton's equations; angular momentum ↔ rotations
