## Block 1 — Chapter 6, Part 1: Continuous Spectra & Position Eigenvalues (11:15–12:15, 60 min)

**Sections:** §6.1–6.2 (~200 lines)  
**Pages:** Chapter 6 lecture notes, first half

### Content to cover:
- **§6.1 Discrete and continuous spectra:**
  - Normalisable eigenvectors (discrete spectrum) vs improper eigenvectors (continuous spectrum)
  - Continuous eigenvalues $\omega$: $\langle\varepsilon_{\omega'}|\varepsilon_\omega\rangle = \delta(\omega'-\omega)$ (delta-normalisation)
  - Resolution of identity: $\int_a^b |\varepsilon_\omega\rangle\langle\varepsilon_\omega|\,d\omega = \hat{I}$
  - **Dirac delta distribution:** $\int f(x)\delta(x)\,dx = f(0)$; Gaussian limit (6.12); Fourier representation (6.13)
- **§6.2 Eigenvalues of the position operator:**
  - Position eigenvalue equation: $\hat{q}|x\rangle = x|x\rangle$
  - Matrix elements in HO basis: recursion relation for $\varepsilon_n(\omega) = \langle n|\varepsilon_\omega\rangle$
  - Solution via Hermite polynomials: $\varepsilon_n(\omega) \propto H_n(\omega/L)$
  - Verification that $\langle\varepsilon(\omega)|\varepsilon(\omega')\rangle = \delta(\omega - \omega')$ using Mehler's formula
  - **Key result:** $\langle x|x'\rangle = \delta(x-x')$, continuous spectrum spans all of $\mathbb{R}$

### Key things to understand deeply:
- Why continuous-spectrum eigenvectors have infinite norm — they are not in the Hilbert space
- The delta distribution is only defined inside an integral — never evaluate $\delta(0)$ alone
- The position operator eigenvalue equation connects the abstract Dirac formalism (Ch 3–4) to wave mechanics (Ch 2)
- Hermite polynomial recursion — same polynomials from Ch 5 reappear here

### Active recall prompt:
> Write the orthonormality condition for continuous eigenvectors. State the resolution of identity for position eigenstates. Explain why $\langle x|x'\rangle = \delta(x-x')$ and not $\delta_{x,x'}$.

## Block 2 — Chapter 6, Part 2 + Chapter 7 §7.1: Representations & HO Wave Functions (12:25–13:25, 60 min)

**Sections:** §6.3–6.4, §7.1 (~150 lines)  
**Pages:** Chapter 6 second half + Chapter 7 first section

### Content to cover:
- **§6.3 Position representation:**
  - Wave function: $\psi(x) := \langle x|\psi\rangle$, with $|\psi\rangle = \int \psi(x)|x\rangle\,dx$
  - Position operator action: $\hat{q}: \psi(x) \mapsto x\psi(x)$
  - Momentum operator action: $\hat{p}: \psi(x) \mapsto -i\hbar\frac{\partial}{\partial x}\psi(x)$
  - Kinetic energy: $\hat{E}_\text{kin} = -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}$
  - **Recovery of Schrödinger's wave equation** from the abstract TDSE
- **§6.4 Change of bases / representation:**
  - Insert resolution of identity to change basis: $\psi(x) = \sum_n \psi_n \varepsilon_n(x)$
  - Position → momentum representation via Fourier transform: $\tilde{\psi}(p) = \int \varepsilon_p^*(x)\psi(x)\,dx$
- **§7.1 Position representation of HO eigenstates:**
  - Ground state from $\hat{a}|0\rangle = 0$ in position rep: $\varepsilon_0(x) = \left(\frac{m\omega}{\pi\hbar}\right)^{1/4}e^{-m\omega x^2/2\hbar}$
  - Excited states: $\varepsilon_n(x) = \frac{1}{\sqrt{n!}}\langle x|(\hat{a}^\dagger)^n|0\rangle$ — expressed via Hermite polynomials

### Key things to understand deeply:
- **This block closes the loop:** abstract QM (Ch 3–5) $\to$ position representation $\to$ wave mechanics (Ch 2)
- The momentum operator $\hat{p} = -i\hbar\partial_x$ is **derived** from $[\hat{q},\hat{p}]=i\hbar$, not postulated
- Fourier transform connects position and momentum representations — exam standard
- Ground state derivation is a classic exam question: solve $\hat{a}|0\rangle = 0$ as a first-order ODE

## Block 3 — Chapter 7, Part 1: General Eigenfunction Properties & Free Particle (13:35–14:35, 60 min)

**Sections:** §7.2–7.3.2 (~200 lines)  
**Pages:** Chapter 7 lecture notes, middle section

### Content to cover:
- **§7.2 Properties of eigenfunctions for general 1D potentials:**
  - Bound states ($\varepsilon \to 0$ at $\pm\infty$) vs continuum states (improper eigenfunctions)
  - Eigenfunctions can be chosen real-valued (proof: $\varepsilon + \varepsilon^*$ is also a solution)
  - No bound states below $V_\text{min}$ (proof: curvature argument)
  - Bound state eigenvalues are **non-degenerate** (proof via Wronskian argument)
  - Ground state has no zeros; $n$-th excited state has $n$ zeros
  - **Parity operator $\hat{S}$:** eigenvalues $\pm 1$; symmetric potentials $\Rightarrow$ even/odd eigenfunctions
- **§7.3.1 Boundary conditions for piecewise constant potentials:**
  - Finite jump in $V \Rightarrow$ $\varepsilon$ and $\varepsilon'$ continuous
  - Infinite jump in $V \Rightarrow$ $\varepsilon$ continuous, $\varepsilon'$ may be discontinuous
- **§7.3.2 The free particle:**
  - Plane wave solutions: $\varepsilon_E(x) = Ae^{ikx} + Be^{-ikx}$, $k = \sqrt{2m(E-V_0)}/\hbar$
  - Continuous spectrum $E \geq V_0$
  - Probability flux: $j = \pm \frac{\hbar k}{m}$
  - Plane waves as momentum eigenstates; delta-normalisation

### Key things to understand deeply:
- The non-degeneracy proof (Wronskian) is short and examinable — learn it
- Parity + non-degeneracy $\Rightarrow$ eigenfunctions of symmetric potentials are automatically even or odd
- The free particle has **only** continuous spectrum — no bound states
- Boundary conditions are the key tool for all piecewise problems in Block 4

### Active recall prompt:
> Prove that 1D bound state eigenvalues are non-degenerate (Wronskian argument). State the parity operator eigenvalues. Write the general free-particle solution and its probability flux.

## Block 4 — Chapter 7, Part 2: Square Wells (15:15–16:15, 60 min)

**Sections:** §7.3.3–7.3.4 (~200 lines)  
**Pages:** Chapter 7 lecture notes, final section

### Content to cover:
- **§7.3.3 Infinite square well ("particle in a box") — full derivation:**
  - Potential: $V(x) = 0$ for $0 \leq x \leq L$, $\infty$ otherwise
  - Boundary conditions: $\varepsilon(0) = \varepsilon(L) = 0$
  - Eigenfunctions: $\varepsilon_n(x) = \sqrt{\frac{2}{L}}\sin\!\left(\frac{n\pi x}{L}\right)$
  - Eigenvalues: $E_n = \frac{n^2\pi^2\hbar^2}{2mL^2}$
  - **Now you see why quantisation arises:** from boundary conditions, not from the equation
- **§7.3.4 Finite square well:**
  - Symmetric potential: $V = 0$ for $|x| \leq L/2$, $V_0$ otherwise
  - Ansatz: oscillatory inside ($A\cos kx + B\sin kx$), exponential decay outside ($Ce^{\kappa x}$, $De^{-\kappa x}$)
  - Parity $\Rightarrow$ separate even and odd solutions
  - Even quantisation condition: $\kappa = k\tan(kL/2)$
  - Odd quantisation condition: $\kappa = -k\cot(kL/2)$
  - Graphical / numerical solution for energy levels

### Key things to understand deeply:
- The infinite well derivation is a **standard exam derivation** — be able to write it from scratch
- Finite well: the key insight is that **wave functions penetrate into the classically forbidden region** ($E < V_0$)
- Quantisation conditions are transcendental equations — cannot be solved in closed form
- This is the archetype for all tunneling and scattering problems

### Active recall prompt:
> Derive the infinite well eigenfunctions and eigenvalues from the TISE + boundary conditions. For the finite well, write the piecewise ansatz and derive one of the quantisation conditions. Why does the wave function not vanish outside the well?

## Block 5 — Chapter 8, Part 1: Heisenberg Equation & Ehrenfest Theorem (16:25–17:25, 60 min)

**Sections:** §8.1 (~150 lines)  
**Pages:** Chapter 8 lecture notes, first half

### Content to cover:
- **Dynamics of expectation values:**
  - From TDSE derive: $\frac{d}{dt}\langle\hat{A}\rangle = \frac{i}{\hbar}\langle[\hat{H},\hat{A}]\rangle + \langle\frac{\partial\hat{A}}{\partial t}\rangle$
  - **Heisenberg equation of motion** — the quantum analogue of classical $\dot{f} = \{f,H\}$
  - Quantum Poisson bracket: $\frac{i}{\hbar}[\cdot,\cdot] \leftrightarrow \{\cdot,\cdot\}$
- **Ehrenfest theorem** for $H = \frac{\hat{p}^2}{2m} + V(\hat{q})$:
  - $\frac{d}{dt}\langle\hat{q}\rangle = \frac{\langle\hat{p}\rangle}{m}$ and $\frac{d}{dt}\langle\hat{p}\rangle = -\langle\frac{\partial V}{\partial \hat{q}}\rangle$
  - Derivation uses $[\hat{p}^2,\hat{q}]$ and $[V(\hat{q}),\hat{p}] = i\hbar V'(\hat{q})$
  - **Classical limit:** these look like Newton's equations for expectation values
  - **Crucial caveat:** $\langle V'(\hat{q})\rangle \neq V'(\langle\hat{q}\rangle)$ in general — equality only for potentials up to order 2
- **Literary excursion:** Dirac's "Poisson bracket" moment — connecting classical and quantum mechanics

### Key things to understand deeply:
- The Heisenberg equation is **equivalent** to the Schrödinger equation — different picture, same physics
- The commutator $[V(\hat{q}),\hat{p}] = i\hbar V'(\hat{q})$ calculation (position-representation trick) is exam-standard
- Ehrenfest says expectation values evolve "classically" — but quantum effects appear through the caveat
- This completes the circle back to Chapter 1: $\{f,H\} \to \frac{i}{\hbar}[f,H]$

### Active recall prompt:
> Derive the Heisenberg equation from the TDSE. Compute $[V(\hat{q}),\hat{p}]$ in position representation. State the Ehrenfest theorem and explain when quantum ≠ classical.

## Block 6 — Chapter 8, Part 2: Lie Groups, Hadamard Lemma & SU(2) (17:35–18:35, 60 min)

**Sections:** §8.2–8.3 (~250 lines)  
**Pages:** Chapter 8 lecture notes, second half

### Content to cover:
- **§8.2 Lie groups and Lie algebras in quantum dynamics:**
  - Groups: associativity, identity, inverses; representations $g \mapsto \hat{D}(g)$
  - Lie groups: continuous groups parametrised by real parameters; generators $\hat{T}_j = -i\frac{\partial\hat{D}}{\partial x_j}\big|_{x=0}$
  - Exponential map: $\hat{D}(x) = e^{ix\hat{T}}$ — just like $\hat{U}(t) = e^{-i\hat{H}t/\hbar}$
  - Lie algebra: vector space of generators closed under commutator (Lie bracket)
  - **Examples:** $[\hat{q},\hat{p}] = i\hbar\hat{I}$ (translations); $[\hat{a},\hat{a}^\dagger]=\hat{I}$, $[\hat{N},\hat{a}]=-\hat{a}$ (HO algebra)
  - Magnus form $\hat{U} = e^{i\sum_j \omega_j \hat{K}_j}$ and Wei-Norman form $\hat{U} = \prod_j e^{i\alpha_j\hat{K}_j}$
- **Theorem 9 (Hadamard lemma):**
  - $e^{s\hat{X}}\hat{Y}e^{-s\hat{X}} = \hat{Y} + s[\hat{X},\hat{Y}] + \frac{s^2}{2!}[\hat{X},[\hat{X},\hat{Y}]] + \cdots$
  - Proof by Taylor expansion of $f(s) = e^{s\hat{X}}\hat{Y}e^{-s\hat{X}}$
  - Application: $e^{ia\hat{p}/\hbar}\hat{q}e^{-ia\hat{p}/\hbar} = \hat{q} + a$ (translation operator)
- **§8.2.2 SU(2):** Pauli matrices $\hat{\sigma}_{1,2,3}$; commutation relations $[\hat{\sigma}_j,\hat{\sigma}_k] = 2i\sum_l \epsilon_{jkl}\hat{\sigma}_l$
- **§8.3 Dynamical tunneling** (conceptual — slides)

### Key things to understand deeply:
- The Hadamard lemma is the **key calculational tool** for Chapter 9 (angular momentum & rotations)
- SU(2) and its Lie algebra su(2) will reappear immediately in angular momentum
- The link: $\hat{H}$ generates time evolution $\leftrightarrow$ $\hat{J}_k$ generates rotations
- Pauli matrices — memorise them; they are the $j=\frac{1}{2}$ angular momentum matrices

## Block 7 — Chapter 9, Part 1: Angular Momentum & Spectral Properties (18:55–19:55, 60 min)

**Sections:** §9.1–9.3 (~260 lines)  
**Pages:** Chapter 9 lecture notes, first two-thirds

### Content to cover:
- **§9.1 Definition of angular momentum:**
  - Classical: $\vec{L} = \vec{q} \times \vec{p}$; components $L_1 = q_2 p_3 - q_3 p_2$, etc.
  - Quantum: $\hat{L}_j$ defined analogously; commutation relations $[\hat{L}_j,\hat{L}_k] = i\hbar\sum_l \epsilon_{jkl}\hat{L}_l$
  - **General angular momentum:** any $\hat{J}_{1,2,3}$ with $[\hat{J}_j,\hat{J}_k] = i\hbar\epsilon_{jkl}\hat{J}_l$
  - Total angular momentum: $\hat{J}^2 = \hat{J}_1^2 + \hat{J}_2^2 + \hat{J}_3^2$, commutes with all $\hat{J}_k$
- **§9.2 Angular momentum and rotations:**
  - $e^{i\theta\hat{L}_j/\hbar}$ generates rotations of coordinate operators (via Hadamard lemma)
  - $e^{i\theta\hat{L}_1/\hbar}\hat{q}_2 e^{-i\theta\hat{L}_1/\hbar} = \hat{q}_2\cos\theta - \hat{q}_3\sin\theta$ — matches $R_1(\theta)$
  - Same pattern for $\hat{J}_k$ acting on $\hat{J}_l$
- **§9.3 Spectral properties — Theorem 10 (the big one):**
  - Orbital angular momentum eigenvalues: $L_3$ eigenvalues are $\hbar(n_x - n_y)$, i.e. integer multiples of $\hbar$
  - **Algebraic derivation:** ladder operators $\hat{J}_\pm = \hat{J}_1 \pm i\hat{J}_2$
  - Commutators: $[\hat{J}_3,\hat{J}_\pm] = \pm\hbar\hat{J}_\pm$; identities: $\hat{J}_-\hat{J}_+ = \hat{J}^2 - \hat{J}_3^2 - \hbar\hat{J}_3$
  - Lemma 4: $m^2 \leq \beta$ (bounded); Lemmas 5–6: $\hat{J}_\pm$ raise/lower $m$ by 1
  - **Theorem 10:** $\beta = j(j+1)$ with $2j \in \mathbb{N}$; $m = -j, -j+1, \ldots, j$; total $2j+1$ states

### Key things to understand deeply:
- The **algebraic derivation of Thm 10 parallels the HO derivation** (Ch 5) — same ladder technique
- Half-integer $j$ is allowed by the algebra but **not** for orbital angular momentum (only for spin)
- The bound $m^2 \leq j(j+1)$ comes from positivity of $\|\hat{J}_{1,2}|\beta,m\rangle\|^2$
- This theorem is **very likely on the exam** — practise until you can reproduce it cold

## Block 8 — Chapter 9, Part 2: Matrix Representations & Quantum Spin (20:20–21:20, 60 min)

**Sections:** §9.4–9.5 (~165 lines)  
**Pages:** Chapter 9 lecture notes, final third

### Content to cover:
- **§9.4 Matrix representation in the "standard basis":**
  - Standard basis: $|j,m\rangle$ with $\hat{J}^2|j,m\rangle = \hbar^2 j(j+1)|j,m\rangle$, $\hat{J}_3|j,m\rangle = \hbar m|j,m\rangle$
  - $\hat{J}^2$ and $\hat{J}_3$ are diagonal in this basis
  - Action of $\hat{J}_\pm$: $\hat{J}_+|j,m\rangle = \hbar\sqrt{j(j+1)-m(m+1)}\,|j,m+1\rangle$
  - $\hat{J}_-|j,m\rangle = \hbar\sqrt{j(j+1)-m(m-1)}\,|j,m-1\rangle$
  - $\hat{J}_1 = \tfrac{1}{2}(\hat{J}_+ + \hat{J}_-)$, $\hat{J}_2 = \tfrac{1}{2i}(\hat{J}_+ - \hat{J}_-)$ — matrix elements follow
  - **$j = 1/2$ case:** angular momentum matrices are $\hat{s}_j = \frac{\hbar}{2}\hat{\sigma}_j$ (Pauli matrices return!)
- **§9.5 Quantum spin:**
  - Spin: internal angular momentum, not related to orbital motion
  - Half-integer spin is possible (unlike orbital angular momentum)
  - Bosons (integer spin) vs Fermions (half-integer spin)
  - **Stern-Gerlach experiment** (1922): beam of silver atoms splits into 2 → evidence for $j=1/2$ spin
  - Historical context: Uhlenbeck & Goudsmit (1925), Pauli matrices, Dirac's relativistic QM
  - Spin as qubit — foundation of quantum information

### Key things to understand deeply:
- The $\hat{J}_\pm$ matrix element formula is the exam workhorse — it constructs all angular momentum matrices
- For $j=1/2$: write out the $2\times 2$ matrices for $\hat{J}_{1,2,3}$ — they are always the Pauli matrices × $\hbar/2$
- Stern-Gerlach: why 2 spots (not 3) proves spin-$\frac{1}{2}$, not $l=1$ orbital angular momentum
- Spin is a purely quantum property with no classical analogue
