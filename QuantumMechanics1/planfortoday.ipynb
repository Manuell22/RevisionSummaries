## Block 1 — Chapter 1: Classical Mechanics (09:00–10:10, 70 min)

**Sections:** §1.1–1.4 (entire chapter, ~250 lines)  
**Pages:** Chapter 1 lecture notes

### Content to cover:
- **§1.1 Newton's second law:** $m\ddot{q} = F(q)$, conservative forces $F = -V'(q)$, energy conservation
- **§1.2 Hamilton's canonical equations:** Phase space $(q,p)$, Hamiltonian $H(q,p) = \frac{p^2}{2m} + V(q)$, Hamilton's equations $\dot{q} = \frac{\partial H}{\partial p}$, $\dot{p} = -\frac{\partial H}{\partial q}$
- **§1.3 Poisson brackets:** Definition $\{f,g\} = \frac{\partial f}{\partial q}\frac{\partial g}{\partial p} - \frac{\partial f}{\partial p}\frac{\partial g}{\partial q}$, canonical relations $\{q,p\}=1$, equation of motion $\dot{f} = \{f,H\}$
- **§1.4 Historic excursion:** Motivation for quantum mechanics (blackbody radiation, photoelectric effect, atomic spectra)

### Key things to understand deeply:
- How classical observables are functions on phase space
- The Poisson bracket as the classical precursor to the quantum commutator $[\hat{q},\hat{p}]=i\hbar$
- Why Hamilton's formulation (not Newton's) is the natural starting point for quantisation
- The harmonic oscillator as a worked example: $H = \frac{p^2}{2m} + \frac{1}{2}m\omega^2 q^2$

## Block 2 — Chapter 2, Part 1: Schrödinger Equation (10:20–11:30, 70 min)

**Sections:** §2.1–2.2 (~220 lines)  
**Pages:** Chapter 2 lecture notes, first half

### Content to cover:
- **§2.1 The Schrödinger equation and the wave function:**
  - Time-dependent Schrödinger equation (TDSE): $i\hbar \frac{\partial}{\partial t}\psi(x,t) = \hat{H}\psi(x,t)$
  - Hamiltonian operator: $\hat{H} = -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x)$
  - Wave function $\psi(x,t)$ as probability amplitude
  - Born interpretation: $|\psi(x,t)|^2$ = probability density
  - Normalisation: $\int_{-\infty}^{\infty} |\psi(x,t)|^2 dx = 1$
- **§2.2 Stationary solutions — the TISE:**
  - Separation of variables: $\psi(x,t) = \phi(x)e^{-iEt/\hbar}$
  - Time-independent Schrödinger equation: $\hat{H}\phi = E\phi$ (eigenvalue problem)
  - Stationary states have definite energy $E$
  - Probability density $|\psi|^2 = |\phi|^2$ is time-independent

### Key things to understand deeply:
- The TDSE is first-order in time → initial condition $\psi(x,0)$ determines everything
- Separation of variables reduces the PDE to an ODE eigenvalue problem
- The link between classical $H(q,p)$ and quantum $\hat{H}$: replace $p \to -i\hbar\frac{\partial}{\partial x}$
- Why normalisation is preserved in time (probability conservation)

### Active recall prompt:
> Write down the TDSE and the TISE from memory. State the Born interpretation. Explain how separation of variables connects the two equations.

## Block 3 — Chapter 2, Part 2: Stationary States & Particle in a Box (11:40–12:45, 65 min)

**Sections:** §2.3–2.4 (~220 lines)  
**Pages:** Chapter 2 lecture notes, second half

### Content to cover:
- **§2.3 Method of stationary states (superposition):**
  - General solution: $\psi(x,t) = \sum_n c_n \phi_n(x) e^{-iE_n t/\hbar}$
  - Coefficients from initial condition: $c_n = \langle \phi_n | \psi(x,0) \rangle$
  - Expansion in energy eigenbasis — fundamental technique for solving QM problems
- **§2.4 Particle in a half box (infinite square well):**
  - Boundary conditions: $\phi(0) = \phi(a) = 0$
  - Eigenfunctions: $\phi_n(x) = \sqrt{\frac{2}{a}}\sin\left(\frac{n\pi x}{a}\right)$
  - Energy levels: $E_n = \frac{n^2 \pi^2 \hbar^2}{2ma^2}$, $n=1,2,3,\ldots$
  - Orthonormality of eigenfunctions
  - Worked example: expanding an initial state in this basis

### Key things to understand deeply:
- The method of stationary states is the **core problem-solving technique** in QM1
- Energy quantisation arises from boundary conditions (not from the equation itself)
- Orthonormality: $\int_0^a \phi_m(x)\phi_n(x)dx = \delta_{mn}$
- How to compute $c_n$ for a given initial state — this is a standard exam calculation

**Sections:** §4.1 (~120 lines)  
**Pages:** Chapter 4 lecture notes, first third

### Content to cover:
- **Principle 1 (States):** The state of a quantum system is described by a vector $|\psi\rangle$ in a Hilbert space $\mathcal{H}$
- **Principle 2 (Observables):** Every physical observable $A$ is represented by a Hermitian operator $\hat{A}$ on $\mathcal{H}$
- **Key operators:**
  - Position: $\hat{q}\psi(x) = x\psi(x)$
  - Momentum: $\hat{p}\psi(x) = -i\hbar\frac{\partial}{\partial x}\psi(x)$
  - Canonical commutation relation: $[\hat{q},\hat{p}] = i\hbar$
- **Quantisation recipe:** Classical $\{q,p\}=1$ → Quantum $[\hat{q},\hat{p}]=i\hbar$ (canonical quantisation)

### Key things to understand deeply:
- The correspondence $\{\cdot,\cdot\} \leftrightarrow \frac{1}{i\hbar}[\cdot,\cdot]$ connects Ch 1 to Ch 4
- States live in Hilbert space; observables are Hermitian operators on that space
- The canonical commutation relation $[\hat{q},\hat{p}]=i\hbar$ is the seed of all quantum mechanics
- Why observables must be Hermitian: links back to Theorems 1–3 (real eigenvalues, complete basis)

## Block 7 — Chapter 4, Part 2: Principles 3–5 (17:15–18:05, 50 min)

**Sections:** §4.2 (~120 lines)  
**Pages:** Chapter 4 lecture notes, second third

### Content to cover:
- **Principle 3 (Born rule / Measurement outcomes):**
  - Measuring $\hat{A}$ on $|\psi\rangle = \sum_n c_n |a_n\rangle$ gives outcome $a_n$ with probability $|c_n|^2$
  - Expectation value: $\langle \hat{A} \rangle = \langle \psi | \hat{A} | \psi \rangle = \sum_n |c_n|^2 a_n$
- **Principle 4 (Collapse):**
  - After measuring $a_n$, the state collapses to $|a_n\rangle$ (or projection onto degenerate eigenspace)
  - Immediate re-measurement gives same result with certainty
- **Principle 5 (Dynamics):**
  - Between measurements, state evolves by the TDSE: $i\hbar \frac{d}{dt}|\psi\rangle = \hat{H}|\psi\rangle$
  - Connects back to Chapter 2

### Key things to understand deeply:
- **The Born rule is the bridge between formalism and experiment** — this is how QM makes predictions
- $|c_n|^2 = |\langle a_n | \psi \rangle|^2$ — the probability is the square of the overlap
- Collapse is discontinuous — fundamentally different from Schrödinger evolution
- Principle 5 = TDSE from Ch 2, now stated in Dirac notation

## Block 8 — Chapter 4, Part 3: Commutators & Uncertainty (18:20–19:10, 50 min)

**Sections:** §4.3 (~130 lines)  
**Pages:** Chapter 4 lecture notes, final third

### Content to cover:
- **Commutators:**
  - $[\hat{A},\hat{B}] = \hat{A}\hat{B} - \hat{B}\hat{A}$
  - Compatible observables: $[\hat{A},\hat{B}] = 0$ → can be simultaneously measured
  - Incompatible observables: $[\hat{A},\hat{B}] \neq 0$ → uncertainty relation
- **Theorem 6 (Generalised Uncertainty Principle):**
  - $\Delta A \cdot \Delta B \geq \frac{1}{2}|\langle [\hat{A},\hat{B}] \rangle|$
  - For $\hat{q},\hat{p}$: $\Delta q \cdot \Delta p \geq \frac{\hbar}{2}$ (Heisenberg)
- **Theorem 7 (CSCO — Complete Set of Commuting Observables):**
  - Commuting Hermitian operators share a common eigenbasis
  - A CSCO uniquely labels all states
  - Importance for quantum numbers and degeneracy resolution

### Key things to understand deeply:
- $[\hat{A},\hat{B}]=0$ is the condition for simultaneous eigenstates — fundamental for CSCO
- The uncertainty principle is a **theorem**, not a postulate — it follows from the commutation relation
- CSCO is the quantum version of "enough coordinates to specify a unique state"
- Proof sketch of Theorem 6: Cauchy-Schwarz on $\langle (\hat{A}-\langle A\rangle)\psi | (\hat{B}-\langle B\rangle)\psi \rangle$

### Active recall prompt:
> State Theorem 6. Derive Heisenberg's uncertainty relation from it using $[\hat{q},\hat{p}]=i\hbar$. Define CSCO and explain why commuting operators share eigenstates.

## Block 9 — Chapter 5, Part 1: Ladder Operators & Algebraic Solution (19:40–20:25, 45 min)

**Sections:** §5.1–5.2 (~150 lines)  
**Pages:** Chapter 5 lecture notes, first half

### Content to cover:
- **§5.1 Algebraic solution of the harmonic oscillator:**
  - Hamiltonian: $\hat{H} = \frac{\hat{p}^2}{2m} + \frac{1}{2}m\omega^2 \hat{q}^2$
  - Dimensionless operators: $\hat{Q} = \sqrt{\frac{m\omega}{\hbar}}\hat{q}$, $\hat{P} = \frac{1}{\sqrt{m\omega\hbar}}\hat{p}$
  - Ladder operators: $\hat{a} = \frac{1}{\sqrt{2}}(\hat{Q} + i\hat{P})$, $\hat{a}^\dagger = \frac{1}{\sqrt{2}}(\hat{Q} - i\hat{P})$
  - Number operator: $\hat{N} = \hat{a}^\dagger \hat{a}$, with $\hat{H} = \hbar\omega(\hat{N} + \frac{1}{2})$
  - **Theorem 8:** Eigenvalues $E_n = \hbar\omega(n + \frac{1}{2})$, $n = 0,1,2,\ldots$
  - $\hat{a}|n\rangle = \sqrt{n}|n-1\rangle$ (lowering), $\hat{a}^\dagger|n\rangle = \sqrt{n+1}|n+1\rangle$ (raising)
  - Ground state: $\hat{a}|0\rangle = 0$ → determines $\phi_0(x)$
- **§5.2 Normalised eigenvectors:**
  - $|n\rangle = \frac{(\hat{a}^\dagger)^n}{\sqrt{n!}}|0\rangle$
  - Ground state wave function: $\phi_0(x) = \left(\frac{m\omega}{\pi\hbar}\right)^{1/4} e^{-m\omega x^2/2\hbar}$ (Gaussian)

### Key things to understand deeply:
- The algebraic method avoids solving the ODE — uses only commutation relations
- $[\hat{a},\hat{a}^\dagger] = 1$ is the key identity that drives everything
- Zero-point energy $E_0 = \frac{1}{2}\hbar\omega$ — a purely quantum effect
- Building excited states from the ground state via $\hat{a}^\dagger$

> Define $\hat{a}$ and $\hat{a}^\dagger$. Show $[\hat{a},\hat{a}^\dagger]=1$. State Theorem 8 (HO eigenvalues). Write the formula for $|n\rangle$ in terms of $|0\rangle$. What is $E_0$?

**Sections:** §5.3–5.4 (~150 lines)

### Content to cover:
- **§5.3 Matrix representations in the HO eigenbasis:**
  - Matrix elements of $\hat{a}$, $\hat{a}^\dagger$, $\hat{Q}$, $\hat{P}$, $\hat{H}$ in $\{|n\rangle\}$ basis
  - $\hat{a}$ is lower-triangular: $(\hat{a})_{mn} = \sqrt{n}\delta_{m,n-1}$

  - Discretising the TISE on a grid
  - Finite-difference approximation for $-\frac{\hbar^2}{2m}\frac{d^2\phi}{dx^2}$ → tridiagonal matrix
  - Eigenvalue problem becomes numerical linear algebra
  - Comparison with analytical HO results

### Key things to understand deeply:
- Matrix representation makes abstract operators concrete and computable
- The HO eigenbasis is a natural "coordinate system" for operators — used heavily later
- $\hat{Q}$ and $\hat{P}$ as combinations of $\hat{a}$ and $\hat{a}^\dagger$ — this is the standard trick
- The numerical method is a practical tool but less likely to be examined than algebraic methods
