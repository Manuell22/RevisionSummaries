# QM1 — Question Methods Guide (Chapters 1–6, Non-Mastery)

> 3 questions, ~20 marks each. Based on 6 years of past exams (2020–2025).
> Each section below is a **complete method recipe** for one question type.

---

# QUESTION 1 — Wave Function (Ch 1–2)
## Appears every year. ~20 marks.

---

### Sub-part: Sketch $|\psi(x)|$
- Find where $|\psi(x)| = 0$ (usually at $x=0$ or boundary)
- Find where it peaks (differentiate and set to zero — same as most probable position)
- Check behaviour as $x \to \pm\infty$ (should decay to zero for normalised $\psi$)

---

### Sub-part: Normalise $\psi$

Set $\int |\psi(x)|^2 dx = 1$ and solve for constant $A$.

**Key integrals:**
$$\int_0^\infty x^n e^{-\lambda x} dx = \frac{n!}{\lambda^{n+1}}, \qquad \int_{-\infty}^\infty e^{-ax^2} dx = \sqrt{\frac{\pi}{a}}$$

Note: $|e^{-i\omega t}|^2 = 1$ so time factors drop out immediately.

---

### Sub-part: Probability in interval $[0, a]$

$$P(x \in [0,a]) = \int_0^a |\psi(x)|^2 dx$$

Usually just write the expression — calculation may not be required.

---

### Sub-part: Most probable position $x^*$

Maximise $|\psi(x)|^2$. Set $\frac{d}{dx}|\psi(x)|^2 = 0$, solve for $x$.

(Tip: maximising $|\psi|^2$ is the same as maximising $|\psi|$ — pick whichever is easier to differentiate.)

---

### Sub-part: Expectation value $\langle x \rangle$

$$\langle x \rangle = \int_{-\infty}^\infty x\,|\psi(x)|^2\,dx$$

Use $\int_0^\infty x^n e^{-\lambda x}dx = n!/\lambda^{n+1}$, or the hint trick $xe^{-\lambda x} = -\frac{d}{d\lambda}e^{-\lambda x}$.

---

### Sub-part: Uncertainty $\Delta x$

$$(\Delta x)^2 = \langle x^2 \rangle - \langle x \rangle^2, \qquad \langle x^2 \rangle = \int_{-\infty}^\infty x^2 |\psi(x)|^2 dx$$

---

### Sub-part: Find potential $V(x)$ from eigenfunction

If $\psi(x)$ is given as an energy eigenfunction with energy $E_0$, rearrange the TISE:

$$V(x) = E_0 + \frac{\hbar^2}{2m}\frac{\psi''(x)}{\psi(x)}$$

Compute $\psi''(x)$ and divide by $\psi(x)$.

# QUESTION 2 — Postulates & Measurement (Ch 3–4)
## Appears every year. ~20 marks. Usually a 2- or 3-level system.

---

### Sub-part: Matrix representation of an operator

If given $\hat{A}|\phi_1\rangle = \ldots$ and $\hat{A}|\phi_2\rangle = \ldots$, the matrix is:

$$A_{ij} = \langle\phi_i|\hat{A}|\phi_j\rangle$$

i.e. the $j$-th column is the vector of components of $\hat{A}|\phi_j\rangle$ in the basis $\{|\phi_i\rangle\}$.

**Check Hermitian:** $A = A^\dagger$ (conjugate transpose equals itself).

**Check commutation:** Compute $AB - BA$ and check if it's zero.

---

### Sub-part: Eigenvalues and eigenvectors of a matrix

For a $2\times2$ matrix: $\det(A - \lambda I) = 0$ gives the characteristic equation.

For a $3\times3$: same, but factor the determinant — often one eigenvalue is obvious by inspection.

Normalise each eigenvector: divide by its norm.

---

### Sub-part: Measurement probabilities

Expand $|\psi\rangle$ in the eigenbasis of the observable being measured:
$$|\psi\rangle = \sum_n a_n |\varepsilon_n\rangle, \qquad a_n = \langle\varepsilon_n|\psi\rangle$$

Then $p(a_n) = |a_n|^2$.

**Expectation value:** $\langle\hat{A}\rangle = \sum_n a_n^2 \cdot (\text{eigenvalue}_n)$, or directly $\langle\psi|\hat{A}|\psi\rangle$.

---

### Sub-part: Subsequent measurement (THE KEY METHOD — appears every year)

> "Measurement of $\hat{A}$ at $t=0$ yields result $a_j$. What is the probability that measurement of $\hat{B}$ at $t>0$ yields result $b_k$?"

**Step 1 — Collapse:** Post-measurement state is $|\alpha_j\rangle$ (normalised eigenstate of $\hat{A}$ for eigenvalue $a_j$).

**Step 2 — Time evolve:** Expand $|\alpha_j\rangle$ in the **energy eigenbasis** $\{|\varepsilon_n\rangle\}$:
$$|\alpha_j\rangle = \sum_n c_n|\varepsilon_n\rangle, \qquad c_n = \langle\varepsilon_n|\alpha_j\rangle$$

Then: $|\psi(t)\rangle = \sum_n c_n e^{-iE_n t/\hbar}|\varepsilon_n\rangle$

**Step 3 — Measure $\hat{B}$:** Compute $p(b_k, t) = |\langle\beta_k|\psi(t)\rangle|^2$

where $|\beta_k\rangle$ is the eigenstate of $\hat{B}$ for eigenvalue $b_k$.

> **Special case:** If the collapsed state IS an energy eigenstate (e.g. you measured $\hat{H}$), then time evolution only adds a global phase $e^{-iEt/\hbar}$ which cancels in $|\cdot|^2$. The subsequent measurement probability is **time-independent**.

---

### Sub-part: Uncertainty $\Delta A$

$$(\Delta A)^2 = \langle\hat{A}^2\rangle - \langle\hat{A}\rangle^2$$

To compute $\langle\hat{A}^2\rangle$: either apply $\hat{A}$ twice, or use $\hat{A}^2 = \hat{A}\cdot\hat{A}$ as a matrix product.

**Uncertainty relation check:** $\Delta A \cdot \Delta B \geq \frac{1}{2}|\langle[\hat{A},\hat{B}]\rangle|$

# QUESTION 3 — Harmonic Oscillator (Ch 5) or Representations (Ch 6)
## Appeared 5 out of 6 years as Q2 or Q3. ~20 marks.

---

## 3A. Harmonic Oscillator — Algebraic (most common)

### Sub-part: Show $\hat{a}|n\rangle$ is eigenvector of $\hat{N}$ with eigenvalue $n-1$

Use $[\hat{N}, \hat{a}] = -\hat{a}$:
$$\hat{N}(\hat{a}|n\rangle) = (\hat{a}\hat{N} - \hat{a})|n\rangle = \hat{a}(n|n\rangle) - \hat{a}|n\rangle = (n-1)\,\hat{a}|n\rangle$$

Same argument for $\hat{a}^\dagger$ using $[\hat{N}, \hat{a}^\dagger] = +\hat{a}^\dagger$.

---

### Sub-part: Show eigenvalues of $\hat{N}$ are non-negative / $\hat{a}|0\rangle = 0$

$$\|\hat{a}|\nu\rangle\|^2 = \langle\nu|\hat{a}^\dagger\hat{a}|\nu\rangle = \langle\nu|\hat{N}|\nu\rangle = \nu$$

Since a norm squared $\geq 0$, we need $\nu \geq 0$. If $\nu = 0$ then $\|\hat{a}|0\rangle\|^2 = 0$ so $\hat{a}|0\rangle = \mathbf{0}$.

---

### Sub-part: Expectation value of position/momentum in eigenstate $|n\rangle$

Express in terms of ladder operators:
$$\hat{Q} = \frac{1}{\sqrt{2}}(\hat{a} + \hat{a}^\dagger), \qquad \hat{P} = \frac{1}{i\sqrt{2}}(\hat{a} - \hat{a}^\dagger)$$

Then e.g. $\langle n|\hat{Q}|n\rangle = \frac{1}{\sqrt{2}}\langle n|(\hat{a}+\hat{a}^\dagger)|n\rangle = \frac{1}{\sqrt{2}}(\sqrt{n}\underbrace{\langle n|n-1\rangle}_{0} + \sqrt{n+1}\underbrace{\langle n|n+1\rangle}_{0}) = 0$

For $\langle\hat{Q}^2\rangle$: use $\hat{Q}^2 = \frac{1}{2}(\hat{a}+\hat{a}^\dagger)^2 = \frac{1}{2}(\hat{a}^2 + \hat{a}\hat{a}^\dagger + \hat{a}^\dagger\hat{a} + (\hat{a}^\dagger)^2)$

Only $\hat{a}\hat{a}^\dagger = \hat{N}+1$ and $\hat{a}^\dagger\hat{a} = \hat{N}$ survive: $\langle n|\hat{Q}^2|n\rangle = \frac{1}{2}(2n+1) = n + \frac{1}{2}$

---

### Sub-part: Expectation value of $\hat{Q}$ in a superposition at time $t$

State: $|\psi(t)\rangle = \sum_n a_n e^{-iE_nt/\hbar}|n\rangle$

$$\langle\hat{Q}(t)\rangle = \langle\psi(t)|\hat{Q}|\psi(t)\rangle = \frac{1}{\sqrt{2}}\sum_{n,m} a_m^* a_n\, e^{i(E_m-E_n)t/\hbar} \langle m|(\hat{a}+\hat{a}^\dagger)|n\rangle$$

Only adjacent terms survive: $\langle m|\hat{a}|n\rangle = \sqrt{n}\,\delta_{m,n-1}$ and $\langle m|\hat{a}^\dagger|n\rangle = \sqrt{n+1}\,\delta_{m,n+1}$

So only pairs $(m,n)$ with $m = n\pm 1$ contribute — everything else is zero by orthogonality.

---

## 3B. Position/Momentum Representation (Ch 6) — also appears

### Sub-part: Switch representation using resolution of identity

Insert $\hat{I} = \int |x\rangle\langle x|\,dx$:
$$\langle n|\hat{q}|m\rangle = \int_{-\infty}^\infty \langle n|x\rangle\, x\, \langle x|m\rangle\, dx = \int_{-\infty}^\infty x\,\phi_n^*(x)\,\phi_m(x)\,dx$$

### Sub-part: Show $\langle p|\hat{p}^2|\psi\rangle = p^2\tilde{\psi}(p)$ or $\langle x|\hat{p}|\psi\rangle = -i\hbar\frac{\partial}{\partial x}\psi(x)$

Standard result — just state and apply. Never derived from scratch in the exam.

### Sub-part: Express eigenvalue equation in position or momentum rep

Position rep: replace $\hat{p} \to -i\hbar\frac{\partial}{\partial x}$, replace $\hat{q} \to x$ (multiplication).

Momentum rep: replace $\hat{q} \to i\hbar\frac{\partial}{\partial p}$, replace $\hat{p} \to p$ (multiplication).

---

# COMMON TRAPS & TIPS

| Trap | Fix |
|------|-----|
| Forgetting $\lvert e^{-i\omega t}\rvert^2 = 1$ | Time factors always vanish in $\lvert\psi\rvert^2$ |
| Using $\lvert a_n\rvert^2$ when $a_n$ is complex | Take modulus squared properly: $a_n^* a_n$ |
| Subsequent measurement — forgetting to time-evolve | Always expand collapsed state in energy basis first |
| Assuming subsequent probability is time-independent | Only true if collapsed state is already an energy eigenstate |
| Wrong sign in $\hat{P} = \frac{1}{i\sqrt{2}}(\hat{a}-\hat{a}^\dagger)$ | Remember the $i$ in the denominator |
| Missing the $\sqrt{n}$ factors | $\hat{a}\lvert n\rangle = \sqrt{n}\lvert n-1\rangle$, NOT $\lvert n-1\rangle$ |
| Not normalising eigenvectors | Always divide by the norm before using as a state |
| Wrong integration limits | Check: half-line $[0,\infty)$ vs full line $(-\infty,\infty)$ |
