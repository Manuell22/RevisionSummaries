**§2.2 — Transformation methods**
- **Inverse CDF (Theorem 2.1):** If $U \sim \text{Unif}(0,1)$, then $X = F_X^{-1}(U) \sim p_X$ — proof is one line: $P(F_X^{-1}(U) \leq x) = P(U \leq F_X(x)) = F_X(x)$
- Algorithm 1: inverse transform sampling
- **General transformation (Remark 2.1):** 1D: $p_Y(y) = p_X(g^{-1}(y))\left|\frac{dg^{-1}}{dy}\right|$; nD: $p_Y(y) = p_X(g^{-1}(y))|\det J_{g^{-1}}(y)|$
- **Box-Müller transform:** $Z_1 = \sqrt{-2\ln U_1}\cos(2\pi U_2)$, $Z_2 = \sqrt{-2\ln U_1}\sin(2\pi U_2)$ gives i.i.d. $N(0,1)$

**§2.3 — Composition**
- **Chain rule for sampling (Prop 2.1):** Sample $X_1 \sim p_{X_1}$, then $X_2|X_1 \sim p_{X_2|X_1}$ → joint $(X_1,X_2) \sim p(x_1,x_2)$
- **Marginalisation by projection (Prop 2.2):** Sample from joint, keep one coordinate → samples from marginal
- **Discrete mixtures (Algorithm 3):** Sample index $k \sim \text{Cat}(w_1,\ldots,w_K)$, then $X \sim q_k(x)$

### §2.3 Composition Method

**Key idea:** If $p(x) = \sum_k w_k p_k(x)$ (mixture), sample the component index $k \sim (w_1,\ldots,w_K)$ then sample $X \sim p_k$.

**Algorithm 4:** (1) draw $k$ from discrete distribution with weights $w_k$; (2) draw $X \sim p_k$.

**Why useful:** Avoids computing intractable CDFs for the full mixture; each component may have a simple inverse CDF.

### §2.4 Multivariate Gaussian

**Result:** If $Z \sim \mathcal{N}(0, I_d)$ and $\Sigma = LL^\top$ (Cholesky), then $X = \mu + LZ \sim \mathcal{N}(\mu, \Sigma)$.

**Algorithm:** (1) decompose $\Sigma = LL^\top$; (2) draw $Z_1,\ldots,Z_d$ i.i.d. $\mathcal{N}(0,1)$; (3) return $\mu + LZ$.

### §2.5 Rejection Sampling (through §2.5.2)

**Setup:** Target $p^*(x)$ (unnormalised ok); proposal $q(x)$ (can sample from); constant $M \geq \sup_x \frac{\bar{p}^*(x)}{q(x)}$.

**Theorem 2.2 (Fundamental Theorem of Simulation):** $(X, U) \sim \mathrm{Uniform}(\{(x,u): 0 \leq u \leq \bar{p}^*(x)\})$ $\Leftrightarrow$ $X \sim p^*(\cdot)$.

**Algorithm 5** (normalised): draw $X \sim q$, $U \sim \mathrm{Uniform}[0,1]$; accept if $U \leq \frac{p^*(X)}{Mq(X)}$.
**Algorithm 6** (unnormalised): draw $X \sim q$, $U \sim \mathrm{Uniform}[0, Mq(X)]$; accept if $U \leq \bar{p}^*(X)$.

**Prop 2.3 (Acceptance rate):** $\Pr(\text{accept}) = \frac{1}{M} \int \bar{p}^*(x)\,dx$ (normalised: $= 1/M$; unnorm: $= Z_{p^*}/M$).

**§2.5.1 — Optimal M:** Minimise $M = \sup_x \frac{\bar{p}^*(x)}{q(x)}$. Set $x^* = \arg\max_x \frac{\bar{p}^*(x)}{q(x)}$, then $M^* = \frac{\bar{p}^*(x^*)}{q(x^*)}$. Find $x^*$ via $\frac{d}{dx}\log\frac{\bar{p}^*(x)}{q(x)} = 0$.

**§2.5.2 — Optimal proposal:** Parameterise $q_\theta$; minimise $\log M_\theta = \log\frac{\bar{p}^*(x^*(\theta))}{q_\theta(x^*(\theta))}$ over $\theta$.

*Worked example:* Target $p^*(x) \propto x^{\alpha-1}e^{-x}$ (Gamma), proposal $q_\lambda(x) = \lambda e^{-\lambda x}$ (Exp). Optimal $\lambda^* = \frac{1}{\alpha}$ (minimise $\log M_\lambda$ over $\lambda$; $M_{\lambda^*} = (\alpha-1)^{\alpha-1}e^{-(\alpha-1)}/\Gamma(\alpha)$).

### §2.6 Curse of Dimensionality

**Key result:** In $d$ dimensions, the acceptance rate of rejection sampling decays exponentially: $\Pr(\text{accept}) = e^{-O(d)}$ in typical cases.

**Intuition:** A $d$-ball of radius $r$ relative to a $d$-cube has volume $\to 0$ as $d \to \infty$. The proposal wastes exponentially more mass in the "corners" the target avoids.

**Formal argument:** If $X \sim \mathcal{N}(0, I_d)$, then $\|X\|^2 \sim \chi^2_d$ concentrates around $d$. A spherical Gaussian proposal centred on 0 has most mass in a thin shell of radius $\sqrt{d}$, while the tails/corners receive negligible mass.

**Consequence for rejection sampling:** $M = \sup_x \frac{\bar{p}^*(x)}{q(x)}$ grows exponentially with $d$, so acceptance probability $1/M \to 0$. Rejection sampling is **not suitable** in high dimensions → motivates importance sampling (Ch 3) and MCMC.

**Key take-away:** Rejection sampling is reliable only in low dimension ($d \leq 3$–$5$). For higher $d$, need IS or MCMC.

### §3.1 Monte Carlo Integration

**Problem:** Estimate $I = \mathbb{E}_{p}[f(X)] = \int f(x)\,p(x)\,dx$ when exact integration is intractable.

**MC estimator:** $\hat{I}_N = \frac{1}{N}\sum_{i=1}^N f(X_i)$, $X_i \overset{\text{i.i.d.}}{\sim} p$.

**By SLLN:** $\hat{I}_N \to I$ a.s. as $N\to\infty$.

**Variance:** $\mathrm{Var}(\hat{I}_N) = \frac{\sigma_f^2}{N}$ where $\sigma_f^2 = \mathrm{Var}_p(f(X))$.

**Key points to note while reading:**
- When is $\sigma_f^2$ finite? (need $\mathbb{E}_p[f^2] < \infty$)
- How does the error scale with $N$? ($\propto N^{-1/2}$ regardless of dimension — key advantage over quadrature)
- CLT for MC: $\sqrt{N}(\hat{I}_N - I) \xrightarrow{d} \mathcal{N}(0, \sigma_f^2)$

---

### §3.2 Error Metrics

**MSE decomposition:** $\mathrm{MSE}(\hat{I}) = \mathrm{Var}(\hat{I}) + \mathrm{Bias}(\hat{I})^2$.

**For unbiased MC estimator:** $\mathrm{MSE} = \mathrm{Var}(\hat{I}_N) = \sigma_f^2 / N$.

**Error metrics to read:**
- Absolute error: $|\hat{I}_N - I|$
- Relative error: $|\hat{I}_N - I|/|I|$
- Root MSE: $\sqrt{\mathrm{MSE}}$; how to estimate $\sigma_f$ from samples?

**Estimating $\sigma_f^2$:** Sample variance $s^2 = \frac{1}{N-1}\sum_{i=1}^N (f(X_i) - \hat{I}_N)^2$; use $s/\sqrt{N}$ as standard error.

**Confidence intervals:** $\hat{I}_N \pm z_{\alpha/2} \cdot s/\sqrt{N}$ (by CLT for large $N$).

### §3.3 Importance Sampling — Overview

**Problem:** $p(x)$ is hard to sample from directly, but $q(x)$ is easy. Rewrite:

$$I = \int f(x)\,p(x)\,dx = \int f(x)\,\frac{p(x)}{q(x)}\,q(x)\,dx = \mathbb{E}_q\!\left[f(X)\,w(X)\right],\quad w(x) = \frac{p(x)}{q(x)}$$

**IS estimator:** $\hat{I}^{\mathrm{IS}}_N = \frac{1}{N}\sum_{i=1}^N f(X_i)\,w(X_i)$, $X_i \overset{\text{i.i.d.}}{\sim} q$.

---

### §3.3.1 Basic Importance Sampling Estimator

**Requirements:** $q(x) > 0$ wherever $f(x)p(x) \neq 0$ (support condition).

**Unnormalised weights:** $w(x) = p(x)/q(x)$ (need $p$ normalised, or use normalised IS — §3.3.3).

**Variance of IS estimator:**

$$\mathrm{Var}_q(\hat{I}^{\mathrm{IS}}_N) = \frac{1}{N}\left[\mathbb{E}_q\!\left(f(X)^2 w(X)^2\right) - I^2\right]$$

**Key points while reading:**
- When is IS variance finite? (need $\mathbb{E}_q[f^2 w^2] < \infty$, i.e. $\chi^2(p\|q) < \infty$)
- Compare to plain MC variance $\sigma_f^2/N$: IS can be better OR worse depending on $q$.
- "Heavy tails" warning: if $q$ has lighter tails than $p$, $w(x)$ can have infinite variance → $\hat{I}^{\rm IS}_N$ unstable.

---

### §3.3.2 Optimal IS Proposal

**Goal:** Find $q^*$ minimising $\mathrm{Var}_q(\hat{I}^{\mathrm{IS}}_N)$.

**Result:** The zero-variance proposal is $q^*(x) = \frac{|f(x)|\,p(x)}{\int |f(x')|p(x')\,dx'}$.

**Why impractical:** Computing the normalising constant $\int |f|p$ is itself the integral we want to estimate.

**Practical consequence:** Use $q$ proportional to $|f(x)|\,p(x)$ as a guide — concentrate proposal where $|f|\,p$ is large (not just where $p$ is large).

**Second-order variance minimisation (if covered):** $\mathrm{Var}_q^* = 0$ iff $f$ is a.e. constant under $p$ (trivial case). In general, the optimal $q$ trades off the tails of $f$ and $p$.
