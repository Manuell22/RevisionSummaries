## 1. Continuous Distributions — PDF, CDF, Inverse

---

### Exponential($\lambda$)
> ✏️ **Memorise** — used as proposal/sampler throughout all years; never written out in 2023–2025 papers.

$$p(x) = \lambda e^{-\lambda x}, \quad x \geq 0$$
$$F(x) = 1 - e^{-\lambda x}$$
$$F^{-1}(u) = -\frac{1}{\lambda}\log(1-u)$$

**Appears as:** Exact sampling (Example 2.2, Lec Notes Ch2); rejection/IS proposal throughout 2023–2025.  
**Special case:** $\lambda=1$: $p(x)=e^{-x}$.

---

### Rayleigh($\sigma$)
> 🎁 **Given in paper** — 2023 Q1 gives $p(x)=xe^{-x^2/2}$; 2024 Q2 gives the general form. CDF and inversion are derived in the exam.

$$p(x) = \frac{x}{\sigma^2}\exp\!\left(-\frac{x^2}{2\sigma^2}\right), \quad x \geq 0$$
$$F(x) = 1 - \exp\!\left(-\frac{x^2}{2\sigma^2}\right)$$
$$F^{-1}(u) = \sqrt{-2\sigma^2 \log(1-u)}$$

**Inversion sample:** Draw $U\sim\text{Unif}(0,1)$, return $X = \sqrt{-2\sigma^2\log(1-U)}$.

---

### Weibull($k$) (scale = 1)
> 🎁 **Given in paper** — 2024 Q1 gives $p_X(x)=kx^{k-1}\exp(-x^k)$. CDF and inversion are derived in the exam (Exercise 2.2).

$$p(x) = k x^{k-1}\exp(-x^k), \quad x \geq 0$$
$$F(x) = 1 - \exp(-x^k)$$
$$F^{-1}(u) = \big(-\log(1-u)\big)^{1/k}$$

**Connection to Exponential:** If $E\sim\text{Exp}(1)$ then $X = E^{1/k}\sim\text{Weibull}(k)$.

---

### Normal / Gaussian $\mathcal{N}(\mu,\sigma^2)$
> ✏️ **Memorise** — appears everywhere as prior, likelihood, MH proposal; assumed known in all years.

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}\exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

No closed-form CDF inverse — use **Box-Müller** (see §5).  
**Standard Normal:** $\mu=0,\sigma^2=1$.

---

### Chi-Squared $\chi^2(k)$ (= Gamma($k/2$, $1/2$))
> 🎁 **Given in paper** — 2025 Q1b explicitly writes out the formula.

$$p(x) = \frac{1}{2^{k/2}\,\Gamma(k/2)}\,x^{k/2-1}\,e^{-x/2}, \quad x > 0$$

**Key identity:** $\Gamma(n) = (n-1)!$ for $n\in\mathbb{N}$ — ✏️ memorise this.  
**2025 Q1c target:** $p(x)\propto x^2 e^{-x/2}$ — recognise as unnormalised $\chi^2(6)$; $Z = \Gamma(3)\cdot 2^3 = 16$.

---

### Gamma($\alpha$, $\beta$) — rate parameterisation
> 🎁 **Given in paper** — written out as prior in 2024 Q3b and 2025 Q2a. You must derive the posterior yourself.

$$p(x) = \frac{\beta^\alpha}{\Gamma(\alpha)}\,x^{\alpha-1}\,e^{-\beta x}, \quad x \geq 0$$

**Mean:** $\alpha/\beta$. **Variance:** $\alpha/\beta^2$.  
**Special cases:** $\alpha=1$: Exponential($\beta$) — $\alpha=k\in\mathbb{N}$: Erlang($k,\beta$).

---

### Erlang / Gamma($k$, $\lambda$) — integer shape
> 🎁 **Given in paper** — 2024 Q1d gives this formula explicitly.

$$p(x) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}, \quad x \geq 0,\; k \in \mathbb{N}^+$$

**Key fact:** Sum of $k$ independent Exp($\lambda$) r.v.s is Erlang($k,\lambda$).

---

### Inverse-Gamma IG($\alpha$, $\beta$)
> 🎁 **Given in paper** — 2023 Q3c explicitly writes out the formula. You must derive the posterior update rule.

$$p(\sigma^2) = \frac{\beta^\alpha}{\Gamma(\alpha)}\left(\frac{1}{\sigma^2}\right)^{\alpha+1}\exp\!\left(-\frac{\beta}{\sigma^2}\right), \quad \sigma^2 > 0$$

**Connection:** If $X\sim\text{Gamma}(\alpha,\beta)$ then $1/X\sim\text{IG}(\alpha,\beta)$.

---

### Cauchy
> ✏️ **Memorise** — appears only in transitional 2022 Q2a and as a lecture-note inversion example (Ch2 Example 2.3). Low priority for 2023–2025 curriculum.

$$p(x) = \frac{1}{\pi(1+x^2)}, \quad x\in\mathbb{R}$$
$$F(x) = \frac{1}{2} + \frac{1}{\pi}\arctan(x), \qquad F^{-1}(u) = \tan\!\left(\pi\!\left(u - \tfrac{1}{2}\right)\right)$$

**Note:** No finite mean or variance.

---

### Multivariate / Bivariate Normal
> ✏️ **Memorise (low priority)** — only in transitional 2022 Q3c; unlikely in 2023–2025 format.

$$f(\mathbf{z}) = \frac{1}{2\pi|\Sigma|^{1/2}}\exp\!\left(-\tfrac{1}{2}(\mathbf{z}-\boldsymbol{\mu})^\top\Sigma^{-1}(\mathbf{z}-\boldsymbol{\mu})\right)$$

Standard bivariate: $\Sigma = \begin{pmatrix}1 & \rho\\ \rho & 1\end{pmatrix}$, $|\Sigma|=1-\rho^2$.  
**Conditional:** $X\mid Y=y \sim \mathcal{N}(\rho y,\;1-\rho^2)$

## 2. Discrete Distributions — PMF, CDF, Inverse

---

### Poisson($\lambda$)
> 🎁 **Given in paper** — 2024 Q3b and 2025 Q2 both write out $P(Y=y\mid x)=\frac{x^y e^{-x}}{y!}$ explicitly.

$$P(Y=y\mid \lambda) = \frac{\lambda^y e^{-\lambda}}{y!}, \quad y = 0,1,2,\ldots$$

**CDF:** $F(k) = e^{-\lambda}\sum_{i=0}^{k}\frac{\lambda^i}{i!}$ — no closed form; inversion uses table-lookup.  
**Conjugate prior:** Gamma($\alpha,\beta$) → posterior Gamma($\alpha+y,\beta+1$). ✏️ Memorise the update rule.

---

### Geometric($q$)
> ✏️ **Memorise** — appears in transitional 2022 Q1b and 2021. Unlikely in 2023–2025 but good for inversion method.

$$P(X=x) = q(1-q)^{x-1}, \quad x = 1,2,3,\ldots$$
$$F(x) = 1-(1-q)^x, \qquad F^{-1}(u) = \left\lfloor\frac{\log(1-u)}{\log(1-q)}\right\rfloor + 1$$

---

### Bernoulli($p$)
> ✏️ **Memorise** — assumed known; standard inversion example.

$$P(X=1)=p,\quad P(X=0)=1-p$$

**Inversion:** Draw $U$; return $X=1$ if $U\leq p$, else $0$.

## 3. Non-Standard / Unnormalised Targets from Exams

These appear as rejection-sampling or MH targets. You must find a normalising constant $Z$ or an upper bound $M$.

---

### 2025 Q1c — Chi-squared(6) unnormalised
$$p^*(x) = x^2\,e^{-x/2}, \quad x > 0$$

**Normalised:** $p(x) = \frac{1}{16}\,x^2\,e^{-x/2} = \frac{1}{2^3\,\Gamma(3)}\,x^{3-1}\,e^{-x/2}$ — this is $\chi^2(6)$.  
**Normalising constant:** $Z = \int_0^\infty x^2 e^{-x/2}\,dx = \Gamma(3)\cdot 2^3 = 2\cdot 8 = 16$.

---

### 2024 Q2 — Rayleigh (normalised but stated as target)
$$p(x) = x\,e^{-x^2/2}, \quad x \geq 0$$

This **is** normalised (Rayleigh with $\sigma=1$). Verify: $\int_0^\infty x e^{-x^2/2}dx = 1$.

---

### 2023 Q3 — Inverse-Gamma target (Bayesian posterior)
$$p(\sigma^2 \mid x) \propto \left(\frac{1}{\sigma^2}\right)^{\alpha+1}\exp\!\left(-\frac{\beta}{\sigma^2}\right)$$

Recognised as IG($\alpha,\beta$). Update rule: data $x$ observed → posterior IG$(\alpha+1/2,\,\beta+x^2/2)$.

---

### 2022 Q1c (transitional) — Truncated Beta-type
$$f^*(x) \propto x^{\theta-1}(1-x)^{\theta-1}, \quad x\in\left(\tfrac{1}{2},1\right)$$

Unnormalised truncated Beta. Use rejection with Unif$(1/2,1)$ proposal if $\theta>1$.

---

### 2020 Q2 (old curriculum — for reference only)
$$f^*(x) \propto e^{-|x|}, \quad x\in(-1,1)$$

Unnormalised truncated Laplace on $(-1,1)$. $Z = \int_{-1}^{1}e^{-|x|}dx = 2(1-e^{-1})$.

---

### 2021 Q2a (old curriculum) — Power-law / Pareto
$$f_X(x) = \beta\,x^{-(\beta+1)}, \quad x\geq 1$$
$$F(x) = 1 - x^{-\beta}$$
$$F^{-1}(u) = (1-u)^{-1/\beta}$$

## 4. Conjugate Pairs

> ✏️ **Memorise** — the individual PDFs (Gamma, IG, Poisson) are given in the paper, but the posterior update rules are **not** given. You must derive them algebraically by recognising the kernel. Know the update rules cold.

---

### Poisson likelihood + Gamma prior → Gamma posterior

**Model** (both PDFs given in paper):
$$p(y\mid x) = \frac{x^y e^{-x}}{y!}, \qquad p(x) = \frac{\beta^\alpha}{\Gamma(\alpha)}\,x^{\alpha-1}e^{-\beta x}$$

**Posterior** ✏️ memorise update rule — one observation $y$:
$$\boxed{p(x\mid y) = \text{Gamma}(x;\,\alpha + y,\;\beta + 1)}$$

**Posterior** — $n$ i.i.d. observations:
$$p(x\mid y_{1:n}) = \text{Gamma}\!\left(x;\;\alpha + \textstyle\sum_{i=1}^n y_i,\;\beta + n\right)$$

---

### Gaussian likelihood + Inverse-Gamma prior → IG posterior

**Model** (IG PDF given in paper in 2023 Q3c):
$$p(x\mid \sigma^2) = \mathcal{N}(x;\,0,\,\sigma^2), \qquad p(\sigma^2) = \text{IG}(\sigma^2;\,\alpha,\,\beta)$$

**Posterior** ✏️ memorise update rule — one observation $x$:
$$\boxed{p(\sigma^2\mid x) = \text{IG}\!\left(\sigma^2;\;\alpha + \tfrac{1}{2},\;\beta + \tfrac{x^2}{2}\right)}$$

**Posterior** — $n$ observations:
$$p(\sigma^2\mid x_{1:n}) = \text{IG}\!\left(\sigma^2;\;\alpha + \tfrac{n}{2},\;\beta + \tfrac{1}{2}\textstyle\sum_{i=1}^n x_i^2\right)$$

---

### Gaussian likelihood + Gaussian prior → Gaussian posterior

**Model** ✏️ memorise — both prior and likelihood assumed known:
$$p(x) = \mathcal{N}(x;\,\mu_0,\,\sigma_0^2), \qquad p(y\mid x) = \mathcal{N}(y;\,x,\,\sigma^2)$$

**Posterior** ✏️ memorise update rule — one observation $y$:
$$\boxed{\mu_p = \frac{\sigma^2\mu_0 + \sigma_0^2\,y}{\sigma^2+\sigma_0^2}, \qquad \sigma_p^2 = \frac{\sigma^2\sigma_0^2}{\sigma^2+\sigma_0^2}}$$

**Appears in:** Lec Notes Ch1 Example 1.1; 2023 Q3.

## 5. Key Formulas to Know Cold

---

### Box-Müller Transform (Normal from Uniform)
> ⚙️ **Derive on the spot** — the transformation formula is given in the exam question (2024 Q1b); you must prove it produces standard normals using the Jacobian method. ✏️ Also memorise Version 1 (standard form) as it may not be given.

**Version 1** ✏️ memorise (standard; $U_1,U_2\sim\text{Unif}(0,1)$ independent):
$$Z_1 = \sqrt{-2\log U_1}\,\cos(2\pi U_2), \qquad Z_2 = \sqrt{-2\log U_1}\,\sin(2\pi U_2)$$

**Version 2** 🎁 given in paper (2024 Q1b; $U_1\sim\text{Unif}(0,1)$, $U_2\sim\text{Unif}(-\pi,\pi)$):
$$X_1 = \cos(U_2)\sqrt{-2\log U_1}, \qquad X_2 = \sin(U_2)\sqrt{-2\log U_1}$$

$Z_1,Z_2$ (and $X_1,X_2$) $\sim\mathcal{N}(0,1)$ independent.

**Key step in proof** ✏️ memorise: $W = X_1^2+X_2^2 \sim \text{Exp}(1/2)$, $\Theta = \arctan(X_2/X_1)\sim\text{Unif}(-\pi,\pi)$; Jacobian $|\det J_{g^{-1}}| = 2$.

---

### Gamma Function Identities
> ✏️ **Memorise** — not given; used to simplify normalising constants (χ², Gamma, IG).

$$\Gamma(n) = (n-1)! \quad \text{for } n\in\mathbb{N}, \qquad \Gamma(1/2) = \sqrt{\pi}, \qquad \Gamma(\alpha+1) = \alpha\,\Gamma(\alpha)$$

**Useful:** $\Gamma(3)=2$; $\Gamma(2)=1$; $\Gamma(1)=1$.

---

### Inversion Formulas — Quick Reference
> ✏️ **Derive, not memorise** — when a PDF is given in the exam, integrate to get $F(x)$ then invert. The results below are the outcomes of that process; know the method, not these by heart (except Exponential).

| Distribution | $F^{-1}(u)$ | Status |
|---|---|---|
| Exponential($\lambda$) | $-\frac{1}{\lambda}\log(1-u)$ | ✏️ Memorise |
| Rayleigh($\sigma$) | $\sqrt{-2\sigma^2\log(1-u)}$ | ⚙️ Derive from given PDF |
| Weibull($k$) | $(-\log(1-u))^{1/k}$ | ⚙️ Derive from given PDF |
| Cauchy | $\tan(\pi(u-\tfrac{1}{2}))$ | ✏️ Memorise (if tested) |
| Geometric($q$) | $\lfloor\log(1-u)/\log(1-q)\rfloor + 1$ | ✏️ Memorise (if tested) |
| Pareto($\beta$) | $(1-u)^{-1/\beta}$ | ✏️ Memorise (if tested) |

**Trick:** $1-U\overset{d}{=}U$ when $U\sim\text{Unif}(0,1)$, so $\log(1-u)$ can be written as $\log(u)$.

---

### Transformation of Random Variables (1D)
> ✏️ **Memorise** — used in Box-Müller proof and Weibull-from-Exponential derivations.

$$p_Y(y) = p_X(g^{-1}(y))\left|\frac{dg^{-1}(y)}{dy}\right| \qquad \text{(2D: replace derivative with }|\det J_{g^{-1}}|\text{)}$$

---

### Normalising Constants — Key Values
> ✏️ **Memorise the method** (use Gamma function); memorise $Z=16$ for the 2025 Q1c target.

| Target $p^*(x)$ | Support | $Z$ | How |
|---|---|---|---|
| $x^2 e^{-x/2}$ | $(0,\infty)$ | **16** | $\Gamma(3)\cdot 2^3 = 2\cdot 8$ |
| $x e^{-x^2/2}$ | $(0,\infty)$ | $1$ | Already normalised (Rayleigh $\sigma=1$) |
| $e^{-\|x\|}$ | $(-1,1)$ | $2(1-e^{-1})$ | Old curriculum only |

---

### Rejection Sampling Bound $M$
> ✏️ **Memorise** — needed for Q1 (exact sampling) every year.

$$M \geq \sup_x \frac{p^*(x)}{q(x)}, \qquad \text{accept with prob. } \frac{p^*(X)}{M\,q(X)}$$

**Efficiency:** acceptance rate $= Z/M$.
