## Key Equations at a Glance

| Equation | Expression |
|----------|------------|
| Bayes' rule | p(x\|y) = p(y\|x)p(x) / p(y) |
| Gaussian posterior (1D) | μ_p = (σ²μ₀ + σ₀²y)/(σ² + σ₀²), σ_p² = σ²σ₀²/(σ² + σ₀²) |
| Gaussian posterior (nD) | Σ_p = (A^TΣ⁻¹A + Σ₀⁻¹)⁻¹, μ_p = Σ_p(A^TΣ⁻¹y + Σ₀⁻¹μ₀) |
| Marginal likelihood (Gaussian) | p(y) = N(y; μ₀, σ₀² + σ²) |
| LCG | x_{n+1} ≡ ax_n + b (mod m), u_n = x_n/m |
| Probability integral transform | F⁻¹_X(U) ~ X when U ~ Unif(0,1) |
| Transformation (1D) | p_Y(y) = p_X(g⁻¹(y)) \|dg⁻¹/dy\| |
| Transformation (nD) | p_Y(y) = p_X(g⁻¹(y)) \|det J_{g⁻¹}(y)\| |
| Box-Müller | Z₁ = √(−2 ln U₁)cos(2πU₂), Z₂ = √(−2 ln U₁)sin(2πU₂) |
| Multivariate Gaussian | Y = μ + Lv, LL^T = Σ, v ~ N(0,I) |
| Rejection acceptance ratio | a(x) = p̄*(x) / (Mq(x)) |
| Acceptance rate (normalised) | â = 1/M |
| Acceptance rate (unnormalised) | â = Z/M |
| Optimal M | M* = sup_x p*(x)/q(x) |
| Curse of dimensionality | â_d = π^{d/2} / Γ(d/2+1) → 0 super-exponentially |
| MC estimator | φ̂_N = (1/N)Σ φ(X^(i)), X^(i)~p* |
| MC variance | Var(φ̂_N) = σ²/N, where σ² = Var_{p*}(φ(X)) |
| IS estimator (unnormalised) | Ê[φ] = (1/N)Σ W(X^(i))φ(X^(i)), W(x) = p̄*(x)/q(x) |
| SNIS estimator | φ̂ = Σ w̄^(i)φ(X^(i)), w̄^(i) = W^(i)/Σ W^(j) |
| Optimal IS proposal | q*(x) ∝ \|φ(x)\|p*(x) |
| ESS (IS) | ESS = (Σ W^(i))² / Σ (W^(i))² |
| MH acceptance ratio | α = min(1, p̄*(x')q(x\|x') / (p̄*(x)q(x'\|x))) |
| Random walk MH (symmetric q) | α = min(1, p̄*(x')/p̄*(x)) |
| MALA proposal | q(x'\|x) = N(x + γ∇log p*, 2γI) |
| ULA update | X_{n+1} = X_n + γ∇log p*(X_n) + √(2γ)V_n |
| SGLD update | X_n = X_{n-1} + γ(∇log p(x) + (n/K)Σ∇log p(y_{i_k}\|X)) + √(2γ)V_n |
| ESS (MCMC) | ESS = N / (1 + 2Σ_k ρ_k) |
| Simulated annealing | p^β ∝ exp(−βf(x)), β_t ↑ ∞ |
| EM update | θ_t = argmax_θ E_{p_{θ_{t-1}}(x\|y)}[log p_θ(x,y)] |
| Fisher's identity | ∇_θ log p_θ(y) = E_{p_θ(x\|y)}[∇_θ log p_θ(x,y)] |
| PMMH acceptance | α̂ = min(1, p̂(y\|θ')p(θ')q(θ\|θ') / (p̂(y\|θ)p(θ)q(θ'\|θ))) |
| SSM joint | π̄_t(x_{0:t},y_{1:t}) = μ(x₀)∏f(x_k\|x_{k-1})g(y_k\|x_k) |
| Filtering predict | ξ_t(x_t) = ∫f(x_t\|x_{t-1})π_{t-1}(x_{t-1})dx_{t-1} |
| Filtering update | π_t(x_t) = ξ_t(x_t)g(y_t\|x_t)/p(y_t\|y_{1:t-1}) |
| BPF incremental weight | W_t^(i) = g(y_t\|x̃_t^(i)) |
| BPF marginal likelihood | p̂(y_{1:t}) = ∏_k (1/N)Σ g(y_k\|x̃_k^(i)) |

## Core Theorems & Results

| # | Name / Topic | Statement (short) |
|---|-------------|--------------------|
| Thm 2.1 | Probability integral transform | F⁻¹_X(U) with U~Unif(0,1) has the same distribution as X |
| Thm 2.2 | Fundamental theorem of simulation | Sampling uniformly under p̄*(x) gives x-marginal ~ p*(x) |
| Prop 1.1 | Conditional Bayes rule | p(x\|y,z) = p(y\|x,z)p(x\|z)/p(y\|z) |
| Prop 2.1 | Chain rule for sampling | Sample X₁~p(x₁), then X₂\|X₁~p(x₂\|x₁) gives joint |
| Prop 2.2 | Marginalisation by projection | Joint sample → discard variables → marginal sample |
| Prop 2.3 | Acceptance rate | â = 1/M (normalised) or Z/M (unnormalised) |
| Prop 3.1 | MC estimator unbiasedness | E[φ̂_N] = E_{p*}[φ(X)]; unbiased for any N |
| Prop 3.2 | MC estimator variance | Var(φ̂_N) = σ²/N; CLT: √N(φ̂_N − μ) → N(0,σ²) |
| Prop 3.3 | IS estimator unbiasedness | E[(1/N)Σ W^(i)φ(X^(i))] = Z·E_{p*}[φ], where Z = ∫p̄*dx |
| Prop 3.5 | Optimal IS proposal | q*(x) ∝ \|φ(x)\|p*(x) minimises IS variance |
| Def 3.1 | Effective sample size (IS) | ESS = (Σ W^(i))² / Σ(W^(i))²; measures weight degeneracy |
| Prop 4.1 | Detailed balance → stationarity | If K(x'\|x)p*(x) = K(x\|x')p*(x') then p* is invariant for K |
| Prop 4.2 | MH detailed balance | MH kernel satisfies detailed balance w.r.t. p* (full proof via kernel decomposition) |
| Prop 4.3 | Gibbs stationarity | Each Gibbs kernel K_m satisfies detailed balance; composite K=K₁…K_d leaves p* invariant |
| Prop 5.1 | EM ascent property | log p_{θ_t}(y) ≥ log p_{θ_{t-1}}(y); EM never decreases likelihood |
| Prop 5.2 | Fisher's identity | ∇_θ log p_θ(y) = E_{p_θ(x\|y)}[∇_θ log p_θ(x,y)] |
| Prop 5.3 | PMMH correctness | Replacing p(y\|θ) by unbiased estimate in MH still targets p(θ\|y) exactly |

## Important Definitions

| Concept | Definition |
|---------|------------|
| Target distribution p* | Distribution we want to sample from |
| Unnormalised density p̄* | p*(x) = p̄*(x)/Z where Z = ∫p̄*(x)dx is intractable |
| Prior p(x) | Encodes prior knowledge about latent variable |
| Likelihood p(y\|x) | Model of observation process |
| Posterior p(x\|y) | Updated belief after observing y |
| Marginal likelihood p(y) | Evidence; normalising constant of posterior |
| Conditional independence | X⊥Y\|Z iff p(x,y\|z) = p(x\|z)p(y\|z) |
| Pseudo-random numbers | Deterministic sequence whose statistics match true randomness |
| LCG | x_{n+1} = ax_n + b (mod m); periodic with T ≤ m |
| CDF / Inverse CDF | F_X(x) = P(X≤x); generalised inverse F⁻(u) = min{x: F(x)≥u} |
| Proposal distribution q(x) | Easy-to-sample distribution used in rejection/IS |
| Acceptance rate â | Expected fraction of accepted samples |
| MC estimator | φ̂_N = (1/N)Σφ(X^(i)) approximates E_{p*}[φ(X)] |
| Importance weights | W(x) = p̄*(x)/q(x); reweight samples from q to target p* |
| Normalised IS weights | w̄^(i) = W^(i)/Σ W^(j); sum to 1 |
| SNIS | Self-normalised IS; uses normalised weights; biased but consistent |
| ESS (IS) | Effective sample size: (Σ W^(i))²/Σ(W^(i))²; diagnoses weight degeneracy |
| SIR | Sampling-importance-resampling: IS + resample to get unweighted samples |
| Markov chain | Sequence X₀,X₁,… where X_t depends only on X_{t-1} |
| Transition kernel K(x'\|x) | Probability of moving from x to x' |
| Irreducibility | Can reach any state from any state in finite steps |
| Invariant distribution | p* s.t. p*(x') = ∫K(x'\|x)p*(x)dx (or p*=p*M discrete) |
| Detailed balance | K(x'\|x)p*(x) = K(x\|x')p*(x'); sufficient for invariance |
| Ergodicity | Positive recurrence + aperiodicity → convergence to p* |
| Burn-in | Initial samples discarded before chain reaches stationarity |
| Full conditional | p_m(x_m\|x_{-m}); distribution of one variable given all others |
| ULA | Unadjusted Langevin algorithm; discretises Langevin SDE without MH correction; biased |
| MALA | Metropolis-adjusted Langevin; ULA + MH correction; exact |
| SGLD | Stochastic gradient Langevin dynamics; uses data subsampling for scalability |
| Autocorrelation | ρ_k = Cov(X_t,X_{t+k})/Var(X_t); measures MCMC mixing |
| ESS (MCMC) | N/(1+2Σρ_k); effective independent samples from correlated chain |
| Simulated annealing | p^β ∝ exp(−βf), raise β → concentrate on minimiser of f |
| MMLE | θ* = argmax_θ log p_θ(y) = argmax_θ log∫p_θ(x,y)dx |
| EM algorithm | Iterates E-step (compute Q(θ,θ_k)) and M-step (maximise Q) |
| State-space model (SSM) | Hidden Markov model: X_t\|X_{t-1}~f, Y_t\|X_t~g |
| Filtering distribution | π_t(x_t\|y_{1:t}); posterior of current state given all observations so far |
| Particle filter | SMC algorithm that approximates filtering distributions with weighted particles |
| Bootstrap PF | Particle filter using transition f as proposal; weights = g(y_t\|x̃_t) |
| Particle MCMC | MH for SSM parameters using BPF's unbiased marginal likelihood estimate |

## Method Toolkit

| Method | When to Use | Key Steps |
|--------|------------|----------|
| Inverse transform | F⁻¹ available in closed form | U ~ Unif(0,1), X = F⁻¹(U) |
| Transformation | Know a transform g and can compute Jacobian | Sample simple RV, apply g, verify via transformation formula |
| Box-Müller | Need standard normal samples | Two uniforms → two independent N(0,1) |
| Composition (joint) | Joint = marginal × conditional | Sample marginal first, then conditional given that sample |
| Composition (mixture) | p* = Σw_k q_k | Sample component index from weights, then sample from that component |
| Multivariate Gaussian | Need N(μ,Σ) samples | Cholesky Σ=LL^T; Y = μ + L·v where v~N(0,I) |
| Rejection sampling | General target, proposal covers it | Sample X~q, accept with prob p̄*(X)/(Mq(X)) |
| Optimal rejection design | Maximise acceptance rate | M* = sup p*/q, then optimise q over parameters |
| MC integration | Estimate E_{p*}[φ(X)] when samples available | φ̂ = (1/N)Σφ(X^(i)); error O(1/√N) by CLT |
| Importance sampling (IS) | Can't sample p* but can evaluate p̄* | Sample X^(i)~q, weight W^(i)=p̄*(X^(i))/q(X^(i)), estimate (1/N)ΣW^(i)φ(X^(i)) |
| SNIS | Only have unnormalised p̄* (don't know Z) | Normalised weights w̄^(i) = W^(i)/ΣW^(j); biased but consistent |
| SIR | Need unweighted approximate samples from p* | IS + multinomial resample from normalised weights |
| Metropolis-Hastings | General target, any proposal | Propose x'~q(·\|x), accept with α = min(1, p̄*q(x\|x')/(p̄*q(x'\|x))) |
| Random walk MH | No gradient info, moderate dimensions | Symmetric proposal q(x'\|x)=q(x\|x'); α = min(1, p̄*(x')/p̄*(x)) |
| MALA | Gradient available, want faster mixing | Proposal N(x+γ∇log p*, 2γI); MH correction ensures exactness |
| Gibbs sampling | Can sample each full conditional | Cycle through variables: x_m ~ p(x_m\|x_{-m}); no accept/reject |
| Metropolis-within-Gibbs | Some full conditionals intractable | Replace exact Gibbs step with MH step for that variable |
| ULA | Langevin without MH; fast but biased | X_{n+1} = X_n + γ∇log p* + √(2γ)V_n; bias → 0 as γ→0 |
| SGLD | Large dataset, gradient too expensive | Subsample data for stochastic gradient; no MH step |
| Simulated annealing | Global optimisation of f(x) | MH on p^β ∝ exp(−βf) with β_t → ∞; concentrates on minimum |
| EM algorithm | MMLE with latent variables | E-step: Q(θ,θ_k) = E[log p_θ(x,y)]; M-step: θ_{k+1} = argmax Q |
| SOUL | MMLE when E-step intractable | ULA to sample p_{θ_k}(x\|y), Fisher's identity for gradient, gradient ascent |
| PMMH | Parameter inference in intractable models | IS estimate of p(y\|θ) replaces exact likelihood in MH; exact for p(θ\|y) |
| Bootstrap PF | Filtering in state-space models | Propagate x̃~f(·\|x), weight by g(y_t\|x̃), resample |
| Particle MCMC | Parameter inference in SSMs | MH for θ using BPF's unbiased p̂(y_{1:T}\|θ) in acceptance ratio |

## Exam Notes

### Chapters 1–2
- **Bayesian inference** is the recurring application context — know Bayes' rule cold, especially the Gaussian conjugate case (1D and multivariate)
- **Inverse transform**: be comfortable deriving F⁻¹ for exponential, Cauchy, Weibull, discrete distributions
- **Transformation of RVs formula**: both 1D (|dg⁻¹/dy|) and 2D (|det J_{g⁻¹}|) — the Box-Müller proof and disk sampling are key examples
- **Rejection sampling** is the main algorithmic topic: know Algorithms 5 & 6, how to compute M, acceptance rate, optimise proposal
- **Curse of dimensionality**: understand why rejection sampling fails in high-d (volume ratio of ball to cube → 0)
- **Proofs to know**: Thm 2.1 (one-liner), Thm 2.2 (x-marginal of uniform under curve), Prop 2.3 (acceptance rate = 1/M), Box-Müller derivation
- **Numerical details**: LCG period, why pseudo-random numbers, log-space for numerical stability

### Chapter 3
- **MC integration**: know Prop 3.1 (unbiasedness), Prop 3.2 (variance σ²/N), CLT statement; be able to compute confidence intervals
- **Bias/Variance/MSE**: MSE = Bias² + Variance; know when each dominates
- **IS estimator**: derive W(x)=p̄*/q formula; know when IS helps (variance reduction) vs hurts (heavy tails)
- **Optimal proposal** (Prop 3.5): q* ∝ |φ|p*; prove it minimises variance
- **SNIS**: biased but consistent; know the normalised weight formula w̄^(i) = W^(i)/ΣW^(j)
- **Implementation**: log-weight trick, SIR algorithm, ESS formula and interpretation, mixture IS for robustness

### Chapter 4
- **Markov chain theory**: definitions (irreducibility, recurrence, aperiodicity), invariant distribution, detailed balance (Prop 4.1)
- **MH algorithm** (Alg 9): acceptance ratio formula; be able to prove detailed balance (Prop 4.2); specialise to independent/random-walk/MALA proposals
- **Gibbs sampling** (Alg 11): full conditionals, no accept/reject step; prove stationarity (Prop 4.3); random scan variant; Metropolis-within-Gibbs
- **Langevin MCMC**: Langevin SDE dX = ∇log p* dt + √2 dB; MALA (discretise + MH); ULA (no correction, biased); know Gaussian ULA analysis (p^γ = N(μ, 2σ⁴/(2σ²−γ)))
- **SGLD**: stochastic gradient with data subsampling; n/K scaling factor
- **Diagnostics**: trace plots, autocorrelation ρ_k, ESS_MCMC = N/(1+2Σρ_k), thinning
- **Simulated annealing** (Alg 13): p^β ∝ exp(−βf), increasing β schedule; acceptance ratio; nonconvex optimisation

### Chapter 5
- **MMLE formulation**: θ* = argmax log∫p_θ(x,y)dx; differs from MLE in intractable integral
- **EM algorithm**: E-step Q(θ,θ_k) = E_{p_{θ_k}(x|y)}[log p_θ(x,y)], M-step θ_{k+1} = argmax Q; **prove ascent** (Prop 5.1, key proof)
- **EM for Gaussian model**: derive recursion θ_{k+1} = rθ_k + b, show convergence to θ* = y/a
- **Fisher's identity** (Prop 5.2): ∇log p_θ(y) = E_{p_θ(x|y)}[∇log p_θ(x,y)]; enables gradient-based MMLE
- **SOUL** (Alg 14): ULA for sampling + Fisher's identity for gradient + gradient ascent; know the algorithm structure
- **PMMH** (Alg 16): unbiased IS estimate of p(y|θ) in MH; prove correctness via extended target (Prop 5.3); noise analysis (Example 5.3)

### Chapter 6
- **State-space models**: X_0~μ, X_t|X_{t-1}~f, Y_t|X_t~g; conditional independence structure; joint distribution formula
- **Filtering problem**: predict (integrate f against π_{t-1}) then update (multiply by g, normalise)
- **SIS** (Alg 17): sequential IS with recursive weight computation; suffers weight degeneracy
- **SISR/General PF** (Alg 18): SIS + resampling; cures degeneracy; incremental weights W_t = f·g/q
- **Bootstrap PF** (Alg 19): proposal = transition f; weights = g(y_t|x̃_t); simplest PF; evolutionary interpretation (mutation + selection)
- **Marginal likelihood**: p̂(y_{1:t}) = ∏(1/N)Σg(y_k|x̃_k^(i)); unbiased; use logsumexp for stability
- **Particle MCMC**: MH for θ in SSMs; uses BPF's unbiased p̂(y_{1:T}|θ) in acceptance ratio; exact sampling from p(θ|y_{1:T})
