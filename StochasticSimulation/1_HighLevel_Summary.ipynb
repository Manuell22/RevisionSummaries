## Key Equations at a Glance

| Equation | Expression |
|----------|------------|
| Bayes' rule | p(x\|y) = p(y\|x)p(x) / p(y) |
| Gaussian posterior (1D) | μ_p = (σ²μ₀ + σ₀²y)/(σ² + σ₀²), σ_p² = σ²σ₀²/(σ² + σ₀²) |
| Gaussian posterior (nD) | Σ_p = (A^TΣ⁻¹A + Σ₀⁻¹)⁻¹, μ_p = Σ_p(A^TΣ⁻¹y + Σ₀⁻¹μ₀) |
| Marginal likelihood (Gaussian) | p(y) = N(y; μ₀, σ₀² + σ²) |
| LCG | x_{n+1} ≡ ax_n + b (mod m), u_n = x_n/m |
| Probability integral transform | F⁻¹_X(U) ~ X when U ~ Unif(0,1) |
| Transformation (1D) | p_Y(y) = p_X(g⁻¹(y)) \|dg⁻¹/dy\| |
| Transformation (nD) | p_Y(y) = p_X(g⁻¹(y)) \|det J_{g⁻¹}(y)\| |
| Box-Müller | Z₁ = √(−2 ln U₁)cos(2πU₂), Z₂ = √(−2 ln U₁)sin(2πU₂) |
| Multivariate Gaussian | Y = μ + Lv, LL^T = Σ, v ~ N(0,I) |
| Rejection acceptance ratio | a(x) = p*(x) / (Mq(x)) |
| Acceptance rate (normalised) | â = 1/M |
| Acceptance rate (unnormalised) | â = Z/M |
| Optimal M | M* = sup_x p*(x)/q(x) |
| Curse of dimensionality | â_d = π^{d/2} / Γ(d/2+1) → 0 super-exponentially |

## Core Theorems & Results

| Thm 2.1 | Probability integral transform | F⁻¹_X(U) with U~Unif(0,1) has the same distribution as X |

## Important Definitions

| Concept | Definition |
|---------|------------|
| Target distribution p* | Distribution we want to sample from |
| Unnormalised density p̄* | p*(x) = p̄*(x)/Z where Z = ∫p̄*(x)dx is intractable |
| Prior p(x) | Encodes prior knowledge about latent variable |
| Likelihood p(y\|x) | Model of observation process |
| Posterior p(x\|y) | Updated belief after observing y |
| Marginal likelihood p(y) | Evidence; normalising constant of posterior |
| Conditional independence | X⊥Y\|Z iff p(x,y\|z) = p(x\|z)p(y\|z) |
| Pseudo-random numbers | Deterministic sequence whose statistics match true randomness |
| LCG | x_{n+1} = ax_n + b (mod m); periodic with T ≤ m |
| CDF / Inverse CDF | F_X(x) = P(X≤x); generalised inverse F⁻(u) = min{x: F(x)≥u} |
| Proposal distribution q(x) | Easy-to-sample distribution used in rejection sampling |
| Acceptance rate â | Expected fraction of accepted samples |

## Method Toolkit

| Method | When to Use | Key Steps |
|--------|------------|----------|
| Inverse transform | F⁻¹ available in closed form | U ~ Unif(0,1), X = F⁻¹(U) |
| Transformation | Know a transform g and can compute Jacobian | Sample simple RV, apply g, verify via transformation formula |
| Box-Müller | Need standard normal samples | Two uniforms → two independent N(0,1) |
| Composition (joint) | Joint = marginal × conditional | Sample marginal first, then conditional given that sample |
| Composition (mixture) | p* = Σw_k q_k | Sample component index from weights, then sample from that component |
| Multivariate Gaussian | Need N(μ,Σ) samples | Cholesky Σ=LL^T; Y = μ + L·v where v~N(0,I) |
| Rejection sampling | General target, proposal covers it | Sample X~q, accept with prob p̄*(X)/(Mq(X)) |
| Optimal rejection design | Maximise acceptance rate | M* = sup p*/q, then optimise q over parameters |

## Exam Notes

- **Bayesian inference** is the recurring application context — know Bayes' rule cold, especially the Gaussian conjugate case (1D and multivariate)
- **Inverse transform**: be comfortable deriving F⁻¹ for exponential, Cauchy, Weibull, discrete distributions
- **Transformation of RVs formula**: both 1D (|dg⁻¹/dy|) and 2D (|det J_{g⁻¹}|) — the Box-Müller proof and disk sampling are key examples
- **Rejection sampling** is the main algorithmic topic: know Algorithms 5 & 6, how to compute M, acceptance rate, optimise proposal
- **Curse of dimensionality**: understand why rejection sampling fails in high-d (volume ratio of ball to cube → 0)
- **Proofs to know**: Thm 2.1 (one-liner), Thm 2.2 (x-marginal of uniform under curve), Prop 2.3 (acceptance rate = 1/M), Box-Müller derivation
- **Numerical details**: LCG period, why pseudo-random numbers, log-space for numerical stability
